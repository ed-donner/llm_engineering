#### 1. Using requests library to use the orberta model which is not an LLM as we have to provide the knowledge base in context based on which roberta replies

```python
import os
import requests
from dotenv import load_dotenv

# Load API key
load_dotenv()
HF_API_KEY = os.getenv("Huggingface_API_Token")

API_URL = "https://api-inference.huggingface.co/models/deepset/roberta-base-squad2"
HEADERS = {"Authorization": f"Bearer {HF_API_KEY}"}

def query_qa(question, context):
    payload = {"inputs": {"question": question, "context": context}}
    response = requests.post(API_URL, headers=HEADERS, json=payload)
    if response.status_code != 200:
        raise Exception(f"Error {response.status_code}: {response.text}")
    return response.json()

# Example
question = "What is the capital of France?"
context = "France is a country in Europe. Its capital city is Paris."

answer = query_qa(question, context)
print("Q:", question)
print("A:", answer["answer"])
```

#### 3. Callling hugging face using requests library

```python
prompt = "When did France receive independence?"
```

```python
import os
import requests

HF_TOKEN = os.environ.get("Huggingface_API_Token")
if not HF_TOKEN:
    raise ValueError("HF_TOKEN not found in environment variables")

API_URL = "https://router.huggingface.co/featherless-ai/v1/completions"
headers = {
    "Authorization": f"Bearer {HF_TOKEN}",
    "Content-Type": "application/json"
}

def query(payload):
    response = requests.post(API_URL, headers=headers, json=payload)
    if response.status_code != 200:
        raise Exception(f"Error {response.status_code}: {response.text}")
    return response.json()

# ✅ Correct payload
payload = {
    "model": "meta-llama/Meta-Llama-3-8B",
    "prompt": "What is the capital of France?",  # <-- must be 'prompt', not 'inputs'
    "max_new_tokens": 5,
    "temperature": 0.7
}

output = query(payload)

# Featherless AI returns 'choices' list
generated_text = output["choices"][0]["text"]

clean_text = generated_text.strip()
print("Generated:", clean_text)
```

#### 4. Runing the tinyllama model locally after downloading from hugging face

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
# Choose the model
model_name = "TinyLlama/TinyLlama_v1.1"  # Base model

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# Load model
# If you have GPU: use device_map="auto"
# If CPU only: use torch_dtype=torch.float16 for faster inference if supported
try:
    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        device_map="auto",     # Automatically put layers on CPU/GPU
        load_in_8bit=True      # Optional quantization, reduces memory footprint
    )
except:
    # Fallback for CPU-only
    model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
# Or get it programmatically
from huggingface_hub import HfFolder
cache_dir = HfFolder.get_cache_dir()
print(f"Cache directory: {cache_dir}")
```

```python
# Choose device
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
```

```python
# Prompt
prompt = "What is the capital of France?"
```

```python
# Tokenize input
inputs = tokenizer(prompt, return_tensors="pt").to(device)

# Generate text
outputs = model.generate(
    **inputs,
    max_new_tokens=50,
    do_sample=True,
    temperature=0.7
)

# Decode output
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("Generated:", generated_text)

```
