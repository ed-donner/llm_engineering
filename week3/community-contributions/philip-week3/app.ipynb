```python
from minute_taker import (
    transcribe_audio,
    generate_minutes,
    send_email_with_attachment,
    export_to_pdf
)

import gradio as gr
```

```python


def process_meeting(audio_file_path, email_address):
    transcript = transcribe_audio(audio_file_path)
    minutes = generate_minutes(transcript)
    pdf_file = export_to_pdf(minutes)
    send_email_with_attachment(
        to_email=email_address,
        subject="Meeting Minutes",
        body="Attached are your AI-generated meeting minutes.",
        attachment_path=pdf_file
    )
    return transcript, minutes, pdf_fil
```

```python
import gradio as gr

with gr.Blocks() as demo:
    gr.Markdown("# 🎙 AI Meeting Assistant")

    audio_input = gr.Audio(
        label="Record Meeting Audio",
        type="filepath",
        tool="microphone"
    )
    email_input = gr.Textbox(label="Email Address")
    btn = gr.Button("Generate & Email Minutes")
    transcript_output = gr.Textbox(label="Transcript", lines=10)
    json_output = gr.JSON(label="Structured Minutes")
    file_output = gr.File(label="Download PDF")

    btn.click(
        process_meeting,
        inputs=[audio_input, email_input],
        outputs=[transcript_output, json_output, file_output]
    )

demo.launch()
```
