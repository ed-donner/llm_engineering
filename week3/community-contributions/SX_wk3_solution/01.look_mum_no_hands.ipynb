```python
# Run the code with Ollama models, NOTE: download the model first before running!!!
from visualiser_ollama import TokenPredictor, create_token_graph, visualise_predictions

message = "In one sentence, describe the color orange to someone who has never been able to see"

# Run downloaded ollama models
model_name = "llama3.2:latest"

predictor = TokenPredictor(model_name)
predictions = predictor.predict_tokens(message)
G = create_token_graph(model_name, predictions)
plt = visualise_predictions(G)
plt.show()
```

```python
# Run the code with Hugging Face models, NOTE: download the model could take a while, check MODEL SIZE before running!!!
from visualiser_HF import TokenPredictor_HF, create_token_graph_HF, visualise_predictions_HF

message = "In one sentence, describe the color orange to someone who has never been able to see"

# Run Hugging Face models
model_name = "meta-llama/Llama-3.2-1B-Instruct"

predictor = TokenPredictor_HF(model_name)
predictions = predictor.predict_tokens(message)
G = create_token_graph_HF(model_name, predictions)
plt = visualise_predictions_HF(G)
plt.show()
```
