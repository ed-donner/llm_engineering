# Week 3 -  Product Reviews Generator.
I created a product review generator that generates reviews for different products online.

```python
import ollama
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv(override=True)

client = OpenAI()

# Generate product reviews using nivea skincare company as an example
response = client.chat.completions.create(
    model="gpt-4",
    messages=[{"role": "user", "content": "Generate 5 realistic customer reviews for Nivea skincare products in Nigeria, the reviews should be 1-2 sentences long and contain a mix of positive and negative comments."}]
)
print(response.choices[0].message.content)
```

```python
# integrate gradio to generate reviews and descriptions
import gradio as gr

def generate_text(model_choice, prompt):
    if not prompt:
        return "Please enter a prompt."
    
    try:
        if model_choice == "Ollama":
            response = ollama.chat(model="llama3.2", messages=[{"role": "user", "content": prompt}])
            return response['message']['content']
        else:
            response = client.chat.completions.create(
                model="gpt-4",
                messages=[{"role": "user", "content": prompt}]
            )
            return response.choices[0].message.content
    except Exception as e:
        return f"Error: {str(e)}"

with gr.Blocks() as demo:
    gr.Markdown("## Product reviews Generator")
    gr.Markdown("Generate realistic product reviews using AI models.")
    
    model_choice = gr.Radio(["OpenAI", "Ollama"], label="Choose Model", value="OpenAI")
    prompt = gr.Textbox(
        label="Generate your reviews",
        placeholder="e.g., Generate 5 realistic customer reviews for Nivea skincare products in Nigeria...",
        lines=3
    )
    output = gr.Textbox(label="Generated Output", lines=10)
    
    btn = gr.Button("Generate", variant="primary")
    btn.click(generate_text, inputs=[model_choice, prompt], outputs=output)

demo.launch()
```
