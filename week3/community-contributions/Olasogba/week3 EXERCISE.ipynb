```python
from openai import OpenAI
client = OpenAI()

# Generate product reviews using nivea skincare company as an example
response = client.chat.completions.create(
    model="gpt-4",
    messages=[{"role": "user", "content": "Generate 5 realistic customer reviews for Nivea skincare products in Nigeria, the reviews should be 1-2 sentences long and contain a mix of positive and negative comments."}]
)
print(response.choices[0].message.content)
```
