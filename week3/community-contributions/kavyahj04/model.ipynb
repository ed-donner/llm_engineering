```python
# Lets create our menu and store it in menu.db 

def create_menu():
    with sqlite3.connect(db) as conn:
        cursor = conn.cursor()
        cursor.execute('CREATE TABLE IF NOT EXISTS menu (name TEXT PRIMARY KEY, category TEXT, Description Text, price REAL)') 
        conn.commit()
```

```python
def set_menu(name, category, description, price):
    with sqlite3.connect(db) as conn:
        cursor = conn.cursor()
        cursor.execute("INSERT INTO menu (name, category, Description, price) VALUES(?, ?, ?, ?) ON CONFLICT(name) DO UPDATE SET price = ?",(name.lower(), category.lower(), description.lower(), price, price))
        conn.commit()
```

```python
def get_item_from_category(category):
    with sqlite3.connect(db) as conn:
        cursor = conn.cursor()
        cursor.execute("SELECT name from menu where category = ?", (category.lower(),))
        res = cursor.fetchall()
        return res
get_item_from_category("drinks")
```

```python
category_function = {
    "name" : "get_item_from_category",
    "description" : "Get all the names of the dish that comes under this category",
    "parameters" : {
        "type" : "object",
        "properties"  :{
            "category" : { 
                "type" : "string",
                "description"  :"category of the food"
            }
        },
        "required" : ["category"],
        "additionalProperties" : False
    }
}
```

```python
def get_details(name):
    with sqlite3.connect(db) as conn:
        cursor = conn.cursor()
        cursor.execute("SELECT description, price from menu where name = ?", (name.lower(),))
        res = cursor.fetchone()
        return res
get_details("soda")
```

```python
details_function = {
    "name" : "get_details",
    "description"  : "Get the description and price for the name",
    "parameters" : {
        "type" : "object",
        "properties" : {
            "name" : {
                "type" : "string",
                "description" : "The name of the dish user wants to order"
            }
        },
        "required" : ["name"],
        "additionalProperties"  :False
    }
}
```

```python
tools = [{"type" : "function", "function" : category_function}, {"type" : "function", "function" : details_function}]
```

```python
system_message = """
You are a friendly restaurant assistant.

STRICT RULES:
- You ONLY know about dishes that exist in our database
- NEVER mention, suggest, or describe any dish not in our menu
- ALWAYS call the get_item_from_category tool when user asks about a category
- ALWAYS call the get_details tool when user asks about a specific dish
- NEVER make up prices, descriptions, or dish names
- If a user asks for something not on the menu, politely say it's not available

FLOW:
1. Greet the customer and tell them categories: Starters, Salads, Soups, Sandwiches, Burgers, Drinks
2. When they pick a category → call get_item_from_category tool and list ONLY those items
3. When they pick an item → call get_details tool and share description and price
4. When they want to order → add it to their order and confirm
5. When they are done → show full order with itemized prices and total

- Be warm and friendly like a great waiter
- Always confirm after adding or removing an item
"""
```

```python
def talker(message):
    response = openai.audio.speech.create(
      model="gpt-4o-mini-tts",
      voice="onyx",    # Also, try replacing onyx with alloy or coral
      input=message
    )
    return response.content
```

```python
def handle_tool_calls(message):
    response = []
    for tool_call in message.tool_calls:
        if tool_call.function.name == "get_item_from_category":
            print("tool 1 called")
            arguments = json.loads(tool_call.function.arguments)
            category = arguments.get("category")
            names = get_item_from_category(category.lower())
            response.append({
                "role": "tool",
                "content": str(names),
                "tool_call_id": tool_call.id
            })
        elif tool_call.function.name == "get_details":
            print("tool 2 called")
            arguments = json.loads(tool_call.function.arguments)
            name = arguments.get("name")
            details = get_details(name.lower())
            response.append({
                "role": "tool",
                "content": str(details),
                "tool_call_id": tool_call.id
            })
    return response
```

```python
def chat(history):
    history = [{"role":h["role"], "content" : h["content"]} for h in history]
    messages = [{"role" : "system", "content" : system_message}] + history
    response = openai.chat.completions.create(model = "gpt-4.1-mini", messages=messages, tools = tools)

    while response.choices[0].finish_reason == "tool_calls":
        message = response.choices[0].message
        responses = handle_tool_calls(message)
        messages.append(message)
        messages.extend(responses)
        response = openai.chat.completions.create(model = "gpt-4.1-mini", messages = messages, tools = tools)
    
    reply = response.choices[0].message.content
    history += [{"role":"assistant", "content":reply}]

    voice = talker(reply)

    return history, voice 


```

```python
import base64
from io import BytesIO
from PIL import Image
import requests

def artist(dish):
    image_response = openai.images.generate(
            model="gpt-image-1",
            prompt=f"A beautiful restaurant style spread of these dishes: {dish}, vibrant colors, appetizing, top down view",
            size="1024x1024",
            n=1,
        )
    image_base64 = image_response.data[0].b64_json
    image_data = base64.b64decode(image_base64)
    return Image.open(BytesIO(image_data))
```

```python
confirm_system_prompt = "Extract the food items the user has ordered from this conversation and return only a comma separated list of dish names, nothing else."

def confirm_order(history):
    history_str = "\n".join([f"{h['role']}: {h['content']}" for h in history])
    messages = [
        {"role": "system", "content": "Extract the food items the user has ordered from this conversation and return only a comma separated list of dish names, nothing else."},
            {"role": "user", "content": history_str}]
    res = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)
    dishes = res.choices[0].message.content
    image = artist(dishes)
    return image, gr.Button(interactive=False, value="✅ Order Confirmed!"), "🍽️ Your order is being prepared! Please wait a few minutes..."
```

```python
price_system_prompt = """
You are a silent order tracker. 

ONLY output a list if the user has EXPLICITLY confirmed adding an item with words like "add", "I want", "order", "yes I'll have".

If the user is just asking about a dish, browsing, or chatting → return BLANK, absolutely nothing.
If an item was removed → remove it from the list silently.
If nothing is ordered → return BLANK, absolutely nothing.

When items ARE ordered, return ONLY this format:
- Item Name: $price
Total: $total

No notes. No explanations. No greetings. No markdown. BLANK or the list. Nothing else.
"""
def get_price(history):
    messages = [{"role": "system", "content": price_system_prompt}] + [{"role": h["role"], "content": h["content"]} for h in history]
    response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)
    reply = response.choices[0].message.content
    return [{"role": "assistant", "content": reply}]
```

```python
def put_message_in_chatbot(message, history):
    return "", history + [{"role":"user", "content":message}]

with gr.Blocks() as ui:
    with gr.Row():
        chatbot = gr.Chatbot(height=500)
        image_output = gr.Image(height=500, interactive=False)
    with gr.Row():
        price_chatbot = gr.Chatbot(height=100)
    with gr.Row():
        audio_output = gr.Audio(autoplay=True)
    with gr.Row():
        message = gr.Textbox(label="Chat with our AI Assistant:")
    with gr.Row():
        confirm_btn = gr.Button("✅ Confirm Order", variant="primary")
        
    message.submit(put_message_in_chatbot, inputs=[message, chatbot], outputs=[message, chatbot]).then(
            chat, inputs=chatbot, outputs=[chatbot, audio_output]
        ).then(get_price, inputs=[chatbot], outputs = [price_chatbot])
    confirm_btn.click(confirm_order, inputs=chatbot, outputs=[image_output, confirm_btn])

ui.launch(inbrowser=True, auth=("kavya", "bananas"), share = True)
```
