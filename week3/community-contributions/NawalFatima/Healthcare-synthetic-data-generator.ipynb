## 🏥 AI Patient Diagnosis Generator using Llama 3.2

This notebook demonstrates running a quantized Llama 3.2 3B Instruct model locally on a GPU to generate structured medical assessments for synthetic patient data.

**What it does:**
- Generates synthetic patient records (age, gender, symptoms) using Faker
- Runs Llama 3.2 3B in 4-bit NF4 quantization using BitsAndBytes
- Prompts the model to produce diagnosis, medication, and doctor notes in JSON
- Parses and merges the LLM output back into the original patient records

**Key concepts covered:** local LLM inference, 4-bit quantization, prompt engineering, structured JSON output, HuggingFace Transformers pipeline

> ⚠️ This is a demonstration project only. Outputs should never be used for actual medical decisions.
