## Audio → Transcript → Meeting Minutes Pipeline

This solution converts an audio file into structured meeting minutes using:

Whisper (Speech-to-Text) → converts audio to transcript

LLM (Phi-3 / other HF model) → converts transcript to structured meeting minutes

Transformers + Quantization → runs efficiently on local GPU or CPU

A complete Audio → Transcript → Meeting Minutes pipeline using Hugging Face Whisper + Hugging Face LLM locally.

This works in Google Colab or local GPU/CPU.
