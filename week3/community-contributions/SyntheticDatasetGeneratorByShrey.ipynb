# SYNTHETIC DATASET GENERATOR

*Local LLM-Powered Synthetic Data Generation with Llama 3.2-1B-Instruct
Generate custom synthetic datasets by simply describing what data you need. This tool runs completely locally using Meta's Llama 3.2-1B-Instruct model on Google Colab — no paid API keys required after setup.*

## ✨ Features
**🦙 Fully Local - Runs on your own machine using Llama 3.2, no API costs**

**🎯 Prompt-Based Generation - Describe your dataset in plain English**

**💬 Chat Interface - Conversational Gradio UI with full message history**

**📊 CSV & JSON Export - Download your generated dataset instantly**

**⚡ 4-bit Quantization - Memory-efficient loading via BitsAndBytes**

**🔒 Private - Your data never leaves your machine**

**🤗 HuggingFace Integration - Model loaded via Transformers**

# 🧠 Model

**AI Model: meta-llama/Llama-3.2-1B-Instruct**

**Task: Structured data generation based on natural language prompts**

**Output Format: CSV or JSON, shown in chat and available for download**

**Quantization: 4-bit (NF4) via BitsAndBytes with bfloat16 compute dtype**

# 🌍 Use Cases

**🧪 Testing & Development - Generate test data for applications**

**📈 Data Science - Create sample datasets for ML experiments**

**🎓 Education - Generate datasets for learning and practice**

**🏢 Prototyping - Quick data mockups for demos**
