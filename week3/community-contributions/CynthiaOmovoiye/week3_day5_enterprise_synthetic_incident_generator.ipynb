# Enterprise Synthetic Incident Report Generator

This Colab notebook builds an enterprise-grade synthetic data generator for incident reports using open-source Hugging Face instruct models.

## What this notebook does
- Generates realistic incident reports in a strict JSON schema
- Supports batch generation (20 to 200+ records)
- Runs with **non-quantized** and **4-bit NF4 quantized** inference
- Validates and repairs malformed JSON outputs
- Compares quality, speed, and GPU memory usage
- Exports data to JSONL and CSV, with Colab download links

## Why this default model?

Default: **`Qwen/Qwen2.5-3B-Instruct`**

Reasoning:
- Instruct-tuned and good at structured generation
- Usually accessible on Hugging Face without special approvals
- Small enough to run on Colab T4, especially in 4-bit mode

A second model is included for comparison: **`meta-llama/Llama-3.2-1B-Instruct`**.
