## Synthetic Data Generator with Llama-3.2-3B-Instruct

This notebook builds a synthetic data generator that:

- Takes a user prompt (e.g., "Generate 100 customer review for a coffee shop")
- Uses the meta-llama/Llama-3.2-3B-Instruct model via Hugging Face Transformers
- Streams the generated output in a Gradio interface
- Saves the final dataset as CSV, JSON, or JSONL

All computations are done on a free T4 GPU using 4-bit quantization to reduce memory usages.
