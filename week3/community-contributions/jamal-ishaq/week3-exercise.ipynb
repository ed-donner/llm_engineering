### 📝 Project Description

**MedSynth-QA** is an automated pipeline designed to generate, validate, and anonymize synthetic medical dialogues. By leveraging the power of **Large Language Models (LLMs)** and **Natural Language Inference (NLI)**, this project creates high-quality, doctor-patient interactions tailored to specific medical niches while ensuring clinical consistency and data privacy.

The project solves a major bottleneck in AI Healthcare: the scarcity of high-quality, privacy-compliant training data. By using a "Model-as-a-Judge" architecture, we ensure that the synthetic data is not only diverse but also medically grounded.

---

### 🚀 Key Features

* **Expert-Grounded Generation:** Uses OpenAI's `gpt-4o-mini` with a "Developer-Role" persona to simulate specialist-level medical consultations.
* **Self-Correcting Validation:** Employs a **Cross-Encoder (DeBERTa-v3)** NLI model to detect and discard hallucinations by checking if clinical advice follows logically from patient symptoms.
* **Privacy-First Design:** Integrates **Microsoft Presidio** to automatically redact names, locations, and contact information, ensuring the dataset is safe for public research.
* **Hugging Face Integration:** Seamlessly pushes validated data to the Hugging Face Hub with proper sharding and metadata.

### 📊 Pipeline Logic

1. **Prompting:** Few-shot expert prompts generate raw dialogue in plain text to avoid formatting noise.
2. **Logic Check:** An NLI model verifies the "Entailment" between patient symptoms and doctor diagnosis.
3. **Sanitization:** Presidio scans for PII and replaces it with placeholders like `<PERSON>`.
4. **Deployment:** Cleaned data is batched, converted to Arrow format, and uploaded to the Hugging Face Hub.

---

### 📝 Dataset Schema

Each record in the generated dataset follows this structure:

* `niche`: The medical specialty (e.g., Cardiology, Neurology).
* `dialogue`: The anonymized, validated text containing the interaction between the patient and the healthcare provider.
