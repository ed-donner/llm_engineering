## Process üîß

1. **Inputs** topic for the song and heavy metal sub-genre
2. **Story** with GPT (`gpt-4o-mini`)
3. **Lyrics** with Hugging Face (`mistralai/Mistral-7B-Instruct-v0.3`) 
3. **Outputs** story and actual generated lyrics streamed to Gradio UI
