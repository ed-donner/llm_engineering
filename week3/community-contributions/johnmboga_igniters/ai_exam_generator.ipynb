# AI Exam & Exercise Generator.ipynb

This colab notebook demonstrates an AI-powered exam and exercise generator. Users can input a subject, select difficulty, and generate multiple-choice questions with explanations. The system supports multiple models via Hugging Face pipelines and enforces structured output to ensure consistency and correctness. A simple Gradio interface allows interactive generation and preview of questions, making it easy to explore AI-assisted assessment creation.
