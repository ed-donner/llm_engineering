# Next Token Prediction Visualizer

A Python-based visualization tool that demonstrates how an open-source Hugging Face LLM predicts the next token during text generation.

The project uses a Hugging Face causal language model (e.g., `Qwen/Qwen2.5-0.5B-Instruct`) to extract token-level probabilities directly from model logits and visualizes the prediction path as a directed graph using NetworkX and Matplotlib.

## Features

- Generate text token-by-token using a Hugging Face LLM
- Extract real next-token probabilities from model logits
- Display top predicted token and alternative candidates
- Visualize token prediction flow as an interactive graph
- Supports any Hugging Face causal language model
- Works locally without requiring paid APIs

## Tech Stack

- Python
- Hugging Face Transformers
- PyTorch
- NetworkX
- Matplotlib
