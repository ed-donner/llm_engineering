```python
"""
  pip i bytez
"""

import os
from bytez import Bytez

key = os.getenv("BYTEZ_API_KEY")   # loads real key
sdk = Bytez(key)

model = sdk.model("openai/gpt-4o")

results = model.run([
    {"role": "user", "content": "Hello"}
])

print({"error": results.error, "output": results.output})
```
