```python
import os
import numpy as np
import pandas as pd
from dotenv import load_dotenv
from openai import OpenAI
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
MODEL = "gpt-4"
```

```python
data = pd.DataFrame({
    "description": [
        "Samsung 55 inch 4K TV",
        "Infinix Android smartphone 128GB",
        "Dell Core i7 laptop 16GB RAM",
        "Wireless Bluetooth headphones"
    ],
    "price": [650000, 180000, 750000, 45000]
})
```

ML BASELINE

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(data["description"])
y = data["price"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)

lr = LinearRegression()
lr.fit(X_train, y_train)

rf = RandomForestRegressor()
rf.fit(X_train, y_train)

print("LR MAE:", mean_absolute_error(y_test, lr.predict(X_test)))
print("RF MAE:", mean_absolute_error(y_test, rf.predict(X_test)))
```

PREDICTION

```python
def gpt_predict(description):
    response = client.chat.completions.create(
        model=MODEL,
        messages=[{
            "role": "user",
            "content": f"Predict realistic Nigerian Naira price for: {description}"
        }]
    )
    return response.choices[0].message.content

gpt_predict("Samsung 65 inch OLED TV")
```
