## Import all necessary libraries

```python
from datasets import load_dataset
import pandas as pd
import plotly.express as px
import os,sys
```

## Load data from Huggingface

```python

dataset = load_dataset("princekhunt19/2025-ai-data-jobs-dataset")

df = pd.DataFrame(dataset["train"])

os.makedirs("../data/raw", exist_ok=True)

df.to_json("../data/raw/jobs_raw.json", orient="records")

print("Saved raw dataset:", len(df))
```
