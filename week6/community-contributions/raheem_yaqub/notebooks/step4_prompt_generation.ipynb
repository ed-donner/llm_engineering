# Step 4 — Prompt Generation
Generate prompts from cleaned dataset.

```python
import json, sys, os
sys.path.append(os.path.abspath(".."))
from src.job_item import JobItem

with open("../data/processed/jobs_cleaned.json") as f:
    data = json.load(f)

items = [JobItem(**row) for row in data]
print("Loaded items:", len(items))
```

```python
for item in items:
    item.make_prompt()

print(items[0].prompt)
```

```python
import pandas as pd
df = pd.DataFrame([i.model_dump() for i in items])
df["prompt_length"] = df["prompt"].apply(len)
df["prompt_length"].describe()
```
