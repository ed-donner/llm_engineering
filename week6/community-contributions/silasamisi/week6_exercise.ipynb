
# Week 6 Exercise: Simple Fine-Tuning Demo

This notebook demonstrates a **minimal fine-tuning style workflow**.

Steps:
1. Create sample dataset (`human_in.csv`, `human_out.csv`)
2. Load the dataset
3. Prepare prompt/completion pairs
4. Train a simple model to simulate fine-tuning
5. Evaluate performance with Mean Absolute Error
6. Plot predicted vs actual prices

No APIs are required for this demo.

```python
import pandas as pd
import json
import os

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

import matplotlib.pyplot as plt
```

```python
# Folder where the notebook lives
base_path = r"C:\Users\user\llm_engineering\week6\community-contributions\silasamisi"

# Example product descriptions
descriptions = [
    "Apple iPhone 13 128GB smartphone",
    "Samsung 55 inch 4K Smart TV",
    "Wireless Bluetooth headphones with noise cancelling",
    "Gaming laptop Intel i7 16GB RAM RTX graphics",
    "USB-C fast charger 65W",
    "Mechanical keyboard RGB gaming",
    "27 inch 4K computer monitor",
    "Portable external SSD 1TB",
    "Smartwatch fitness tracker waterproof",
    "Budget Android smartphone 64GB"
]

# Example prices
prices = [800, 650, 200, 1500, 40, 120, 400, 180, 150, 250]

# Create dataframes
df_in = pd.DataFrame({"description": descriptions})
df_out = pd.DataFrame({"price": prices})

# Save CSV files
df_in.to_csv(os.path.join(base_path, "human_in.csv"), index=False)
df_out.to_csv(os.path.join(base_path, "human_out.csv"), index=False)

print("CSV files created successfully")
```

```python
# Load dataset
inputs = pd.read_csv(os.path.join(base_path, "human_in.csv"))
outputs = pd.read_csv(os.path.join(base_path, "human_out.csv"))

df = pd.DataFrame({
    "description": inputs.iloc[:,0],
    "price": outputs.iloc[:,0]
})

# Create prompt/completion pairs
df["prompt"] = "Estimate the price of the product: " + df["description"]
df["completion"] = df["price"].astype(str)

df.head()
```

```python
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

vectorizer = TfidfVectorizer()

X_train = vectorizer.fit_transform(train_df["description"])
X_val = vectorizer.transform(val_df["description"])

y_train = train_df["price"]
y_val = val_df["price"]

model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_val)

mae = mean_absolute_error(y_val, predictions)

print("Mean Absolute Error:", mae)
```

```python
plt.scatter(y_val, predictions)

plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Predicted vs Actual Prices")

plt.show()
```
