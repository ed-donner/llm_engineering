# Step 6 — Fine‑Tuning

```python
from openai import OpenAI
client = OpenAI()
```

## Create a Job file to fine tune

```python
training_file = client.files.create(
file=open("../data/fine_tune/train.jsonl","rb"),
purpose="fine-tune"
)
training_file
```

## Start the fine tuning job

```python
job = client.fine_tuning.jobs.create(
    training_file=training_file.id,
    model="gpt-3.5-turbo",
    seed=42,
    hyperparameters={
        "n_epochs": 2
    },
    suffix="salarysense"
)

job
```

## List the fine tuning Job

```python
jobs = client.fine_tuning.jobs.list(limit=5)
jobs
```

## Get the latest Job id

```python
job_id = client.fine_tuning.jobs.list(limit=1).data[0].id
job_id
```

## Retrive the Job status

```python
client.fine_tuning.jobs.retrieve(job_id)
```

## View Training Events

```python
client.fine_tuning.jobs.list_events(
    fine_tuning_job_id=job_id,
    limit=10
).data
```

## Check when training finishes

```python
client.fine_tuning.jobs.retrieve("ftjob-WkiZAYhbIFn0rAnGjBl5O3Xr")
```

## Get Your Fine-Tuned Model

```python
job = client.fine_tuning.jobs.retrieve(job_id)

fine_tuned_model = job.fine_tuned_model
fine_tuned_model
```
