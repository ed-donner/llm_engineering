```python

import gradio as gr
from openai import OpenAI
import dotenv
```

```python
api_key = dotenv.get_key('.env', 'OPENAI_API_KEY')
client = OpenAI(
    api_key=api_key,
    base_url="https://openrouter.ai/api/v1"
)

models = ["gpt-4o-mini", "gpt-4o"]
```

```python
system = """
You are a senior software developer, specialize in typescript and python.
youre task is to covert the user input to typescript code based on user prompt, and the code should be clean and efficient.
dont generate any explanation, only code. if the user prompt is not clear, ask for clarification.
"""

def generate_code(user_prompt, model):
    print(f"Prompt: {user_prompt}")
    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": system},
            {"role": "user", "content": user_prompt}
        ]
    )
    
    data = response.choices[0].message.content
    print(data)
    return data
```

```python
with gr.Blocks( title=f"Port from Python to Typescript") as ui:
    gr.Markdown(f"## Port from Python to Typescript")
   
    
    with gr.Row():
        prompt_box = gr.Textbox(label="Enter your Python code here", lines=26)
        output_box = gr.Markdown(label="Generated Code", container=True, show_label=True, min_height=550, max_height=600,   elem_id="output-box")
    
    select_model = gr.Dropdown(models, value=models[0], label="Select Model")
    generate_button = gr.Button("Generate Code")
    generate_button.click(fn=generate_code,  inputs=[prompt_box, select_model], outputs=output_box)

ui.launch(inbrowser=True)
```
