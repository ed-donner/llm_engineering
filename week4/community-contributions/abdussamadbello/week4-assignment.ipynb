# Week 4 Exercise – Python to Go (Golang) code converter (abdussamadbello)

Uses a frontier model to generate **idiomatic Go** code from **Python** code. Same “code generator” idea as the course (Python→C++) but with Go as the target language.

```python
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()
openai = OpenAI()
MODEL = "gpt-4o-mini"

SYSTEM_PROMPT = """You are a code translator. Convert the given Python code into idiomatic Go (Golang).
Preserve logic and behavior. Use Go conventions (error returns, types, package main when appropriate).
Output only the Go code, no markdown fences or extra explanation."""


def python_to_go(python_code: str) -> str:
    response = openai.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": python_code},
        ],
        max_tokens=800,
    )
    return response.choices[0].message.content.strip()
```

```python
example_python = """
def fib(n: int) -> int:
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b
"""

go_code = python_to_go(example_python)
print("--- Generated Go ---")
print(go_code)
```
