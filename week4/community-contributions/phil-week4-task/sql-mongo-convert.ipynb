## Gradio app using this fixed OpenRouter setup so you can convert SQL → MongoDB interactively.

```python
import gradio as gr
from converter import convert_sql_to_mongo

def convert(sql_query):
    if not sql_query.strip():
        return "Please enter a SQL query."
    return convert_sql_to_mongo(sql_query)


with gr.Blocks() as demo:
    gr.Markdown("# AI SQL → MongoDB Converter")

    sql_input = gr.Textbox(
        label="SQL Query",
        lines=8,
        placeholder="SELECT name FROM users WHERE age > 30;"
    )

    convert_btn = gr.Button("Convert")

    output = gr.Code(
        label="MongoDB Query",
        language="javascript"
    )

    convert_btn.click(
        fn=convert,
        inputs=sql_input,
        outputs=output
    )

demo.launch()
```
