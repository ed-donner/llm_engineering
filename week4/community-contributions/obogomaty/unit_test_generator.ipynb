```python
import subprocess

# Path to your Ollama CLI
OLLAMA_PATH = "/usr/local/bin/ollama"
# Model to use
MODEL_NAME = "llama3.2:latest"

def translate_python_to_cpp(python_code):
    """
    Translate Python code to C++ using Ollama CLI (positional prompt).
    """
    prompt = f"Translate the following Python code to C++:\n{python_code}"
    
    try:
        result = subprocess.run(
            [OLLAMA_PATH, "run", MODEL_NAME, prompt],
            capture_output=True,
            text=True,
            timeout=60
        )
        return result.stdout.strip()
    except subprocess.TimeoutExpired:
        return "Error: Ollama model took too long to respond."
```

```python
python_code = """
def add(a, b):
    return a + b

print(add(5, 3))
"""

cpp_code = translate_python_to_cpp(python_code)
print("===== Generated C++ Code =====")
print(cpp_code)
```

```python
import gradio as gr

ui = gr.Interface(
    fn=translate_python_to_cpp,
    inputs=gr.Textbox(lines=10, placeholder="Paste your Python code here..."),
    outputs=gr.Textbox(lines=10, placeholder="C++ output will appear here..."),
    title="Python → C++ Translator",
    description="Enter Python code and get the C++ translation using Ollama locally."
)

ui.launch()
```
