# Python Refactorer – AI-Powered Code Modularizer

## Overview

This notebook implements an AI-powered system that refactors messy Python code into clean, modular, and maintainable code using a locally hosted Large Language Model (LLM). Instead of relying on cloud APIs, the system uses **Ollama** to run the language model locally.

The application provides an interactive interface built with **Gradio**, allowing users to paste Python code and receive a refactored version along with an explanation of the improvements.

## Key Features

- AI-powered Python code refactoring
- Code modularization into functions and classes
- Improved readability and maintainability
- Explanation of the refactoring process
- Real-time streaming output
- Interactive browser interface including CSS

## Technologies Used

- **Python**
- **Ollama** – local LLM runtime
- **Gradio** – interactive web interface

## Workflow

1. User pastes messy Python code into the interface.
2. The code is sent to a locally running LLM through Ollama.
3. The LLM analyzes and refactors the code following Python best practices.
4. The improved code and explanation are streamed back to the interface.

This approach allows developers and students to improve Python scripts quickly without needing external APIs or internet-based AI services.

```python
import gradio as gr
import ollama

# Choose your model installed in Ollama
MODEL = "llama3.2"

# Example messy Python code to demonstrate refactoring
example_code = """
a=[1,2,3,4,5]
s=0
for i in a:
    s=s+i
print("Total:",s)

b=[10,20,30]
t=0
for i in b:
    t=t+i
print("Total:",t)
"""

def refactor_code(code):

    prompt = f"""
You are an expert Python software engineer.

Refactor the following Python code.

Goals:
- Improve readability
- Break repeated logic into reusable functions
- Follow Python best practices (PEP8)
- Preserve original functionality

Return the response in this format:

### Refactored Code
<clean modular python code>

### Explanation
- bullet points explaining improvements

Python Code:
{code}
"""

    try:
        response = ollama.chat(
            model=MODEL,
            messages=[{"role": "user", "content": prompt}]
        )

        return response["message"]["content"]

    except Exception as e:
        return f"Error communicating with Ollama: {str(e)}"


# Build the interface
with gr.Blocks() as app:

    gr.Markdown("# Python Refactorer – AI Code Modularizer")

    gr.Markdown(
        "Paste messy Python code and the AI will refactor it into cleaner, modular, and maintainable Python code."
    )

    with gr.Row():

        with gr.Column():

            input_code = gr.Code(
                label="Original Python Code",
                language="python",
                value=example_code,
                lines=20
            )

            refactor_button = gr.Button("Refactor Code")

        with gr.Column():

            output = gr.Markdown(
                "### Refactored code and explanation will appear here."
            )

    refactor_button.click(
        fn=refactor_code,
        inputs=input_code,
        outputs=output
    )
```

```python
app.launch()
```
