```python
# !pip install openai gradio
import openai
import gradio as gr
import os
```

```python
openai.api_key = os.getenv('OPENAI_API_KEY')

def python_to_cpp(python_code):
    response = openai.ChatCompletion.create(
        model='gpt-4o',
        messages=[
            {'role': 'system', 'content': 'Convert Python code to high-performance C++ code.'},
            {'role': 'user', 'content': python_code}
        ]
    )
    return response['choices'][0]['message']['content']
```

```python
def code_converter_interface(python_code):
    return python_to_cpp(python_code)

with gr.Blocks() as demo:
    gr.Markdown("# Python to C++ Code Converter (Week 4)")
    py_code = gr.Textbox(label="Python Code", lines=6)
    cpp_code = gr.Textbox(label="C++ Code", lines=8)
    submit = gr.Button("Convert")
    submit.click(code_converter_interface, inputs=py_code, outputs=cpp_code)
demo.launch()
```
