# JSDoc Generator
Paste JavaScript code and get JSDoc comments added automatically using `deepseek-coder:1.3b` via Ollama.

```python
import gradio as gr
from openai import OpenAI

MODEL = 'deepseek-coder:1.3b'
client = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
```

```python
SYSTEM = """You are a code tool. Add JSDoc comments to JavaScript code.
Output ONLY the original code with JSDoc blocks inserted above functions. No explanations. No markdown fences. No extra text."""

def add_jsdoc(js_code):
    if not js_code.strip():
        yield "Please paste some JavaScript code first."
        return
    response = client.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": SYSTEM},
            {"role": "user", "content": js_code},
        ],
        stream=True,
    )
    result = ""
    for chunk in response:
        delta = chunk.choices[0].delta.content or ""
        result += delta
        yield result
```

```python
with gr.Blocks(title="JSDoc Generator") as demo:
    gr.Markdown("## JSDoc Generator\nPaste JavaScript code on the left, get JSDoc-annotated code on the right.")
    with gr.Row():
        input_code  = gr.Code(language="javascript", label="Your JavaScript", lines=25)
        output_code = gr.Code(language="javascript", label="With JSDoc", lines=25)
    btn = gr.Button("Add JSDoc", variant="primary")
    btn.click(fn=add_jsdoc, inputs=input_code, outputs=output_code)

demo.launch()
```
