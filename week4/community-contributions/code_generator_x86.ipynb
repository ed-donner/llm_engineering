# VS Tools Compile learnings for Windows x86 machine 
- The suggestion from GPT-5 didn't work, VS Tools compile requires env setup
- Create a batch file with following code to set environment and then compile

```python
# Contents of build_main.bat
@echo off
setlocal
"%ProgramFiles(x86)%\Microsoft Visual Studio\Installer\vswhere.exe" -latest -products * -requires Microsoft.VisualStudio.Component.VC.Tools.x86.x64 -property installationPath > "%TEMP%\vs.txt"
set /p VS=<"%TEMP%\vs.txt"
del "%TEMP%\vs.txt"
if not defined VS ( echo vswhere did not find VS & exit /b 1 )
call "%VS%\VC\Auxiliary\Build\vcvars64.bat"
cl /nologo /O2 /GL /EHsc /std:c++20 /permissive- /MD /arch:AVX2 main.cpp /Fe:main.exe /link /LTCG
```
