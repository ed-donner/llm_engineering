```python
groq=Groq()
website_url="https://en.wikipedia.org/wiki/Website"
website_content=fetch_website_contents(website_url)
print(website_content[:500])
```

```python
prompt=[
    {"role": "system", "content": "You are a good website content summarizer."},
    {"role":"user","content": "Summarize the following website content in less than 30 words" + website_content}
]

response=groq.chat.completions.create(model="openai/gpt-oss-120b",messages=prompt)
```

```python
response.choices[0].message.content
```

```python
def display_summary(url):
    website_content=fetch_website_contents(url)
    prompt=[
        {"role": "system", "content": "You are a good website content summarizer."},
        {"role":"user","content": "Summarize the following website content in less than 30 words" + website_content}
    ]
    response=groq.chat.completions.create(model="openai/gpt-oss-120b",messages=prompt)
    summary=response.choices[0].message.content
    display(Markdown(f"### Summary of {url}\n\n{summary}"))
```

```python
display_summary("https://en.wikipedia.org/wiki/Website")
```
