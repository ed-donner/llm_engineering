# Week 1 Lab: LLM Architecture Benchmarking for Value Investing
## Executive Summary: Model Performance on Financial Mispricings

### Overview
This notebook evaluates the capacity of local vs. cloud-based LLMs to apply a disciplined, quantitative framework (Benjamin Graham's Margin of Safety) to market dislocations, specifically utilizing the pandemic turnaround of Rolls-Royce (LSE: RR) as a case study.

### Key Insights
* **Cloud Infrastructure (GPT-4o-mini):** Demonstrates superior conceptual understanding of corporate finance metrics (WACC, terminal growth, cash flow floors). Best suited for generating structured financial commentary and advanced reasoning frameworks.
* **Local Processing (Llama 3.2 via Ollama):** Effective for local text processing and script generation, but subject to high hardware latency on laptop components and prone to factual hallucinations regarding specific operational leadership and historical cutoffs.


```python
# imports
import os
from dotenv import load_dotenv
from openai import OpenAI
import ollama
from IPython.display import display, Markdown
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

```python
# here is the question; type over this to ask something new

question = """
Act as an expert value investing analyst following Benjamin Graham's 'The Intelligent Investor' framework.

1. Explain the deep psychological difference between a stock with a 'low market price' relative to its intrinsic value, versus a genuinely 'bad business.' 

2. Provide historic and modern examples of massive market dislocations where an S&P 500 or FTSE 100 company was priced for bankruptcy by 'Mr. Market,' but its core intrinsic value remained intact, eventually proving the market wrong. 

3. Specifically analyze a case study like Rolls-Royce (which collapsed under pandemic pressures 5 years ago but has since undergone a massive operational turnaround, surging from under 70p to over £13.00). Detail how a disciplined 10-year DCF analyst could have spotted the Margin of Safety when the crowd was panicking.

Format your response using clear Markdown headings, bullet points, and a concluding analytical summary.
"""
```

```python
# Get gpt-4o-mini to answer, with streaming
stream = client.chat.completions.create(
    model=MODEL_GPT,
    messages=[{"role": "user", "content": question}],
    stream=True,
)

print("--- GPT-4o-mini Response ---")
full_response = ""
for chunk in stream:
    if chunk.choices[0].delta.content is not None:
        full_response += chunk.choices[0].delta.content

# This is where the Markdown structure is applied
display(Markdown(full_response))
```

```python
# Get Llama 3.2 to answer
import ollama

response = ollama.chat(
    model=MODEL_LLAMA,
    messages=[{'role': 'user', 'content': question}],
)

print("--- Llama 3.2 Response ---")
# Renders the response with headers and bullet points
display(Markdown(response['message']['content']))
```
