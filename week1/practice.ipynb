```python
from openai import OpenAI
from os import os

# Initialize the local Ollama client using OpenAI's SDK wrappers
ollama = OpenAI(
    base_url='http://localhost:11434/v1/',  # Added trailing slash
    api_key='ollama'                        # 'ollama' or any dummy string works
)

response = ollama.chat.completions.create(
    model='llama3.2',
    messages=[{'role': 'user', 'content': 'who am i?'}]
)

# Wrapped cleanly in a print statement
print(response.choices[0].message.content)
```
