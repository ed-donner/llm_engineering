# Game Review Bot

```python
from openai import OpenAI

OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="anything")
# Step 1: Create your prompts

system_prompt = """
    You are an experienced video game reviewer with deep knowledge of gameplay mechanics, storytelling, graphics, sound design, and industry trends. 
    Your role is to provide balanced, insightful, and engaging reviews that highlight both strengths and weaknesses. 
    You write in a clear, structured, and entertaining style, blending professional critique with relatable commentary. 
    Always consider the target audience of the game (casual players, hardcore gamers, genre fans) and compare it to similar titles when relevant. 
    Avoid spoilers unless explicitly requested.
"""
user_prompt = """
    Can you review "Elden Ring" as if you were writing for a gaming magazine? Focus on gameplay mechanics, difficulty, and how it compares to other FromSoftware titles. 
    Keep it engaging but professional.
"""

# Step 2: Make the messages list

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
] # fill this in

# Step 3: Call OpenAI
response = ollama.chat.completions.create(model="llama3.2", messages=messages)

# Step 4: print the result
print(response.choices[0].message.content)
```
