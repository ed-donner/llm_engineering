# Simple bedtime story teller (Week 1)

```python
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
openai = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")
```

```python
system_prompt = "You are a top tier bed time story teller"
user_prompt = "Tell me an amazing story about the balkan wars"

messages = [
    {
        "role":"system",
        "content":system_prompt
    },
    {
        "role":"user",
        "content":user_prompt
    }
]

response = openai.chat.completions.create(model="gpt-oss", messages=messages)

result = response.choices[0].message.content

display(Markdown(result))
```
