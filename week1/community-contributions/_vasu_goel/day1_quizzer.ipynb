```python
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openrouter import OpenRouter
```

```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')
```

```python
client = OpenRouter(
    api_key=os.getenv("OPENROUTER_API_KEY")
)
messages = [
    {"role": "system", "content": "You are a professor. Ask one beginner-friendly multiple-choice question at a time on the given topic. Give 4 options. If the student gives a correct answer, move to a new question. If the answer is wrong, give a small hint and ask the same question once. Do not repeat the same question over and over. After 5 questions, stop and show the result."}
]

def quizzer(prompt: str) -> str:
    messages.append({"role": "user", "content": "Topic is: " + prompt})
    response = client.chat.send(
        model="tencent/hy3:free",
        messages=messages
    )
    question = response.choices[0].message.content
    messages.append({"role": "assistant", "content": question})
    return question

print(quizzer("House of the dragons"))


def answerQuestion(prompt: str) -> str:
    messages.append(
        {"role": "user", "content": "Correct option is " + prompt}
    )
    response2 = client.chat.send(
        model="tencent/hy3:free",
        messages=messages
    )
    reply = response2.choices[0].message.content
    messages.append({"role": "assistant", "content": reply})
    return reply
```

```python
print(answerQuestion("C"))
```

```python
print(answerQuestion("A"))
```
