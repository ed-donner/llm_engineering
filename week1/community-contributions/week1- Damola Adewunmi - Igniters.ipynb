```python
# imports
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'  # Ollama model tag (use the name from `ollama list`; pull with `ollama pull llama3.2`)
```

```python
# set up environment
load_dotenv(override=True)
api_key = os.getenv('LLAMA_API_KEY')
OLLAMA_BASE_URL = "http://localhost:11434/v1"
openai = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
else:
    print("API key found and looks good so far!")

```

```python
# here is the question; type over this to ask something new
system_prompt = """
You are a helpful assistant that can answer questions about literature.
"""
question = """
Analyse all the works of George R.R. Martin and provide a summary of the key themes and characters.
"""
```

```python
# Get gpt-4o-mini to answer, with streaming
```

```python
# Get Llama 3.2 to answer
def generate_summary():
    response = openai.chat.completions.create(
        model=MODEL_LLAMA,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": question}
        ],
        response_format={"type": "json_object"}
    )
    result = response.choices[0].message.content
    
    return result
```

```python
generate_summary()
```
