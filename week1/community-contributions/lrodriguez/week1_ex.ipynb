```python
# imports
from openai import OpenAI
import os
from dotenv import load_dotenv
import json
from IPython.display import Markdown, display, update_display
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
OLLAMA_BASE_URL = 'http://localhost:11434/v1'
```

```python
# set up environment
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

#open_ai = OpenAI()
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
def get_question():
    question = """from {book.get("author") for book in books if book.get("author")}"""
    return question
```

```python
# here is the question; type over this to ask something new
def get_user_prompt():
    user_prompt = f"""
    Please explain what this code does and why:
    {get_question()}
    """
    return user_prompt
```

```python
# Get gpt-4o-mini to answer, with streaming
```

```python
# Get Llama 3.2 to answer
system_prompt = """You are an expert at answering technical questions regarding software engineering/programming.
Your task is read the code that is input by the user and explain it in detail.
If ther are any errors in the code, please point them out and explain how to fix them.
Assume that the user has, at most, a limited experience with programming.
"""

question = get_user_prompt()

user_prompt = get_user_prompt()

stream = ollama.chat.completions.create(
    model="llama3.2",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt},
    ],
    stream=True
)
response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    update_display(Markdown(response), display_id=display_handle.display_id)
```
