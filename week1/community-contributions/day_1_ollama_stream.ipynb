```python
import os
import sys
from dotenv import load_dotenv

# Add parent directory to path so we can import scraper
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from scraper import fetch_website_contents
from IPython.display import Markdown, display
```

```python
# Setup Ollama client
from openai import OpenAI

OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
# Define our user prompt

user_prompt_prefix = """
Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.

"""
```

```python
def summarize(url):
    website = fetch_website_contents(url)
    response = ollama.chat.completions.create(
        model = "gemma3:latest",
        messages = [{"role": "user", "content": user_prompt_prefix + website}],
        stream=True
    )
    
    full_response = ""
    for chunk in response:
        if chunk.choices[0].delta.content:
            content = chunk.choices[0].delta.content
            print(content, end='', flush=True)  # Stream to screen
            full_response += content  # Collect for return
    
    print()  
```

```python
summarize("https://www.google.com/")
```
