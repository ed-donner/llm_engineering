Day-1: Project - AI=Powered Resume Reviewer
What it does is it takes a resume text and give structured feedback like Strengths, Weakness, ATS improvements, Skill gaps, Clear role definition and controlled output. 

```python
#imports

import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
#LoadEnvironment and connect to openai

load_dotenv()

client = OpenAI(
    api_key = os.getenv("OPENAI_API_KEY")
)
```

```python
system_prompt = """
You are an expert technical recruiter and career coach
specializing in AI and software engineering roles.

Your task is to analyze resumes objectively and provide:
-Clear strengths
-Honest weakness
-Actionable improvement suggestions
-ATS(Applicant Tracking System) optimization tips

Your response must be professional, structured, and concise,
Respond in Markdown
"""
```

```python
user_prompt = """
Here is my resume.

Please review it and suggest improvements to make it
suitable for an AI Engineer role.

Resume:
"""
```

```python
#Input Data

resume_text = """
Name: Sai Chandra
Location: India

Summary:
Java Backend Engineer with 2 years of experience building REST APIs using Spring Boot.
Currently transitioning into Artificial Intelligence and Machine Learning.

Skills:
- Java, Python, SQL
- Spring Boot, Hibernate, REST APIs
- MySQL, PostgreSQL
- NumPy, Pandas, Scikit-learn (Beginner)
- Git, GitHub, IntelliJ IDEA

Experience:
Software Engineer – ABC Technologies (2023–Present)
- Built scalable REST APIs using Spring Boot
- Integrated MySQL using Hibernate ORM
- Wrote unit tests with JUnit and Mockito
- Worked in Agile environment

Projects:
- AI Resume Analyzer (Python + NLP)
- Stock Price Prediction using Scikit-learn
- Banking API using Spring Boot

Education:
B.Tech in Computer Science
"""

```

```python
#Message Builder

def messages_for(resume):
    return[
        {"role": "system", "content": system_prompt},
        {"role":"user", "content": user_prompt + resume}
    ]
```

```python
#Calling OpenAi Api

def review_resume(resume):
    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages= messages_for(resume)
    )

    return response.choices[0].message.content
```

```python
#Displaying the output in nicely

def display_review():
    review = review_resume(resume_text)
    display(Markdown(review))
```

```python
display_review()
```
