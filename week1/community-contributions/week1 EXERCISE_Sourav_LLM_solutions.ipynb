```python
# imports
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')

openai = OpenAI()
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
## Prompts

system_prompt = """ 
You are a Technical Assistant who is experienced in writing code reviewing codebases and understanding the code.
You always give the explanantion in a structured manner.

IMPORTANT Rule : 
1. Always ask the user if you require further clarifications
2. Before giving the output you have to validate the response and if you are confident then only you can publish the output
3. You will give the output STRICTLY Based on FACTS
4. You always suggest the user for a follow up topics related to the query that the user is asking
5. The output should be clear and concise.
"""

user_prompt = """ Please explain me on the topic""" + question
```

```python
message = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
```

```python
# Get gpt-4o-mini to answer, with streaming
stream = openai.chat.completions.create(
    model=MODEL_GPT,
    messages=message,
    stream=True
)
response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    update_display(Markdown(response), display_id=display_handle.display_id)
```

```python
# Get Llama 3.2 to answer

stream = ollama.chat.completions.create(
    model=MODEL_LLAMA,
    messages=message,
    stream=True
)
response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    update_display(Markdown(response), display_id=display_handle.display_id)
```
