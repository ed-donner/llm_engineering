```python
from bs4 import BeautifulSoup
from IPython.display import Markdown, display
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
```

```python
class Website:
    url: str
    title: str
    text: str

    def __init__(self, url):
        self.url = url

        # 1. Setup Options
        options = Options()
        # options.add_argument("--headless") # Uncomment this to run without a visible window
        options.add_argument("--no-sandbox")
        options.add_argument("--disable-dev-shm-usage")

        # 2. Start Driver (No PATH needed! Selenium handles it automatically)
        # driver = webdriver.Chrome(options=options)
        
        service = Service() 
        driver = webdriver.Chrome(service=service, options=options)  

        try:
            driver.get(url)

            # 3. Handle JavaScript/Verification
            # Since the site is JS-heavy, wait for the user or a specific element
            input("Complete verification in the browser, then press Enter here...")

            # 4. Grab the rendered HTML
            page_source = driver.page_source
        finally:
            driver.quit()

        # 5. Parse with BeautifulSoup
        soup = BeautifulSoup(page_source, 'html.parser')
        
        self.title = soup.title.string if soup.title else "No title found"
        
        # Clean up unwanted tags
        for irrelevant in soup(["script", "style", "img", "input"]):
            irrelevant.decompose()
            
        self.text = soup.get_text(separator="\n", strip=True)
```

```python
site = Website("https://www.nba.com/game/det-vs-cha-0022501171")

site.text
```

```python
print(site.text)
```

```python
system_prompt = "You are a basketball game analyst who analyzes the game match from a game website of NBA that analyzes the contents of a NBA match from the website and gives a short summary ignoring text that might be navigation related. \
Respond in markdown. Do not include any other text than the summary."

def user_prompt_for(website):
    user_prompt = f"You are looking at a website titled {website.title}"
    user_prompt += "\nThe contents of this website is as follows; \
please provide a short summary of this NBA game in markdown.\n\n"
    user_prompt += website.text
    return user_prompt
```

```python
print(user_prompt_for(site))
```

```python

OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')

def summary(url):
    response = ollama.chat.completions.create(
        model= "qwen3:14b",
        messages=[
            {"Role": "System", "Content": system_prompt},
            {"Role": "USer", "Content": user_prompt_for(url)}
        ]
    )
    return response.choices[0].message.content
```

```python
def clear_summary(url):
    summar = summary(url)
    display(Markdown(summar))
```
