```python
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
load_dotenv(override=True)
api_key = os.getenv("GEMINI_API_KEY")
if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python
GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"
google_api_key = api_key
gemini = OpenAI(base_url=GEMINI_BASE_URL, api_key=google_api_key)
```

```python
user_prompt = input("What would you like to learn? ")

system_prompt = f"""You are a jolly helpful assistant for finding the best online courses available, list 3 top courses, with a short description, the platform it's on, price of the course, duration, and a link to the course. The user is looking for courses on {user_prompt}"""

messages = [{"role": "system", "content": system_prompt}, {"role": "user", "content": user_prompt}];
response = gemini.chat.completions.create(model="gemini-2.5-flash", messages=messages, max_tokens=10000);
display(Markdown(response.choices[0].message.content))
```
