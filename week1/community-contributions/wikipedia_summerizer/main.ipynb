```python
import os
from openai import OpenAI
from IPython.display import Markdown, display
import wikipediaapi
from dotenv import load_dotenv
load_dotenv(override=True)
```

```python
# Creating prompts
user_prompt = input('Enter the topic you wanna search on Wikipedia')

system_prompt = """
You are a snarkyassistant that analyzes the contents of a wikipedia,
and provides a helpfull summary. Respond in markdown. Do not wrap the 
markdown in a code block - respond just with the markdown.
"""
```

```python
wiki_wiki = wikipediaapi.Wikipedia(
    user_agent='MyProjectName (johndoe@example.com)',
    language='en',
    extract_format=wikipediaapi.ExtractFormat.WIKI
)

if not user_prompt:
    user_prompt = 'Honda Civic'
    
#  Wikipedia topic fetch
p_wiki = wiki_wiki.page(user_prompt)
p_wiki.text
```

```python
# Prepare messages
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
```

```python
GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"
google_api_key = os.getenv("GOOGLE_API_KEY")

# Call the model
gemini = OpenAI(base_url=GEMINI_BASE_URL, api_key=google_api_key)
response = gemini.chat.completions.create(model="gemini-2.5-flash-preview-05-20", messages=messages)
```

```python
# Step 5: Display the summary
display(Markdown(response.choices[0].message.content))
```
