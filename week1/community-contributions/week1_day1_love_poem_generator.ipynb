````python
# imports
import os
import requests
import json
from typing import List
from dotenv import load_dotenv
from bs4 import BeautifulSoup
from IPython.display import Markdown,display,update_display
from openai import OpenAI
````

````python
# constants
MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'

LANGUAGES = ['English', 'Spanish', 'French', 'German', 'Italian', 'Nigerian Pidgin', 'Yoruba', 'Hausa', 'Igbo']
````

````python
# set up environment
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

if not api_key:
    print("No API key was found!")
else:
    print("API key found and looks good so far!")
````

````python
# function to input a valid language
def inputLanguage():
    print("Select a language from the following options:")
    for idx, lang in enumerate(LANGUAGES, start=1):
        print(f"{idx}. {lang}")
    choice = int(input("Enter the number corresponding to your choice: "))
    if 1 <= choice <= len(LANGUAGES):
        return LANGUAGES[choice - 1]
    else:
        print("Invalid choice. Defaulting to English.")
        return "English"
    
# function to input a valid lover's name
def inputLoversName():
    name = input("Enter your lover's name: ").strip()
    if name:
        return name
    else:
        print("Name cannot be empty. Please try again.")
        return inputLoversName()
````

````python
# ask for user inputs
language = inputLanguage()
lover_name = inputLoversName()
````

````python
system_prompt = "You are a world class poet who can create beautiful and evocative poetry of maximum 10 lines from any word given. Each line of the poem should end rhyming with the given lover's name."
user_prompt = "Please create a poem with each line end rhyming with the lover's name: " + lover_name + ". The poem should be in " + language + " language."
````

````python
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
````

````python
# Get gpt-4o-mini to answer, with streaming
openai = OpenAI()
stream = openai.chat.completions.create(model=MODEL_GPT, messages=messages, stream=True)
response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    response = response.replace("```","").replace("markdown", "")
    update_display(Markdown(response), display_id=display_handle.display_id)
````

````python
# Get Llama 3.2 to answer
response = ollama.chat(model=MODEL_LLAMA, messages=messages)
result = response['message']['content']
display(Markdown(result))
````
