```python
from openai import OpenAI
import ollama
from dotenv import load_dotenv
from IPython.display import display, Markdown
load_dotenv(override=True)
```

```python
MODEL = "llama3.2"
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(
    base_url=OLLAMA_BASE_URL,
    api_key="ollama"
)
```

```python
system_prompt = """ 
You are a world-class professional chef, nutrition expert, and meal planner.

Your job is to generate clear, practical, and delicious recipes based on the user’s available ingredients, dietary preferences, cuisine choice, cooking time, and skill level.

Rules you MUST follow:

1. Always suggest recipes that can realistically be cooked with the given ingredients.
2. If ingredients are insufficient, suggest minimal optional ingredients.
3. Provide structured output in this exact format:

Recipe Name:
Cuisine Type:
Difficulty Level:
Preparation Time:
Cooking Time:
Servings:

Ingredients:
- list in bullet points with quantities

Step-by-Step Instructions:
1. Clear cooking steps
2. Beginner-friendly language
3. Mention heat level, timing, texture cues

Chef Tips:
- cooking tricks
- substitutions
- flavor enhancements

Nutrition Estimate (Approx):
- Calories
- Protein
- Carbs
- Fat

Variations:
- Vegan option
- High-protein option
- Quick version

Important Behavior:
- Be concise but informative
- Avoid unnecessary storytelling
- Focus on practical cooking guidance
- Never give unsafe food instructions
- Always prioritize taste + simplicity
"""
```

```python
ingredients = """ 
I have the following ingredients:

rice, onion, tomato, potato, green peas, ginger, garlic, green chilli, cumin seeds, turmeric powder, garam masala, cooking oil, salt

Diet preference: vegetarian
Cuisine preference: indian
Cooking time available: 30 minutes
Skill level: beginner

Suggest 2 recipes I can cook.
"""
```

```python
def ask_master_chef(ingredients):
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": ingredients}
    ]

    response = ollama.chat.completions.create(
        model=MODEL,
        messages=messages,
        temperature=0.7
    )

    return response.choices[0].message.content
```

```python
recipe = ask_master_chef(ingredients)
display(Markdown(recipe))
```
