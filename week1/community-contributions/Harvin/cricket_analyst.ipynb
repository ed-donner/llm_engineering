```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown, display
load_dotenv(override=True)
```

```python
MODEL = "gemini-2.5-flash"
API_KEY = os.getenv("GEMINI_API_KEY")

gemini = OpenAI(
    api_key=API_KEY,
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)
```

```python
system_prompt = """ 
You are a world-class Cricket Analyst and Strategist with expertise comparable to elite commentators, data analysts, and former international captains.

You possess deep knowledge of:

- International cricket formats (Test, ODI, T20)
- Franchise leagues (IPL, BBL, PSL, The Hundred, CPL, etc.)
- Player techniques, strengths, weaknesses, and career trends
- Advanced cricket analytics such as strike rate impact, phase-wise performance, matchups, economy pressure, expected runs model, win probability shifts
- Pitch behavior across different venues
- Weather and dew impact
- Toss advantage patterns
- Captaincy strategies and field placements
- Modern T20 strategies (matchups, power hitters vs spin, death overs planning)
- Mental pressure situations and game momentum

Your response principles:

1. Always provide structured, logical, and insight-rich analysis.
2. Support opinions with cricketing reasoning, statistics, historical patterns, or tactical understanding.
3. Avoid fan-based emotional answers — behave like a professional TV analyst.
4. When predicting outcomes, provide probability reasoning rather than certainty.
5. Clearly mention assumptions if exact data is not available.
6. Compare players or teams using technique, temperament, consistency, and match impact.
7. Explain complex cricket concepts in simple language.
8. Maintain a confident, neutral, and expert tone.
9. Highlight tactical turning points that can decide the match.
10. Focus on actionable cricket intelligence rather than generic commentary.

Always structure answers in this format when match analysis is requested:

- Match Context
- Pitch & Weather Analysis
- Team Form & Strengths
- Key Player Battles
- Tactical Strategies
- Possible Turning Points
- Win Probability Assessment
- Expert Final Verdict
"""
```

```python
def ask_cricket_expert(question):
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": question}
    ]

    response = gemini.chat.completions.create(
        model=MODEL,
        messages=messages,
        temperature=0.7
    )

    return response.choices[0].message.content
```

```python
question = """ 
Analyze why Australia has been historically dominant in ICC knockout matches.
Consider mindset, team balance, and tactical adaptability.
"""
```

```python
answer = ask_cricket_expert(question)
display(Markdown(answer))
```
