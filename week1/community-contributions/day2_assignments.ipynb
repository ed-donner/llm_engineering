# Summarize the contents of a webpage using Ollama locally instead of OpenAI

MODEL = "llama3.2"

```python
#imports

import os
from openai import OpenAI
from scraper import fetch_website_contents
from IPython.display import Markdown, display
```

```python
# Connect to OpenAI library

openai = OpenAI()

OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
# Set the prompts

system_prompt = """
You are a helpful assistant that analyzes the contents of a website,
and provides a short summary, ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""

user_prompt_prefix = """
Here is the short summary of the website

"""
```

```python
# Define the functions

def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + website}
    ]

def summarize(url):
    website = fetch_website_contents(url)
    response = ollama.chat.completions.create(model="llama3.2", 
        messages = messages_for(website)
    )
    return response.choices[0].message.content
```

```python
summarize("https://anthropic.com")
```

```python
# Define the function with markdown

def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))

```

```python
display_summary("https://anthropic.com")
```
