To demonstrate familiarity with OpenAI API, and also Ollama, this tool that takes a technical question,  
and responds with an explanation.

```python
# imports
import os
import requests
import json
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment

load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')
openrouter_api_key = os.getenv('OPENROUTER_API_KEY')

# Check the keys
if openai_api_key:
    print(f"OpenAI API Key exists and begins with '{openai_api_key[:8]}'")
else:
    print("OpenAI API Key not set")
    
if openrouter_api_key:
    print(f"OpenRouter API Key exists and begins with '{openrouter_api_key[:7]}'")
else:
    print("OpenRouter API Key not set")
```

```python
# here is the question; type over this to ask something new

question = """
What was the name of the first President of Zambia? How old was he took tenure and when he died?
"""
```

```python
# Get gpt-4o-mini to answer, with streaming
openai = OpenAI(api_key=openai_api_key)
openai_stream = openai.chat.completions.create(model=MODEL_GPT, messages=[{"role": "user", "content": question}], stream=True)
openai_response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in openai_stream:
    openai_response += chunk.choices[0].delta.content or ''
    update_display(Markdown(openai_response), display_id=display_handle.display_id)
```

```python
# Get Llama 3.2 to answer
ollama = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")
ollama_response = ollama.chat.completions.create(model=MODEL_LLAMA, messages=[{"role": "user", "content": question}])
ollama_response.choices[0].message.content
```
