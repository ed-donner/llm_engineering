To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown,display,update_display
```

```python
# constants
load_dotenv(override=True)

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
OPENROUTER_API_KEY=os.getenv("OPENROUTER_API_KEY")
openRouter_url = "https://openrouter.ai/api/v1"
```

```python
# set up environment

if OPENROUTER_API_KEY and OPENROUTER_API_KEY.startswith("sk-") and len(OPENROUTER_API_KEY) > 10:
    print("API key looks good so far")
else:
    print("There might be a problem with your API key? Please check your environment variable!")
```

```python
# here is the question; type over this to ask something new

question = input("Enter your question: ")

system_prompt = """
You are an expert Computer Science tutor and software engineer.

Your role is to answer both theoretical and programming-related questions clearly and accurately.

The user may ask about:
- Programming (Python, Java, C++, JavaScript, Go, Rust, etc.)
- Data Structures & Algorithms
- Object-Oriented Programming
- Operating Systems
- Computer Networks
- Database Management Systems
- Software Engineering
- System Design
- Web Development
- Cloud Computing
- DevOps
- Artificial Intelligence
- Machine Learning
- Deep Learning
- Large Language Models (LLMs)
- Agentic AI
- Prompt Engineering
- Mathematics for Computer Science
- or any other Computer Science topic.

Instructions:
1. Detect whether the question is theoretical or code-related.
2. If it contains code:
   - Identify the programming language.
   - Explain what the code does.
   - Explain the syntax.
   - Explain important concepts used.
   - Explain the logic step-by-step.
   - Mention time and space complexity if applicable.
   - Point out potential improvements or best practices.
3. If it is a theoretical question:
   - Start with a simple definition.
   - Explain the core concepts.
   - Explain how it works.
   - Give real-world intuition or analogy where appropriate.
   - Provide practical use cases.
   - Mention advantages, disadvantages, and limitations if relevant.
4. Keep explanations beginner-friendly but technically accurate.
5. Use examples whenever they improve understanding.
6. If the user asks for a detailed explanation, provide an in-depth answer.
7. If the question is ambiguous, explain the most likely interpretation and mention any assumptions.
"""

user_prompt = f"""
Please answer the following question in detail:

{question}
"""
```

```python
# Get gpt-4o-mini to answer, with streaming
messages=[{"role":"system","content":system_prompt},{"role":"user","content":user_prompt}]
openai=OpenAI(base_url=openRouter_url,api_key=OPENROUTER_API_KEY)
stream=openai.chat.completions.create(model=MODEL_GPT,messages=messages,stream=True)
response=""

display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ""
    update_display(Markdown(response), display_id=display_handle.display_id)
```

```python
# Get Llama 3.2 to answer
openai = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama"
)
stream=openai.chat.completions.create(model=MODEL_LLAMA,messages=messages,stream=True)
response=""

display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ""
    update_display(Markdown(response), display_id=display_handle.display_id)
```
