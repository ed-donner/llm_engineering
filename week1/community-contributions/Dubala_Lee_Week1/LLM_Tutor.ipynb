# A Tutot for LLM Engineering Courses

## This is your personal tutor help to answer questions for LLM Engineering Courses

### BUSINESS Scope:

A learning tutor for my LLM Engineering class helps me understand course concepts, debug assignments, and connect theory to practical implementation. It answers my questions clearly, explains code and model behavior step by step, and guides me toward solutions without simply doing the work for me.

````python
# imports

from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
import ollama
````

````python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
````

````python
# set up environment

load_dotenv()
openai = OpenAI()
````

````python
# here is the question; type over this to ask something new

question = """
What is git and how to use it. 
"""
````

````python
# prompts

system_prompt = """You are a helpful technical tutor who answers questions for my LLM Enginerring learning course. 
You should be able to explain the concepts in a way that is easy to understand and follow, explain the code and model behavior step by step, 
and guide me toward solutions without simply doing the work for me."""
user_prompt = "Please give a detailed explanation to the following question: " + question
````

````python
# messages

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
````

````python
# LLM response by GPT 4o, with streaming

stream = openai.chat.completions.create(model=MODEL_GPT, messages=messages, stream=True)

#display the response

response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    response = response.replace("```","").replace("markdown", "")
    update_display(Markdown(response), display_id=display_handle.display_id)
        
````

````python
# Get Llama 3.2 to answer

response = ollama.chat(model=MODEL_LLAMA, messages=messages)
reply = response['message']['content']
display(Markdown(reply))
````
