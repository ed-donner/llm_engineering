To demonstrate your familiarity with GOOGLE GEMINI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports

from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
import ollama
import os
```

```python
# constants
MODEL_LLAMA = 'llama3.2:1b'
```

```python
# set up environment

load_dotenv(override=True)
GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"
google_api_key = os.getenv("GOOGLE_API_KEY")

if not google_api_key:
    print("No API key was found - please be sure to add your key to the .env file, and save the file!")
elif not google_api_key.startswith("AIz"):
    print("An API key was found, but it doesn't start AIz. Please check your key.")
else:
    print("API key found and looks good so far!")
```

```python
# here is the question; type over this to ask something new

question = """
Refactor the following code to improve its time complexity from O(N^2) to O(N):

def find_pairs_with_sum(arr, target_sum):
    pairs = []
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            if arr[i] + arr[j] == target_sum:
                pairs.append((arr[i], arr[j]))
    return pairs
"""
```

```python
# prompts

system_prompt = "You are a helpful technical tutor who answers questions about python code, software engineering, data science and LLMs."
user_prompt = "Please give a detailed explanation and complete refactored code for the following question: " + question
```

```python
# messages

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
```

```python
# Get gemini-2.5-flash-lite to answer, with streaming

gemini = OpenAI(base_url=GEMINI_BASE_URL, api_key=google_api_key)

stream = gemini.chat.completions.create(model="gemini-2.5-flash-lite", messages=messages, stream=True)
    
response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    if chunk.choices and len(chunk.choices) > 0:
        content = chunk.choices[0].delta.content
        if content is not None:
            response += content
            update_display(Markdown(response), display_id=display_handle.display_id)
```

```python
import ollama  # This resets the variable 'ollama' back to the imported library

try:
    response = ollama.chat(model=MODEL_LLAMA, messages=messages)
    reply = response['message']['content']
    display(Markdown(reply))
except ollama.ResponseError as e:
    print(f"Ollama Error: {e}")
    print(f"Please verify if you have pulled the model using: !ollama pull {MODEL_LLAMA}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
