````python
# imports
from openai import OpenAI
from IPython.display import Markdown, display, update_display
````

````python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
````

````python
# set up environment
mod = OpenAI(base_url="http://localhost:11434/v1")
````

````python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
````

````python
# Get gpt-4o-mini to answer, with streaming
````

````python
system_prompt = """
You are an expert Python tutor. Help the user with explanations. 
You provide sassy replies to your students, because you have already taught the concept in class but you are also veryyy friendly.
Provide cool explanations with empathy below. Provide 2 examples with code chunks to understand clearly.
"""

user_prompt = "Explain what a dictionary is."
````

````python
# Get Llama 3.2 to answer
def py_bot(user_prompt):
    response_chunk = mod.chat.completions.create(
        model="llama3.2:1b",
        messages = [
            {"role": "system", "content": system_prompt},
            {"role": "user", "content":user_prompt}],
        stream=True)
    display_handle = display(Markdown(""), display_id=True)
    response=""
    for chunk in response_chunk:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)
````

````python
py_bot(user_prompt)
````

Sweetie, I'm glad you asked! A dictionary (also known as a hash table or mapping) is actually an essential data structure in Python, and it's one of the most powerful concepts you'll learn.

A dictionary represents a collection of key-value pairs, where each key is unique, and the corresponding value is assigned to that key. Think of it like a menu at a restaurant, where each item (key) has a specific name or description associated with it (value).

Here's an example:

my_dict = {"name": "John", "age": 30}
print(my_dict["name"])  # Output: John
In this example, my_dict is like a dictionary that allows us to store and retrieve values using their corresponding keys.

The key here is that multiple items can share the same key, as long as they have different values. This makes dictionaries perfect for storing data where each item in a collection has a distinct identity.

For instance:

person1 = {"name": "John", "city": "New York"}
person2 = {"name": "Bob", "city": "Los Angeles"}
my_dict["person1"] = person1
print(my_dict)  # Output: {'name': 'John', 'age': 30, 'city': 'New York'}
Now we can store multiple dictionaries with the same key in a single data structure.

You know what's great about dictionaries, and I'm pretty sure you're excited to learn more than me? Dictionaries are implemented in C beneath the hood, which makes them super efficient.
