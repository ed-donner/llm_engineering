# Playing with tokens

```python
import tiktoken

encoding = tiktoken.encoding_for_model("gpt-4.1-mini")

tokens = encoding.encode("Hi my name is Shubhasmita")
```

```python
tokens
```

```python
for token_id in tokens:
    token_text = encoding.decode([token_id])
    print(f"{token_id} = {token_text}")
```

```python
encoding.decode([1955])
```

# Exploring memory

```python
from openai import OpenAI

mod = OpenAI(base_url="http://localhost:11434/v1")
```

```python
messages = [
    {"role": "system", "content": "You are a curious assistant"},
    {"role": "user", "content": "Hi! I'm Shubhasmita!"}
    ]
```

```python
response = mod.chat.completions.create(model="llama3.2", messages=messages)
response.choices[0].message.content
```

```python
messages = [
    {"role": "system", "content": "You are a curious assistant"},
    {"role": "user", "content": "What's my name?"}
    ]
```

```python
response = mod.chat.completions.create(model="llama3.2", messages=messages)
response.choices[0].message.content
```
