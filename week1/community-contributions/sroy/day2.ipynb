## Trying out fetch_website_contents instead of directly passing url

```python
import os
from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt_prefix+fetch_website_contents("https://edwarddonner.com")}
]

deepseek = OpenAI(base_url="http://localhost:11434/v1", api_key="something")
response = deepseek.chat.completions.create(model="llama3.2", messages=messages)
response.choices[0].message.content
```

OUTPUT:

'### Website Summary\n\nThis website appears to be the personal homepage of Edward Donner, a prominent figure in the AI industry. He\'s the co-founder and CTO of Nebula.io and has experience as the founder of untapt and Managing Director at JPMorgan. The site showcases his expertise in LLMs (Large Language Models) through Udemy courses that have achieved significant success, with over 600,000 enrolled students across 194 countries.\n\n### News/Announcements\n\n* Edward Donner announced new resources on his blog:\n\t+ "Vibe Coder to Agentic Engineer" - February 17, 2026\n\t+ "AI Builder with n8n" - January 4, 2026\n\t+ "AI Engineering MLOps Track" - September 15, 2025\n\t+ A suggested order for taking his AI courses is mentioned in a separate post.'


```python
def get_message(url):
    return [{"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt_prefix+fetch_website_contents(url)}]
```

```python
def summarise(url):
    mod = OpenAI(base_url="http://localhost:11434/v1", api_key="something")
    response = mod.chat.completions.create(
        model ="llama3.2",
        messages=get_message(url)
    )
    return response.choices[0].message.content

def summary_md(url):
    summary = summarise(url)
    display(Markdown(summary))
```

```python
summary_md("https://edwarddonner.com")
```

OUTPUT:

This is a website centered around the ego trip of its creator, Edward Donner. He writes about AI, self-congratulates for his own success stories (500k Udemy course enrollments!), and shares what amounts to thinly veiled self-promotion on "AI Builder" articles that provide resources (read: overpriced online courses).

There's no actual content or updates on the site; it's just Edward rambling about himself.
