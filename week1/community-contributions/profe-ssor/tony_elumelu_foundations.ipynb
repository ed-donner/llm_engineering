# Tony Elumelu Foundation – Website Summarizer

The **Tony Elumelu Foundation** gives **grants and support to entrepreneurs who wish to start or grow their own business**, especially across Africa. This notebook summarizes their website (mission, programs, impact, key numbers) using the OpenAI API or **OpenRouter** (OpenAI-compatible).

**Requirements:** `.env` in the project root with `OPENAI_API_KEY` (OpenAI key starting with `sk-proj-` or OpenRouter key starting with `sk-or-v1-`).

**Note:** Run the notebook from the repo root or from `week1/` so that `scraper` can be imported; or run this from `profe-ssor/` and the first code cell will add the path.

```python
# Imports (add week1 to path so scraper can be found)
import os
import sys
from pathlib import Path

cwd = Path.cwd()
if cwd.name == "profe-ssor":
    week1_dir = cwd.parent.parent
elif cwd.name == "week1":
    week1_dir = cwd
else:
    week1_dir = cwd / "week1"  # repo root
if str(week1_dir) not in sys.path:
    sys.path.insert(0, str(week1_dir))

from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
# Load API key from .env (project root or week1)
load_dotenv(override=True)
if not os.getenv("OPENAI_API_KEY") and Path.cwd().name == "profe-ssor":
    load_dotenv(Path.cwd().parent.parent.parent / ".env", override=True)
api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    print("No API key found. Add OPENAI_API_KEY to your .env file.")
elif not (api_key.startswith("sk-proj-") or api_key.startswith("sk-or-v1-")):
    print("API key should start with sk-proj- (OpenAI) or sk-or-v1- (OpenRouter).")
elif api_key.strip() != api_key:
    print("API key has leading/trailing spaces. Remove them in .env.")
else:
    print("API key OK" + (" (OpenRouter)" if api_key.startswith("sk-or-v1-") else ""))
```

```python
# Use OpenRouter if key is sk-or-v1-; otherwise OpenAI
if api_key and api_key.startswith("sk-or-v1-"):
    openai = OpenAI(api_key=api_key, base_url="https://openrouter.ai/api/v1")
    model = "openai/gpt-4o-mini"
else:
    openai = OpenAI()
    model = "gpt-5-nano"
```

```python
system_prompt = """
You are a concise analyst that summarizes organizational websites.
You focus on: mission and vision, main programs and impact, key numbers (funding, beneficiaries, countries), and notable partners or initiatives.
Always make clear who the organization gives grants or funding to (e.g. entrepreneurs who wish to start their own business).
Use clear, neutral language. Respond in markdown. Do not wrap the markdown in a code block—respond with markdown only.
"""

user_prompt = """
Below is the content from a website. Summarize this organization in a short, structured way. Include:

1. **What they do** – mission and main focus. Clearly state who they give grants or funding to (e.g. entrepreneurs who wish to start their own business).
2. **Key programs** – flagship initiatives (e.g. how entrepreneurs can apply for grants or training).
3. **Impact** – numbers on funding, beneficiaries, jobs, countries if mentioned
4. **Notable facts** – founding year, founder, partners, or recent updates

Keep it under 300 words. Use markdown headings and bullet points.

Website content:
"""
```

```python
def summarize_website(website):
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt + website}
    ]
    response = openai.chat.completions.create(model=model, messages=messages)
    return response.choices[0].message.content
```

```python
# Fetch Tony Elumelu Foundation and display summary
url = "https://www.tonyelumelufoundation.org/"
content = fetch_website_contents(url)
display(Markdown(summarize_website(content)))
```
