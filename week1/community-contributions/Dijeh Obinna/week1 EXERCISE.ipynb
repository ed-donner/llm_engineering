```python
# imports
import os
import json
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from scraper import fetch_website_links, fetch_website_contents
from openai import OpenAI
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

if api_key and api_key.startswith('sk-proj-') and len(api_key)>10:
    print("API key looks good so far")
else:
    print("There might be a problem with your API key? Please visit the troubleshooting notebook!")
    

openai = OpenAI()
```

```python
# here is the question; type over this to ask something new

system_prompt = """
Role: You are an expert technical educator and senior engineer.
Task: Your goal is to take a technical question and provide a comprehensive, clear, and structured explanation.
Instructions:
Analyze the Question: Determine the core technical concept being asked.
Reasoning: Before answering, "think out loud" to identify key components and potential pitfalls.
The Explanation:
Use analogies to explain complex concepts to non-experts.
Use Markdown (headers, bolding, and bullet points) for high scannability.
Provide a concrete code example or diagram description if applicable.
Structure:
Direct Answer: Start with a 1-2 sentence summary.
Deep Dive: Detailed explanation broken into logical sections.
Summary/Best Practices: A brief "key takeaway" section.
"""
question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": question}
]
```

```python
# Get gpt-4o-mini to answer, with streaming
response = openai.chat.completions.create(model=MODEL_GPT, messages=messages)
gpt_response = response.choices[0].message.content
gpt_response
```

```python

display(Markdown(gpt_response))
```

```python
# Get Llama 3.2 to answer
ollama = OpenAI(
    base_url='http://localhost:11434/v1',
    api_key='ollama', # required but ignored for local calls
)

ollama_response = ollama.chat.completions.create(
  model="llama3.2:1b",
  messages=messages
)
ollama_response_raw = ollama_response.choices[0].message.content
ollama_response_raw
```

```python
display(Markdown(ollama_response_raw))
```
