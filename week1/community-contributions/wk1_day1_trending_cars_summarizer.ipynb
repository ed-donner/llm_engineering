```python
import os
from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
openai = OpenAI()
system_prompt = """you are a automobile enthusiast and have good knowledge of cars."""
user_prompt = """For the given website, please let me know the top 5 trending car models. """
```

```python
def messages(website):
    return [
            {"role":"system", "content":system_prompt},
            {"role":"user", "content":user_prompt + website} 
        ] 
```

```python
def summarize(url):
    site = fetch_website_contents(url)
    response = openai.chat.completions.create(
        model = "gpt-4.1-mini", 
        messages = messages(site)
    )

    return response.choices[0].message.content
```

```python
display(Markdown(summarize("https://www.teambhp.com")))
```
