```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown, display
```

```python
MODEL_TO_USE = 'gpt-4o-mini'
```

```python
api_key=os.getenv('OPENAI_API_KEY')
client=OpenAI(
    api_key=api_key,
    base_url="https://openrouter.ai/api/v1"
)
```

```python
system_prompt = """
You are a highly technical assistant.
Respond using precise programming terminology.
Avoid casual language.
Be specific and concise.
"""

# user_prompt = """
# Explain line by line the syntax of a Python program that prints a star pattern using the '*' character.
# """
```

```python
def message_for(programming_query):
    return [{"role":"system","content":system_prompt},
            {"role":"user","content":programming_query}]
```

```python
def genarate_response(programming_query):
    response=client.chat.completions.create(model=MODEL_TO_USE,messages=message_for(programming_query))
    return response

programming_query="""
Explain line by line the syntax of a Python program that prints an arrow shaped made in stars pattern using the '*' character.
"""

response = genarate_response(programming_query)
print(response.choices[0].message.content)
```

```python
# Test the output

def print_arrow(height):
    # Loop to create the upper part of the arrow
    for i in range(height):
        # Print spaces to center the arrow
        print(' ' * (height - i - 1), end='')
        # Print the '*' characters
        print('*' * (2 * i + 1))
    
    # Loop to create the lower part of the arrow
    for i in range(height - 1):
        # Print spaces to center the arrow
        print(' ' * (i + 1), end='')
        # Print the '*' characters
        print('*' * (2 * (height - i - 1) - 1))

# Call the function with a height parameter
print_arrow(5)
```
