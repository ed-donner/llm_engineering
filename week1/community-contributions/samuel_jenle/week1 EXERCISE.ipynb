To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display, update_display
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2:1b'
OLLAMA_BASE_URL = "http://localhost:11434/v1"
```

```python
# set up environment
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

if api_key and api_key.startswith('sk-proj-') and len(api_key)>10:
    print("API key looks good so far")
else:
    print("There might be a problem with your API key. Please check it and try again.")
```

```python
# here is the system prompt
system_prompt = "You are a helpful assistant that takes a technical question and responds with an explanation."
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what is Quantization in machine learning and how it can be used to optimize model performance.
"""
```

```python
# here are the messages
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": question}
]
```

```python
def display_streaming_response(stream):
    response = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)
```

```python
# Get gpt-4o-mini to answer, with streaming
gpt = OpenAI()
stream = gpt.chat.completions.create(messages=messages, model=MODEL_GPT, stream=True)
display_streaming_response(stream)
```

```python
# Get Llama 3.2 to answer
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")
stream = ollama.chat.completions.create(messages=messages, model=MODEL_LLAMA, stream=True)
display_streaming_response(stream)
```
