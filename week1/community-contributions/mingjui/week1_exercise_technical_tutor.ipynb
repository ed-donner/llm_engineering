```python
# imports
import os
import json
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

if api_key and api_key.startswith('sk-proj-') and len(api_key)>10:
    print("API key looks good so far")
else:
    print("There might be a problem with your API key? Please visit the troubleshooting notebook!")

openai = OpenAI()
```

```python
# here is the question; type over this to ask something new
question = "Please explain what are token, tokenizer, and tokenization"
```

```python
# Prompt
system_prompt = """
You are an AI expert specialized in Large Language Models (LLMs).
Explain technical concepts clearly and structured. 
Use markdown format with proper headings, bold text, and bullet points to ensure readability.
"""
```

```python
# Get gpt-4o-mini to answer, with streaming
def stream_answer(model, tech_question):
    print(f"🚀 Asking {model}...\n")
    
    # User Prompt
    user_prompt = f"Here is my question:\n{tech_question}"
    
    stream = openai.chat.completions.create(
        model=MODEL_GPT,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
        stream=True
    )    
    
    # Stream
    response = ""
    display_handle = display(Markdown(""), display_id=True)
    
    for chunk in stream:
        # Make sure chunk has content
        token = chunk.choices[0].delta.content or ''
        response += token
        # Real-time up Jupyter Notebook
        update_display(Markdown(response), display_id=display_handle.display_id)
        
    return response
```

```python
# Get gpt-4o-mini to answer, with streaming
final_explanation = stream_answer(MODEL_GPT, question)
```

```python
# Get Llama 3.2 to answer
final_explanation = stream_answer(MODEL_LLAMA, question)
```
