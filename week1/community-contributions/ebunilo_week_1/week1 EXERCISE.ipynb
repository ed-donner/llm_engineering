To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports

import os
from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment
load_dotenv(override=True)
api_key = os.getenv("OPENAI_API_KEY")

# OpenAI API client (uses OPENAI_API_KEY from env)
client = OpenAI()

# Ollama via OpenAI-compatible API (for Llama 3.2)
client_ollama = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
# Prompt

messages = [
    {"role": "system", "content": "You are a helpful technical tutor who answers questions about python code, software engineering, data science and LLMs."},
    {"role": "user", "content": question}
]
```

```python
# Get gpt-4o-mini to answer, with streaming

stream = client.chat.completions.create(model=MODEL_GPT, messages=messages, stream=True)

response = ""
for chunk in stream:
    if chunk.choices[0].delta.content:
        response += chunk.choices[0].delta.content
display(Markdown(response))
```

```python
# Get Llama 3.2 to answer
stream = client_ollama.chat.completions.create(model=MODEL_LLAMA, messages=messages, stream=True)
response = ""
for chunk in stream:
    if chunk.choices[0].delta.content:
        response += chunk.choices[0].delta.content
display(Markdown(response))
```
