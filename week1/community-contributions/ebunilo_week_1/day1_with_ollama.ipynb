```python
# imports

import os
from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI
```

# Connect to Ollama


```python
load_dotenv()
OLLAMA_BASE_URL = os.getenv("OLLAMA_BASE_URL", "http://localhost:11434/v1").rstrip("/")
OLLAMA_MODEL = os.getenv("OLLAMA_MODEL", "llama3.2")  # Run `ollama list` and set this to a model you have
openai = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")
```

```python
# To give you a preview -- calling OpenAI with these messages is this easy. Any problems, head over to the Troubleshooting notebook.

message = "Hello, GPT! This is my first ever message to you! Hi!"

messages = [{"role": "user", "content": message}]

messages
```

```python
response = openai.chat.completions.create(model=OLLAMA_MODEL, messages=messages)
response.choices[0].message.content

# This assumes that ollama is running on localhost:11434/v1
# And that a model is installed. Run `ollama list` to see what models are available.
# To install a model, run `ollama pull <model_name>` (example: `ollama pull llama3.2`)
```
