```python
# imports
import os
from dotenv import load_dotenv
from typing import Optional
from openai import OpenAI
from IPython.display import Markdown, display, update_display
```

```python
# constants
MODEL_GPT = 'gpt-5'
MODEL_LLAMA = 'llama3.2'
LLAMA_BASE_URL = "http://localhost:11434/v1"

system_prompt = """
    You are a helpful teaching assistant in programming. 
    You will be given a question about a code snippet.
    Your task is to explain what the code does and why it works that way.
"""
```

```python
# set up environment
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')

if openai_api_key and openai_api_key.startswith('sk-proj-') and len(openai_api_key) > 10:
    print("OpenAI API key is set correctly.")
else:
    print("OPENAI_API_KEY might not be set correctly. Please check your .env file.")
```

```python
# here is the question; type over this to ask something new
def format_question(content: str, code: Optional[str] = None) -> str:
    return f"""
        Question content: {content}
        Code: {code}
    """
```

```python
# Get gpt-4o-mini to answer, with streaming
def get_gpt_answer(question: str, code: Optional[str] = None):
    openai_client = OpenAI()
    stream = openai_client.chat.completions.create(
        model=MODEL_GPT,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": format_question(question, code)}
        ],
        stream=True
    )

    content = ""
    display_content = display(Markdown(content), display_id=True)
    
    for chunk in stream:
        content += chunk.choices[0].delta.content or ""
        update_display(Markdown(content), display_id=display_content.display_id)
            
```

```python
get_gpt_answer(
    question="How to learn programming in AI era?"
)
```

```python
# Get Llama 3.2 to answer
!ollama pull llama3.2

ollama_client = OpenAI(base_url=LLAMA_BASE_URL, api_key='ollama')

def get_llama_answer(question: str, code: Optional[str] = None):
    stream = ollama_client.chat.completions.create(
        model=MODEL_LLAMA,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": format_question(question, code)}
        ],
        stream=True
    )

    content=""
    display_content = display(Markdown(content), display_id=True)
    for chunk in stream:
        content += chunk.choices[0].delta.content or ""
        update_display(Markdown(content), display_id=display_content.display_id)
```

```python
get_llama_answer(
    question="How to learn programming in AI era?"  
)
```
