```python
# imports
# If these fail, please check you're running from an 'activated' environment with (llms) in the command prompt

import os
import json
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
```

```python
# Initialize and constants

load_dotenv(override=True)
```

```python
api_key = os.getenv('OPENAI_API_KEY')

if api_key and api_key.startswith('sk-proj-') and len(api_key)>10:
    print("API key looks good so far")
else:
    print("There might be a problem with your API key? Please visit the troubleshooting notebook!")
    
MODEL = 'gpt-4o-mini'
openai = OpenAI()
```

```python
# prompts
system_prompt = """
        You are an expert Computer Science Technical Tutor specialized in Software Development and Aritifical Intelligence.
        Your goal is to simplify complex concepts while maintaining technical accuracy at a beginner level with practical examples and real-world applications.
"""

user_prompt = """I'm confused about the difference between a regular computer program and Artificial Intelligence. 
Doesn't a programmer just tell the computer exactly what to do in both cases? 
I heard about something called a 'Neural Network'—is that like a digital brain? 
I'd love a simple explanation and a link to a video or article that won't go over my head.
"""
```

```python
def tutor_chat():
    stream = openai.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
          ],
        stream=True
    )    
    response = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)
```

```python
tutor_chat()
```
