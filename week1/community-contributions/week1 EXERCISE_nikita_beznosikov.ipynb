```python
# imports

import os
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
OLLAMA_BASE_URL = "http://localhost:11434/v1"
```

```python
# set up environment

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

if api_key and api_key.startswith('sk-proj-') and len(api_key)>10:
    print("API key looks good so far")
else:
    print("There might be a problem with your API key? Please visit the troubleshooting notebook!")
    
```

```python
openai = OpenAI()
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
system_prompt = """
You are personal tutor. Your task is to help understand asked question and provide a detailed answer. Try to explain it as if you were explaining to a 10 year old, using simple language and examples. Always respond in markdown format.
"""
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
# function to answer the question, streaming the response as it comes in
def answer_the_question(client, model, q):
    stream = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": q}
          ],
        stream=True
    )    
    response = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)
```

```python
# answer the question using GPT-4o-mini
answer_the_question(client=openai,  model=MODEL_GPT, q=question)
```

```python
# answer the question using LLaMA 3.2
answer_the_question(client=ollama, model=MODEL_LLAMA, q=question)
```
