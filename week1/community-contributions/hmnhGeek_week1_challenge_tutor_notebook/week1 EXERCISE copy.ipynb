To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

### Importing required packages

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import display, Markdown

load_dotenv(override=True)
```

### Loading environment variables

```python
MODEL_GPT = os.getenv("AI_MODEL")
MODEL_OLLAMA = os.getenv("GEMMA3_1B")
```

```python
api_key = os.getenv("OPENAI_API_KEY")
base_url_gpt = os.getenv("BASE_URL")
base_url_ollama = os.getenv("OLLAMA_LOCAL_URL")
```

### Define GPT and OLLAMA clients

```python
gpt = OpenAI(base_url=base_url_gpt, api_key=api_key)
ollama = OpenAI(base_url=base_url_ollama, api_key="gemma")
```

### System and user prompts

```python
system_prompt = """
    You are an expert tutor and a well qualified teacher. You specialize in technical fields like
    physics, chemistry and mathematics along with computer science fundamentals. You take great pride
    in answering technical doubts. You explain doubts in a professional, simple and to-the-point manner.
    
    Whenever you reply, you ensure that it is in markdown format outside code blocks. However, for questions
    related to computer science or programming you're allowed to type code in back-ticks (`). You provide 
    step by step solution for every problem. Equations, derivations, pseudo-codes, etc are child's play for
    you.

    IMPORTANT MARKDOWN FORMATTING RULES:

    - Never use \\( ... \\) for inline math.
    - Never use \\[ ... \\] for display math.
    - Use $ ... $ for inline math.
    - Use $$ ... $$ for display math.
    - The output will be rendered by Jupyter/IPython Markdown.
    - Any occurrence of \\(, \\), \\[, or \\] is invalid.
    - Before returning the response, convert all LaTeX delimiters to dollar-sign delimiters.

    Examples:

        Correct:
        Einstein's famous relation $E = mc^2$

        Correct:
        $$
        \\gamma = \\frac\{1\}{\\sqrt{1-\\frac{v^2}{c^2}}}
        $$

        Incorrect:
        \\(E = mc^2\\)

        Incorrect:
        \\[
        \\gamma = \\frac{1}{\\sqrt{1-\\frac{v^2}{c^2}}}
        \\]
"""
```

```python
def get_user_prompt(subject, doubt):
    user_prompt = f"""
        I have a doubt related to {subject} and I want you to answer and resolve my query. Please explain
        in step by step manner and assume that I am total beginner in the context of this question/doubt.\n\n
    """
    user_prompt += doubt
    return user_prompt
```

### Using clients to answer doubts

```python
def get_answer_from_gpt(subject, doubt):
    response = gpt.chat.completions.create(
        model=MODEL_GPT,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": get_user_prompt(subject, doubt)}
        ]
    )
    return response.choices[0].message.content


def get_answer_from_ollama(subject, doubt):
    response = ollama.chat.completions.create(
        model=MODEL_OLLAMA,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": get_user_prompt(subject, doubt)}
        ]
    )
    return response.choices[0].message.content
```

### Displaying markdown results

```python
def show_answer_of_gpt(subject, doubt):
    message = get_answer_from_gpt(subject, doubt)
    display(Markdown(message))


def show_answer_of_ollama(subject, doubt):
    message = get_answer_from_ollama(subject, doubt)
    display(Markdown(message))
```

### Examples

```python
show_answer_of_gpt("Physics", "Can you derive Einstein's equation E=mc^2?")
```

```python
show_answer_of_ollama("Computer Science", "Can you write and then explain the code of Fibonacci series in Python?")
```

```python
show_answer_of_gpt("Math", "Derive pythagoras theorem!")
```

```python
show_answer_of_ollama("English", "What is a verb?")
```

```python
show_answer_of_gpt("Chemistry", "What is benzene?")
```
