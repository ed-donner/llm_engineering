## Upgrade Summrize Website Assignment to Summrize Website with Local Ollama Model 

```python
#set kernal
import os
import sys
from dotenv import load_dotenv

# append path to get scrapper
sys.path.append(".../week1")
# Now import your scraper
import scraper
from scraper import fetch_website_contents
from IPython.display import Markdown, display
# use OpenAI but use Ollama Model
from ollama import OpenAI

```

### Initilize OpenAI with Local Ollama Model gemma3:latest

```python


OLLAMA_BASE_URL = "http://localhost:11434"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama");
```

### Create Prompts

```python

system_prompt = """
You are a snarky assistant that analyzes the contents of a website,
and provides a short, snarky, humorous summary, ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""

user_prompt_prefix = """
Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.

"""
```

### Get Contents of websites using website scraper method

```python

ed = fetch_website_contents("https://edwarddonner.com")
print(ed)
```

### Now Create a Useful Message Method to use it for gemma3:latest Ollama

```python
def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + website}
    ]
```

```python
messages_for(ed)
```

### Create Summrize method which will bring all together to summerize the website

```python
def summarize(url):
    website = fetch_website_contents(url)
    response = ollama.chat.completions.create(
        model = "gemma3:latest",
        messages = messages_for(website)
    )
    return response.choices[0].message.content
```

### Finally summerize the website with nice display

```python
def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))
```

```python
display_summary("https://edwarddonner.com")
```
