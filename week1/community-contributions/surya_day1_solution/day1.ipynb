# LLM Powered Interview Prep Question Generator


### Built on Week 1, Day 1 — using GPT
**Use case:** Enter any topic (e.g. “binary search trees”, “system design basics”, “LLM engineering”) and get a small set of interview-style questions with concise model answers — ideal for students preparing for tech interviews and coding rounds.

This extends the Day 1 course recommender by:

- Turning the model into an interview coach instead of a course finder

- Generating 5 focused interview questions for a given topic

- Providing short, high-level answers that can be used for quick revision

- Returning the output as clean Markdown, easy to read and save as notes



```python
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')


if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
    

openai = OpenAI()
```

```python
system_prompt = """
    You are an interview coach for software engineering and ML roles.
    When the user gives you a topic (for example: 'binary search trees', 'system design basics', 'LLM engineering', or 'SQL joins'),
    you must generate exactly 5 interview-style questions about that topic.
    For each question, also provide a short, high-level model answer that would be acceptable in a real interview.

    Format your response in Markdown as:

    1. **Question 1**
    Answer: ...

    2. **Question 2**
    Answer: ...

    Keep the answers concise (2–4 sentences).
"""
```

```python
user_prompt = """
    Here are the questions of the interview topics mentioned below.
    Focus on clarity and correctness rather than long theory.
"""
```

```python
def messages_for(topic):
    return [
        {"role": "system", "content": system_prompt.strip()},
        {"role": "user", "content": user_prompt.strip() + topic}
    ]
```

```python
def interview_questions(topic):
    response = openai.chat.completions.create(
        model = "gpt-4.1-mini",
        messages = messages_for(topic)
    )

    return response.choices[0].message.content
```

```python
interview_questions("LLM")
```

```python
def display_summary(topic):
    result = interview_questions(topic)
    display(Markdown(result))
```

```python
display_summary('React-Redux')
```
