```python
# imports
import os
from dotenv import load_dotenv
from openai import OpenAI

# constants

MODEL_GPT = 'gpt-4o-mini'

# set up environment
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

openai = OpenAI(api_key=api_key)

# here is the question; type over this to ask something new

question = """
Please explain what this code does and give alternative ways to implement it and choose the best way to implement it:
history = [{"role":h["role"], "content":h["content"]} for h in history]
"""

system_prompt = """You are an experience software engineer who explains code in a very detailed way. Itemise your responses and give a conclusive resopnse when done.
"""

# Get gpt-4o-mini to answer, with streaming]

message = [{"role":"system","content": system_prompt},{"role":"user","content": question}] 

stream = openai.chat.completions.create(model=MODEL_GPT,messages=message,stream=True)

for chunk in stream:
    response = chunk.choices[0].delta.content
    if response:
        print(response, end="", flush=True)
```
