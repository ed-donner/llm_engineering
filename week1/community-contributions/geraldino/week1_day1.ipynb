## Day 1 exercise - summarize a webpage




Get's a webpage and summarizes it

```python
# imports dependencies
import os
from dotenv import load_dotenv, find_dotenv
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
#instantiate openai object
client = OpenAI()

#set the system and user prompts 
system_prompt = """
You are a snarky assistant that analyzes the contents of an email,
and provides a short, snarky, humorous summary with a perfect subject line,
ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""

user_prompt = """
Summarize the website with a perfect subject line.
Please let your response be short, snarky and humorous.
"""

def summarize(url):
    website = fetch_website_contents(url)

    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": website + "\n\n" + user_prompt}
    ]
    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=messages
    )
    return response.choices[0].message.content

# retireve and display result
result = summarize("https://geedino07.github.io/cybersecurity-scam-alert/")
display(Markdown(result))
```
