# Market Research Synthesizer
## One API Call to Extract Market Intelligence

Input: Industry keyword  
Output: Trends, competitors, market gaps, growth signals—all synthesized from web articles + LLM reasoning

This is overengineered for: market research, due diligence, product strategy, competitive analysis.

```python
import os
import json
from dotenv import load_dotenv
from market_research_synthesizer import MarketResearchSynthesizer
from IPython.display import Markdown, display, JSON

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

if not api_key:
    print("❌ No API key found. Set OPENAI_API_KEY in .env")
else:
    print("✅ API key loaded")

synthesizer = MarketResearchSynthesizer()
```

## Example 1: AI in Healthcare Market

```python
result1 = synthesizer.synthesize_market_intelligence("AI in healthcare diagnostics 2025")

print(f"\n📊 Market: {result1['keyword']}")
print(f"📰 Articles analyzed: {result1['articles_analyzed']}\n")

intel = result1['intelligence']

if 'error' not in intel:
    print(f"📈 Market Summary:\n{intel['market_summary']}\n")
    
    print("🎯 Key Trends:")
    for trend in intel.get('key_trends', []):
        print(f"  • {trend['trend']} ({trend['signal_strength']})")
    
    print("\n🏢 Emerging Competitors:")
    for comp in intel.get('emerging_competitors', []):
        print(f"  • {comp['name']} - {comp['positioning']} (threat: {comp['threat_level']})")
    
    print("\n💡 Market Gaps (Opportunities):")
    for gap in intel.get('market_gaps', []):
        print(f"  • {gap['gap']} (opportunity: {gap['opportunity_size']})")
    
    print("\n📊 Growth Signals:")
    for signal in intel.get('growth_signals', []):
        print(f"  • {signal['signal']} (confidence: {signal['confidence']})")
    
    print("\n🚀 Recommended Business Angles:")
    for angle in intel.get('recommended_angles', []):
        print(f"  • {angle}")
else:
    print(f"⚠️ Error: {intel['error']}")
    print(f"Raw response: {intel['raw_response'][:500]}...")
```

## Example 2: Fitness Tech Market

```python
result2 = synthesizer.synthesize_market_intelligence("personalized fitness nutrition app market 2025")

intel2 = result2['intelligence']
print(f"\n📊 Market: {result2['keyword']}\n")

if 'error' not in intel2:
    print(f"📈 {intel2.get('market_summary', 'N/A')}\n")
    
    if 'key_trends' in intel2:
        print("🎯 Trends:")
        for t in intel2['key_trends'][:3]:
            print(f"  • {t['trend']}")
    
    if 'market_gaps' in intel2:
        print("\n💡 Opportunities:")
        for gap in intel2['market_gaps'][:3]:
            print(f"  • {gap['gap']}")
    
    if 'recommended_angles' in intel2:
        print("\n🚀 Angles for YOUR product:")
        for angle in intel2['recommended_angles']:
            print(f"  • {angle}")
```

## How This Works

1. **Scrape**: Attempts to scrape articles from news, Medium, TechCrunch, HN
2. **Prompt**: Sends scraped content + keyword to GPT with a system prompt that acts like a senior analyst
3. **Parse**: Extracts JSON with structured market intelligence
4. **Output**: Trends, competitors, gaps, growth signals, business angles

**Cost**: ~$0.01-0.02 per query (using gpt-4.1-mini)

**Use cases**:
- Due diligence before entering a market
- Competitive analysis for product strategy
- Identify white space / unmet needs
- Spot emerging competitors before they blow up
- Guide GTM strategy
- Investor pitch research
