#### Import necessary modules

```python
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown, display
from scraper import fetch_website_contents
import os
```

#### Load the environment variables

```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')
MODEL = os.getenv("AI_MODEL")
BASE_URL = os.getenv("BASE_URL")
```

#### Give the system prompt

```python
system_prompt = """
    You are a keen news analyst who likes to go into depth of any news content shared with you.
    You always try to summarize the news in simple words and try to keep it within 3 paragraphs, 
    ignoring text that might be navigation related.
    
    You are also a great observer; as soon as you see that the text shared with you doesn't look
    like a news, you simply don't summarize and respond back by saying that the text does not 
    seem to be a news article. 

    Other than that, once you summarize, you also tell at the end, the tone of the news, if it is
    positive, negative or just neutral. If there are any characters involed, at the end you must list them 
    as well, each one in a separate line.

    Remember, respond ONLY and ONLY in a valid markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""
```

#### Define a user prompt

```python
user_prompt_prefix = """
    Here are the contents of a possible news article.
    Provide a short summary of this news (if it is a news).
"""
```

#### Create `OpenAI` instance

```python
def get_message_for(news):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + news}
    ]
```

```python
openai = OpenAI(
    base_url=BASE_URL,
    api_key=api_key
)

def summarize_news(news_url):
    news_content = fetch_website_contents(news_url)
    response = openai.chat.completions.create(
        model = MODEL,
        messages = get_message_for(news_content)
    )
    return response.choices[0].message.content
```

```python
def display_summary(url):
    summary = summarize_news(url)
    display(Markdown(summary))
```

#### Running the model

```python
display_summary("https://www.theguardian.com/world/2026/may/22/us-arms-sales-taiwan-pause-iran-war-says-acting-navy-chief")
```

```python
display_summary("https://github.com/ed-donner/llm_engineering")
```
