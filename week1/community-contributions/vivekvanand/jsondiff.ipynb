Summarize differences between jsons

```python
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI
```

Change models to see the differences it make to the summary. I only want the differences in the json to be listed. 
Even though the prompt explicitly state not to summarize similaries and key order differences statesgpt-4o-mini will still lists those in the summary.
gpt-5.4-mini does a much better job.

```python
MODEL_GPT = 'gpt-4o-mini'
# MODEL_GPT = 'gpt-5.4-mini'
```

```python
system_prompt = "You are a json expert who is great at finding differences between two pieces of jsons. You will be given two pieces of json, and you will respond with a summary of the differences. Provide the differences in natural language. List the points that are differences between the two pieces of json. Be concise, but informative. Produce the summary in a clear and structured format. If there are no differences, say 'There are no differences between these two pieces of json. Do not summarize similarities. Do not summarize key order differences."
user_prompt = """
    Here are the two pieces of json:
    Provide a concise but informative summary of the differences between these two pieces of json:
"""
```

```python
json1 = """ {"test": 1, "test2": 2, "test3": 4 } """
json2 = """ {"test2": 2, "test": 3, "test3": 4 } """
```

```python
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt + "\n" + json1 + "\n" + json2}
]
```

```python
openai = OpenAI()
response = openai.chat.completions.create(model=MODEL_GPT, messages=messages)

# Step 4: print the result
summary = response.choices[0].message.content
display(Markdown(summary))
```
