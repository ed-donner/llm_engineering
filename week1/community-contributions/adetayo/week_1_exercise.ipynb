To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

My Modification: I get reponse from multiple models: Open AI and gemma through Olama, then I have the models score each other

```python
MODEL_GPT = 'gpt-4o-mini'
MODEL_GEMMA = 'gemma3:1b'
from openai import OpenAI
```

First, we setup the system and user prompts for the question

```python
system_prompt = """
You are a technical assistant, your role is to answer technical questions by providing a detailed explanation.
"""
user_prompt = """
What is idempotency and how is it used in API design?
"""

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
```

Now lets make actual request with OpenAI

```python
gpt4o = OpenAI()
response_from_gpt4o = gpt4o.chat.completions.create(
    model=MODEL_GPT,
    messages = messages
)

print(response_from_gpt4o.choices[0].message.content)
```

lets make another request with Ollama

```python
gema = OpenAI(base_url="http://localhost:11434/v1")
response_from_gemma = gema.chat.completions.create(
    model=MODEL_GEMMA,
    messages = messages
)
print(response_from_gpt4o.choices[0].message.content)
```

Now lets race these response against each other, we will have each model rate the response of the other

```python
response_scoring_system_prompt = """
You are a scoring system that will rate the response of the other model to a technical question. You will be presented with question and the model's response and you will reply with a score over 100.
"""
response_scoring_user_prompt = f"""
Here is the question: {user_prompt}
Here is the response from the other model: 
"""
```

```python
gpt_scorer = OpenAI()
gpt_scoring_response = gpt_scorer.chat.completions.create(
    model=MODEL_GPT,
    messages = [
        {"role": "system", "content": response_scoring_system_prompt},
        {"role": "user", "content": response_scoring_user_prompt+response_from_gemma.choices[0].message.content}
    ]
)

print(gpt_scoring_response.choices[0].message.content)
```

```python
gemma_scorer = OpenAI(base_url="http://localhost:11434/v1")
gema_scoring_response = gemma_scorer.chat.completions.create(
    model=MODEL_GEMMA,
    messages = [
        {"role": "system", "content": response_scoring_system_prompt},
        {"role": "user", "content": response_scoring_user_prompt+response_from_gpt4o.choices[0].message.content}
    ]
)

print(gema_scoring_response.choices[0].message.content)
```
