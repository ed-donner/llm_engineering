```python
# imports
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI

load_dotenv()
```

```python
MODEL = "gpt-5-nano"
API_KEY = os.getenv("OPENAI_API_KEY")

client = OpenAI(api_key=API_KEY)
```

```python
system_prompt = """
You are a technical tutor.

Respond with clear, concise, practical answers.
Use short examples when helpful.
Avoid unnecessary explanations.
Assume the user has technical experience.
"""

user_prompt = """
Explain event-driven architecture in simple terms and give one real-world example.
"""
```

```python
# Create prompt function
def create_prompt(system_prompt, user_prompt):
    return [
        {"role": "system", "content": system_prompt.strip()},
        {"role": "user", "content": user_prompt.strip()},
    ]
```

```python
messages = create_prompt(system_prompt, user_prompt)
```

```python
# Create a function to make the API call
def make_api_call(messages):
    response = client.chat.completions.create(
        model=MODEL,
        messages=messages,
    )
    return response.choices[0].message.content
```

```python
result = make_api_call(messages)
```

```python
result = make_api_call(messages)
```

```python
def display_result(result):
    display(Markdown(result))
```

```python
display_result(result)
```
