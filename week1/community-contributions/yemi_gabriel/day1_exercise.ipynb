### Day 1 Exercise: Email Subject Line (for marketing campaigns)
**Takes a marketing email campaign and generates a subject line.**

`Write something that will take the contents of an email, and will suggest an appropriate short subject line for the email. That's the kind of feature that might be built into a commercial email tool.`

```python
# imports

import os
from dotenv import load_dotenv
from openai import OpenAI
```

```python
# Load environment variables in a file called .env

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python
openai = OpenAI()
```

Type in your email or use the default email body

```python
# Step 1: Create your prompts

system_prompt = """
You are a smart assistant in sales and marketing department of an e-commerce startup. 
You analyse and craft an appropriate short subject line for marketing email campaigns. 
"""

default_body = """
Hi Leo,
I noticed you haven’t had a chance to start your first deep work session with FocusFlow yet.
Often, the hardest part of a big project is just finding that first 20 minutes of momentum. To make it a bit easier to commit to your goals this month, I wanted to give you a nudge.
I’ve created a special invite for you to join FocusFlow Premium. You’ll get unlimited distraction blocking, our full ambient sound library, and deep-dive productivity analytics.
Use the code FLOW30 at checkout to get 30% off your first year: focusflow.app
This code is active for the next 48 hours.
Even if you aren't ready to upgrade, I'd still love for you to try the "2-Minute Entry" feature today. It’s the fastest way to hit a flow state without the struggle.
Launch your dashboard here: focusflow.app
Let's get some work done,
Sarah
Founder, FocusFlow
"""
user_input = input('Enter marketing email body (or press Enter for default):').strip() or default_body
prompt_prefix = """
Come up with a concise, impactful, attention-grabbing, yet personal, 
one-liner subject for this marketing email body:
"""

user_prompt = f"{prompt_prefix} {user_input}"

# Step 2: Make the messages list

messages = [
    {'role': 'system', 'content': system_prompt},
    {'role': 'user', 'content': user_prompt}
]

# Step 3: Call OpenAI
response = openai.chat.completions.create(model='gpt-5-nano', messages=messages)

# Step 4: print the result
print(response.choices[0].message.content)
```
