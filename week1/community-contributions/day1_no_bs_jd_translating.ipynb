```python
# Step 1: Create your prompts

system_prompt = """ You are a no-BS job description interpreter.

Your job is to read a job description and translate it into what the role would ACTUALLY be like in real life.

Rules:
- Be concise, practical, and slightly snarky (but not rude).
- Translate corporate jargon into plain English.
- Focus on day-to-day reality, not aspirational fluff.
- Ignore or downplay company values, culture statements, DEI slogans, brand marketing, and “we are a family” language unless it clearly affects daily work.
- Do not hype the role. Be honest and grounded.
- Do not invent responsibilities that are not implied by the description.

Your output must include the following sections:

1) **What this job really is**
   - 2 to 4 sentences explaining the core purpose of the role in plain English.
   - Answer: “What will I actually spend most of my time doing?”

2) **What you will do on a typical week**
   - Bullet list of concrete activities (meetings, analysis, building decks, managing stakeholders, executing campaigns, etc.)
   - Translate vague phrases like “drive alignment” or “partner cross-functionally” into actual actions.

3) **Corporate jargon → Plain English**
   - A short list translating the most common buzzwords from the posting into what they really mean.

4) **Skills that actually matter**
   - Bullet list of practical skills needed to succeed (tools, thinking skills, soft skills).
   - Separate “must-have” vs “nice-to-have” if implied.

5) **Keywords for ATS / resumes**
   - A compact list of keywords and phrases someone should mirror on their resume.
   - Prioritize hard skills, role-specific terms, and measurable outcomes.

Tone guidelines:
- Sound like a smart friend who is read 100 job descriptions and is tired of the nonsense.
- Dry humor is allowed; sarcasm should be subtle.
- Clarity > cleverness.


Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""
user_prompt_prefix = """
    Here is the job description. Tell me what this job is really like.
    Cut the fluff, explain the day-to-day, and list real skills + ATS keywords.
    Job description:
"""

# Step 2: Make the messages list

# PASTE THE JOB DESCRIPTION HERE
job_description = """
Job Description:

The Integrated Marketing Planning team partners with our business units, channels and marketing specialists to design world-class and award-winning marketing strategies and campaigns that drive shareholder value and client loyalty. Our team acts as leaders in all aspects of marketing, leading strategies while ensuring TD brand is protected and delivered in the client experience.

Job Description 

The Integrated Marketing Planning Manager will lead the development of marketing strategies and programs for TD Wealth Advice (Financial Planning and Private Wealth Management lines of businesses). Collaborating with business, marketing and agency partners in an agile environment, the Integrated Marketing Planning Manager is responsible for designing and delivering innovative, best-in-class marketing strategies across the marketing funnel in paid and owned channels.

Key Accountabilities:

Marketing Strategies and Plans: Understands market and growth drivers, contributes to a vision of market, channel, and segment potential, develops value propositions and proof points that further our competitive advantage, and translates business strategies into marketing plans. Delivers integrated marketing programs that meet business and marketing strategies and understand the role of different media options and their impact on achieving marketing objectives.

Consumer Insights and Trends: Leverage market research and apply customer insights to marketing campaigns/programs. Interpret the industry environment, recommend a course of action and best practices to improve products, processes or services. Understand the nuances of marketing to mass affluent and high net worth segments.

Influencing and relationship management: Build and manage strong relationships with business units, marketing partners and other internal departments to deliver effective marketing programs.

Deliver and Monitor Results:   Ensure stated objectives are delivered on plan (monitor/ early indicators) and where required, take corrective actions to achieve targets.  Analyze past results and make recommendations to maximize ROI. 

Build and Protect the Brand; Ensure integrated marketing campaigns are designed for a positive client experience and support the brand.  Understand client journeys and support the business to deliver the brand through client experience.

Builds Innovative Solutions: Actively contributes to driving the spirit of innovation among the team. Brings external thinking/ideas to problems.

Budget and Project Management: Manage projects in the most effective way (on time, on target, on budget) from initiation to completion. Coordinate multi-functional teams and agency partners through the project.  Track and manage budget against plan, meeting all financial reporting requirements.

Analysis and Problem Solving: Help to define the problem to solve, engages stakeholders from different groups to gather data and draw insights from it. Leverage data combined with judgment and intuition to make recommendations and drive action.  Deliver and monitor results; analyze results and makes recommendations to maximize return on investment while evaluating effectiveness of in market activity.

Job Requirements

Undergraduate degree or relevant professional certifications, designations, or equivalent required

7+ years relevant experience

Experience leading marketing strategy in similar industry.

Digital marketing experience strategy and execution is ideal.

Advanced knowledge of multiple marketing principles and disciplines, strategy development, consulting principles, lines of businesses supported, and product/customer markets and proven ability to drive business results.

Innovative, customer-obsessed mindset with strong financial acumen to advocate for customer advice and experience.

A self-motivator with excellent organizational and communications skills to influence, prioritize and multi-task in a collaborative or independent setting.

Experienced in relationship management, work collaboratively and build relationships across teams and functions including agency and internal partners.

A self-motivator with excellent organizational and communications skills to prioritize and multi-task in a collaborative or independent setting.

"""
messages = [
   {'role': 'system', 'content': system_prompt},
   {'role': 'user', 'content': user_prompt_prefix + job_description}
] 

# Step 3: Call OpenAI
def summarize():
    response = openai.chat.completions.create(
        model = "gpt-4.1-mini",
        messages = messages
    )
    return response.choices[0].message.content
    
#response = openai.chat.completions.create(model="gpt-5-nano", messages=messages")
#response.choices[0].message.content

# Step 4: print the result
def display_summary():
    summary = summarize()
    display(Markdown(summary))

display_summary()
```
