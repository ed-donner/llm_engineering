```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python
openai = OpenAI()

# Step 1: Create your prompts
system_prompt = (
    "You are a Senior Wall Street Institutional Analyst. Analyze the provided Zacks report "
    "on the SpaceX IPO. You must structure your output strictly according to these Markdown rules:\n"
    "- Use '#' for the document main title.\n"
    "- Use '## 1-1.' for the core topic title.\n"
    "- Use '###' for subsections like metrics, TAM breakdown, and strategic AI partnerships.\n"
    "- Use '#### Note' for risk assessments or market outlook.\n"
    "- You MUST provide financial metrics in a standard markdown table, and additionally provide "
    "a secondary table backup using raw pipe-separated syntax (|)."
)

# Paste the core content scraped from the Zacks SpaceX IPO guide
user_prompt = """
Here are the contents of the Zacks SpaceX IPO report:
- Ticker & Listing: Debuting on Nasdaq under the ticker 'SPCX' at a $1.75 Trillion valuation, raising $75 Billion.
- Total Addressable Market (TAM): Identified a $28.5 Trillion quantifiable TAM ($370B Space, $1.6T Connectivity via Starlink Broadband/Mobile, $26.5T AI Infrastructure).
- Strategic AI Move: Disclosed an AI compute agreement with Anthropic utilizing ~325,000 Nvidia chips, generating $1.25 Billion in monthly recurring revenue through May 2029.
- Retail Allocation: Reserved up to 5% of Class A shares for employees and insider affiliates without lock-up restrictions.
- Industry Impact: Serves as the first large-cap benchmark for the commercial space economy, re-pricing pure-plays like Intuitive Machines (LUNR).
"""

# Step 2: Make the messages list
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]

# Step 3: Call OpenAI
# Using gpt-4o-mini as a robust and cost-effective model for text processing
response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)

# Step 4: print the result
print(response.choices[0].message.content)
```
