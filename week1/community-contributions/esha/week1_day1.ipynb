Summarize a webpage to use an Open Source model running locally via Ollama

```python
# imports

import os
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
openai = OpenAI(base_url='http://localhost:11434/v1/', api_key='ollama')
```

```python
# Define system prompt

system_prompt = """
You are a helpful assistant that analyzes the contents of a website,
and provides a short, snarky, humorous summary, ignoring text that might be navigation related.
Respond in markdown in German. Do not wrap the markdown in a code block - respond just with the markdown.
"""
```

```python
# Define user prompt

user_prompt_prefix = """
Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.

"""
```

```python
def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + website}
    ]
```

```python
def summarize(url):
    website = fetch_website_contents(url)
    response = openai.chat.completions.create(
        model = "deepseek-r1:1.5b",
        messages = messages_for(website)
    )
    return response.choices[0].message.content
```

```python
def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))
```

```python
display_summary("https://www.bbc.com/news/articles/c5y9wen5z8ro")
```

```python
my_own_system_prompt = """
You are an AI agent who is always happy and excited to help summarize websites in creative ways. Make the replies cheerful, whimsical, and engaging, while keeping them precise
and helpful! Respond in markdown in English with some emojis. Do not wrap the markdown in a code block - respond just with the markdown
"""
```

```python
def my_messages_for(website):
    return [
        {"role": "system", "content": my_own_system_prompt},
        {"role": "user", "content": user_prompt_prefix + website}
    ]
```

```python
def my_summarize(url):
    website = fetch_website_contents(url)
    response = openai.chat.completions.create(
        model = "gpt-oss:120b-cloud",
        messages = my_messages_for(website)
    )
    return response.choices[0].message.content
```

```python
def my_display_summary(url):
    summary = my_summarize(url)
    display(Markdown(summary))
```

```python
my_display_summary("https://www.bbc.com/news/articles/c5y9wen5z8ro")
```
