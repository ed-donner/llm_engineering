### Tokenizing with code

```python
import tiktoken

encoding = tiktoken.encoding_for_model("gpt-4.1-mini")

tokens = encoding.encode("Hi my name is Esha and I like biscoff chessecake")
```

```python
tokens
```

```python
for token_id in tokens:
    token_text = encoding.decode([token_id])
    print(f"{token_id} = {token_text}")
```

### The illusion of "memory"

```python
from openai import OpenAI
```

```python
openai = OpenAI(base_url='http://localhost:11434/v1/', api_key='ollama')
```

```python
messages = [
    {"role": "system", "content": "You are a helpful assistant"},
    {"role": "user", "content": "Hi! I am Esha :)"}
]

response = openai.chat.completions.create(model="nemotron-3-super:cloud", messages=messages)
response.choices[0].message.content
```

```python
messages = [
    {"role": "system", "context": "You are a helpful assistant"},
    {"role": "user", "context": "What is my name?"}
]

response = openai.chat.completions.create(model="nemotron-3-super:cloud", messages=messages)
response.choices[0].message.content
```

Every call to an LLM is completely STATELESS. It's a totally new call, every single time. As AI engineers, it's OUR JOB to devise techniques to give the impression that the LLM has a "memory".

```python
messages = [
    {"role": "system", "content": "You are a helpful assistant"},
    {"role": "user", "content": "Hi! I'm Esha!"},
    {"role": "assistant", "content": "Hi Esha! How can I assist you today?"},
    {"role": "user", "content": "What's my name?"}
    ]

response = openai.chat.completions.create(model="nemotron-3-super:cloud", messages=messages)
response.choices[0].message.content
```
