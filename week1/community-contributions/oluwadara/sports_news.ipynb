# SkySports News Scraper

This notebook does the following:
- Scrapes the SkySports news website
- Extracts the text
- Sends the information to an LLM (in this case, Open AI) to summarize and categorize the latest news from the sports world

## Scrape Text from SkySports

I will start with all necessary imports

```python
import os
from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI
```

Now, you must ensure you have the `OPENAI_API_KEY` set in your `.env` file

```python
content = fetch_website_contents("https://www.skysports.com/")

display(Markdown(content))
```

## Setup Prompts

```python
system_prompt = """
You are a top flight sports resporter that analyzes the contents of a sports website,
and provides a short, professional summary of all that is happening in the world of sports.
Respond in markdown. Do not wrap the markdown in a code block, respond just with the markdown.
"""

user_prompt = f"""
    Here is the content that I got from this website:
    {content}
"""
print(user_prompt)
```

## Load needed variable and setup LLM

```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

if not api_key:
    print("No API key found!")
else:
    print("API key found!")

# Initialize the OpenAI client
client = OpenAI(api_key=api_key)
```

## Run Summary

```python
model = "gpt-4o-mini"
response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt},
        ],
        temperature=0.3,
    )

print(response.choices[0].message.content)
```
