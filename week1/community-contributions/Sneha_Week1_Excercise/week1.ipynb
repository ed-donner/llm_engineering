```python
# imports
import os
from openai import OpenAI
from dotenv import load_dotenv
```

```python
# constants
#  'llama3.2' is corrupted in my local ollama and not able to tested in my local machine, so I am using 'gemma3:270m' and gemini-2.5-flash-lite instead.
# MODEL_GPT = 'gpt-4o-mini' --- IGNORE ---
# MODEL_LLAMA = 'llama3.2'


MODEL_GEMMA = 'gemma3:270m'
GEMINI_MODEL = 'gemini-2.5-flash-lite'
```

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1/"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python

load_dotenv(override=True)

google_api_key = os.getenv('GOOGLE_API_KEY')

if not google_api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not google_api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python
question = """
Please explain what is reason monsoon rain not started in India yet?
Tell me in simple words.
How we can improve the situation and what are the reasons for delay in monsoon rain in India?
"""
```

```python
# Get gemini to answer, with streaming

GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"
ollama = OpenAI(base_url=GEMINI_BASE_URL, api_key=google_api_key)

stream = ollama.chat.completions.create(
    model=GEMINI_MODEL,
    messages=[{"role": "user", "content": question},
                {"role": "system", "content": "Do you want me to explain that in more detail?"},
                {"role": "user", "content": "Yes, please explain in more detail."},
              ],
    stream=True
)
for event in stream:
    print(event.choices[0].delta.content, end="", flush=True)

stream = ollama.chat.completions.create(
    model=GEMINI_MODEL,
    messages=[
                {"role": "assistant", "content": "Sure. I can explain the reasons for the delay in monsoon rains in India in more detail."},
                {"role": "user", "content": "Can you give me some examples?"}
              ],
    stream=True
)
for event in stream:
    print(event.choices[0].delta.content, end="", flush=True)
```

```python
# Get Gemma 3.2 to answer without streaming
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
response = ollama.chat.completions.create(model=MODEL_GEMMA, messages=[{"role": "user", "content": question}    ])

print(response.choices[0].message.content)

response = ollama.chat.completions.create(model=MODEL_GEMMA, messages=[
                                                                       {"role": "assistant", "content": "Do you want me to explain that in more detail?"},
                                                                       {"role": "user", "content": "Yes, please explain in more detail."},
                                                                       {"role": "assistant", "content": "Sure. I can explain the reasons for the delay in monsoon rains in India in more detail."},
                                                                       {"role": "user", "content": "Can you give me some examples?"}
                                                                       ])
print(response.choices[0].message.content)
```
