```python

from pypdf import PdfReader

def extract_text_pypdf(pdf_path):
    """
    Extracts text from a digitally generated PDF file using pypdf.
    """
    text = ""
    try:
        # Create a pdf reader object
        reader = PdfReader(pdf_path)
        # Iterate through pages and extract text
        for page in reader.pages:
            text += page.extract_text() + "\n"
    except Exception as e:
        print(f"Error reading PDF with pypdf: {e}")
        return None
    return text

def get_text():
    return """
            Have you been wondering what the process of getting into selective is? If you have, this explanation is perfect for you. The three main things are preparation, application, and test day.  

            The first point is that if you are taking the test, you should prepare for it. You can do this by researching the selective test and see some other tips on how to sit the test. Once you know more about it, you should study for the test. This will get you more used to what to do in the real exam. You can practice by using old test papers so you can get an idea of what type of questions they give. After you have answered all of them right in a row, you should add in a time limit. When it is your first test with a limit, you should stick to the limit of the test you are doing. Once it starts to get easier, you should lower the time down by five minutes to increase your speed. If you practice, you shall be able to sit the test easily. 

            Before the test day, you should know where you will go. Some schools offer notes. In the note the student has to put in their local school, or their preferred school. If it is not your local school, you will need to explain why you want to go to another school. You will need to know more about the school. The best way to do this is to go on a school tour. You will have to give the office your details but remember to keep it private. After a while, you should receive a letter of confirmation. If not, contact the NSW department of education immediately.  

            The end result is the most important thing. The exam administrator will give you more information on how to sit the test. Once they give you the time limit and the number of questions, you should start thinking about how long you should take to answer each one. A timer will be given to help you do this. You should keep in mind that some schools allow people with certain grades to come there, so you may not get into the school you applied for. E.g. The school you applied for may only allow people with 25 percentiles, but you might have gotten 50. This is why you should practice properly. 

            A selective test needs a lot of time and effort. The test taker can contribute by practicing, letting the department of education know about where you are going to learn (parent or guardian), and applying all your skills to the test day. This ensures you feel ready for the test. 
    """
```

```python
# here is the question; type over this to ask something new

system_prompt = """You are an expert tutor in Non Narrative writing for Selective Tests in Australia. You give prcise guidance on Content, form organisation, style and expression.
You MUST STRICTLY follow the following rubrics while assessing a piece of writing and suggesting gaps and providing improvement clues. You MUST also score the writing out of 25 and include it in the output

# IMPORTANT Rubric for Content, Form, Organisation and Style
  1. The response clearly and consistently addresses the stimulus. All 3 main ideas are relevant to the audience and purpose of the text.
  2. The response contains 3 different main ideas. Each main idea is supported by at least 3 distinct sub-points/evidence which focus on a particular theme. 
     In total there should be around 9 sub points.
  3. The ideas are organised coherently using appropriate paragraphs and a range of cohesive devices.
  4. The ideas are highly interesting and mature. Mature ideas demonstrate an awareness of current affairs(if applicable) and show that the writer is knowledgeable about different topics.
  5. The written response clearly and consistently uses an appropriate form and addresses all features of the given text type.
     This includes aim/audience, structure and appropriate language conventions of the text type (narrative voice, tense, formality, modality).
  6. There is a variety of sophisticated vocabulary and expressions used in the correct context.
  7. The written response uses a deliberately chosen style and contains a variety of special writing techniques appropriate for the text type.

# IMPORTANT Rubric for Expression
  1. There are varied sentence lengths(compound/complex) and structures which engage the reader.
  2. There is a variety of sentence types(simple, complex,compund) that creates flow and tension.
  3. There is a logical flow of ideas within paragraphs.There are no redundant words and no verbose or ambiguous phrases.
  4. Technical skills (grammar, punctuation and spelling) were implemented correctly with minor slip-ups that do not distract from the overall quality

  OUTPUT ** MAKE SURE you respond in valid json object only without code blocks. The response MUST adhere to properly indented JSON structure as shown in the following example:
  {
     "Overall score out of 25":"20",
     "Details" : [
         {"Line number which violates rubric item":"5","Extract which violates rubric":"The three main things are preparation,application, and test day.","Which rubric item it violates":"Coherent organisation","Reason for violating rubric":"Does not tell the reader that you will explain the whole process from start to finish, so they can see how your points fit together.","Improvement with example": "You can make your explanation stronger by using clearer linking words to show how each part connects to the next. At the moment, the steps are present, but sometimes they feel a bit separate, so adding words like 'after that', \"because of this\" or \"as a result\" will help the reader see how one step leads to another. For example, you could change a sentence like \"After a while, you should receive a letter of confirmation\" to \"After you finish your application, you should receive a letter of confirmation, and as a result you will know your place for the test.\""}  
      ]
     "Detailed Feedback" :{"Good":"You organised your explanation into clear sections that match the steps in the process, which makes it easy for the reader to follow. By having an introduction, then parts on preparation, application, and what happens with the test, and finally a short ending, you show you understand how to set out an explanation. For example, you start with a question to hook the reader, then move into paragraphs that begin with lines like 'The first point is that if you are taking the test, you should prepare for it' and later finish with a short conclusion about time and effort. You keep a calm and helpful tone all the way through, which suits an explanation piece about school very well. This friendly and steady voice helps readers feel guided rather than scared about a big test. For instance, sentences like 'If you practise, you will be able to sit the test more easily' and 'This helps you feel ready for the test' sound encouraging and supportive instead of bossy or worrying.","Improve": "You can make your explanation stronger by using more clear linking words to show how each part connects to the next. At the moment, the steps are there, but sometimes they feel a bit separate, so adding words like 'after that', 'because of this' or 'as a result' will help the reader see how one step leads to another. For example, you could change a sentence like 'After a while, you should receive a letter of confirmation' to 'After you finish your application, you should receive a letter of confirmation, and as a result you will know your place for the test.' You could also improve your writing by making some of your sentences longer and more detailed, instead of lots of short ones that sound similar. When many sentences start the same way, such as 'You will need to know more about the school. The best way to do this is to go on a school tour. You will have to give the office your details', it can feel a bit choppy, so try joining ideas together to create smoother sentences. For example, you might write 'You will need to know more about the school, so the best way to do this is to go on a school tour, where you will have to give the office your details.'"}
  }
  Keep the tone always encouraging and enthusiatic. Make the suggestions concise and precise with clear examples relevant to the piece.
"""

question = f"""
Stimulus: 'Write an explanation explaining the process of getting into a selective school'
Writing to be marked using the rubric : {get_text()}
"""
```

```python
# Get gpt-4o-mini to answer, with streaming
def getGuidanceFromGPT():
    openai = OpenAI()

    response = openai.chat.completions.create(model="gpt-5-nano",
     messages=[{"role":"system","content":f"{system_prompt}"},{"role":"user","content":f"{question}"}],
     response_format={'type': 'json_object'})
    return response.choices[0].message.content

result = getGuidanceFromGPT()
print(json.dumps(result, indent=4))
print("==="*60)
display(Markdown(result))
```
