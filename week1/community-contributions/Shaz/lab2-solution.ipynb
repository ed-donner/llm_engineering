```python
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
email_text = """
Subject: Meeting Schedule

Hi Sarah,

I hope you're doing well.

I wanted to check if you're available for a project meeting sometime next week. I'd like to discuss the latest updates and review the remaining tasks before our deadline.

Please let me know which day and time works best for you.

Looking forward to hearing from you.

Best,
John
"""

system_prompt = """
You are an expert executive assistant who writes natural, human-sounding email responses.

Your responses should:
- Match the tone of the original email.
- Be concise and easy to read.
- Be grammatically correct.
- Avoid unnecessary repetition.
- Never fabricate information.
- Preserve professionalism unless the email is clearly casual.
- Return only the email body.
"""

user_prompt = f"""
Respond to the email below.

Email:
{email_text}

Write only the reply.
"""

message = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt},
]
```

```python
response = ollama.chat.completions.create(model="llama3.2", messages=message)

result = response.choices[0].message.content

display(Markdown(result))
```
