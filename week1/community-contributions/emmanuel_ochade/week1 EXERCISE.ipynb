```python
# imports

import os
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_OLLAMA = 'llama3.2:1b'
```

```python
# set up environment

load_dotenv(override=True)

gpt_api_key = os.getenv('OPENROUTER_API_KEY')
gpt_base_url = os.getenv('OPENROUTER_BASE_URL')
ollama_api_key = os.getenv('OLLAMA_API_KEY')
ollama_base_url = os.getenv('OLLAMA_BASE_URL')


if(gpt_api_key):
    print(f"GPT MODEL APi KEY FOUND, and GPT_MODEL_BASE_URL {gpt_base_url}")
else:
    print(f"GPT MODEL APi KEY NOT FOUND, and GPT_MODEL_BASE_URL {gpt_base_url}")

if(ollama_api_key):
    print(f"OLLAMA MODEL APi KEY FOUND, and OLLAMA_MODEL_BASE_URL {ollama_base_url}")
else:
    print(f"OLLAMA MODEL APi KEY NOT FOUND, and OLLAMA_MODEL_BASE_URL {ollama_base_url}")
```

```python
# here is the question; type over this to ask something new

question = """
What techniques would you use to structure an LLM system prompt that ensures content-aware summarization, excludes non-semantic UI elements (like navigation text), enforces tone constraints (snarky/humorous), and guarantees Markdown-only output formatting?
"""

system_message = """
You are a content analysis assistant designed to process raw website text.

Your responsibilities:

Extract and focus only on the primary semantic content of the page (e.g., main article body, product descriptions, blog content).

Ignore non-content elements such as:

Navigation menus

Headers and footers
"""
```

```python

def stream_api_response(base_url, api_key, model):
    openai = OpenAI(base_url=base_url, api_key=api_key)
    stream = openai.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": system_message},
            {"role": "user", "content": question}
          ],
        stream=True
    )    
    response = ""
    display_handle = display(Markdown(""), display_id=True)

    for chunk in stream:
        if chunk.choices:
            content = chunk.choices[0].delta.content or ''
            response += content
            update_display(Markdown(response), display_id=display_handle.display_id)
```

```python
# Get gpt-4o-mini to answer, with streaming

stream_api_response(gpt_base_url, gpt_api_key, MODEL_GPT)
```
