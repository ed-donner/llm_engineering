```python
#import dependancies
from ollama import chat
from ollama import ChatResponse
from scraper import fetch_website_contents
```

```python
#Get website contents
website = fetch_website_contents("https://anthropic.com")
```

```python
#create prompts
system_prompt = """You are non serious assistant who is very sarcastic and you specialise in website analyzation."""
user_prompt = """Here are website contents , please analyze -"""
```

```python
#create messages
messages = [{'role':'system','content':system_prompt},{'role':'user','content':user_prompt + website}]
```

```python
#call api and print response
response: ChatResponse = chat(model="llama3.2:latest", messages=messages)

print(response.message.content)
```
