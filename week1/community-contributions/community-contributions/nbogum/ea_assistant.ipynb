This LLM notebook example demonstrates a simple, practical use of LLM to translate the rough drafts of meeting notes into a concise one-page executive summary.

```python
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

Example meeting notes

```python
SOURCE_NOTES = """
- App is old, monolith, runs on on-prem VMs, IIS + Windows
- Scaling issues during peak hours, especially month-end
- Deployments are painful, manual, require downtime
- Ops mentioned alerts come after users complain
- Logging scattered, some logs local, some in shared folders

- Team wants to move to cloud (Azure or AWS), not decided yet
- Goal is infra + ops improvements first, not app rewrite
- Maybe break out some services later, but phase 1 = lift + stabilize
- Containers came up, AKS/EKS likely
- DB is SQL Server, big risk area, needs careful migration

- CI/CD basically doesn’t exist right now
- People manually RDP and deploy
- Want automated pipelines, blue/green if possible

- DR is weak, backups exist but no real DR testing
- Cloud should improve this but needs design
- Security team will need to review networking and access

- Skills gap: team hasn’t used containers much
- Training or support needed
- Cost concern raised, don’t want surprise bills

- High-level timeline thrown out:
  - Few weeks to assess/design
  - 2–3 months for initial platform + pilot
  - Rest of year to migrate fully

- Open questions:
  - Azure vs AWS
  - SQL licensing
  - Who owns platform long-term
"""
```

Create Prompts

```python
system_prompt = "You are an enterprise architect's executive-summary assistant."
user_prompt = f"""
Create a one-page executive summary using the following template:

Title
Purpose
Current State (3 bullets)
Proposed Approach (3 bullets)
Business Value (3 bullets)
Key Risks & Mitigations (3 bullets)
Decisions Needed (up to 3 bullets)
Next Steps / Timeline (3 bullets)
Open Questions (only if needed)

Constraints:
- Max 550 words
- Short bullets (max 14 words each)
- Plain business language
- Do NOT invent facts or decisions
- Use only the information provided

Source notes:
{SOURCE_NOTES}
"""
```
