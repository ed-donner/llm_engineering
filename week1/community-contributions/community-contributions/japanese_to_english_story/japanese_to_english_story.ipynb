```python
Fetch a story in japanese text and convert it to english.
```

from scraper import fetch_website_contents

japanese_story = fetch_website_contents("https://www.aozora.gr.jp/cards/000035/files/301_14912.html")

# japanese_story

OLLAMA_BASE_URL = "http://localhost:11434/v1"
from openai import OpenAI
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
for line in japanese_story.split("\n\n"):
    response = ollama.chat.completions.create(model="llama3.2", messages=[{"role": "user", "content": "just convert this to english from japanese. do not add any extra text. just the translation. "+line}])
    print(response.choices[0].message.content)
