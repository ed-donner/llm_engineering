```python
import sys
sys.path.insert(0, '..')

from openai import OpenAI
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
from scraper import fetch_website_contents

def summarize(url):
    website = fetch_website_contents(url)
    user_prompt_prefix = "Summarize the following website without missing any important details "
    response = ollama.chat.completions.create(model="deepseek-r1:1.5b", messages=[ {"role": "user", "content": user_prompt_prefix + website}])
    return response.choices[0].message.content
```

```python
summarize("https://finance.yahoo.com/news/warren-buffetts-final-373-billion-024100554.html")
```
