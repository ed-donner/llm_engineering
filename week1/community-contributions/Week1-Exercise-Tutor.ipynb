Build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI
import requests
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python
system_prompt="""
You are a teachnical tutor. Your responsibility is to help the user understand the technical code provided assuming they have no prior knowledge of the subject.
"""

user_prompt="""
Here is the code provided by the user, please explain what the code does and why.
"""
```

```python
def explain_code(code):
    client = OpenAI(api_key=api_key)
    #user_prompt = f"Here is the code provided by the user: {code}"
    response = client.chat.completions.create(
        model=MODEL_GPT,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt + code}
        ]
    )
    return response.choices[0].message.content
```

```python
def pretty_explanation(code):
    explanation = explain_code(code)
    display(Markdown(explanation))
```

```python
explain_code('yield from {book.get("author") for book in books if book.get("author")}')
```

```python
pretty_explanation('yield from {book.get("author") for book in books if book.get("author")}')
```
