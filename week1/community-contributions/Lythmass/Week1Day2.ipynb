```python
from openai import OpenAI
from IPython.display import Markdown, display
from bs4 import BeautifulSoup
import requests

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
}

OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL)
```

```python
system_message = {"role": "system", "content": "You're an expert at roasting a summarized version of a website's content. You will be given the title and main text of a website, and your task is to provide a concise summary that captures the essence of the content while adding a humorous and critical twist. Your summary should be engaging, witty, and should not hold back on the roasting aspect. Make sure to highlight any absurdities, contradictions, or noteworthy points in the content in a way that entertains the reader. Answer in clear markdown."}
user_message_prefix = {"role": "user", "content": "Summarize the following website contents:\n\n"}
```

```python
class Website:
  def __init__(self, url):
    self.url = url
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    self.title = soup.title.string if soup.title else "No Website Title"
    for irrelevant in soup.body(["script", "style", "img", "input"]):
      irrelevant.decompose()
    self.content = soup.body.get_text()
```

```python
def summarize_website(title, content):
  contents = f"Title: {title}\n\nContent:\n{content}"
  response = ollama.chat.completions.create(
    model="llama3.2", 
    messages=[system_message, user_message_prefix, {"role": "user", "content": contents}]
  )
  return response.choices[0].message.content
```

```python
def display_summary(url):
  website = Website(url)
  summary = summarize_website(website.title, website.content)
  display(Markdown(summary))
```

```python
display_summary("https://www.anthropic.com/")
```
