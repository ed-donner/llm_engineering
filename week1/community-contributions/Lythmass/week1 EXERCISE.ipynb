```python
# imports
import os
from openai import OpenAI
import dotenv
from IPython.display import Markdown, display, update_display
```

```python
# constants

MODEL_GPT = 'gpt-5.4-mini'
MODEL_LLAMA = 'llama3.2'
OLLAMA_BASE_URL = 'http://localhost:11434/v1'
```

```python
# set up environment
dotenv.load_dotenv(override=True)
openai = OpenAI()
ollama = OpenAI(base_url=OLLAMA_BASE_URL)

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

if not OPENAI_API_KEY:
  print("Please set the OPENAI_API_KEY environment variable.")
else:
  print("OPENAI_API_KEY is set.")
```

```python
system_prompt = """
You are an expert coding tutor.
Your task is to explain my questions in a clear and concise way.
Make sure to provide simple but comprehensive examples.
You will provide one or two analogies to make the concept easier to understand.

Lastly, you will tell me how I can apply this concept in my own coding projects.

Make sure you respond in markdown format, using code blocks for any code snippets and bullet points for lists.
"""
```

```python
# Get gpt-4o-mini to answer, with streaming
def get_gpt_answer(question):
  stream = openai.chat.completions.create(
    model=MODEL_GPT,
    messages=[{"role": "system", "content": system_prompt},{"role": "user", "content": question}],
    stream=True
  )
  answer = ""
  display_answer = display(Markdown(""), display_id=True)
  for chunk in stream:
    answer += chunk.choices[0].delta.content or ''
    update_display(Markdown(answer), display_id=display_answer.display_id)
    
```

```python
# Get Llama 3.2 to answer
def get_llama_answer(question):
  stream = ollama.chat.completions.create(
    model=MODEL_LLAMA, 
    messages=[{"role": "system", "content": system_prompt},{"role": "user", "content": question}],
    stream=True
  )
  answer = ""
  display_answer = display(Markdown(""), display_id=True)
  for chunk in stream:
    answer += chunk.choices[0].delta.content or ''
    update_display(Markdown(answer), display_id=display_answer.display_id)
```

```python
question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
#get_gpt_answer(question)
get_llama_answer(question)
```
