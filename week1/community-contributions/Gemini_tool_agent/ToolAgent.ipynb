```python
# imports

import os
from dotenv import load_dotenv
# from IPython.display import Markdown, display
import google.generativeai as genai
from fib import fib,printout,call_api

# If you get an error running this cell, then please head over to the troubleshooting notebook!
```

```python
# Load environment variables in a file called .env

load_dotenv(override=True)
api_key = os.getenv('GEMINI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")

elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python


# message = "Hello, Gemini! This is my first ever message to you! Hi!"

# messages = [
#     {
#         "role": "user", 
#         "parts": [message]
#     }
# ]

# messages
model = genai.GenerativeModel(
    model_name='gemini-3-flash-preview',
    tools=[fib,printout,call_api]
)

chat = model.start_chat(enable_automatic_function_calling=True)
```

```python

# model = genai.GenerativeModel('gemini-3-flash-preview')

# response = model.generate_content(messages)

# print(response.text)


response = chat.send_message("just call the call_api function and print the response body")
# After your chat.send_message call:
for message in chat.history:
    for part in message.parts:
        if part.function_call:
            print(f"🔹 Gemini called: {part.function_call.name}({part.function_call.args})")
        
        if part.function_response:
            # .response is a dictionary containing your function's return value
            output = part.function_response.response
            print(f"✅ Function returned: {output}")
print(response.text)
```
