```python
# Step 1: Create your prompts

system_prompt = "You are a job application consultant and you are to help me apply for jobs"
user_prompt = """
   help me check this website and tailor a job application letter for a backend engineer position
"""

# Step 2: Make the messages list

messages = [{"role":"system", "content":system_prompt}, {"role":"user", "content":user_prompt }] # fill this in

# Step 3: Call OpenAI
response = openai.chat.completions.create(model="gpt-4.1-nano", messages=messages)

# Step 4: print the result
response.choices[0].message.content
```

```python
def application_for_site(website):
    return [
        {"role":"system", "content":system_prompt},
        {"role":"user", "content":user_prompt + website}
    ]
```

```python
def job_application_consultant(url):
    website = fetch_website_contents(url)
    response = openai.chat.completions.create(model="gpt-4o-mini", messages=application_for_site(website))
    return response.choices[0].message.content
```

```python
def display_job_application_consultant(url):
    job_apps = job_application_consultant(url)
    display(Markdown(job_apps))
```

```python
display_job_application_consultant("https://chowdeck.com")
```
