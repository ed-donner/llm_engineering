```python
# imports

import os
from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
#Setting up the prompt

system_prompt = """
You are a snarky assistant that analyzes the contents of a website,
and provides a short, snarky, humorous summary,
"""

user_prompt_prefix = """
Here are the contents of a website.
Provide a short summary of this website.
"""

def model_messages(url):
    return[
        {"role":"system", "content":system_prompt},
        {"role":"user", "content":user_prompt_prefix + fetch_website_contents(url)}
    ]

```

```python
#Defining base URL and model (Using a LLM model from locally)
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL,api_key="ollama")


def analyze_website(url):
    response = ollama.chat.completions.create(
        model="deepseek-r1:1.5b",
        messages = model_messages(url)
    )
    return response.choices[0].message.content



def display_summary(url):
    summary = analyze_website(url)
    print(summary)

# display_summary("https://anthropic.com")
display_summary("https://edwarddonner.com")
```
