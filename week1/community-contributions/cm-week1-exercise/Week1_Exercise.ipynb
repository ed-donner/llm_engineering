To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
from openai import OpenAI
import os
from dotenv import load_dotenv
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment
load_dotenv(override=True)
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
# Get gpt-4o-mini to answer, with streaming
messages = [{"role":"user", "content":question}]

openai = OpenAI()
response = openai.chat.completions.create(model= MODEL_GPT, messages= messages, stream=True)

full_reply = ""
for chunk in response:
    # For the new Chat Completions API, content is in choices[0].delta.content
    delta = chunk.choices[0].delta
    if delta.content:
        text = delta.content
        full_reply += text

print(full_reply)  # final newline
```

```python
# Get Llama 3.2 to answer

OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
response = ollama.chat.completions.create(model=MODEL_LLAMA, messages=messages, stream=False)
response.choices[0].message.content
```
