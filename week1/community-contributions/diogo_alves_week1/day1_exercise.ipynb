```python
import os

from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI



load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

system_prompt = """You are a brutally honest UX reviewer.

Analyze the website and provide:
- First impression
- What works well
- What confuses users
- Conversion killers
- Design red flags
- Actionable fixes

Be witty but constructive."""

user_prompt = """
Analyze the following website as a brutally honest UX reviewer.

Your job is to identify what helps users and what gets in their way. Be witty, memorable, and entertaining, but make sure every criticism is constructive and actionable.

Please evaluate:

1. First Impression
   - What does the site communicate in the first 5 seconds?
   - Is the value proposition clear?

2. What's Working
   - Strong design choices
   - Effective UX patterns
   - Good copywriting
   - Trust-building elements

3. UX Crimes Against Humanity
   - Confusing navigation
   - Poor information hierarchy
   - Unclear calls to action
   - Accessibility concerns
   - Mobile usability issues
   - Any design decisions that make users work harder than necessary

4. Conversion Killers
   - Elements likely to reduce signups, sales, or engagement
   - Sources of friction or hesitation

5. Roast Level
   - Deliver a funny but fair roast of the website in 3–5 paragraphs.

6. Action Plan
   - List the top 10 improvements in order of impact.

7. Scores (1–10)
   - Clarity
   - Visual Design
   - Navigation
   - Trust
   - Accessibility
   - Mobile Experience
   - Conversion Potential

Important:
- Be direct, not mean.
- Criticize the design, not the people behind it.
- Explain why each issue matters.
- Focus on practical improvements, not personal preferences.
"""

messages = [
    {
        "role": "system",
        "content": system_prompt
    },
    {
        "role": "user",
        "content": user_prompt + "https://www.apple.com/"
    }
]

response = OpenAI().chat.completions.create(model="gpt-4.1-nano", messages=messages)

display(Markdown(response.choices[0].message.content))
```
