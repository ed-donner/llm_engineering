```python
# Load environment variables in a file called .env

import os

from dotenv import load_dotenv


load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python
import openai

system_prompt = """
Your are a nutritionist with many years of experience.
You provide a detailed meal plan for a week, inspired by Dorian Yates' training philosophy.
"""
user_prompt = """
Create a sample meal plan for a week to complement a 4-day split bodybuilding routine inspired by Dorian Yates.
This is for a 230 pound male aiming to maintain muscle mass while losing 30 pounds of fat over up to six months.
Focus the types of meals, portion sizes, and timing around workouts.
Please provide macronutrient breakdowns for each meal and daily totals.
Also specify daily caloric intake.
"""
def messages():
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ]

def create_custom_diet_plan():
    response = openai.chat.completions.create(model="gpt-4.1-nano", messages=messages())
    return response.choices[0].message.content

print(create_custom_diet_plan())
```
