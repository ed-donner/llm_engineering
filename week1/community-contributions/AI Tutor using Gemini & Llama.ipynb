Build a tool that takes a technical question, and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
import os
import json
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display, update_display 
```

```python
MODEL_Gemini = 'gemini-3.1-flash-lite'
MODEL_llama = 'llama3.2:1b'
```

```python
load_dotenv(override=True)

google_api_key=os.getenv('GOOGLE_API_KEY')

gemini_base_url="https://generativelanguage.googleapis.com/v1beta/openai/"

gemini=OpenAI(base_url=gemini_base_url,api_key=google_api_key)
```

```python
system_prompt='''
You are an expert AI Tutor. Given a technical question, provide a clear, accurate explanation. Start with a simple overview, then explain the concept step-by-step, include a practical example if helpful, define important terms, and keep the explanation appropriate for the engineering graduate audience's level.
All questions are in context of GenAI.'''

user_prompt ='''
Please explain the following technical question for engineering graduate audience. All questions are in context of GenAI.
yield from {book.get("author") for book in books if book.get("author")}
'''
```

```python
response = gemini.chat.completions.create(model=MODEL_Gemini,
messages =[{"role":"system","content":system_prompt}, {"role":"user","content":user_prompt}])


print(response.choices[0].message.content)
```

```python
# Get gemini model to answer, with streaming

stream = gemini.chat.completions.create(model=MODEL_Gemini,
messages =[{"role":"system","content":system_prompt}, {"role":"user","content":user_prompt}],
stream=True)
response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    update_display(Markdown(response), display_id=display_handle.display_id)

```

```python
# Get Llama 3.2 to answer
OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama=OpenAI(base_url=OLLAMA_BASE_URL,api_key='ollama')

response=ollama.chat.completions.create(model=MODEL_llama,
messages=[{"role":"system","content":system_prompt},
{"role":"user","content":user_prompt}])

print(response.choices[0].message.content)
```

```python
#prompt to take question as an input

user_prompt_prefix = """
Please explain the following technical question for engineering graduate audience. All questions are in context of GenAI.


"""
```

```python
def messages_for(question):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + question}
    ]
```

```python
def tutor(question):
    
    response = gemini.chat.completions.create(model=MODEL_Gemini, messages =messages_for(question))
    
    return response.choices[0].message.content
```

```python
display(Markdown(tutor("What is RAG?")))
```

```python

def tutor_llama(question):
    
    response = ollama.chat.completions.create(model=MODEL_llama, messages =messages_for(question))
    
    return response.choices[0].message.content
```

```python
display(Markdown(tutor_llama("What is Retrieval Augmented Generation?")))
```
