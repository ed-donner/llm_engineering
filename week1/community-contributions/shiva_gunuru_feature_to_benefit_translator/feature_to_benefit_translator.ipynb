## 📌 Project Overview
This project is an automated pipeline that uses a Large Language Model (LLM) which is run locally(in my case it's gemma3:4b) to read dense, technical product pages and translate them into compelling, benefit-driven social media hooks.

Tech companies often struggle to explain their complex products in plain English. This tool demonstrates how generative AI can quickly translate technical jargon into consumer-friendly marketing content, saving hours of copywriting time.

### *Step 1: Importing all the necessary Libraries*

```python
import os  # this is the package to interact with the operating system
from dotenv import load_dotenv   # this is the package to load the environment variables
from scraper import fetch_website_contents # this is a custom function to fetch the website contents
from openai import OpenAI   # this is the package to exactly call the http endpoint

load_dotenv(override=True) # this is the function to load the .env file for the api key which is stored in the root directory and hidden from the github repo( we dont have one as we are runing the model locally)
api_key = os.getenv("GEMMA_API_KEY") 
```

### *Step 2: Craft the System Prompt*
This is the most important part of the project. We need to give the AI strict instructions on its persona and the exact output format we want.

```python
system_prompt = """
You are an elite B2B technology copywriter. Your job is to read dense, technical product pages, API documentation, or software update logs and translate them into plain English.

First, identify the core technical feature. Then, translate that feature into a human benefit (e.g., instead of "features end-to-end encryption," use "keeps your team's private data safe from hackers").

Output exactly three compelling, benefit-driven social media hooks that a non-technical audience would click on. Format the output strictly as:

Hook 1: [Text]
Hook 2: [Text]
Hook 3: [Text]

Do not include any conversational filler.
"""
```

### *Step 3: Build the Execution Function*
Now, create the function that pulls the website data, constructs the message package, and talks to the model.

```python
def  generate_pitch(url):
    # Fetch the website content
    website_content = fetch_website_contents(url)

    # Construct the message package
    message = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": f"Translate the following website content into a benefit-driven social media hook:\n\n {website_content}"}
    ]

    openai = OpenAI(base_url='http://localhost:11434/v1',api_key='gemma')

    # Call the model
    response = openai.chat.completions.create(
        model="gemma3:4b",
        messages=message
    )

    # Extract the output
    return response.choices[0].message.content.strip()
```

### *Step 4: Test the pipeline*

```python
# Let's test it on an AWS cloud computing service page
print(generate_pitch("https://aws.amazon.com/ec2/"))
```
