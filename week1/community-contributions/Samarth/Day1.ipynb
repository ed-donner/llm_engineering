```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown, display
```

```python
system_prompt = '''You are a helpful and friendly AI assistant.Answer questions clearly and in simple language.
Keep responses short.Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.'''

user_prompt = "Explain what is meant by an AI engineer"

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
```

```python
load_dotenv()
api_key = os.getenv("GEMINI_API_KEY")
openai = OpenAI(
    api_key=api_key,
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)

response = openai.chat.completions.create(model = "gemini-3.1-flash-lite-preview", messages = messages)

display(Markdown(response.choices[0].message.content))

```
