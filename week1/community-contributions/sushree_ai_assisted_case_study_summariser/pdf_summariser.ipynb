```python
openai = OpenAI()
```

```python
pdf_extract=extract_pdf_text(r'D:\projects\llm_engineering\week1\community-contributions\sushree\CASE_STUDY_TATA_NANO.pdf')
#print(pdf_extract)
```

```python
system_prompt="""
You are a highly qualified and helpful professor and a case study expert who is highly skilled  \
in writing case study summaries which can help students to understand the structure, content and enable  \
them to understand and solve case studies. You can prefer the structure: Executive Summary, Problem Statement, \
Situation Analysis, Identify and Evaluate Alternatives, Recommendation, Implementation Plan, Risk and Contingency, \
Conclusion. Respond in markdown. Do not wrap the markdown in a code block \
- respond just with the markdown.
"""
```

```python
user_prompt = """
Here are the contents of a pdf.
Provide a structured summary of this pdf.
If it includes news or announcements, then ignore them.
"""

user_prompt += pdf_extract
```

```python
def messages_for(pdf_file_path):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content":user_prompt}
    ]
```

```python
def case_study_summary(file_path):
    extract=extract_pdf_text(file_path)
    response=openai.chat.completions.create(
        model="gpt-4.1-mini",
        messages=messages_for(extract)
    )
    return response.choices[0].message.content
```

```python
case_study_summary(r"D:\projects\llm_engineering\week1\community-contributions\sushree\CASE_STUDY_TATA_NANO.pdf")
```

```python
def display_case_study_summary(file_path):
    display_analysis = case_study_summary(file_path)
    display(Markdown(display_analysis))
```
