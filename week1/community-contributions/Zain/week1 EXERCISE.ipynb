## Agentic AI Expert Tutor Using OpenAI

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from groq import Groq
```

```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

if api_key and api_key.startswith('sk-proj-') and len(api_key)>10:
    print("API key looks good so far")
else:
    print("There might be a problem with your API key? Please visit the troubleshooting notebook!")
    
MODEL_GPT = 'gpt-4o-mini'
openai = OpenAI()
```

```python
agentic_ai_expert_system_prompt = """
You are an Agentic AI Expert Tutor with deep expertise in AI agents, LLMs, Agentic AI frameworks (LangGraph, CrewAI, AutoGen, OpenAI Agents SDK, PydanticAI, DSPy, Semantic Kernel, MCP, A2A), and production-grade agent architectures.
Answer every question accurately, clearly, and with educational depth, adapting explanations to the user's skill level from beginner to expert.
When appropriate, provide architecture diagrams, code examples, best practices, trade-offs, debugging guidance, and real-world use cases.
Never hallucinate APIs, framework features, or technical details; explicitly state uncertainty when information is unknown.
Your goal is not only to answer questions but to teach users how Agentic AI systems work, how to build them, and how to think like an Agentic AI engineer.
"""
```

```python
def agentic_ai_expert_user_prompt(question):
    user_prompt = f"""
    I want to learn and build Agentic AI systems. For every question, explain the concept clearly, then provide technical details, practical examples, architecture insights, and code snippets when relevant. Compare alternatives, discuss trade-offs, and share industry best practices. If a topic is advanced, break it down step-by-step and connect it to real-world Agentic AI applications. Assume I am an AI engineer seeking production-grade knowledge. Give me a comprehensive answer to the following question: "{question}"
    """
    return user_prompt
```

```python
def stream_answer(question):
    stream = openai.chat.completions.create(
        model=MODEL_GPT,
        messages=[
            {"role": "system", "content": agentic_ai_expert_system_prompt},
            {"role": "user", "content": agentic_ai_expert_user_prompt(question)}
          ],
        stream=True
    )    
    
    for chunk in stream:
        print(chunk.choices[0].delta.content or '', end='', flush=True)
```

```python
stream_answer("How can I build Agentic AI applications without writing code?")
```

## Chemistry Expert Tutor using Llama

```python
load_dotenv(override=True)
api_key = os.getenv('GROQ_API_KEY')

if api_key and api_key.startswith('gsk_') and len(api_key)>10:
    print("API key looks good so far")
else:
    print("There might be a problem with your API key? Please visit the troubleshooting notebook!")
    
MODEL_LLAMA = 'llama-3.3-70b-versatile'
groq = Groq()
```

```python
chemistry_expert_system_prompt = """
You are a Chemistry Expert Tutor with deep expertise in Organic, Inorganic, Physical, Analytical, Computational, Medicinal, Industrial, and Materials Chemistry, capable of answering questions from high-school to PhD and research level.
Provide scientifically accurate, evidence-based explanations, adapting the depth and complexity of answers to the user's knowledge level.
When appropriate, explain reaction mechanisms, molecular structures, thermodynamics, kinetics, spectroscopy, calculations, and laboratory techniques step-by-step.
Include chemical equations, reaction pathways, examples, real-world applications, and problem-solving methods to build deep understanding.
Never fabricate chemical facts, data, mechanisms, or safety information; explicitly state uncertainty and prioritize scientific accuracy and safe laboratory practices.
"""
```

```python
def chemistry_expert_user_prompt(question):
    user_prompt = f"""
    I want to learn Chemistry from basic to advanced level. For every question, explain clearly with step-by-step reasoning, include equations, mechanisms, and real-world applications where relevant. Adjust depth based on difficulty (beginner to PhD level). Provide examples, problem-solving methods, and when needed, link concepts across Physical, Organic, and Inorganic Chemistry. Keep explanations accurate, structured, and exam/research oriented. Give me a comprehensive answer to the following question: "{question}"
    """
    return user_prompt
```

```python
def stream_answer(question):
    stream = groq.chat.completions.create(
        model=MODEL_LLAMA,
        messages=[
            {"role": "system", "content": chemistry_expert_system_prompt},
            {"role": "user", "content": chemistry_expert_user_prompt(question)}
          ],
        stream=True
    )    
    
    for chunk in stream:
        print(chunk.choices[0].delta.content or '', end='', flush=True)
```

```python
stream_answer("Please explain me methan hydroxide and phosphate hydroxide with their structures?")
```
