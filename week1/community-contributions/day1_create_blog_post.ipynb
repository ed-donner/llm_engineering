Snarky Blog Post Generator
This notebook builds a lightweight AI writing assistant that transforms a rough blog idea into a short, catchy post. It uses OpenAI chat completions with a custom “snarky but concise” tone prompt, then outputs polished markdown content with a title in under 50 words. Great for quickly drafting engaging blog intros.

```python
# Step 1: Create your prompts
# imports
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI


# Load environment variables in a file called .env

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key
if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print(f"API key found and looks good so far! key length: {len(api_key)}")

openai = OpenAI()

system_prompt = """ as a snarky assistant, you will help me with my tasks. 
                    keep your responses concise and to the point.
                    Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
                    """
user_prompt = """
    below is the content, I would like to refine the content to make it more engaging and interesting.
    keep it to 50 words. give it a catchy title.
    {content}
"""
user_content = """ 
    i would like to write a blog post about the benefits of using cursor for coding.
"""

def refine_content(content):
    user_prompt_full = user_prompt.format(content=content)
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_full}
    ]
    response = openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages
    )
    return response.choices[0].message.content
    # print("user_prompt_full: ", user_prompt_full)

# Step 3: Call OpenAI

# Step 4: print the result
display(Markdown(refine_content(user_content)))
```
