# Email subject-line suggester (OpenRouter)

Commercial-style example: given an email body, suggest a short subject line (e.g. as in an email client).

Uses OpenRouter with `openai/gpt-4o-mini`. Set `OPENROUTER_API_KEY` in your `.env`.

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
```

```python
load_dotenv(override=True)
api_key = os.getenv("OPENROUTER_API_KEY")
if not api_key:
    raise ValueError("Add OPENROUTER_API_KEY to your .env file")

client = OpenAI(
    api_key=api_key,
    base_url="https://openrouter.ai/api/v1",
)
```

```python
# Commercial example: suggest a short subject line for an email

system_prompt = """You are an assistant that suggests concise, professional email subject lines.
Given the body of an email, reply with only a short subject line (under 10 words). No explanation, no quotes—just the subject line."""

email_body = """
Hi team,

Quick update on the Q3 launch: we're moving the release to September 15 to allow
extra time for the payment integration tests. Marketing will adjust the campaign
dates accordingly. Let me know if this affects your deliverables.

Thanks,
Jordan
"""

user_prompt = f"Suggest a subject line for this email:\n\n{email_body.strip()}"

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt},
]

response = client.chat.completions.create(
    model="openai/gpt-4o-mini",
    messages=messages,
)

suggested_subject = response.choices[0].message.content.strip()
print("Suggested subject line:", suggested_subject)
```
