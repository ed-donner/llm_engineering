```python
from openai import OpenAI
from IPython.display import Markdown, display
from dotenv import load_dotenv
import os
```

```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python
import random
openai = OpenAI()
# Step 1: Create your prompts
prescriptions = [{
    "patient_name": "JS",
    "age": 20,
    "weight": 70,
    "prescription": "Paracetamol 1g tds x 3/7"
}, {
    "patient_name": "JD",
    "age": 30,
    "weight": 80,
    "prescription": "Ibuprofen 1g tds x 3/7"
},
{
    "patient_name": "JH",
    "age": 40,
    "weight": 90,
    "prescription":["Carbamazepine 200mg bd x 1/52", "Bromazepam 10mg nocte x 3/7"]
}
]

system_prompt = """You are a helpful assistant for pharmacists in a community pharmacy.
 You are able to review prescriptions and recommend to the pharmacist wether to dispence or not.
 You recommend and summarise to the pharmcist some safety measures like "contact prescriber" if there is an issue with the precription.
 Always consider drug -drug interaction and call it out.
  """
user_prompt = f"""
    Please review the following prescription and recommend whether to dispense or not.
    {random.choice(prescriptions)}
"""

# Step 2: Make the messages list

messages = [
  {"role":"system", "content": system_prompt},
  {"role":"user", "content": user_prompt}
] # fill this in

# Step 3: Call OpenAI
response = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages
)

# Step 4: print the result
display(Markdown(response.choices[0].message.content))
```
