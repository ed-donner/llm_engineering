```python
# imports

# use sys when getting ModuleNotFoundError: No module named 'scraper', 
# this notebook lives one directory deeper

import sys
sys.path.insert(0, '..')

import os
from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
## use an Open Source model running locally via Ollama rather than OpenAI

OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
# This website is a clone of Zillow.com

zillow = fetch_website_contents("https://appbrewery.github.io/Zillow-Clone/")    

# Define our system prompt 

system_prompt = """
You are a trusted advisor, blending deep local market expertise that analyses the contents of a website,
and provides a short summary of the website, ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
""" 

# Define our user prompt

user_prompt_prefix = """
Here are the contents of a website.
Provide a list of the apartments and details, for this one webpage only.
If it includes a map, then summarize that too.
"""
```

```python
# Form the messages

def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + website}
    ]
```

```python
# And now: call the Ollama

def summarize(url):
    website = fetch_website_contents(url)
    response = ollama.chat.completions.create(
        model = "gemma4",
        messages = messages_for(website)
    )
    return response.choices[0].message.content
```

```python
# A function to display this nicely in the output, using markdown

def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))
```

```python
summarize("https://appbrewery.github.io/Zillow-Clone/")
```

```python
display_summary("https://appbrewery.github.io/Zillow-Clone/")
```
