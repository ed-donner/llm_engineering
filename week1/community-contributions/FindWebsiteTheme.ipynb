Upgrade the day 1 project to summarize a webpage to use an Open Source model running locally via Ollama rather than OpenAI

**Instead of summarize a webpage, i just find the main theme of the webpage in shortly:**



```python
from openai import OpenAI

OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')


system_prompt = "you are an assistant that can answer questions about the world, and you are a genius in giving short answers"
user_prompt = """
   give me a short answer to the question: what is the theme of the website?
"""
website = "https://www.sport5.co.il"

messages = [{"role": "system", "content": system_prompt}, {"role": "user", "content": user_prompt + website}]

response = ollama.chat.completions.create(model="llama3.2", messages=messages)
response.choices[0].message.content
```
