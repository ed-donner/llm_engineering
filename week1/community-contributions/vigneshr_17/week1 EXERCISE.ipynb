```python
# imports

import os
import sys
import json
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
from scraper import fetch_website_contents, fetch_website_links
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')
if api_key and api_key.startswith('sk-proj') and len(api_key) > 10:
    print("API key loaded successfully.")
else:
    print("Error: API key not found or invalid. Please set the OPENAI_API_KEY environment variable.")

openai = OpenAI()
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""

def answer_question(question):
    response = openai.chat.completions.create(
        model = MODEL_GPT,
        messages = [
            {"role": "system", "content": "You are a helpful assistant that explains code snippets."},
            {"role": "user", "content": question}
        ],
        max_tokens = 200,
    )

    return response.choices[0].message.content.strip()

print("Question:")
display(Markdown(f"**{question.strip()}**"))
print("Answer:")
display(Markdown(f"**{answer_question(question)}**"))
```

```python
# Get gpt-4o-mini to answer, with streaming

def stream_answer(qn):
    stream = openai.chat.completions.create(
        model = MODEL_GPT,
        messages= [
            {"role": "system", "content": "You are a Snarky assistant that explains code snippets."},
            {"role": "user", "content": qn}
        ],
        stream= True
    )
    response = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)

stream_answer(question)

```
