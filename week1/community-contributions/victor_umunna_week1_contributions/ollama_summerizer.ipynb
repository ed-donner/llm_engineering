```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import requests


```

```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python
# Create OpenAI client
openai = OpenAI()
```

```python
email = """
Dear Ed,

I hope you are doing well. I’m writing to inform you that I’ve been unwell and was recently admitted to the hospital, which has made it impossible for me to take the scheduled exam.

I sincerely apologize for the inconvenience this may cause and kindly request an opportunity to retake the exam once I have recovered and been discharged. I will provide any necessary medical documentation if required.

Thank you very much for your understanding and support.

Warm regards,
Victor Umunna
"""
# Step 1: Create your prompts

system_prompt = """
You are an helpful assistant that analyzes the contents of an email and
provide an appropriate short subject line for the email.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""
user_prompt = """
    Here are the contents of an email.
    Provide an appropriate short subject line for the email.
"""

# Step 2: Make the messages list

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt + email}
] 
```

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
# Step 3: Call OpenAI
response = ollama.chat.completions.create(
        model="llama3.2",
        messages =messages)
result = response.choices[0].message.content

# Step 4: print the result
print(result)
```
