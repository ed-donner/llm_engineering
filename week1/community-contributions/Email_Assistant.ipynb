## AI Email Assistant

```python
from IPython.display import Markdown, display
from openai import OpenAI
# Step 1: Create your prompts

system_prompt = "You are a helpful assistant who goes through emails and give a brief summary of the email including only the important information. You tell the user the most appropriate subject. Make sure that you include any timelines if mentioned. "
user_prompt = """
    Here is the overview of the email along with the suitable subject line.
    "I hope this email finds you well. I am reaching out to share the comprehensive project proposal and implementation roadmap for our upcoming Q3 marketing campaign, which we discussed during last week's leadership alignment session.Over the past ten days, our creative and data analytics teams have collaborated closely to refine our target audience segments, optimize the digital ad spend allocation, and design fresh visual assets. Based on your valuable feedback, we have shifted 15% of the total budget away from static display ads and redirected it toward short-form video content. This strategic adjustment aims to significantly increase our user engagement rates and lower our overall customer acquisition costs across all major digital marketing channels.The fully detailed strategy document—complete with specific weekly milestones, content calendars, asset guidelines, and performance metrics—is attached to this email for your final review. I would highly appreciate it if you could pay close attention to the vendor launch schedule outlined on page four, as we must finalize our external contracts before the end of this business week.Could you please review the attached PDF and provide your feedback or formal approval by Thursday afternoon at 4:00 PM? This timeline will ensure we stay perfectly on track for our projected launch. If you would prefer to jump on a quick 10-minute conference call to review any specific line items together, please let me know your availability for tomorrow morning.Thank you very much for your leadership, strategic guidance, and continued support on this project."


"""

# Step 2: Make the messages list

messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt

}
    ] 
    

# Step 3: Call OpenAI
openai = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")
response = openai.chat.completions.create(

        model = "llama3.2",
        messages = messages
    )

# Step 4: print the result
display(Markdown(response.choices[0].message.content))
```
