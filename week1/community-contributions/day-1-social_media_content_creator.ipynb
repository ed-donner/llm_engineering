```python
# imports

import os
import requests
from dotenv import load_dotenv
from bs4 import BeautifulSoup
from IPython.display import Markdown, display
from openai import OpenAI

# Load environment variables in a file called .env

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

openai = OpenAI()
```

```python
system_prompt = """You are a professional content creator. You have an instagram account./
Everyday, you share a post on your instagram account. You have a following of 1000 people./
Your responses should be formatted in Markdown."""

user_prompt = """
Help me to create contents for my social media platforms. I want to create contents for my Instagram.
Create 7 post ideas for the next 7 days. Decribe what you will post about. Create an energic text and a picture about my day. 
The text should include motivational quotes. and share possible hashtags.
"""

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt},
]
    
response = openai.chat.completions.create(
        model = "gpt-4o-mini",
        messages = messages
    )

# Step 4: print the result in markdown format
pretty_response = Markdown(response.choices[0].message.content)
display(pretty_response)
```
