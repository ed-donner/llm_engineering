# Snarky Landing Page & Messaging Auditor
An AI tool that scrapes a company's landing page, cuts through the corporate buzzwords, and gives a brutally honest audit of their messaging and value proposition.

```python
import os
import requests
from bs4 import BeautifulSoup
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
# Load environment variables
load_dotenv(override=True)
api_key = os.getenv('GOOGLE_API_KEY')

if not api_key:
    print("Warning: GOOGLE_API_KEY not found in environment. Please add it to your .env file.")
else:
    print("Google/Gemini API key loaded successfully!")
```

```python
# Setup client pointing to Gemini OpenAI-compatible endpoint
GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"
openai = OpenAI(base_url=GEMINI_BASE_URL, api_key=api_key)
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
}

def fetch_landing_page(url):
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    title = soup.title.string if soup.title else "No title found"
    
    # Strip unnecessary tags
    if soup.body:
        for irrelevant in soup.body(["script", "style", "img", "input", "footer", "nav"]):
            irrelevant.decompose()
        text = soup.body.get_text(separator="\n", strip=True)
    else:
        text = ""
        
    return title, text[:3000] # Limit to 3000 chars
```

```python
system_prompt = """
You are a cynical, experienced Venture Capitalist and branding critic.
You analyze landing page copy and provide a brutally honest, snarky, yet insightful messaging audit.
Your output must be in Markdown and follow these sections:
1. ## Plain English Translation
   Translate the marketing speak into 1-2 sentences explaining what this product actually does.
2. ## The Buzzword Bingo Score
   List the top 3-5 corporate buzzwords found (e.g. 'synergy', 'disruptive', 'seamlessly') and explain why they are meaningless in this context. Grade the copy from A (clean) to F (buzzword soup).
3. ## Brutally Honest SWOT
   Provide 1 bullet point for each: Strength, Weakness, Opportunity, Threat.
4. ## Suggested Fixes
   Give 2 snarky, practical suggestions on how they can improve their messaging.
"""
```

```python
def audit_landing_page(url):
    print(f"Scraping and auditing: {url}...\n")
    title, text = fetch_landing_page(url)
    
    user_prompt = f"Page Title: {title}\n\nPage Copy:\n{text}"
    
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ]
    
    response = openai.chat.completions.create(
        model="gemini-2.5-flash",
        messages=messages
    )
    
    display(Markdown(response.choices[0].message.content))
```

```python
# Try auditing a site!
audit_landing_page("https://openai.com")
```
