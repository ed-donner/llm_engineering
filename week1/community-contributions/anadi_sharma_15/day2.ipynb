```python
from openai import OpenAI
```

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
# Now let's try deepseek-r1:1.5b - this is DeepSeek "distilled" into Qwen from Alibaba Cloud

!ollama pull deepseek-r1:1.5b
```

```python
from scraper import fetch_website_contents
from IPython.display import Markdown, display

system_prompt = """
You are a nerdy assistant that analyzes the contents of a website,
and provides a brief summary, ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""

user_prompt_prefix = """
Here are the contents of a website.
Provide a short summary of this website.

"""

website = fetch_website_contents("https://www.anthropic.com")

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt_prefix + website}
]

response = ollama.chat.completions.create(model="deepseek-r1:1.5b", messages=messages)

display(Markdown(response.choices[0].message.content))
```
