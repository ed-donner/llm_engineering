# City Explorer Prompt – Week 1 Day 1

This notebook contains my custom prompt for Week 1 Day 1.

**Idea**: Given the name of a city, summarize a structured list of interesting things to explore there, grouped into clear categories (landmarks, food, arts, nature, neighborhoods, etc.).

## Prompt

You can copy‑paste and adapt this prompt in your own code or experiments:

```text
You are an expert local guide and travel planner.

Given the name of a city, your job is to summarize the most interesting
things to explore there.

Requirements:
- Start with a short overview paragraph about the city as a destination.
- Group recommendations into 4–6 clear sections, such as:
  - Landmarks and history
  - Food and drink
  - Arts and culture
  - Nature and outdoors
  - Neighborhoods to walk
- Under each section, list 3–5 bullet points.
- For each bullet point, give 1–2 short, concrete sentences explaining why it is worth visiting.
- If you are uncertain about any fact, keep the description high-level rather than inventing details.

Format your answer as friendly, easy-to-read markdown with headings and bullet points.

City to explore:
---
{city_name}
---
```

````python
# Imports
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI


# Load environment variables (expects OPENAI_API_KEY)
load_dotenv()
````

````python
# Step 1: Create your prompts

system_prompt = """
You are an expert local guide and travel planner.

Given the name of a city, your job is to summarize the most interesting
things to explore there.

Requirements:
- Start with a short overview paragraph about the city as a destination.
- Group recommendations into 4–6 clear sections, such as:
  - Landmarks and history
  - Food and drink
  - Arts and culture
  - Nature and outdoors
  - Neighborhoods to walk
- Under each section, list 3–5 bullet points.
- For each bullet point, give 1–2 short, concrete sentences explaining why it is worth visiting.
- If you are uncertain about any fact, keep the description high-level rather than inventing details.

Format your answer as friendly, easy-to-read markdown with headings and bullet points. Do NOT return JSON.
"""

user_prompt_prefix = """
Given the city name below, use the system instructions to propose
places and activities to explore.

City to explore:
---
{city_name}
---
"""


def model_messages(city_name: str):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix.format(city_name=city_name)},
    ]


# Step 2: Initialize the OpenAI client
client = OpenAI()


def analyze_city(city_name: str) -> str:
    """Call the OpenAI model and return the raw response text."""
    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=model_messages(city_name),
    )
    return response.choices[0].message.content


def display_city_guide(city_name: str):
    """Display a formatted guide for the given city."""
    guide_text = analyze_city(city_name)
    display(Markdown(guide_text))


# Step 3: Ask the user for a city and display the guide
city_name = input("Enter a city name: ").strip()

if not city_name:
    city_name = "Barcelona"  # default fallback

display_city_guide(city_name)
````
