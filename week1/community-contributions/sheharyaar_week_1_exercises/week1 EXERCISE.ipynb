To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports

from IPython.display import Markdown, display
from openai import OpenAI
```

```python
# constants

OLLAMA_BASE_URL = "http://localhost:11434/v1"
OLLAMA_MODEL = "llama3.2"
```

```python
# set up environment

# Ollama must be running locally on http://localhost:11434
# This notebook only uses the local Ollama model, not the OpenAI API.

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
# here is the question; type over this to ask something new

question = """
Please explain this Python code thoroughly, including what each part does,
why it works, and any relevant edge cases:

yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
question_1 = """
Please explain this code thoroughly, including what each part does,
why it works, and any relevant edge cases:

function checkUserLocation() {
  return new Promise((resolve, reject) => {
    if (!navigator.geolocation) {
      reject("Geolocation is not supported by this browser.");
      return;
    }

    navigator.geolocation.getCurrentPosition(
      (position) => {
        const { latitude, longitude } = position.coords;
        console.log(`Location confirmed: ${latitude}, ${longitude}`);
        resolve({ latitude, longitude });
      },
      (error) => {
        reject(`Unable to retrieve location: ${error.message}`);
      },
      { enableHighAccuracy: true, timeout: 10000, maximumAge: 0 }
    );
  });
}

// Usage
checkUserLocation()
  .then((coords) => alert(`Your location: ${coords.latitude}, ${coords.longitude}`))
  .catch((err) => alert(err));
"""
```

```python
# Define a helper function to explain Python code using Llama via Ollama

def explain_code(code: str, model: str = OLLAMA_MODEL) -> str:
    system_prompt = (
        "You are an Senior Python developer and code explainer. "
        "Explain the code in depth, line by line, including behavior, purpose, and any subtle edge cases."
    )
    user_prompt = (
        "Explain the following Python code in detail:\n\n" + code +
        "\n\nInclude a clear explanation of how it works and why each part is used."
    )

    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt},
    ]

    response = ollama.chat.completions.create(
        model=model,
        messages=messages
    )
    return response.choices[0].message.content
```

```python
# Get Llama 3.2 to answer using Ollama only

answer = explain_code(question)

display(Markdown(answer))
```

```python
answer = explain_code(question_1)

display(Markdown(answer))
```
