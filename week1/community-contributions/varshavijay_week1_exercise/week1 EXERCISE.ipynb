```python
# imports
import os
import json
from openai import OpenAI
from IPython.display import Markdown, display, update_display
from scraper import fetch_website_contents, fetch_website_links
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment
llama = OpenAI(base_url='http://localhost:11434/v1', api_key = 'something')
```

```python
system_prompt = """
You are an expert technical tutor with deep knowledge of software engineering, AI, and computer science.

Your task:
Given a technical question, provide a clear, accurate, and structured explanation.

Instructions:
1. Start with a simple, intuitive explanation (ELI5 level if possible).
2. Then provide a more detailed technical explanation.
3. Include examples where relevant (code, analogies, or real-world scenarios).
4. If applicable, break down the concept step-by-step.
5. Mention common mistakes or misconceptions.
6. If the question is ambiguous, ask clarifying questions before answering.
7. If you are uncertain about any part, explicitly say "I don’t know".

Output format:
- Short summary (2–3 lines)
- Detailed explanation
- Example(s)
- Key takeaways

Constraints:
- Do NOT hallucinate facts.
- Prefer accuracy over completeness.
- Keep explanations concise but informative.
"""
```

```python
# here is the question; type over this to ask something new

# question = """
# Please explain what this code does and why:
# yield from {book.get("author") for book in books if book.get("author")}
# answer in words and don't write new code
# """

print("User Prompt \n(Note: Once done please double click on 'Enter' Key)")
print()
lines = []
while True:
    k = input()
    if k == "": 
        break
    lines.append(k)

question = "\n".join(lines)
print("Question: "+question)
```

```python
# Get Llama 3.2 to answer, with streaming
response = llama.chat.completions.create(model = MODEL_LLAMA, messages = [
    {"role": "system", "content": system_prompt},
    {"role":"user", "content": question}
],
stream=True)

collected = ""
display_handle = display(Markdown(""), display_id = True)

# now iterate over chunks as they arrive
for chunk in response:
    text = chunk.choices[0].delta.content or ""
    collected += text
    update_display(Markdown(collected), display_id = display_handle.display_id)
```
