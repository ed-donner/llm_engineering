```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment
import os
import json
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from scraper import fetch_website_links, fetch_website_contents
from openai import OpenAI
system_prompt = """
    you are a helpful assistant that can read the code block and explain it in a way that is easy to understand.
"""
user_prompt = """
    Please explain what this code, how it works, and what it does in a clear way that is easy to understand for a human that doesn't have any background in programming.
"""
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
# Get gpt-4o-mini to answer, with streaming
user_prompt += "\n".join(question)
openai = OpenAI()
stream = openai.chat.completions.create(model=MODEL_GPT, messages=[{"role": "system", "content": system_prompt}, {"role": "user", "content": user_prompt}], stream = True)

response = ""
display_handle = display(Markdown(""), display_id = True)

for chunk in stream: 
    response += chunk.choices[0].delta.content or ''
    update_display(Markdown(response), display_id = display_handle.display_id)
```

```python
# Get Llama 3.2 to answer
OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
user_prompt += "\n".join(question)
openai = OpenAI()
stream = ollama.chat.completions.create(model=MODEL_LLAMA, messages=[{"role": "system", "content": system_prompt}, {"role": "user", "content": user_prompt}], stream = True)

response = ""
display_handle = display(Markdown(""), display_id = True)

for chunk in stream: 
    response += chunk.choices[0].delta.content or ''
    update_display(Markdown(response), display_id = display_handle.display_id)
```
