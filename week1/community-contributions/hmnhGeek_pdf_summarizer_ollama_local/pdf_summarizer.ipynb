#### Importing necessary modules

```python
from utils import extract_pdf_text
from openai import OpenAI
from IPython.display import Markdown, display
from dotenv import load_dotenv
import os
```

#### Loading environment variables

```python
load_dotenv(override=True)
BASE_URL = os.getenv("OLLAMA_LOCAL_URL")
GEMMA3_1B = os.getenv("GEMMA3_1B")
GEMMA3_270M = os.getenv("GEMMA3_270M")
```

#### System and user prompts

```python
system_prompt = """
    You are a professional PDF summarizer who summarizes in just 1 line any text 
    that is given to you. Therefore, read and analyze the PDF with as much sincerity
    as possible and try to give the theme or summary of the PDF text in just one and
    only one line. And always reply in simple english language, irrespective of the 
    text language. Never respond with a followup or a question, just conclude with 
    summary always.

    Reply in markdown and do not include markdown in a code block. Remember, only respond
    in markdown format with only 1 line summarization.
"""
```

```python
user_prompt_prefix = """
    Summarize this PDF text in 1 line.
    
"""
```

```python
def get_messages(pdf_path):
    text = extract_pdf_text(pdf_path)
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + text}
    ]
```

#### Calling the model

```python
gemma = OpenAI(
    base_url=BASE_URL,
    api_key="gemma"
)
```

```python
def summarize(pdf_path, model):
    response = gemma.chat.completions.create(
        model=model,
        messages=get_messages(pdf_path)
    )
    return response.choices[0].message.content
```

```python
def summarize_pdf(pdf_path, model):
    summary = summarize(pdf_path, model)
    print(f"Summarization done by {model}")
    display(Markdown(summary))
    print()
```

#### Example runs

```python
summarize_pdf(r"kafka_streams.pdf", GEMMA3_270M)
summarize_pdf(r"kafka_streams.pdf", GEMMA3_1B)
```
