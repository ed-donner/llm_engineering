```python
#jianan 
# imports

import os
import httpx
from dotenv import load_dotenv
from scraper import fetch_website_contents, fetch_website_links
from IPython.display import Markdown, display, update_display
from openai import OpenAI
import truststore
truststore.inject_into_ssl()
```

```python
# constants

MODEL_GPT = 'gpt-5-nano'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment
load_dotenv(override=True)
# ollama 
OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')

# openai 


api_key = os.getenv('OPENAI_API_KEY')
openai=OpenAI()
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
#step up prompt


system_prompt="You are a helpful technical tutor who answers questions about python code, software engineering, data science and LLMs"
user_prompt="please give well explained with few layer/levels of detials answer, here is the question :"+question

message=[{"role":"system","content":system_prompt},{"role":"user","content":user_prompt}]
```

```python
# Get gpt-4o-mini to answer, with streaming

stream = openai.chat.completions.create(model=MODEL_GPT, messages=message, stream=True)

answer_chunks = []
display_id = "stream_out"
display(Markdown(""), display_id=display_id)

for chunk in stream:
    delta = chunk.choices[0].delta
    #handle those delat content is none
    if delta and delta.content:
        answer_chunks.append(delta.content)
        update_display(Markdown("".join(answer_chunks)), display_id=display_id)

answerfromGPT = "".join(answer_chunks)
```

```python
# Get gpt-4o-mini to answer, without  streaming
response=openai.chat.completions.create(model=MODEL_GPT, messages=message)
answerfromGPT=response.choices[0].message.content
display(Markdown(answerfromGPT))
```

```python
# Get Llama 3.2 to answer
response=ollama.chat.completions.create(model=MODEL_LLAMA, messages=message)
answerfromLLama=response.choices[0].message.content
display(Markdown(answerfromLLama))
```
