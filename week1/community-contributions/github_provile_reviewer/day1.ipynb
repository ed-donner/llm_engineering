```python
# Step 1: Create your prompts
from IPython.display import Markdown, display
import openai

github_username = ""  # Replace with your actual GitHub username
if len(github_username.strip()) == 0:
    raise ValueError("Please set your GitHub username in the 'github_username' variable.")
github_username = github_username.strip()  # Remove any leading/trailing spaces

system_prompt = """
You are a senior software engineer, code reviewer, and technical mentor. Your task is to analyze my public GitHub repositories and provide structured, actionable, and honest feedback. You will be given a link to my GitHub profile and/or specific repositories, including README files and source code, and possibly context about my goals (such as job searching, portfolio improvement, startup work, or learning). Assume these repositories are part of my public technical portfolio and that your feedback may influence hiring decisions.

First, provide a high-level portfolio assessment. Evaluate my overall skill level (beginner, junior, mid, senior — justify your reasoning), identify primary technical strengths, highlight gaps or weaknesses, and explain what signals this portfolio sends to recruiters or collaborators.

Then, for each repository, evaluate the following areas: purpose and clarity (is the problem clearly defined, is the README professional, is setup documented clearly, are screenshots or demos included where appropriate); code quality (structure and architecture, separation of concerns, naming conventions, readability, maintainability, consistency, comments, documentation); technical depth (does it demonstrate real understanding, are advanced concepts used properly, are tradeoffs visible, is error handling robust); testing and reliability (presence and quality of tests, edge case handling, defensive programming); and production readiness (environment configuration, security considerations, logging, CI/CD setup, deployment readiness).

After the evaluation, provide concrete improvement suggestions. Be specific and actionable. Suggest code-level improvements, refactoring ideas, architectural upgrades, and feature additions that would significantly increase impact. Improve vague feedback into precise recommendations (for example, instead of “improve documentation,” suggest adding a Quick Start section with installation commands and a usage example).

Next, provide career-oriented feedback. State whether this portfolio would likely pass a technical screening, what types of roles it best fits, what two to three projects would dramatically level it up, and what is missing for the next career level.

Finally, include a section titled “If I were a strict hiring manager…” and list the biggest red flags without softening the message.

Structure your response clearly with the following sections: Executive Summary, Overall Portfolio Assessment, Repository-by-Repository Breakdown, Strategic Improvements, Career Impact Assessment, and Strict Hiring Manager Mode. Be structured, detailed, practical, and prioritize usefulness over politeness.

"""
user_prompt = f"""
Analyze my public GitHub repositories and provide detailed, structured feedback. Here is my GitHub profile: https://github.com/{github_username}?tab=repositories. If needed, focus especially on these repositories: none. My current goal is: [e.g., get a mid-level backend engineering role / improve my portfolio for FAANG / prepare for startup technical interviews]. Please evaluate my overall skill level, the strength of each repository, code quality, architecture, technical depth, testing practices, and production readiness. Be specific and actionable and do not give vague advice. I want honest, direct feedback, especially what would prevent me from getting hired. Include an overall portfolio assessment, per-repository breakdown, concrete improvements, a career impact evaluation, and a “strict hiring manager” section. If something is weak, explain exactly why and how to fix it.

"""

# Step 2: Make the messages list

messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt }
    ]

# Step 3: Call OpenAI
response = openai.chat.completions.create(model="gpt-4.1-nano", messages=messages)
# Step 4: print the result
formatText = display(Markdown(response.choices[0].message.content))
print(formatText)
```
