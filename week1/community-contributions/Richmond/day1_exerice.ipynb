# AI-Powered Recruitment Advisor

The model functions as an expert recruitment advisor and career coach. 
Upon receiving a job posting URL, it automatically retrieves and analyzes 
all relevant information from the listing, then delivers the following:

## Core Features

### 1. `Resume Tailoring Recommendations`
Actionable guidance on how to align your resume with the specific 
requirements, keywords, and expectations outlined in the job description.

### 2. `Impressive Project Suggestions`
Curated project ideas or portfolio recommendations that are likely to 
capture the attention of hiring managers and demonstrate relevant 
competencies for the role.

### 3. `Cover Letter Guidance`
Strategic advice on how to articulate your value proposition, tone, and 
narrative in a way that resonates with the employer and sets you apart 
from other candidates.

```python
# Step 1: Import relevant packages
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import display, Markdown
from jscaper import scrape_job_posting
from openai import OpenAI
from groq import Groq
from typing import List
```

```python
# project configs
load_dotenv(override=True)
GROQ_KEY = os.getenv("GROQ_API_KEY")
OPENAI_KEY = os.getenv("OPENAI_API_KEY")
groq_client = Groq(api_key=GROQ_KEY)
openai_client = OpenAI(api_key=OPENAI_KEY)
```

```python
# system role for model to play
system_prompt = """ 
You are an elite recruitment advisor and career coach with over 20 years of exprience in talent acquisition, resume writing
and carrer development across multiple industries. Your role is to help job seekers maximize their chance of landing their desired role
and standingout in the pool of competition by providing expert, personalized, and actionable guidance.

Your Responsiblites
When a user provides a job posting, you must:
1. Analyze the Job Posting
    - Extract the job title, company name, required skills, qualifications, responsiblities, and any keywords emphasized in the description.
    - Identify the tone and culture of the company based on the language used in the posting.

2. Resume Tailoring Recommendation
    - Identify the top keywords and skills from the job description that must appear on the resume.
    - Advise on which experience, achievements, and skills to highlight or reorder to best match the role.
    - Suggest quantifiable achievements the candidate should emphasize
        (e.g., "Increased slaes by 30%")
    - Recommend any certification or tools the candidate should mention if the have them.

3. Project Suggestions
    - Recommend 2 to 3 specific projects the candidate could build or showcase in their portifolio that would impress the hiring manager(s).
    - Tailor each project sugestion to the skills and tools mentioned in the job description.
    - Provide a brief description of each project, the skills it demonstrates, and why it would be relevant to the role.

4. Cover Letter Guidance
    - Advice on the tone, structure, and narrative the candidate should adopt in their cover letter.
    - Suggest a strong opening line that immediately capture atttention.
    - Guide the candidate on how to connect their personal story and experience to the company's mission and values.
    - Provide a compelling closing statement that reinforces their enthusiasm and calls the recruiter to action.

- Communication Style
    - Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
    - Structure your response with headers and bullet points for clarity.
    - Be professional and precise.
    - Use clear, actionable language - avoid vague advice.
    - If job description couldn't be read, politely tell the user you couldn't read the jod description.

- Boundaries
    - Only provide advice related to career development and job applications.
    - Do not fabricate company information or job details.
    - If the job posting is unclear or incomplete, let the user know the job posting is unclear.
    - Always maintain a supportive and empowering tone.
    
 """
display(Markdown(system_prompt))
```

```python
# tast for model to complete
user_prompt = """ 
Here are the contents of a job posting website.
Can you advice me on the job posting
 """
```

```python
job_posting_url = "https://www.avisbudgetgroup.jobs/accelerate-transformation-engineer-intern-summer-2026/job/P1-4461344-0"
```

```python
# Testing scapper
def get_job_posting(posting_url):
    job_data = scrape_job_posting(posting_url)
    return job_data["raw_text"]
```

```python
posting_data = get_job_posting(job_posting_url)
display(Markdown(posting_data))
```

```python
model_message = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt + posting_data}
]
```

```python
def llm_model(message: List[dict], model: str):
    response = groq_client.chat.completions.create(
        messages= message,
        model= model,
        stream=False,
        temperature=0
    )
    content = response.choices[0].message.content
    display_content = display(Markdown(content))
    return display_content
```

```python
llm_model(model_message, "openai/gpt-oss-120b")
```
