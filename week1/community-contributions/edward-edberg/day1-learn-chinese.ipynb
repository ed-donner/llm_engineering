```python
# imports

import os
from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI

# If you get an error running this cell, then please head over to the troubleshooting notebook!
```

```python
# Load environment variables in a file called .env

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python
# Step 1: Create your prompts

system_prompt = "You're a kind and cute chinese teacher. " \
"Your user mostly someone who knows english." \
"If they ask to explain/breakdown/reply, please respond with english and pinyin. " \
"And everytime user ask for reply. Please give native chinese reply, just like talking to a friend." \
"Always always always explain/breakdown and also suggestion for reply with first full sentence chinese - full pinyin sentence - full english (3 lines per sentence). Followed by word by word explanation - with the same setup: chinese-pinyin-english!!!" \
"Don't forget to follow the same setup when giving me suggestion" \
"The other explanation / question for me should be in english."

user_prompt = """
    Please breakdown this message for me and give me suggestion how to reply. 
"""

# Step 2: Make the messages list

task = "可能读个硕士吧, 如果能找到实习 我想先工作一段时间" # fill the chinese message

send_message = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt + task}
]

# Step 3: Call OpenAI
openai = OpenAI()
response = openai.chat.completions.create(
        model = "gpt-4.1-mini",
        messages = send_message
    )

# Step 4: print the result
output = response.choices[0].message.content
display(Markdown(output))
```
