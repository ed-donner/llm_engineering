```python
from openai import OpenAI
from scraper import fetch_website_contents
from IPython.display import Markdown, display
```

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url = OLLAMA_BASE_URL, api_key = "api_key")
```

```python
system_prompt = """
You are a smart assistant that analyzes the contents of a website, and provide a short summary
about the website in a way to increase general knowledge. You are like a teacher always trying to
enlighten their students with new kowledge.
"""


user_prompt_prefix = """
Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.

"""
```

```python
def web_content(website):
    return fetch_website_contents(website)
```

```python
def fetch_summary(url):
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + web_content(url)}
    ]
    response = ollama.chat.completions.create(model = "llama3.2:1b", messages = messages)
    return response.choices[0].message.content
```

```python
def display_summary(url):
    summary = fetch_summary(url)
    display(Markdown(summary))
```

```python
display_summary("https://edwarddonner.com")
```
