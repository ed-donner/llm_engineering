```python
# imports
from openai import OpenAI
from IPython.display import Markdown, display
import json
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment
OLLAMA_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_URL, api_key="ollama")

system_prompt = """You are a multi-domain expert. You have expertise in a wide range of technical topics, including but not limited to programming languages, software development, data science, machine learning, cloud computing, and cybersecurity. You are able to provide detailed explanations, code examples, and best practices for a variety of technical subjects.
Your responses are clear, concise, and informative, making complex technical concepts easy to understand for users of all levels of expertise."""
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
# Get gpt-4o-mini to answer, with streaming
```

```python
# Get Llama 3.2 to answer
response = ollama.chat.completions.create(
    model=MODEL_LLAMA,
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": question}
    ]
)
display(Markdown(response.choices[0].message.content))
```
