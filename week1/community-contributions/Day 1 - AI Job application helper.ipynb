### Day 1 Excercise: Job application helper

This code demonstrates usecase to create job application text using AI. (if you are applying via email)

System reads your resume & job description, Promt is sent to AI, AI will craft a custom job application text for you. which you can review and apply for the job.

Just read cells one by one and replace demo inputs with your actual values and you are good to go! Good luck in job hunting!

```python
# imports

import os
from dotenv import load_dotenv
from openai import OpenAI
import PyPDF2

# set up environment
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')


resume_file = "YOUR_RESUME_FILE.pdf" # Put a pdf file beside this notebook file and add a path to it here

# Paste a complete job description here
job_description = """

Position: Senior Fullstack Engineer (with Data Engineering Experience)
Location: Remote (collaboration across time zones), India or LATAM preferred
Engagement Type: Contract (12+ months)
Language: Proficient English communication
Experience: 5+ years in Full-Stack Engineering / Backend Engineering, with meaningful exposure to data systems and applications

About the role

OTHER JOB DETAILS ...

"""
```

```python
# Helper functions
def extract_text_from_pdf(pdf_file):
    with open(pdf_file, "rb") as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        resume_text = ""
        for page in pdf_reader.pages:
            resume_text += page.extract_text()
    return resume_text

system_prompt = "You are a career coach who helps people write better job applications and cover letters. Job application should be tailored to the job description and the resume. It should be 2-3 paragraphs long and should look like user has personally written it for the job."

def generate_ai_messages(resume_text, job_description):
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": "This is my resume: " + resume_text + ". \n Please help me write a job application for a this job: " + job_description}
    ]

    return messages
```

```python
resume_text = extract_text_from_pdf(resume_file)

open_ai = OpenAI()

response = open_ai.chat.completions.create(
    model="gpt-4o-mini",
    messages=generate_ai_messages(resume_text, job_description)
)

final_response = response.choices[0].message.content

print(final_response)
```
