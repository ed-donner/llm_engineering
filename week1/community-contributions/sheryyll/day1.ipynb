```python
import os
import sys
from pathlib import Path

from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI

sys.path.append(str(Path.cwd().parents[1]))
from scraper import fetch_website_contents

load_dotenv( override=True)

GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"
api_key = os.getenv("GOOGLE_API_KEY")

gemini = OpenAI(
    base_url=GEMINI_BASE_URL,
    api_key=api_key
)

system_prompt = """
You are Sherlock Holmes.

Analyze website content and present your findings as deductions.
Be observant, witty, and mildly sarcastic.
Ignore navigation menus, headers, footers, and boilerplate text.
Focus on what the website truly does and any important announcements.

Respond in markdown only.
"""

user_prompt = """
My dear Holmes,

The following text was recovered from a website.

Kindly investigate and provide your deductions regarding:
- The website's true purpose
- Its main offerings
- Any notable announcements

The evidence:
"""

url = "https://edwarddonner.com"
website = fetch_website_contents(url)

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": f"{user_prompt}\n\n{website}"}
]

response = gemini.chat.completions.create(
    model="gemini-2.5-flash",
    messages=messages
)

display(Markdown(response.choices[0].message.content))
```
