```python
import os
import json
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from scraper import fetch_website_links, fetch_website_contents
from openai import OpenAI
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="anything")
openai = OpenAI()
```

```python
# system prompt

system_prompt = """
You are an expert in the field of computer science and programming and you have the ability to explain the lines of code that are provided to you in a simple nature, without overcomplications.
Don't give examples unless necessary to understand the code, don't use big words without explaining them consecutively or beforehand,
for example, when explaining inheritance: a child of a parent class which has all the properties from the parent.

You should be able to break down line by line of code and the give the full picture of what the code snippet is doing.
"""
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
# Get gpt-4o-mini to answer, with streaming
def stream_response(question, provider, model):
    stream = provider.chat.completions.create(
        model=model,
        messages=[{"role": "system", "content":system_prompt},
        {"role": "system", "content":question}],
        stream=True
    )
    response=""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)
```

```python
# Get GPT to answer
stream_response(question, openai, MODEL_GPT)
```

```python
# Get ollama to answer
stream_response(question, ollama, MODEL_LLAMA)
```
