This code will fetch content of a portfolio websites and Summarize it in professional format

```python
import os
from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
def scrape(url):
    content = fetch_website_contents(url)
    return content

def messages_creation(website, user_prompt):
    system_prompt = "This model will create summary of a webpage with separate sections for About, Roles, Experience and external links, important points"
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt + website}
    ]
```

```python
def create_portfolio_summary(url):
    content = scrape(url)
    messages = messages_creation(content, "create the summary of this webpage")
    OLLAMA_BASE_URL = "http://localhost:11434/v1"
    ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
    response = ollama.chat.completions.create(model="llama3.2:1b", messages=messages)
    return response.choices[0].message.content
```

```python
print(create_portfolio_summary("https://edwarddonner.com"))
```
