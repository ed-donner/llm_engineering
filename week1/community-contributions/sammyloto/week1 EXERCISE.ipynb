To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports

import os
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment

load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')

# Validate the keys
if openai_api_key:
    print("OpenAI API Key exists and begins with '{openai_api_key[:8]}'")
else:
    print("OpenAI API Key not set")
```

```python
# here is the question; type over this to ask something new

question = """
Please explain why evolution favours certain traits over others. Especially in crustaceans.
"""
```

```python
# Get gpt-4o-mini to answer, with streaming

openai = OpenAI(api_key=openai_api_key)
openai_stream = openai.chat.completions.create(model=MODEL_GPT, messages=[{"role": "user", "content": question}], stream=True)
openai_response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in openai_stream:
    openai_response += chunk.choices[0].delta.content or ''
    update_display(Markdown(openai_response), display_id=display_handle.display_id)
```

```python
# Get Llama 3.2 to answer

client = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")

def ask_ollama(model: str, question: str) -> str:
    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": question}],
    )
    return response.choices[0].message.content

# usage
answer = ask_ollama(MODEL_LLAMA, question)
print(answer)
```
