To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
from openai import OpenAI
from dotenv import load_dotenv

import os
```

```python

# constants
MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
OLLAMA_BASE_URL = "http://localhost:11434/v1"
```

```python

# set up environment
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')
openai = OpenAI()

system_prompt = "You are coding tutor please explain what this code does to a programmer in short and brief explanation "
```

```python
# here is the question; type over this to ask something new
question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""

response = openai.chat.completions.create(
    model=MODEL_GPT,
    messages= [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content":  question},
    ]
)

print(response.choices[0].message.content)
```

```python
# Get gpt-4o-mini to answer, with streaming
response_stream = openai.chat.completions.create(
    model=MODEL_GPT,
    messages= [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content":  question},
    ],
    stream=True
)
full_response = ""

for chunk in response_stream:
    delta = chunk.choices[0].delta.content
    if delta:
        full_response += delta
        print(delta, end="", flush=True)
```

```python
# Get Llama 3.2 to answer

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")
response = ollama.chat.completions.create(
    model=MODEL_LLAMA,
    messages= [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": question},
    ]
)

print(response.choices[0].message.content)

```

```python
# Get Llama 3.2 to answer with streaming

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")
response_stream = ollama.chat.completions.create(
    model=MODEL_LLAMA,
    messages= [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": question},
    ],
    stream=True
)


full_response = ""

for chunk in response_stream:
    delta = chunk.choices[0].delta.content
    if delta:
        full_response += delta
        print(delta, end="", flush=True)
```
