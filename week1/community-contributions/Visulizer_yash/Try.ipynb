| Component | Purpose | Can be Changed? | Example |
|-----------|---------|-----------------|---------|
| **System Prompt** | Defines the AI's role, behavior, personality, and response style. | ✅ Yes (change the AI's persona or instructions) | `"You are The Professor. Explain programming concepts step by step."` |
| **User Prompt** | Contains the user's request or task for the AI to perform. | ✅ Yes (change the task) | `"Explain the following Python code."` |
| **Question** | The specific input or content the AI should work on. It is usually inserted into the user prompt. | ✅ Yes (change the code, concept, or question) | `"Please explain what this code does:\nyield from {book.get('author') for book in books if book.get('author')}"` |

````python
#message go ask ed what is this no markdown for this 
messages = [
 {"role": "system", "content": system_prompt},
 {"role": "user", "content": user_prompt}
    ]
````

````python
#steaming
stream = client.chat.completions.create(model=MODEL_GPT, messages=messages,stream=True)
    
response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    response = response.replace("```","").replace("markdown", "")
    update_display(Markdown(response), display_id=display_handle.display_id)
````

````python
# Get Llama 3.2 to answer

response = ollama.chat(model=MODEL_LLAMA, messages=messages)
reply = response['message']['content']
display(Markdown(reply))
````
