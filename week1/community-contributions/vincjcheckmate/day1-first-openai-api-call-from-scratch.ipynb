# First OpenAI API Call: Explaining a Security Concept in Simple Language

A Week 1 exercise from Ed Donner's AI Engineer Core Track.

This notebook builds the core structure of an OpenAI chat completion call from scratch:
a system prompt, a user prompt, a messages list, an API call, and a printed response.
No web scraping. No parsing. Just the fundamental pattern every LLM application is built on.

The model is asked to explain one security concept — why API keys should never be
committed to GitHub — in plain language a beginner can understand.

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
```

```python
load_dotenv(override=True)
api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    print("No API key found. Add OPENAI_API_KEY to your .env file.")
elif not api_key.startswith("sk-proj-"):
    print("Key found but doesn't start with sk-proj- — check you're using the right key.")
elif api_key.strip() != api_key:
    print("Key found but has leading or trailing whitespace — remove it from your .env file.")
else:
    print("API key found and looks good.")

openai = OpenAI()
```

```python
# Step 1: Create your prompts

system_prompt = "You are a helpful teacher who explains technical concepts in plain, beginner-friendly English. Keep your answers concise and clear."

user_prompt = "Can you explain why API keys should never be committed to GitHub, in simple language that a beginner can understand?"
```

```python
# Step 2: Make the messages list

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user",   "content": user_prompt}
]
```

```python
# Step 3: Call OpenAI

response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)
```

```python
# Step 4: Print the result

print(response.choices[0].message.content)
```

## What I learned

- The OpenAI API expects messages in a specific structure: a list of dictionaries,
  each with a `role` and `content` field
- The system prompt sets the model's behaviour privately — the user never sees it
- The user prompt is the actual question or request
- The API response is a structured object — the model's text lives at
  `response.choices[0].message.content`
- Python counts from zero, so `[0]` means the first item in a list
- This four-step pattern is the foundation of every LLM application
