```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
system_prompt = """
You are a mathematician. You are provided with mathematical question. Provide steps on how you solved the problem.
Solve it with utmost accuracy. Ask back question if you have a doubt.
"""
```

```python
def user_prompt(question):
    user_prompt = f"""
Solve the following question with 100% accuracy. """ + question 
    return user_prompt
```

```python
question = "5 divided by 10"
```

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
# Get a fun fact

response = ollama.chat.completions.create(model="llama3.2", messages=[{"role": "user", "content": "Tell me a fun fact"}])

response.choices[0].message.content
```

```python
def math_genius(question):
    response = ollama.chat.completions.create(
        model="llama3.2",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt(question)}
        ],
    )
    result = response.choices[0].message.content
    display(Markdown(result))
```

```python
math_genius(question)# Get Llama 3.2 to answer
```
