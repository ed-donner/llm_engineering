# Let's make a quick call to Groq to get started, as a preview!


````python
# To give you a preview -- calling Groq with these messages is this easy.

message = "Hello, Groq! This is my first ever message to you! Hi!"

messages = [{"role": "user", "content": message}]

messages
````

````python
client = Groq()  # Initialize the Groq client using the official groq SDK

response = client.chat.completions.create(
    model="llama-3.1-8b-instant",  # Specify the fast Llama-3.1 8B model hosted by Groq
    messages=messages  # Pass the messages list containing the user prompt
)  # Send the request to Groq's completions endpoint

response.choices[0].message.content  # Access the generated text directly from the first completion choice

# HOW THIS CODE WORKS:
# 1. client = Groq() initializes the Groq client using the official groq python SDK library.
# 2. client.chat.completions.create sends the payload to the completions endpoint to generate a text response.
# 3. model="llama-3.1-8b-instant" specifies the fast, lightweight text generation model.
# 4. messages takes a list of message dictionaries detailing the conversation history.
# 5. response.choices[0].message.content extracts the text of the first completed response choice returned by the API.
````

## Types of prompts

You may know this already - but if not, you will get very familiar with it!

Models like Groq have been trained to receive instructions in a particular way.

They expect to receive:

**A system prompt** that tells them what task they are performing and what tone they should use

**A user prompt** -- the conversation starter that they should reply to

````python
# Define our system prompt - you can experiment with this later, changing the last sentence to 'Respond in markdown in Spanish."

system_prompt = """
You are a very humiliating assitant, your job is to describe the website content in a humiliating and agonizing way. 
"""
````

````python
# Define our user prompt

user_prompt_prefix = """
Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.

"""


#  System prompt is the personality/instructions, user prompt is the specific request. This pattern is used everywhere in LLM engineering.
# The system prompt stays the same. It's like setting up the AI's personality once at the start.
# But: Each time you call the API, you send a fresh list of messages. It's not one continuous chat that remembers everything.
````

## Messages

The API from Groq expects to receive messages in a structure like:

```python
[
    {"role": "system", "content": "system message goes here"},
    {"role": "user", "content": "user message goes here"}
]
```
To give you a preview, the next 2 cells make a rather simple call.


````python
client = Groq()  # Initialize the Groq client

messages = [
    {"role": "system", "content": "You are a helpful assistant"},
    {"role": "user", "content": "What is 2 + 2?"}
]  # Define a messages list containing system and user prompts
 
response = client.chat.completions.create(
    model="llama-3.1-8b-instant",  # Specify the fast Llama-3.1 model
    messages=messages  # Pass the messages list directly to the completions endpoint
)  # Send the completions request to Groq
 
response.choices[0].message.content  # Extract the text answer from the response choices

# HOW THIS CODE WORKS:
# 1. client = Groq() initializes the Groq interface using the official groq python SDK package.
# 2. messages is configured as a list of dictionaries with 'role' (e.g. system or user) and 'content' (raw text string).
# 3. client.chat.completions.create takes the model configuration and the messages list to generate the completion response.
# 4. model="llama-3.1-8b-instant" configures the active LLM to generate the output content.
# 5. response.choices[0].message.content drills into the first choice in the response object to retrieve the text payload.
````

## And now let's build useful messages for Groq llama3-8b-8192, using a function


````python
# See how this function creates a Groq-compatible structure

def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + website}
    ]  # Return standard system and user messages in a list

# HOW THIS CODE WORKS:
# 1. def messages_for(website) defines a helper function to compile a structured request list for the webpage summary.
# 2. The function returns a standard flat list containing two distinct message dictionaries: system and user.
# 3. {"role": "system", "content": system_prompt} sets the persona/behavior of the assistant.
# 4. {"role": "user", "content": user_prompt_prefix + website} appends the prefix instructions to the scraped webpage text.
````

````python
# Try this out, and then try for a few more websites

messages_for(ed)
````

## Time to bring it together - the API for Groq is very simple!


````python
# And now: call the Groq API!

def summarize(url):
    website = fetch_website_contents(url)  # Fetch the contents of the website
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",  # Specify the Groq Llama model
        messages=messages_for(website)  # Pass the messages generated for this webpage
    )  # Request completion from Groq
    return response.choices[0].message.content  # Return the summarized text

# HOW THIS CODE WORKS:
# 1. def summarize(url) takes a URL, scrapes its contents using the fetch_website_contents helper, and generates a summary.
# 2. client.chat.completions.create takes the specified model and compiles the messages payload using messages_for.
# 3. The request is transmitted to Groq's low-latency text completion endpoint.
# 4. response.choices[0].message.content parses the API response payload to extract the text summary of the scraped webpage.
````

````python
summarize("https://edwarddonner.com")
````

````python
# A function to display this nicely in the output, using markdown

def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))
````

````python
display_summary("https://www.codewithharry.com/")
````
