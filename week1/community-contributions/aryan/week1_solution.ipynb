```python
# imports
import os
import json
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from scraper import fetch_website_links, fetch_website_contents
from openai import OpenAI
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# set up environment
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")

openai = OpenAI()    
```

```python
# here is the question; type over this to ask something new

question = """
You are a tech job assistant. A fresher to the field of tech is asking you for advice on how to get a job in tech in today's market as of 2026 May 24.
What will you advice the fresher to do to study and prepare for the job market in 2026?
Which skills will you advice the fresher to learn to get the job which would be in demand in 2026 and easy to learn?
"""
user_prompt = """ The fresher is an AIML Graduate and want a job in the next month. The fresher is from India and currently knows Python, SQL, Excel and PowerBI.
He has no work experience but is a quick learner and has a passion to learn and work in tech.
"""
```

```python
# Get gpt-4o-mini to answer, with streaming
def stream_brochure(question):
    stream = openai.chat.completions.create(
        model=MODEL_GPT,
        messages=[
            {"role": "system", "content": question},
            {"role": "user", "content": user_prompt}
          ],
        stream=True
    )    
    response = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)
```

```python
stream_brochure(question)
```
