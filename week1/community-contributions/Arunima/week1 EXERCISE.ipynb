To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# constants
from openai import OpenAI
OLLAMA_BASE_URL = "http://localhost:11434/v1"
```

```python
# set up environment
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="anything")
```

```python
concepts=[]
good_example=""
bad_example=""
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
"""
context="""
Utilise the given context to refine your answer. Provided includes examples of good answers, bad answers and user knowledge. If user already knows something don't include it in the explanation. List down at the end of your explanation the concepts already known by the user.
"""
```

```python
question+="""
n = 4
for i in range(0, n):
    print(i)
"""
```

```python
# Get Llama 3.2 to answer

response=ollama.chat.completions.create(
    model="llama3.2",
    messages=[
        {"role":"user",
         "content":question},
         {"role":"system",
          "content":context}
    ]
)

answer=response.choices[0].message.content

print(answer)
```

```python
response=ollama.chat.completions.create(
    model="llama3.2",
    messages=[
        {"role":"user",
         "content":f"What concept is being discussed in this code: {question}"}
    ]
)

concept=response.choices[0].message.content
concepts.append(concept)
```

```python
context+=f""""
Concepts known by user: {concepts}
"""
```

```python
good_example+=answer
```

```python
context+=f"""Heres a good exmaple: {good_example}"""
```
