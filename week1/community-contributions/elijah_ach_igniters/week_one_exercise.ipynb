```python
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
llm = OpenAI(base_url="http://127.0.0.1:11434/v1", api_key="")

system_prompt = "You are a help AI assistant. Your task is to answer user questions. Only respond in markdown format."
```

```python
def answer_question(query):
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": query}
    ]

    res = llm.chat.completions.create(model="gemma3:270m", messages=messages)

    return res.choices[0].message.content
```

```python
question = "Who is the richest man on earth"
answer = answer_question(question)
display(Markdown(answer))
```
