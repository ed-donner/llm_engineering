To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
from openai import OpenAI
from IPython.display import Markdown, display, update_display
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
OLLAMA_BASE_URL = 'http://localhost:11434/v1'
```

```python
# set up environment
open_ai = OpenAI()

ollama_ai = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")

```

```python
# here is the question; type over this to ask something new

# question = "How does the display function of IPython.display work?"

def stream_gpt_answer(question):
  stream = open_ai.chat.completions.create(
    model=MODEL_GPT,
    messages= [
      {"role": "system", "content": "You are an expert LLM engineer. You are given a technical question and you need to explain it in a way that is easy to comprehend. You will use relevant analogies and examples to help explain the concept. You will also use markdown formatting to make the explanation more readable. You will keep your response concise and succinct."},
      {"role": "user", "content": question}
      ],
    stream=True
    )

  response = ""
  display_handle = display(Markdown(""), display_id=True)
  for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    update_display(Markdown(response), display_id=display_handle.display_id)



def stream_llama_answer(question):
  stream = ollama_ai.chat.completions.create(
    model=MODEL_LLAMA,
    messages= [
      {"role": "system", "content": "You are an expert LLM engineer. You are given a technical question and you need to explain it in a way that is easy to comprehend. You will use relevant analogies and examples to help explain the concept."},
      {"role": "user", "content": question}
      ],
    stream=True,
  )

  response = ""
  display_handle = display(Markdown(""), display_id=True)
  for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    update_display(Markdown(response), display_id=display_handle.display_id)
```

```python
# Get gpt-4o-mini to answer, with streaming
stream_gpt_answer(question=input("Enter a question: "))
```

```python
# Get Llama 3.2 to answer
stream_llama_answer(question=input("Enter a question: "))
```
