This code notebook is all abiout creating an open API from scratch and summarizing a document sent for business and legal review

```python
#All Imports

import os
from openai import OpenAI
from IPython.display import Markdown, display
from dotenv import load_dotenv
import PyPDF2
```

```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')
if not api_key:
    print("OPENAI_API_KEY is not set")
elif not api_key.startswith('sk-proj-'):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key is set")
```

```python
def build_messages(document_text, system_prompt, user_prompt):
    return [
            {
                'role' : 'system',
                'content' : system_prompt
            },
            {
                'role' : 'user',
                'content' : user_prompt+document_text
            }
        ]
```

```python
#We will use PyPdf2 to read the entire content of the document

def read_pdf(file_path):
    full_txt = ""
    with open(file_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text = page.extract_text()
            if text:
                full_txt += text + "\n"
    return full_txt
```

```python
import tkinter as tk
from tkinter import filedialog

def get_file_from_user():
    # We create a tiny 'root' window and immediately hide it
    root = tk.Tk()
    root.withdraw()
    
    # Open the file explorer pop-up
    file_selected = filedialog.askopenfilename(
        title="Select a Document for the AI Agent",
        filetypes=[("PDF files", "*.pdf"), ("Text files", "*.txt"), ("All files", "*.*")]
    )
    
    return file_selected

# Usage
path = get_file_from_user()
if path:
    print(f"Agent is now reading: {path}")
else:
    print("User cancelled the selection.")
```

```python
def summarize_document(system_prompt, user_prompt):
    openai = OpenAI()    
    file_path = get_file_from_user()
    document_text = read_pdf(file_path)
    messages = build_messages(document_text, system_prompt, user_prompt)
    response = openai.chat.completions.create(
        model='gpt-4o-mini',
        messages=messages
    )
    display(Markdown(response.choices[0].message.content))
```

```python
system_prompt1 = """
You are a seasoned legal, HR and business review expert. Depending on the document type.
You will be given a document and you will need to review it and provide a concise professional summary of the document.
The summary should be in a way that is easy to understand and can be used by a business or legal team to review the document.
"""

user_prompt1 = """"
Here are the contents of the document.
Review the document and provide a concise professional summary of the document.
If there are any legal terms, important dates, or other key information, make sure to include that in the summary.
"""
```

```python
summarize_document(system_prompt1, user_prompt1)
```
