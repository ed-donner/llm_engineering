```python
#System Prompt

system_prompt1 = """
You are a sarcastic assistant that can summarize the content of a website in a sarcastic, humorous and concise manner.
You would do that by analyzing the content of the website. Ignoring the navigation links, headers if necessary.
You would also make sure that the summary is not too long and is easy to understand.
Respond in markdown format. Do not wrap the response in a code block - just the entire markdown response.
"""

#User Prompt

user_prompt1 = """
Here are the contents of the website.
Provide a short summary of the website.
If it includes news or announcements, make sure to include that in the summary.
"""
```

```python
def messages_for(website_url, system_prompt, user_prompt):
    return [
        {
            'role' : 'system',
            'content' : system_prompt
        },
        {
            'role' : 'user',
            'content' : user_prompt + website_url
        }
        ]
```

```python
messages_for('https://medium.com/@jigarjoshi7/why-your-data-is-probably-lying-to-you-and-how-not-to-kill-anyone-0e6091444d0c', system_prompt1, user_prompt1)
```

```python
def summarize_website(url,system_prompt,user_prompt):
    website_contents = fetch_website_contents(url)
    openai = OpenAI()
    response = openai.chat.completions.create(
        model='gpt-4o-mini',
        messages=messages_for(url, system_prompt, user_prompt)
    )
    return response.choices[0].message.content
```

```python
def display_summary(url,system_prompt,user_prompt):
    summary = summarize_website(url,system_prompt,user_prompt)
    display(Markdown(summary))
```

```python
display_summary('https://medium.com/@jigarjoshi7/why-your-data-is-probably-lying-to-you-and-how-not-to-kill-anyone-0e6091444d0c',system_prompt1,user_prompt1)
```
