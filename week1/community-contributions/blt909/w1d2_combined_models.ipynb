# SEO optimizer

| First through OSS models then rating and improval with frontier models

## Load script

```python
import w1d2_optimizer
from IPython.display import Markdown, display
```

## Get data from a page

```python
optimizer = w1d2_optimizer.Optimizer()
original_content = optimizer.scrape(url="https://fr.wikipedia.org/wiki/Mod%C3%A8le_de_fondation")

original_content
```

## Get optimized content from OSS models

```python
llama_content = optimizer.optimize_content(content=original_content, model="llama3.2")

print("llama3.2 response:")
display(Markdown(llama_content))
```

```python
deepseek_content = optimizer.optimize_content(content=original_content, model="deepseek-r1:1.5b")

print("deepseek-r1:1.5b response:")
display(Markdown(deepseek_content))
```

## Compare results using frontier models and improve if necessary

```python
evaluation = optimizer.evaluate_content(model="gpt-5-nano",original_content=original_content, llama3_content=llama_content, deepseek_content=deepseek_content)

print("GPT response:")
display(Markdown(evaluation))
```

```python
evaluation = optimizer.evaluate_content(model="gemini-2.5-flash-lite", original_content=original_content, llama3_content=llama_content, deepseek_content=deepseek_content)

print("Gemini response:")
display(Markdown(evaluation))
```
