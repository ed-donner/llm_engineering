## Exercise Coach who finds open swim hours with Selenium


The goal of this contribution is to have the bot scrape a local pool hours site (https://recreation.gocrimson.com/sports/2021/5/14/facility-hours.aspx) and return a schedule for the week for marathon training. 

### Imports!

```python
import os 
from dotenv import load_dotenv
from scrape_pool_times import scrape_pool_hours
from openai import OpenAI
```

### Grab Key

```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

### Run Scraper file

```python
swimming_hours = scrape_pool_hours()
print(swimming_hours)
print(type(swimming_hours))
```

### Feed scraper file into LLM and get workout plan! 

```python
openai = OpenAI()

system_prompt = """You're a dedicated workout coach who's goal is to help me train for Marathons
You're given a list of open swim times for the week. 
Your job is to create a workout plan according to the user's needs. 
The pool is 4 miles away and there's a gym there, at the office, and at home.
Give a brief schedule for each week with 1 rest day and summarize the goal of the week in a sentence.
"""

user_prompt="My buffalo marathon is in 4 months. I work 9-5 and want to swim twice a week and run ~40mpw, what's the plan? Here's the schedule of the pool"

messages = [
       {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt + swimming_hours}
]

response = openai.chat.completions.create(model="gpt-4.1-nano", messages=messages)
response.choices[0].message.content
```
