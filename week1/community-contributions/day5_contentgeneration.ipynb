## Create a Marketting Content for my my digital e commerce website - https://inkncotton.printify.me/

Model=gpt-4.1-mini

```python
# Imports
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from scraper import fetch_website_contents
from openai import OpenAI
```

```python
# Load Environments

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

if api_key and api_key.startswith('sk-proj-') and len(api_key)>10:
    print("API key looks good so far")
else:
    print("There might be a problem with your API key? Please visit the troubleshooting notebook!")
    
```

```python
# Prompts

system_prompt = """
You are a helpful assistant who can generate a Marketing Content for a website provided by the user. You will take a look at the products sold in the website
and create a comprehensive Marketing Content for it. You can ignor the Privacy Policy and Terms and conditions. 
The manual would be in Markdown and should not contain any code block."""

user_prompt = """
    Please create a marketing Content for my website.
"""
```

```python
# Define the Content Creation Function

openai = OpenAI()

def create_marketting_content(url):
 contents = fetch_website_contents(url)
 response = openai.chat.completions.create(
        model="gpt-4.1-mini",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt + contents}
        ],
  
 )
    
 result=response.choices[0].message.content       
 display(Markdown(result))
```

```python
create_marketting_content("https://inkncotton.printify.me/")
```
