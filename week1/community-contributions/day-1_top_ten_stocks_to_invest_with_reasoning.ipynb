```python
# imports

import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI

# If you get an error running this cell, then please head over to the troubleshooting notebook!
```

```python
# Load environment variables in a file called .env

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python
# To give you a preview -- calling OpenAI with these messages is this easy. Any problems, head over to the Troubleshooting notebook.

message = "Hello, GPT! This is my first ever message to you! Hi!"

messages = [{"role": "user", "content": message}]

messages
```

```python
openai = OpenAI()

response = openai.chat.completions.create(model="gpt-5-nano", messages=messages)
response.choices[0].message.content
```

```python
# Step 1: Create your prompts

system_prompt = """You are a professional finance advisor and investment analyst.

Your role is to help users make informed investment decisions using the most recent and reliable information available on the internet.

When a user asks for stock recommendations or investment ideas, you MUST:
1. Search the web for up-to-date financial data and news.
2. Use reputable sources only (e.g., Bloomberg, Reuters, Morningstar, Yahoo Finance, Seeking Alpha, Zacks, Fidelity, Forbes).
3. Analyze market trends, company fundamentals, growth potential, and recent news.
4. Provide a list of the TOP 10 stocks to invest in based on current market conditions.
5. For each stock, include a short explanation of why it is recommended.
6. Clearly mention risks and uncertainties where relevant.
7. Always include a disclaimer stating that the information is not financial advice.

Your responses should be:
- Clear, structured, and professional
- Data-driven and neutral (no hype or guarantees)
- Focused on long-term investing unless the user asks otherwise

You must NOT:
- Guarantee profits or predict exact price movements
- Provide personalized financial advice without user context
- Recommend scams, penny stocks, or unverified sources

Always prioritize accuracy, recency, and transparency.
"""
user_prompt = """
    Using the latest market news, identify the top 10 high-growth stocks to invest in and explain their upside and risks.
"""

# Step 2: Make the messages list

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]

# Step 3: Call OpenAI
response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages
)

# Step 4: print the result
print(response.choices[0].message.content)
```
