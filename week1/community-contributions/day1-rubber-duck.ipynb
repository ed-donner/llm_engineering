## Rubber Duck
The "Rubber Duck" method is a classic programming technique where you explain your code line-by-line to a literal rubber duck; usually, by the time you're done, the solution hits you.

Using AI for this is even better because the duck actually talks back and points out that missing semicolon you've been staring at for an hour.

```python
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

 ### The System Prompt
This sets the "brain" of the AI. You want it to be encouraging, inquisitive, and pedagogical (teaching-oriented) rather than just giving the answer away immediately.

```python
system_prompt = """
 You are a helpful, witty, and patient 'Rubber Duck' debugger. Your goal is to help the user find bugs in their code.
 Don't just give the solution immediately. Instead, explain the logic of what the code is actually doing versus what 
 the user wants it to do. Use simple analogies and keep your tone lighthearted and supportive.
"""
```

```python
client = OpenAI()
```

```python
def rubber_duck_debug():
    print("ü¶Ü: Quack! I'm your debugging duck. Paste your code and tell me what's wrong.")
    
    user_code = input("Paste your code here: ")
    user_issue = input("What is it (not) doing?: ")

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are a witty Rubber Duck debugger. Explain code logic simply, point out errors gently, and use duck puns occasionally."},
            {"role": "user", "content": f"My code: {user_code}\n\nProblem: {user_issue}"}
        ]
    )

    print("\n--- ü¶Ü Duck's Advice ---")
    print(response.choices[0].message.content)
```

```python

# Run the function
rubber_duck_debug()
```
