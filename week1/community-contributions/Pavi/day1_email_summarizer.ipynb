```python
# imports

import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
# using anothropic (claude)
ANTHROPIC_BASE_URL = "https://api.anthropic.com/v1/"
CLAUDE_MODEL = "claude-haiku-4-5"
anthropic_api_key = os.getenv("ANTHROPIC_API_KEY")
claude = OpenAI(base_url=ANTHROPIC_BASE_URL, api_key=anthropic_api_key)
# Step 1: Create your prompts

system_prompt = """You are an AI assistant designed to analyze internal business emails.
Your task is to:
1. Summarize the provided email content in one concise sentence (maximum 20 words).
2. Generate a professional and clear subject line for the email."""
user_prompt = """
    Hi team,
    I’m reaching out to share an update on the Project Apollo timeline. Due to a delay in receiving the API documentation from the vendor, we are pushing the UAT phase back by three days.
    The new UAT start date is now Thursday, October 26th.
    I have updated the master project schedule in SharePoint. Please ensure your respective teams are aware of this shift to avoid scheduling conflicts. Let me know if this creates a critical blocker for any of your deliverables.
    Best regards,
    Sarah J.
    Project Manager
"""

# Step 2: Make the messages list

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
] # fill this in

# Step 3: Call OpenAI (using claude)
response = claude.chat.completions.create(
    model = CLAUDE_MODEL,
    messages = messages
)

# Step 4: print the result
print(response.choices[0].message.content)
```
