Project: AI English → Any Language Converter

Goal:
Input English text + target language → AI outputs translated text.

```python
# import import libraries

import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI

# If you get an error running this cell, then please head over to the troubleshooting notebook!
```

```python
# Load environment variables in a file called .env and load the api key

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python
# The system prompt defines behavior of the AI.
system_prompt = """
You are a professional translator.

Rules:
1. The user will provide English text.
2. Translate the text into the requested language.
3. Preserve meaning and tone.
4. Only return the translated text.
5. Do not include explanations.
"""
```

```python
# User Prompt
text = "Artificial Intelligence will transform the future of work."
language = "Hindi"

user_prompt = f"""
Translate the following English text into {language}.

Text:
{text}
"""
```

```python
# OpenAI API Call
openai = OpenAI()

response = openai.chat.completions.create(
    model="gpt-4.1",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ]
)

print(response.choices[0].message.content)
```

```python
# Turn It Into a Function

def translate_text(text, language):
    
    system_prompt = """
    You are a professional translator.
    Translate English text into the requested language.
    Only return the translated text.
    """
    
    user_prompt = f"""
    Translate this text into {language}:

    {text}
    """
    
    response = openai.chat.completions.create(
        model="gpt-4.1",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    )

    return response.choices[0].message.content
```

```python
# Test Multiple Languages

text = "Machine learning is transforming businesses."

print("Hindi:", translate_text(text, "Hindi"))
print("French:", translate_text(text, "French"))
print("Japanese:", translate_text(text, "Japanese"))
print("German:", translate_text(text, "German"))
print("Korean:", translate_text(text, "Korean"))
```
