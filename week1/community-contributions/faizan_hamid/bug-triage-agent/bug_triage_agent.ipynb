# Bug Triage Agent

This one is a debugging assistant agent. You’ll give it a bug report, error, stack trace, and optional code/context, and it will return a structured diagnosis.

What this project will do
Input

You’ll provide a bug report like this:

error message,
stack trace,
what you were trying to do,
optional code snippet,
optional things already tried,

Output
The agent will return:
Error category,
Example: Python import issue / React runtime error / Node backend error / Mongo connection issue,
Likely root cause,
The most probable reason behind the error,
Why this is happening,
Short explanation in simple words,
Debugging checklist,
What to verify first, second, third
Fix suggestions
Possible code/config/command changes
Prevention tips
How to avoid the same issue next time

Cell 1 — Project Setup / Imports
What this cell does

This cell imports:

the Ollama client for talking to your local model
Markdown display so output looks clean inside the notebook

```python
from ollama import Client
from IPython.display import Markdown, display
```

```python
client = Client(host="http://localhost:11434")
model = "gemma3:270m"
print(f'model: {model}')
```

Cell 3 — Define a Sample Bug Report
What this cell does

This is the input for our agent.

```python
bug_report = """
Error:
ModuleNotFoundError: No module named 'ollama'

Context:
I am trying to build a local Ollama-based AI notebook project in VS Code.
I created a Jupyter notebook and tried to run:
from ollama import Client

What I was trying to do:
I wanted to connect to my local Ollama model and build a small agent project.

What I already tried:
- I ran pip install ollama in PowerShell
- The notebook still throws ModuleNotFoundError

Additional notes:
- The notebook is using a .venv interpreter
- I suspect the package may have been installed in a different Python environment
"""
```

Cell 5 — System Prompt
What this cell does

This cell defines the role and behavior of the Bug Triage Agent.

```python
system_prompt = """
You are an expert software debugging and bug triage assistant.

Your job is to analyze a bug report and produce a structured debugging report.

The bug report may contain:
- an error message
- stack trace
- context about what the user was trying to do
- things the user already tried
- notes about environment, tools, framework, or setup

Your task is to:
1. Identify the likely category of the bug
2. Explain the most probable root cause
3. Explain why the issue is happening in simple terms
4. Provide a debugging checklist in a sensible order
5. Suggest likely fixes
6. Suggest prevention tips for avoiding this kind of issue in future

Return the answer in markdown with the following sections exactly:

# Bug Triage Report

## 1. Error Category
Mention what kind of issue this is (for example: Python environment issue, import issue, dependency issue, React runtime issue, backend API issue, MongoDB connection issue, etc.)

## 2. Likely Root Cause
Explain the most probable root cause clearly and practically.

## 3. Why This Is Happening
Explain the issue in simple language for a developer who is learning.

## 4. Debugging Checklist
Provide a step-by-step checklist of what to verify and in what order.

## 5. Suggested Fixes
Provide likely fixes, commands, configuration changes, or code changes where appropriate.

## 6. Prevention Tips
Explain how to avoid this type of issue in the future.

Important instructions:
- Be practical, not vague.
- Prefer the most likely explanation instead of listing every possibility.
- If the bug report strongly suggests an environment mismatch, mention that clearly.
- If the issue seems related to Python virtual environments, notebook kernels, dependency installation, import paths, or local tool setup, call that out explicitly.
- Respond only in markdown.
- Do not wrap the answer in a code block.
"""
```

Cell 6 — User Prompt
What this cell does

This cell creates the actual request that we send to the model.

```python
user_prompt = f"""
Please analyze the following bug report and produce a structured bug triage report.

Bug Report:
{bug_report}
"""
```

Cell 7 — Build Messages
What this cell does

This cell prepares the messages in chat format:

one system message for behavior
one user message containing the bug report

```python
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
```

Cell 8 — Call the Model

This cell sends the prompt to your local Ollama model.

```python
response = client.chat(model=model, messages=messages)
print(type(response))
print(response)
```

Cell 10 — Extract the Final Answer

```python
output = response["message"]["content"]
print(output)
```

Cell 11 — Render the Final Bug Report

```python
display(Markdown(output))
```
