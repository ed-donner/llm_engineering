```python
# import
import json
from openai import OpenAI
from IPython.display import Markdown, display, update_display
from content import fetch_website_links, fetch_website_contents;
```

```python

website = "https://edwarddonner.com"
links = fetch_website_links(website)
links
# display(Markdown(links))
```

```python
link_system_prompt = """
You are provided with a list of links found on a webpage.
You are able to decide which of the links would be most relevant to include in a brochure about the company,
such as links to an About page, or a Company page, or Careers/Jobs pages.
You should respond in JSON as in this example:

{
    "links": [
        {"type": "about page", "url": "https://full.url/goes/here/about"},
        {"type": "careers page", "url": "https://another.full.url/careers"}
    ]
}
"""
```

```python
# this is hard coded user prompt 

# link_user_prompt = f"""
# Here is the list of links on the website {"https://edwarddonner.com"} -
# Please decide which of these are relevant web links for a brochure about the company, 
# respond with the full https URL in JSON format.
# Do not include Terms of Service, Privacy, email links.

# Links (some might be relative links):
# {links}
# """
# link_user_prompt

# this is dynamic user prompt 

def user_prompt_with_links(website_url):
    user_prompt = f""" Here is the list of links on the website {website_url} -
Please decide which of these are relevant web links for a brochure about the company, 
respond with the full https URL in JSON format.
Do not include Terms of Service, Privacy, email links.

Links (some might be relative links)"""
# again fetching links 
    links = fetch_website_links(website_url)
    user_prompt += "\n".join(links)
    return user_prompt
```

```python
user_prompt_with_links(website)
```

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL,api_key='ollama')
MODEL="llama3.2:1b"
```

```python
messages= [
    {'role':'system','content':'link_system_prompt'},
    {'role':'user','content':user_prompt_with_links(website)},
    
]
```

```python
response = ollama.chat.completions.create(model=MODEL,messages=messages,response_format={"type": "json_object"})
result = response.choices[0].message.content
Related_link= json.loads(result)
print(Related_link)
```

Create the Brochure


```python
brochure_system_prompt = """
You are an assistant that analyzes the contents of several relevant pages from a company website
and creates a short brochure about the company for prospective customers, investors and recruits.
Respond in markdown without code blocks.
Include details of company culture, customers and careers/jobs if you have the information.
"""
```

```python

def brochure_user_prompt(company_website):
    user_prompt=f"""
    You are looking at a company called: {company_website}
    Here are the contents of its landing page and other relevant pages;
    use this information to build a short brochure of the company in markdown without code blocks.\n\n
     """
    user_prompt += str(Related_link)
    # user_prompt += user_prompt[:5_000] # Truncate if more than 5,000 characters
    return user_prompt
```

```python
brochure_user_prompt("https://edwarddonner.com")
```

```python
messages=[
    {'role':'system','content':brochure_system_prompt},
    {'role':'system','content':brochure_user_prompt(website)}
    ]
```

```python
#normal way
response = ollama.chat.completions.create(model=MODEL,messages=messages)
result = response.choices[0].message.content
display(Markdown(result))
```

```python
#different output way
stream = ollama.chat.completions.create(model=MODEL,messages=messages,stream=True)
response=""
display_different = display(Markdown(""),display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    update_display(Markdown(response),display_id=display_different.display_id)
```
