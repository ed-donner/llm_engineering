# Resume Review Agent

This is a very practical one for you.

What it does
You give it your resume text / PDF
It reviews it like a recruiter
Returns:
summary of profile
weak bullet points
missing keywords
ATS improvement suggestions
improved intro / headline

Agent flow

resume pdf -> extract text -> send to model -> get structured review

Prompt example

Ask it to return:

Professional summary
Top 5 strengths
Top 5 improvements
ATS keyword suggestions
Better project bullet points

Cell 1 — Imports

```python
from ollama import Client
from IPython.display import Markdown, display
from openai import OpenAI
import pdf_extractor
```

Cell 2 — Create local Ollama client

```python
# client = Client(host=""http://localhost:11434"")
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL,api_key='ollama')
model = "llama3.2:1b"

print(f"Using model: {model}")
```

Cell 3 — Extract resume text

```python
resume_text = pdf_extractor.extract_text_from_pdf("sample_resume.pdf")
print(resume_text[:2000])
```

Cell 4 — System prompt

```python
system_prompt = """
You are an expert resume reviewer, recruiter, and ATS evaluator.

Your task is to review the candidate's resume and provide a structured, practical analysis.

Analyze the resume and return the response in markdown with these sections:

# Resume Review

## 1. Candidate Profile Summary
Give a 4-6 line summary of the candidate's profile, experience, strengths, and likely target roles.

## 2. Key Strengths
List the strongest aspects of the resume in bullet points.

## 3. Weak Areas / Gaps
List unclear, weak, or missing parts in the resume.

## 4. ATS & Keyword Suggestions
Suggest missing keywords, tools, technologies, or phrases that could improve ATS relevance.

## 5. Bullet Improvement Suggestions
Pick 3-5 resume bullet points that could be rewritten better and suggest improved versions.

## 6. Recruiter Verdict
Give an honest recruiter-style verdict:
- what level this candidate seems to be
- what kind of roles they fit
- whether the resume is interview-ready or needs work

Important instructions:
- Be practical, specific, and direct.
- Do not invent fake experience.
- If some section is missing, clearly say so.
- Focus on improving the resume for software engineering / frontend / full-stack roles.
- Respond only in markdown, not in a code block.
"""
```

Cell 5 — User prompt

```python
user_prompt = f"""
Here is the extracted text from the candidate's resume.

Review this resume thoroughly and provide a structured resume review.

Resume Text:
{resume_text}
"""
```

Cell 6 — Messages

```python
messages = [
    {'role':'system', 'content':system_prompt},
    {'role':'user', 'content':user_prompt}
]
```

Cell 7 — Call local Ollama model

```python
# response = client.chat(model=model, messages=messages)
response = ollama.chat.completions.create(model=model,messages=messages)
```

Cell 8 — Display output

```python
# output = response['message']['content']
output = response.choices[0].message.content
display(Markdown(output))
```
