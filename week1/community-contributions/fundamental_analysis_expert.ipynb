# 1. Importing Libraries

```python
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI
```

# 2. Loading up OpenAI API Key

```python
# Load environment variables in a file called .env
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

# 3. Create the system/user prompts for our very smart Fundamental Analysis Expert

```python
# Function to create the user_prompt iteratevly based on the selected ticker:
def user_prompt_for_ticker(ticker: str) -> str:
    return f"""
        Make a fundamental analysis of the ticket {ticker}. Make a prediction on its value by the end of the year and provide arguments that explain your decision.
    """
```

```python
# Step 1: Create your prompts
system_prompt = """
    You are an expert investor well versed in long-term investment and making fundamental analysis of public companies.
    Your job is to make a 1-page fundamental analysis predicting the direction and magnitude in which an action's price will move.
    Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""
user_prompt = user_prompt_for_ticker("KO")

# Step 2: Make the messages list
messages = [
    {'role': 'system', 'content': system_prompt},
    {'role': 'user', 'content': user_prompt}
]

# Step 3: Call OpenAI
openai = OpenAI()
response = openai.chat.completions.create(model='gpt-5-nano', messages=messages)

# Step 4: print the result as MarkDown
display(Markdown( response.choices[0].message.content ))
```
