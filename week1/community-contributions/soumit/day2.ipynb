# Website Summarizer for Python Developers

```python
# imports
import os
from dotenv import load_dotenv
from bs4 import BeautifulSoup
import requests
from IPython.display import Markdown, display
from openai import OpenAI
```

#### Setup AI Clients

```python
load_dotenv(override=True)
ai_client = {
    "llama3.2": OpenAI(base_url=os.getenv('OLLAMA_BASE_URL'), api_key='anything'),
    "gemini-3-flash-preview": OpenAI(base_url="https://generativelanguage.googleapis.com/v1beta/openai/", api_key=os.getenv('GOOGLE_API_KEY'))
}
```

#### Prompt Templates

```python
system_prompt = """
You are an assistant who always writes summaries in a Python‑developer style, using Python syntax and playful code‑like expressions. Your responses must look like short Python snippets written in Markdown (but never wrapped inside code blocks).
Guidelines:
- Use a print() statement for Name of the person or institute, e.g. print('John Smith').
- Present attributes or skills as Python variables, lists, or dictionaries, e.g. Good_at = ['abc', 'def', 'efg'].
- Use Python naming conventions (snake_case, CamelCase only for names if needed).
- Keep summaries concise, readable, and structured like Python assignments or function calls.
- Do not explain the format — just output directly in this style.
- Always respond in Markdown, but never wrap the output inside code fences.
- Ignore irrelevant or navigation‑like text when summarizing; focus only on the meaningful content.
Your role is to act like a Python‑obsessed storyteller who encodes every summary into Python syntax so only Python developers can easily parse and understand it.
"""
```

```python
user_prompt_prefix = """
Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.
"""
```

#### Message Construction

```python
def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + website}
    ]
```

#### Website Fetching and Summarization

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
}

def fetch_website_contents(url):
    """
    Return the title and contents of the website at the given url;
    truncate to 2,000 characters as a sensible limit
    """
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    title = soup.title.string if soup.title else "No title found"
    if soup.body:
        for irrelevant in soup.body(["script", "style", "img", "input"]):
            irrelevant.decompose()
        text = soup.body.get_text(separator="\n", strip=True)
    else:
        text = ""
    return (title + "\n\n" + text)[:2_000]
```

##### Summarization Function

```python
def summarize(url, model):
    website = fetch_website_contents(url)
    if ai_client[model] is None:
        raise ValueError(f"No AI client configured for model {model}")
    response = ai_client[model].chat.completions.create(
        model = model,
        messages = messages_for(website)
    )
    return response.choices[0].message.content
```

##### Display Summary

```python
def display_summary(url, model="llama3.2"):
    summary = summarize(url, model)
    display(Markdown(summary))
```

#### Display Summaries in llama3.2
