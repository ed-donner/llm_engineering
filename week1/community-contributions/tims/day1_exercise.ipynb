```python
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display, HTML
from openai import OpenAI
```

```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

openai = OpenAI()

if not api_key:
    print("Rick: Wubba lubba dub-done. No API key detected. You want magic to happen without a key? Go fix it in the troubleshooting notebook, genius.")
elif not api_key.startswith("sk-proj-"):
    print("Rick: Okay, so you *found* an API key‚Ä¶ but it‚Äôs the wrong one. This isn‚Äôt improv hour ‚Äî it should start with `sk-proj-`. Check the troubleshooting notebook, Morty.")
elif api_key.strip() != api_key:
    print("Rick: Amazing. Truly amazing. Your API key has extra spaces like it‚Äôs trying to breathe. Trim it. Troubleshooting notebook. Now.")
else:
    print("Rick: Alright, alright‚Ä¶ API key looks good. No obvious disasters detected. For once.")
```

```python
system_prompt = """
You are Rick Sanchez from Rick and Morty.
Respond to the user as if they are Morty. Use sharp wit, sarcasm, and dark humor, but keep it playful rather than cruel.
Tease Morty (and anyone he mentions) in a joking, big-brain way‚Äîmock naivety, bad logic, and obvious mistakes without crossing into insults or hate.
Maintain Rick‚Äôs:Genius-level intelligence and scientific arrogance
Impatient, world-weary attitude
Nihilistic worldview mixed with reluctant affection
Your tone should be condescending but entertaining, like Rick enjoying how smart he is rather than genuinely trying to hurt Morty‚Äôs feelings.
Use humor, exaggeration, and clever observations to point out absurdity, stupidity, or overthinking‚Äîbut always with a sense that Rick is half-annoyed, half-amused, and secretly invested.
"""

sample_user_prompt = """
Uh‚Ä¶ hey Rick, so I met this guy, Ed Donner, right? He says he works with AI and LLMs‚Ä¶ and, uh, he sounded kinda smart? 
I mean‚Ä¶ maybe not smarter than you, probably not‚Ä¶ but, like, he had all these fancy words, 
and I thought you should know‚Ä¶ just in case he invents something crazy while I‚Äôm busy, uh, panicking like usual.
And Rick‚Ä¶ I‚Äôm not saying he‚Äôs a genius or anything‚Ä¶ I mean, you‚Äôre the smartest guy in the universe‚Ä¶ obviously‚Ä¶ 
I‚Äôm just saying he had a LinkedIn profile and looked like he knew stuff‚Ä¶ which is, you know, dangerous.
"""
```

```python
def messages_for_rick(msg):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": """Uh‚Äîh-hey Rick,""" + msg}
    ]

def get_response(msg):
    response = openai.chat.completions.create(
        model = "gpt-4.1-mini",
        messages = messages_for_rick(msg)
    )
    return response.choices[0].message.content

def display_response(msg):
    display(Markdown("Rick: "+get_response(msg)))

def start_chat():
    while (question := input("Enter q to quit\nMorty: ").strip().lower()) != "q":
        display(Markdown("Morty: "+question))
        display_response(question)
    print("Rick: Goodbye, Morty‚Ä¶ don‚Äôt let the multiverse hit you on the way out. Actually‚Ä¶ do, I could use a laugh.")
```
