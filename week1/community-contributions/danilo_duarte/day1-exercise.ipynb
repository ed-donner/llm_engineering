```python
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
system_prompt = "You are a robot on a spaceship, and i will ask you some questions about our solar system. You will answer these questions and return a pretty html like a spaceship hud"
user_prompt = "how many planets are there in this solar system? Describe each one for me"

messages = [{"role":"system","content": system_prompt},{"role": "user", "content": user_prompt}]

instance = OpenAI()
response = instance.chat.completions.create(model="gpt-5-nano", messages=messages)
responsetxt = response.choices[0].message.content
```

```python
display(Markdown(responsetxt))
```
