```python
import os
import requests
from openai import OpenAI
from bs4 import BeautifulSoup
```

```python
#make a class for parsing web pages
class AnyWebPage:
    
    def __init__(self, url):
        self.url = url
        
        #Generic headers widely compatible
        self.headers = {
        "User-Agent": "Mozilla/5.0",
        "Accept": "*/*",
        "Connection": "keep-alive",
        }
        self._html = None
        self._soup = None
        
    def get_header(self):
        return self.headers
    
    def get_html(self):
        if self._html is None:
            #construct the request 
            req = requests.Request("GET", self.url, headers=self.headers)
            
            #prepare the request
            prepared_req = req.prepare()
            
            #verify the full URL
            print(f"Verifying URL: {prepared_req.url}\n..")
            
            #create an Session 
            with requests.Session() as session:
                #send the request
                response = session.send(prepared_req)
            
                #check for error
                response.raise_for_status()
                self._html = response.text
            
        #always return
        return self._html
    
    #define a parser
    def get_soup(self):
        if self._soup is None:
            self._soup = BeautifulSoup(self.get_html(), 'html.parser')
        return self._soup
    
    def get_title(self):
        title_tag = self.get_soup().find("title")
        return title_tag.get_text(strip=True) if title_tag else None
    
    def get_body(self):
        body_tag = self.get_soup().find("body")
        return body_tag.get_text(strip=True) if body_tag else None
```

```python
page = AnyWebPage("https://edwarddonner.com")
print(page.get_title())
print(page.get_body())
```

```python
def system_prompt(self, language):
    return (
        f"You are an assistant that analyzes the content of a webpage and provides "
        f"a concise summary. Ignore navigation elements such as menus, headers, and footers. "
        f"Respond in {language}."
    )
```

```python
def user_prompt(self, webpage_text):
    return (
        "Here is the content of a webpage:\n\n"
        f"{webpage_text}\n\n"
        "Please provide a concise summary of the main content."
    )
```
