and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import display, clear_output, Markdown
import sys
```

```python
load_dotenv(override=True)
OPENROUTER_API_KEY = os.getenv('OPENROUTER_API_KEY')

client = OpenAI(base_url="https://openrouter.ai/api/v1", api_key=OPENROUTER_API_KEY)
```

```python
def explain_stream_pretty(question, model="openai/gpt-4o-mini", level="intermediate"):
    
    stream = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": f"Explain clearly at {level} level."},
            {"role": "user", "content": question}
        ],
        stream=True
    )
    
    full_text = ""
    
    for chunk in stream:
        if chunk.choices and chunk.choices[0].delta.content:
            full_text += chunk.choices[0].delta.content
            clear_output(wait=True)
            display(Markdown(full_text))
    
    return full_text
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
# Get gpt-4o-mini to answer, with streaming
explain_stream_pretty(question=question)
```

```python
# Get Llama 3.2 to answer
explain_stream_pretty(question=question, model='meta-llama/llama-3.2-3b-instruct')
```
