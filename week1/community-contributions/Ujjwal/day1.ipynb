```python
import os
from openai import OpenAI
from dotenv import load_dotenv

openai = OpenAI()
```

```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key: display("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python
# Step 1: Create your prompts

system_prompt = "You are a code generator that generates code for a given prompt"
user_prompt_prefix = """
    Generate code
"""
```

```python
# Step 2: Make the messages list
def messages_for(user_prompt):
    return [
        {"role": "system", "content": system_prompt}, 
        {"role": "user", "content": user_prompt_prefix + user_prompt}
        ]
```

```python
# Step 3: Call OpenAI
def generate_code(user_prompt):
    response = openai.chat.completions.create(
        model="chatgpt-4o-latest",
        messages=messages_for(user_prompt)
    )
    return response.choices[0].message.content
```

```python
# Step 4: print the result
print(generate_code("where is the nearest starbucks?"))
```
