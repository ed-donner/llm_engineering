```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key=api_key)
```

```python
# Define our system prompt - you can experiment with this later, changing the last sentence to 'Respond in markdown in Spanish."

system_prompt = """
You are a data analyst always looking for trends in the data and ways to take complex data and summarize it.
Respond in markdown. Do not wrap the markdown in a code block.
"""

# Define our user prompt

user_prompt_prefix = """
Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.
"""

website = "https://finance.yahoo.com/quote/SW/"
```

```python
messages_web = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt_prefix + website}
]
```

```python
# summarize the website

response = ollama.chat.completions.create(
    model="llama3.2",
    messages = messages_web
)
response.choices[0].message.content
```
