```python
# imports
import os
from pathlib import Path
from openai import OpenAI
from dotenv import load_dotenv
```

```python
# constants

MODEL = 'gpt-4o-mini'
KNOWLEDGE_BASE_PATH = Path('knowledge_base/fruits')

FRUITS = ['Mango', 'Apple', 'Banana', 'Pawpaw', 'Watermelon']

TEMPLATE = """
Generate a detailed markdown file about the nutritional profile of {fruit}.
Use exactly this structure:

# {fruit}

## Overview
A brief description of the fruit, its origin, and general characteristics.

## Nutritional Profile (per 100g)
A table with the following rows: Calories, Carbohydrates, Protein, Fat, Fibre, Sugar, Water

## Vitamins
A table listing key vitamins and their amounts per 100g.

## Minerals
A table listing key minerals and their amounts per 100g.

## Health Benefits
A bullet point list of 5 evidence-based health benefits.

## Common Uses
A bullet point list of common culinary and non-culinary uses.

Respond with only the markdown content. No explanations, no code fences.
"""
```

```python
# set up environment

load_dotenv()

openai_api_key = os.getenv('OPENAI_API_KEY')

if not openai_api_key:
    print("❌ OpenAI API key not found")
else:
    print("✅ OpenAI API key found!")
    client = OpenAI(api_key=openai_api_key)
```

```python
# generate and save markdown files

KNOWLEDGE_BASE_PATH.mkdir(parents=True, exist_ok=True)

for fruit in FRUITS:
    filepath = KNOWLEDGE_BASE_PATH / f"{fruit.lower()}.md"

    if filepath.exists():
        print(f"⏭️  Skipping {fruit} — file already exists")
        continue

    print(f"🔄 Generating {fruit}...")

    response = client.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": "You are a nutrition expert who writes detailed, accurate markdown documents about fruits."},
            {"role": "user",   "content": TEMPLATE.format(fruit=fruit)}
        ]
    )

    content = response.choices[0].message.content.strip()
    filepath.write_text(content, encoding='utf-8')
    print(f"✅ Saved {fruit} to {filepath}")

print("\n🎉 Knowledge base ready!")
```
