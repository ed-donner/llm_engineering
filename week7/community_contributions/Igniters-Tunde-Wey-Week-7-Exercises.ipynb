## Exercise 1 (Day 1) — Quantization in QLoRA

**Task:** In one sentence, what does "quantization" mean in QLoRA and why does it help?

**Solution:** Quantization means storing the model’s weights in lower precision (e.g. 4-bit instead of 16-bit), which greatly reduces memory use and often allows fine-tuning a large model on a single consumer GPU that couldn’t hold the full-precision model.
