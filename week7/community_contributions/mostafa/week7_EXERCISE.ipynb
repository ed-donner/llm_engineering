
# Price Prediction with Transformer + LoRA + Attention Pooling

This notebook replaces the Word2Vec+TF‑IDF feature extractor with a pre‑trained transformer encoder (e.g., DeBERTa, MiniLM) fine‑tuned using **LoRA** (Low‑Rank Adaptation). LoRA adds a small number of trainable parameters while keeping the base model frozen, making fine‑tuning efficient even on a single GPU.

**Key improvements over the hybrid model:**
- Contextual embeddings capture word order and phrase‑level meaning.
- The model can learn to focus on important words via attention pooling (optional).
- Fine‑tuning adapts the representations specifically to price prediction.

**Pipeline:**
- Load train/val/test splits.
- Tokenise summaries using the transformer's tokenizer.
- Load a pre‑trained transformer and add LoRA adapters.
- Add a regression head (with optional attention pooling).
- Train with log‑transformed prices and L1 loss.
- Evaluate on the test setPrice Prediction with Transformer + LoRA + Attention Pooling
