# A Product Price Prediction Model
In this notebook, I fine-tuned the Llama-3.2-3B model to predict the price of products based on their descriptions.

The model was trained using the `ed-donner/items_prompts_lite` dataset, which contains structured Amazon product descriptions and their corresponding prices. The task is framed as a completion prompt where the model reads the product details and predicts the expected price.

Training was performed in Google Colab using QLoRA, a parameter-efficient fine-tuning technique that quantizes the base model to 4-bit precision and trains lightweight LoRA adapters, allowing the model to learn pricing patterns without retraining the full 3 billion parameters.

The evaluation notebook compares the raw base model against the fine-tuned model side by side, and includes an interactive "Price is Right" game where you can guess the price of a product and see how you stack up against both models.

**Results:** $65.16 MAE, beating the human baseline of $87.62 and approaching the GPT-4.1-nano fine-tuned benchmark of $62.51.
