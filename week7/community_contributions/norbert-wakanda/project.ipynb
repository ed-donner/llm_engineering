# Week 7 Project

The **Jumia AI Product Advisor** is an agentic shopping assistant that leverages a fine-tuned Llama 3.2-3B model, a Retrieval Augmented Generation (RAG) knowledge base, and live web search. It estimates fair product prices and evaluates deals on Jumia's e-commerce platform. The system is orchestrated by Claude Haiku in a tool-calling loop and presented through a Gradio chat UI.
