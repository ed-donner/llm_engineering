# A Price Prediction Model

In this notebook, I fine-tuned the TinyLlama-1.1B-Chat-v1.0 model to predict the price of products based on their descriptions.

The model was trained using the items_prompts_lite dataset by Ed, which contains structured product descriptions and their corresponding prices. The task is framed as a prompt where the model reads the product details and predicts the expected price.

Training was performed in Google Colab using parameter-efficient fine-tuning so the model can learn the pricing patterns without retraining the full model.
