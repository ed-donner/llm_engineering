# Week 7 Exercise

# Fine-Tune Llama 3.2 3B for Electronics Pricing

Fine-tune an open-source model using QLoRA to predict the price of **Electronics** products,
trained exclusively on Electronics data from `inginia/items_lite_plus`.

## Goal

Beat every metric achieved by the category-aware models in the Week 6 exercise
(`week6/community-contributions/tobenna/category_aware_pricing.ipynb`).

## Target Metrics to Beat (Electronics category)

| Model | MAE | MSE | r² |
|-------|-----|-----|----|
| General XGBoost | $74.97 | — | — |
| Per-Category XGBoost | $72.48 | 12,767 | 55.5% |
| Fine-tuned GPT-4.1-nano | $72.77 | 21,794 | 24.0% |

## Approach

1. **Data Preparation** — Filter the `inginia/items_lite_plus` dataset to Electronics only (~18,780 train / ~918 test), create prompt-completion pairs, and push to HuggingFace.
2. **QLoRA Training** — Load Llama 3.2 3B with 4-bit quantization, apply LoRA adapters, and fine-tune with SFTTrainer on the Electronics prompts.
3. **Evaluation** — Run inference on the Electronics test set, compute MAE / MSE / r², and compare against the target metrics.

## Why This Should Work

- 18,780 focused training samples (vs. 500 used for the GPT-4.1-nano fine-tune)
- Semantic understanding from a 3B-parameter LLM (vs. bag-of-words for XGBoost)
- Category-focused training — no capacity wasted on other product domains
- The week 7 full fine-tune achieved $39.85 MAE across *all* categories; a category-focused fine-tune should do even better within Electronics
