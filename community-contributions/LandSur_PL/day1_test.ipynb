```python

# pobierz wartości poufne z pliku .env
load_dotenv(override=True)

api_key = os.getenv('OPENAI_API_KEY') 

# Sprawdź klucz

if not api_key:
    print("Brak klucza API - nie znalazłem klucza API w pliku .env")
elif not api_key.startswith("sk-proj-"):
    print("Znalazłem klucz API, ale nie zaczyna się od sk-proj-; proszę sprawdzić, czy używasz poprawnego klucza")
elif api_key.strip() != api_key:
    print("Nie znalazłem klucza API w pliku .env")
else:
    print("Znalazłem klucz API i wygląda OK!")

```

```python
# To give you a preview -- calling OpenAI with these messages is this easy. Any problems, head over to the Troubleshooting notebook.

message = "Witam, GPT! To jest moja pierwsza wiadomość do ciebie! Hi!"

messages = [{"role": "user", "content": message}]

messages
```

```python
openai = OpenAI()
print(openai)
response = openai.chat.completions.create(model="gpt-5-nano", messages=messages)
response.choices[0].message.content
response.usage[0]
print(response)
```
