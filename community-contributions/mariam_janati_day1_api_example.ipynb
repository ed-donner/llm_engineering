# Day 1 Exercise - Simple API Explanation Generator

A quick example of using OpenAI's API with system and user prompts to generate a clear, structured explanation.

By Mariam Janati Idrissi

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv(override=True)
openai = OpenAI()
```

```python
# Step 1: Create your prompts
system_prompt = "You are a helpful assistant that writes clear, structured, and concise answers."
user_prompt = """
Explain what an API is in simple terms.
Give an example related to everyday life.
Keep it short.
"""

# Step 2: Make the messages list
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]

# Step 3: Call OpenAI
# temperature controls randomness:
# 0.0-0.3 -> very predictable, consistent (good for coding)
# 0.5-0.8 -> balanced (best default)
# 0.9+    -> creative but less reliable

response = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages,
    temperature=0.7
)

# Step 4: print the result
print(response.choices[0].message.content)
```
