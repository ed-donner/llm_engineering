```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

if not api_key:
    print("No API key was found")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start with sk-proj-; please check you're using the right key")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them")
else:
    print("Api key found")
```

Quick model call

```python
message = "Hello, GPT! I just want to know if you are ready to go?"

messages = [{"role": "user", "content": message}]

messages
```

```python
openai = OpenAI()

response = openai.chat.completions.create(model="gpt-5-nano", messages=messages)
response.choices[0].message.content
```

Fetching website, providing system and user prompts

```python
magazine = fetch_website_contents("https://aimagazine.com/")

system_prompt = """ 
You are an assistant that analyzes the contents of a website, and provides the titles of the articles available on the website,
ignoring text that might be navigation related. Respond in markdown in 2 languages English and Polish. Do not wrap the markdown in a code block - respond just with the markdown.
"""
user_prompt = """ 
Please return the titles of top 10 articles, if along with the sort description of each article. 
"""
```

```python

def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt + website}
    ]

messages_for(magazine)

def titles_des(url):
    Website = fetch_website_contents(url)
    response = openai.chat.completions.create(
        model = "gpt-4.1-mini",
        messages = messages_for(Website)
    )
    return response.choices[0].message.content
```

```python
titles_des("https://aimagazine.com/")
```

```python

def display_titles_des(url):
    title = titles_des(url)
    display(Markdown(title))

display_titles_des("https://aimagazine.com/")
```
