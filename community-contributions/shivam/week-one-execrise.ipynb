This exercise will  get you familiar with the concepts of prompts, messages, and the Google Gemini API. Build a tool that take a technical question and response with a clear and concise answer

You will be using the `google.generativeai` library to make the API call.



```python
# imports

import os
from dotenv import load_dotenv
# from scraper import fetch_website_contents
from IPython.display import Markdown, display
import google.generativeai as genai

load_dotenv()
```

```python
# constants

MODEL = 'gemini-flash-latest'
GOOGLE_API_KEY = os.getenv('GOOGLE_API_KEY')
```

```python
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
for m in genai.list_models():
    if 'generateContent' in m.supported_generation_methods:
        print(m.name)
```

```python
 # Create a function to make the API call
def make_api_call(system_prompt, user_prompt):
    model = genai.GenerativeModel(model_name=MODEL, system_instruction=system_prompt)
    response = model.generate_content(user_prompt)
    return response.text

# Create a function to print the result
def print_result(result):
    print(result)

# Create a function to display the result
def display_result(result):
    display(Markdown(result))
```

```python
system_prompt = """
You are a Web 3 and Blockchain assistant.

Help with:
- Ethereum, Solidity, and smart contracts
- Decentralized applications (dApps)
- Cryptocurrencies and DeFi
- NFTs and token standards
- Blockchain security and best practices

Provide clear, practical, and production-ready answers.
Use examples when helpful.
Avoid unnecessary explanations.
Assume the user has technical experience.
"""

user_prompt = """
Give me a Solidity smart contract for a simple ERC-20 token.
"""

result = make_api_call(system_prompt, user_prompt)

print_result(result)
```

```python
# Display the result in markdown
display_result(result)
```
