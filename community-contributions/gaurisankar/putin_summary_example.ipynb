# Vladimir Putin Wikipedia Summary Example

This notebook demonstrates how to use the OpenAI API to summarize content from a Wikipedia page using a structured system prompt, user prompt, and messages list.

```python
import os
from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI

load_dotenv(override=True)
openai = OpenAI()
```

```python
# Step 1: System Prompt — tells the model its role and output style
system_prompt = """
You are a concise political analyst. When given information about a world leader,
provide a factual, neutral, and well-structured summary covering:
- Early life and rise to power
- Key political positions and ideology
- Major domestic and foreign policy decisions
- International controversies and legacy
Respond in markdown. Do not wrap the markdown in a code block.
"""
```

```python
# Step 2: User Prompt — fetch Putin's Wikipedia page and build the prompt
putin_wiki = fetch_website_contents("https://en.wikipedia.org/wiki/Vladimir_Putin")

user_prompt = (
    "Below is the Wikipedia content about Vladimir Putin.\n"
    "Please provide a concise, structured summary of his life, "
    "political career, and legacy.\n\n"
    + putin_wiki
)
```

```python
# Step 3: Messages — the structure the OpenAI API expects
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user",   "content": user_prompt}
]
```

```python
# Step 4: Response — call the OpenAI API and display the result
response = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages
)

result = response.choices[0].message.content
display(Markdown(result))
```
