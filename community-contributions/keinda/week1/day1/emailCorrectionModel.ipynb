#Email Correction Model
This example shows how to use OpenAI prompts to generate a polite correction email.

## Overview
In this notebook, I demonstrate how to use OpenAI's chat completion API with a system prompt and user prompt to generate a professional correction email. The model is instructed to write an email requesting that a professor update an incorrect date on a previously submitted PowerPoint slide.


```python
import openai

system_prompt = "You are an AI assistant that specializes in writing and correcting professional emails."

user_prompt = """
I accidentally wrote 'Fall 2025' instead of 'Fall 2024' on a slide in the PowerPoint presentation
I already sent to my professor. Please write a polite email asking my professor to correct the mistake,
and provide a short subject line for the email.
"""

# Step 2: Make the messages list
messages = [
    {"role": "system", "content": system_prompt},
    {"role":"user", "content": user_prompt}
]

# Step 2: Make the messages list

messages = [{"role": "system", "content": system_prompt},
{"role":"user", "content": user_prompt}] # fill this in

# Step 3: Call OpenAI
response = openai.chat.completions.create(model="gpt-4.1-nano", messages=messages)
print(response.choices[0].message.content)
# Step 4: print the result

```
