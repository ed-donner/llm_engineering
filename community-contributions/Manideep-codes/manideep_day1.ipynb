```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import openai
client = OpenAI()

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Step 1: Create your prompts

system_prompt = "You are a helpful assistant to suggest subject of the emails"
user_prompt = """
    Dear Jasmine,

I hope you're doing well. I came across your job postings for the Full Stack (Java/React) Developer and Backend (Java) Developer roles, and I am very interested in exploring these opportunities.

With my experience in Java development, particularly with Spring Boot, Microservices, API development, and cloud platforms, I believe I would be a strong fit for either role. I have previously worked on enterprise-level applications, handling Kafka for messaging, and deploying solutions on AWS and OpenShift.

I would love to discuss how my skills align with your needs. Please let me know a convenient time for a quick chat. I’ve attached my resume for your reference.

Looking forward to your response!


"""

# Step 2: Make the messages list

messages = [{"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}] # fill this in

# Step 3: Call OpenAI
response = openai.chat.completions.create(
    model = "gpt-4.1-mini",
    messages = messages
)
# response =

# Step 4: print the result
print(response.choices[0].message.content)
# print(
```
