To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

````python
# imports

from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
import ollama
````

````python
# constants

MODEL_GPT = 'openai/gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
````

````python
# set up environment

load_dotenv()
openai = OpenAI()
````

````python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
````

````python
# prompts

system_prompt = "You are a helpful technical tutor who answers questions about python code, software engineering, data science and LLMs"
user_prompt = "Please give a detailed explanation to the following question: " + question
````

````python
# messages

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
````

````python
# Get openai/gpt-4o-mini to answer, with streaming

stream = openai.chat.completions.create(max_tokens=1000, model=MODEL_GPT, messages=messages,stream=True)
    
response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    response = response.replace("```","").replace("markdown", "")
    update_display(Markdown(response), display_id=display_handle.display_id)
````

````python
# Get Llama 3.2 to answer

response = ollama.chat(model=MODEL_LLAMA, messages=messages)
reply = response['message']['content']
display(Markdown(reply))
````

You could make it better by taking in the question using  
`my_question = input("Please enter your question:")`

And then creating the prompts and making the calls interactively.

````python
# Create an interactive tutor that takes a question from the user
my_question = input("Please enter your question: ")

if my_question.strip():
    # Construct prompts
    interactive_system_prompt = "You are a helpful technical tutor who answers questions about python code, software engineering, data science and LLMs"
    interactive_user_prompt = "Please give a detailed explanation to the following question: " + my_question
    
    interactive_messages = [
        {"role": "system", "content": interactive_system_prompt},
        {"role": "user", "content": interactive_user_prompt}
    ]
    
    print("\n--- Answer from GPT-4o-Mini ---")
    interactive_stream = openai.chat.completions.create(max_tokens=1000, model=MODEL_GPT, messages=interactive_messages, stream=True)
    response_gpt = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in interactive_stream:
        response_gpt += chunk.choices[0].delta.content or ''
        response_gpt = response_gpt.replace("```","").replace("markdown", "")
        update_display(Markdown(response_gpt), display_id=display_handle.display_id)
        
    print("\n--- Answer from Llama 3.2 (Local) ---")
    response_llama = ollama.chat(model=MODEL_LLAMA, messages=interactive_messages)
    display(Markdown(response_llama['message']['content']))
````
