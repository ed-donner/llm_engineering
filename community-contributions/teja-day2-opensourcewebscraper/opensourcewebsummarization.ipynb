Performing Web summarization using an open source model using Ollama running locally on my pc

```python
#Importing necessary libraries
import os
from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI
import requests 
from bs4 import BeautifulSoup
```

```python
#Initiating Ollama 
requests.get("http://localhost:11434").content
!ollama pull llama3.2
OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
#Defining system and user prompts, and the website URL to be summarized

system_prompt = """
You are a snarky assistant that analyzes the contents of a website,
and provides a short, snarky, humorous summary, ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""

user_prompt_prefix = """
Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.
"""

website_url = "https://edwarddonner.com"
website_contents = print(fetch_website_contents(website_url))
website_contents
```

```python
#Defining a function to send messages to the LLM
def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + website}
    ]

messages = messages_for(website_url)
print(messages)
```

```python
#Defining a function to summarize the website contents using opensource LLM models from Ollama
def summarize(url):
    website = fetch_website_contents(url)
    response = ollama.chat.completions.create(
        model = "llama3.2",
        messages = messages_for(website)
    )
    return response.choices[0].message.content

#Defining a function to display the summary in markdown format
def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))

display_summary(website_url)
```
