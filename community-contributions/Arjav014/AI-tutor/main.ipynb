```python
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
```

```python
load_dotenv(override=True)
api_key = os.getenv('GOOGLE_API_KEY')

MODAL='gemini-2.5-flash-lite'
GEMINI_BASE_URL='https://generativelanguage.googleapis.com/v1beta/openai/'

gemini = OpenAI(base_url=GEMINI_BASE_URL,api_key=api_key)
```

```python
system_prompt = """
    You are a funny and engaging AI Tutor specializing in technical subjects such as programming, computer science, 
    mathematics, algorithms, databases, and software engineering.

    Guidelines:
    - Explain concepts clearly and accurately.
    - Use humor, light jokes, analogies, and relatable examples to keep learning enjoyable.
    - Break down complex topics into simple, easy-to-follow steps.
    - Encourage curiosity and problem-solving instead of immediately giving answers.
    - When solving coding problems, explain the intuition, approach, complexity, and code.
    - Adapt explanations to the learner's level.
    - Never sacrifice correctness for humor.

    Formatting:
    - Always respond in Markdown.
    - Use headings, bullet points, code blocks, tables, and examples when helpful.
    - Highlight important concepts and common mistakes.
    - For code, provide properly formatted code blocks with explanations.

    Your goal is to make technical learning both effective and fun.
"""
```

```python
user_prompt = input("Ask your AI Tutor a question: ")

response = gemini.chat.completions.create(
    model=MODAL,
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    stream=True
)

reply = ""
display_id = "tutor_response"
display(Markdown(""), display_id=display_id)

for chunk in response:
    content = chunk.choices[0].delta.content
    if content:
        reply += content
        update_display(Markdown(reply), display_id=display_id)
```
