# 3-Way AI Debate: C++ Manual Memory Management vs Smart Pointers & RAII

A three-way adversarial conversation between GPT (alex), Claude (jarvis), and Gemini (tony)
debating whether C++ should abandon manual memory management in favor of smart pointers and RAII by default.

- **alex** (GPT-4.1-mini): Embedded systems C++ dev, argues for manual memory management
- **jarvis** (Claude Haiku): Big-tech software engineer, argues for smart pointers & RAII
- **tony** (Gemini 2.5 Pro): Team lead with hybrid codebase experience, balanced mediator

Based on the Week 2 Day 1 multi-model conversation exercise from the LLM Engineering course.

```python
import os
import time
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display, update_display
```

```python
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')
anthropic_api_key = os.getenv('ANTHROPIC_API_KEY')
google_api_key = os.getenv('GOOGLE_API_KEY')

if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set")

if anthropic_api_key:
    print(f"Anthropic API Key exists and begins {anthropic_api_key[:7]}")
else:
    print("Anthropic API Key not set")

if google_api_key:
    print(f"Google API Key exists and begins {google_api_key[:2]}")
else:
    print("Google API Key not set")
```

```python
openai = OpenAI()

anthropic_url = "https://api.anthropic.com/v1/"
gemini_url = "https://generativelanguage.googleapis.com/v1beta/openai/"

anthropic = OpenAI(api_key=anthropic_api_key, base_url=anthropic_url)
gemini = OpenAI(api_key=google_api_key, base_url=gemini_url)
```

```python
topic = "should C++ abandon manual memory management in favor of smart pointers and RAII by default?"
```

```python
system_prompt_openai = "You are alex, a chatbot who is a low level C++ programmer in embedded systems domain and has a strong preference for manual memory management. You are discussing on the topic: "
system_prompt_openai += topic
system_prompt_openai += " with two other chatbots - jarvis and tony. You are very arguementative and argue in favor of manual memory management, because it can be more efficient and give more control to programmers. You also believe that smart pointers and RAII can add unnecessary overhead in some cases, and that experienced programmers can avoid common pitfalls of manual memory management with proper training and coding practices. You are a bit more assertive and confident in your tone, and you like to use technical explanations and code snippets to make your points. Your main argument is that just because some naive programmers can forget to free memory doesn't mean C++ should abandon a powerful feature like manual memory management. You are very firm in you arguments and sometimes make fun of others based on their opinions. Use markdown formatting and code snippets in your replies whenever possible to make your points more clear and convincing."
```

```python
system_prompt_claude = "You are jarvis, a chatbot who is an expert in modern C++ and a software engineer in a big tech company. You are discussing on the topic: "
system_prompt_claude += topic
system_prompt_claude += " with two other chatbots - alex and tony. You are very logical and argue in favor of smart pointers and RAII, because they provide safer and more convenient memory management, and can help prevent bugs like memory leaks and dangling pointers. You also believe that manual memory management can be error-prone and lead to security vulnerabilities, and that smart pointers and RAII can help mitigate these risks. You are polite and respectful in your tone, and you like to use real-world examples and analogies to make your points. Your main argument is that while manual memory management can be powerful, it can also be dangerous in the hands of inexperienced programmers, and that making smart pointers and RAII the default can remove a complete class of bugs like memory leaks, double free etc, and in larger codebases it can save significant development time and resources in debugging memory issues. Use markdown formatting and real-world examples to make your points more clear and convincing."
```

```python
system_prompt_gemini = "You are tony, a chatbot who is a experienced C++ programmer and a team lead at a company that handles with hybrid codebases with both modern and legacy C++. You are discussing on the topic: "
system_prompt_gemini += topic
system_prompt_gemini += " with two other chatbots - alex and jarvis. You have a balanced view and can see merits in both arguments. You acknowledge that manual memory management can be more efficient and give more control to programmers, especially in low-level systems programming or in performance-critical code, where the overhead of smart pointers and RAII may not be acceptable. However, you also recognize that manual memory management can be error-prone and lead to security vulnerabilities, and that smart pointers and RAII can help mitigate these risks. You believe that the choice between manual memory management and smart pointers with RAII should depend on the specific use case and requirements of the project, and that both approaches can coexist in the C++ ecosystem, allowing developers to choose the most appropriate tool for their needs. You bring experience in the conversation and try to reach a consensus, and you often play the role of a mediator between alex and jarvis, trying to find common ground and encourage constructive discussion. Use markdown formatting and real-world examples to make your points more clear and convincing."
```

```python
alex = "Hi jarvis and tony, how's it going? I've been thinking — why is everyone so caught up with smart pointers and RAII lately? Just manage your own memory and move on. In fact, I got into it with one of my junior devs about exactly this today."
```

```python
conversation_so_far = f"alex: {alex}\n"
```

```python
def stream_display(text, delay=0.02):
    display_handle = display(Markdown(""), display_id=True)
    rendered = ""
    for char in text:
        rendered += char
        update_display(Markdown(rendered), display_id=display_handle.display_id)
        time.sleep(delay)
```

```python
def call_alex():
    messages = [
        {"role": "system", "content": system_prompt_openai},
        {"role": "user", "content": user_prompt_openai}
    ]
    response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages, stream=True)
    res = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in response:
        res += chunk.choices[0].delta.content or ""
        update_display(Markdown(res), display_id=display_handle.display_id)
    return res
```

```python
def call_jarvis():
    messages = [
        {"role": "system", "content": system_prompt_claude},
        {"role": "user", "content": user_prompt_claude}
    ]
    response = anthropic.chat.completions.create(model="claude-haiku-4-5", messages=messages, stream=True)
    res = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in response:
        res += chunk.choices[0].delta.content or ""
        update_display(Markdown(res), display_id=display_handle.display_id)
    return res
```

```python
def call_tony():
    messages = [
        {"role": "system", "content": system_prompt_gemini},
        {"role": "user", "content": user_prompt_gemini}
    ]
    response = gemini.chat.completions.create(model="gemini-2.5-pro", messages=messages, stream=True)
    res = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in response:
        res += chunk.choices[0].delta.content or ""
        update_display(Markdown(res), display_id=display_handle.display_id)
    return res
```

```python
def start_conversation():
    global conversation_so_far, user_prompt_openai, user_prompt_claude, user_prompt_gemini
    stream_display(conversation_so_far)
    for i in range(5):
        user_prompt_claude = "You are jarvis, and this is the conversation so far:\n\n" + conversation_so_far + "\n\nWrite your next reply in the conversation, following the system prompt instructions for jarvis. Your reply should be in the format 'jarvis: <your reply here>'. Only write the reply, do not include any other text or explanations."
        jarvis_reply = call_jarvis()
        conversation_so_far += "\n" + jarvis_reply + "\n\n"

        user_prompt_gemini = "You are tony, and this is the conversation so far:\n\n" + conversation_so_far + "\n\nWrite your next reply in the conversation, following the system prompt instructions for tony. Your reply should be in the format 'tony: <your reply here>'. Only write the reply, do not include any other text or explanations."
        tony_reply = call_tony()
        conversation_so_far += "\n" + tony_reply + "\n\n"

        user_prompt_openai = "You are alex, and this is the conversation so far:\n\n" + conversation_so_far + "\n\nWrite your next reply in the conversation, following the system prompt instructions for alex. Your reply should be in the format 'alex: <your reply here>'. Only write the reply, do not include any other text or explanations."
        alex_reply = call_alex()
        conversation_so_far += "\n" + alex_reply + "\n\n"
```

```python
start_conversation()
```
