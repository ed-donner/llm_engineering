# Top heading summary of todays news

Create a summary report using promt running in locally using ollama and deepseek3.2 as a model.


```python
import os
from dotenv import load_dotenv

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python
from bs4 import BeautifulSoup
from openai import OpenAI
import requests

headers = {"Content-Type": "application/json"}

OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')

def fetch_website_contents(url):
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    title = soup.title.string if soup.title else "No title found"
    if soup.body:
        for irrelevant in soup.body(["script", "style", "img", "input"]):
            irrelevant.decompose()
        text = soup.body.get_text(separator="\n", strip=True)
    else:
        text = ""
    return (title + "\n\n" + text)
ed = fetch_website_contents("https://indianexpress.com/")
print(ed)
```

# Creating params for the API request

```python
user_promt = "here is the content of the latest website news. Please provide me the top news headlines from the content for today. Also, provide a short summary of it."
message = [{
    "role": "user",
    "content": user_promt + "\n\n" + ed
},
{
    "role": "system",
    "content": "You are a helpful assistant that summarizes the news content and provides top headlines."

}
]
```

# Display for the today's latest news.

```python
from IPython.display import Markdown, display

def summarize_news(message):
    result = ollama.chat.completions.create(model="deepseek-r1:1.5b", messages=message)
    return result.choices[0].message.content

summarize_news(message)

def display_news_summary(message):
    summary = summarize_news(message)
    print("News Summary and Headlines:")
    display(Markdown(summary))

display_news_summary(message)
```
