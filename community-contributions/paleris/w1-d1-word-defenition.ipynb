```python
# imports

import os
import sys
from pathlib import Path
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI

# Load environment variables in a file called .env

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

### Words Defenitions

It takes an input from the user and provide a definition for it.

```python
# Step 1: Create your prompts

system_prompt = """
You are professional linguist and lexicographer. You provide clear and detailed definitions of words, etimology and examples of usage. In your response use only language user used for input. Correct mistakes in user's input if any. Don't interact with the user in any form.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""
user_prompt_prefix = """
Here are the words I want you to define:

"""

# Step 2: Make the messages list
def messages_for():
    user_words = input("Enter words you want to define ")
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + user_words}
    ]

# Step 3: Call OpenAI
openai = OpenAI()

def words_definition():
    response = openai.chat.completions.create(
        model = "gpt-4.1-mini",
        messages = messages_for()
    )
    return response.choices[0].message.content

# Step 4: print the result
def display_definition():
    definition = words_definition()
    display(Markdown(definition))

display_definition()
```
