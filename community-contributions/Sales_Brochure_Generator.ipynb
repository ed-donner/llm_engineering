```python
import os
from openai import OpenAI  # OpenRouter-compatible client

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=os.getenv("OPENROUTER_API_KEY")
)
```

```python
product_name = "AI Productivity Suite"
target_audience = "Small business owners"
key_features = "Automated reports, Smart scheduling, AI-powered insights"

prompt = f"""
Create a professional and persuasive sales brochure.

Product Name: {product_name}
Target Audience: {target_audience}
Key Features: {key_features}

Include:
- Catchy headline
- Engaging introduction
- Bullet points of benefits
- Strong call-to-action
"""
```

```python
response = client.chat.completions.create(
    model="openai/gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are a professional marketing copywriter."},
        {"role": "user", "content": prompt}
    ],
)

brochure = response.choices[0].message.content
print(brochure)
```
