```python
import playsound3
from playsound3 import playsound
```

```python
def self_talk():
    while True:
        playsound3.playsound("Self talk.mp3")
```

```python
import gradio as gr
while True:
    gr.Interface(
        fn=self_talk, 
        inputs=None, 
        outputs
            =gr.Audio("My self talk")
    ).launch()
```
