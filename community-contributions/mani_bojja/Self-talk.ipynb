```python
import playsound3
from playsound3 import playsound
```

```python
def self_talk():
    while True:
        playsound3.playsound("https://drive.google.com/file/d/1Wzn2nWcL9SDO5ADSi4xcRPTDltb2kEyF/view?usp=sharing")
```

```python
import gradio as gr
while True:
    gr.Interface(
        fn=self_talk, 
        inputs=None, 
        outputs
            =gr.Audio("My self talk")
    ).launch(server_port=7900)
```
