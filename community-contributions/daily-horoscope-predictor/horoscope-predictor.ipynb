```python
from dataCollector import scrape_horoscopes_ht, scrape_horoscopes_vedicrishi, scrape_horoscopes_indiatv
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
openai = OpenAI()
```

```python
vr = scrape_horoscopes_vedicrishi(headless=False, timeout=5)
ht = scrape_horoscopes_ht(headless=False, timeout=3)
itv = scrape_horoscopes_indiatv(headless=False, timeout=3)
```

```python
ht
```

```python
data = [ht, vr, itv] 
```

```python
data
```

```python
system_prompt = """
You are an astrology assistant that analyzes the contents of provided data,
and provides a short summary, ignoring text that might not be relevant to the horoscope analysis.
The data is a list of dictionaries, where each dictionary contains the following keys:
- 'sign': The zodiac sign (e.g., "Aries", "Taurus", etc.). and corresponding prediction for the day.
Analyze the data and provide a concise summary of the horoscope predictions for each zodiac sign, highlighting any common themes or notable differences across the signs.
Cover the key points: Lover, Career, Finance and Heaalth at last provide Key Advice for the day.  
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""
```

```python
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": str(data)}
]
```

```python
messages
```

```python
# Load environment variables in a file called .env

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")
```

```python
response = openai.chat.completions.create(
        model = "gpt-4.1-mini",
        messages = messages
    )
```

```python
display(Markdown(response.choices[0].message.content))
```
