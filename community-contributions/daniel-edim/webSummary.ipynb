```python
from pathlib import Path
import json
import random

# Read and parse repo_tree.json, then keep a random 3 projects
repo_tree_path = Path("repo_tree.json")
repo_tree_data = json.loads(repo_tree_path.read_text(encoding="utf-8"))
children = repo_tree_data.get("children", [])
sampled_children = random.sample(children, k=min(3, len(children)))
sampled_repo_tree = {**repo_tree_data, "children": sampled_children}
repo_tree_json_string = json.dumps(sampled_repo_tree, ensure_ascii=False)

```

```python
system_prompt = """
You analyze a repository tree exported from community contributions.
The input is a JSON tree where file nodes may include a `content` field.
Use only the available `content` fields to infer what each project does.
Output requirements:
- Return only a markdown table (no extra text before/after).
- Table columns must be exactly: user name | project path | summary
- One row per project.
- `user name` is the contributor folder name under `community-contributions`.
- `project path` is the directory path for that project.
- `summary` should be concise (1-2 sentences) and based on project files.
- If there is not enough readable content, say: "Insufficient readable content to summarize."
- Do not wrap the markdown in a code block.
"""

user_prompt = """
Analyze this repo tree content.
For each contributor project, read available file `content` fields and summarize what the project is about.
Return the result as a markdown table with exactly these columns:
user name | project path | summary
Here is the repo tree:

"""

```

```python
def messages_for(fileContent):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt + fileContent},
    ]
```

```python
openai = OpenAI()

def summarize_large_input(fileContent):
    messages = messages_for(fileContent)
    response = openai.chat.completions.create(
        model="gpt-5.5",
        messages=messages,
    )
    return response.choices[0].message.content
```

```python
def display_summary(fileContent):
    summary = summarize_large_input(fileContent)
    display(Markdown(summary))

display_summary(repo_tree_json_string)
```
