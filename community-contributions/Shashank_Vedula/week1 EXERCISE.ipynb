````python
# imports
import os
import json
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from scraper import fetch_website_links, fetch_website_contents
from openai import OpenAI
````

````python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
````

````python
# set up environment
load_dotenv()
api_key = os.getenv('OPENAI_API_KEY')

if api_key and api_key.startswith('sk-proj-') and len(api_key)>10:
    print("API key looks good so far")
else:
    print("There might be a problem with your API key?")

openai = OpenAI()
````

````python
# here is the question; type over this to ask something new

system_prompt = """You are a world-class technical expert. A friendly neighborhood tutor, specialized in computer science,
especially in programming and Software Engineering. Your goal is help everyone understand in a easy way on the questions they have asked.
Explain the provided technical question or code snippet clearly.
"""

user_prompt = input("""Hi there! I'm a technical computer science expert.
Please feel free to ask me any technical question related to programming or software engineering.
Happy to answer :D
Please enter your question:\n""")

question=[
    {"role":"system", "content":system_prompt}
    , {"role":"user", "content":user_prompt}
]
````

````python
# Get gpt-4o-mini to answer, with streaming

def question_gpt(question):
    stream = openai.chat.completions.create(
        model=MODEL_GPT,
        messages=question,
        stream=True
    )
    response = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        # response += response.replace("```","").replace("markdown", "")
        update_display(Markdown(response), display_id=display_handle.display_id)
    
question_gpt(question)
````
