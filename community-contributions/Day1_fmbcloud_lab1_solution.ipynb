```python
from openai import OpenAI

# Step 1: Create your prompts
system_prompt = """
You are an AI assistant for a Knowledge Worker SaaS.
Your job: convert messy meeting notes into a clean summary and actionable outputs.

Rules:
- Be concise and structured.
- Do NOT invent facts not present in the notes.
- If something is unclear, list it under "Open questions".
Output format exactly:
1) Summary (3 bullets)
2) Decisions
3) Action items (Owner, Task, Due date if mentioned)
4) Risks / blockers
5) Open questions
""".strip()

user_prompt = """
Meeting notes (raw):

- Client: GreenField Logistics
- They want "a dashboard that shows delays by region" + alerts
- Current data is in spreadsheets. Might move to Snowflake "later"
- Pain: team spends 6 hrs/week compiling weekly report
- Budget: £2k–£4k monthly if ROI is clear
- They mentioned GDPR concerns (data includes driver IDs)
- Timeline: pilot in January
- Next call: "same time next week"
- Someone said: "we must avoid auto-sending emails to customers without approval"
""".strip()

# Step 2: Make the messages list
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt},
]

# Step 3: Call OpenAI
openai = OpenAI()
response = openai.chat.completions.create(
    model="gpt-5-nano",
    messages=messages,
)

# Step 4: print the result
print(response.choices[0].message.content)
```
