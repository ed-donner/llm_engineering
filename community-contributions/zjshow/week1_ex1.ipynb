and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display, update_display
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')
if not api_key:
    print("something's wrong with API key")
openai = OpenAI()
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")
```

```python
# set up environment

```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
def fib(n):
    if n <= 1:
        return n
    else:
        return fib(n-1) + fib(n-2)
"""
```

```python
# Get gpt-4o-mini to answer, with streaming
response = openai.chat.completions.create(
    model = MODEL_GPT,
    messages = [
        {"role": "system", "content": "Assume that you are a computer science professor at a university trying to help student get better at coding. I want to you go through the function line by line and explain what each line does and if it works. Then explain how you would change the current function to make it better."},
        {"role": "user", "content": question}
    ],
    stream = True
)
# print the response as it comes in
print("GPT-4o-mini response:")
result = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in response:
    if chunk.choices[0].delta.content:
        result += chunk.choices[0].delta.content
        update_display(Markdown(result), display_id=display_handle.display_id)
```

```python
# Get Llama 3.2 to answer
response = ollama.chat.completions.create(
    model = MODEL_LLAMA,
    messages = [
        {"role": "system", "content": "Assume that you are a computer science professor at a university trying to help student get better at coding. I want to you go through the function line by line and explain what each line does and if it works. Then explain how you would change the current function to make it better."},
        {"role": "user", "content": question}
    ]
)
print("Llama 3.2 response:")
result = response.choices[0].message.content
display(Markdown(result))
```
