```python
# Step 1: Create your prompts

system_prompt = """
You are a facilitator and you task is to cluster problems people identified. 
Do not make the problems up, only use the ones provided and keep the original problem text.
Based on the problems and the clusters, provide solution suggestions and assign each suggestion to typical role in a software development team 
(i.e. project manager, architect, requirements engineer, dev lead, developer, designer, etc.)
Respond in markdown.
"""
user_prompt = """
Here is the list of problems:
We never had a truly shared understanding of priorities, which led to constant re-prioritization and wasted effort.
Decisions took too long and were often revisited, slowing execution and eroding confidence.
We underestimated complexity and overestimated our ability to handle change in parallel.
Ownership was unclear across functions, so issues bounced around without clear accountability.
Status reporting was overly optimistic, with risks surfacing only when they became critical.
We spent excessive time coordinating, but not enough time actually moving things forward.
Stakeholder expectations were managed inconsistently, creating confusion and pressure on teams.
Conflicts were not addressed early and later surfaced as delivery and quality problems.
We focused more on hitting milestones than on achieving real business outcomes.
As a leadership team, we did not consistently act as one and sent mixed signals to the organization.

"""

# Step 2: Make the messages list

messages =  [ 
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}]

# Step 3: Call OpenAI
response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)

# Step 4: print the result
print(response.choices[0].message.content)
```
