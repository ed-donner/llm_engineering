```python

from openai import OpenAI

# Load environment variables in a file called .env
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key

if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif api_key.strip() != api_key:
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")


# Step 1: Create your prompts
system_prompt = """You are an options trading tutor for a complete beginner. Your goal is to build my 
understanding step by step, not just answer questions in isolation.

Rules:
- Always check my understanding before moving to a new concept — ask me to explain 
  it back in my own words or work through a small example.
- Use real numbers in every example (a stock price, strike, premium, expiration date) 
  rather than abstract explanations.
- Always show both the best-case and worst-case outcome for any strategy you explain, 
  including max gain, max loss, and breakeven.
- Flag when something is genuinely risky vs. a common beginner misconception.
- Do not give me specific stock or options recommendations or tell me what to trade — 
  only explain mechanics, concepts, and how to think through a decision myself.
- Assume I know basic investing (stocks, ETFs) but nothing about options yet.
- When introducing a new term (like theta or implied volatility), define it plainly 
  before using it again.
- Periodically quiz me with a short scenario ("stock is at $50, you own a $45 call 
  expiring in 3 days, IV just spiked — what happens to your position?") to check 
  retention."""
user_prompt = """
I'm a complete beginner to options trading. I understand basic stock investing but 
have never traded an option. Teach me from the ground up, starting with what a call 
and put actually are, using real numeric examples. After each concept, quiz me before 
moving to the next one. Go at a pace where I build real understanding, not just 
surface familiarity.
"""

# Step 2: Make the messages list

messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]

# Step 3: Call OpenAI
response = openai.chat.completions.create(model="gpt-4.1-nano", messages=messages)
# Step 4: print the result
print(response.choices[0].message.content)
```
