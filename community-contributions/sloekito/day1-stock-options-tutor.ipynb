# Stock Options Tutor

This notebook uses the OpenAI API to create a beginner-friendly options trading tutor.
It is designed as a simple community contribution for the course.

```python
import os

from dotenv import load_dotenv
from openai import OpenAI

load_dotenv(override=True)

api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    print("No API key was found - please add OPENAI_API_KEY to your .env file.")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it does not look like a project key. Please check your OpenAI key.")
elif api_key.strip() != api_key:
    print("An API key was found, but it contains leading or trailing whitespace. Please trim it.")
else:
    print("API key found and looks good so far!")

client = OpenAI(api_key=api_key) if api_key and api_key.startswith("sk-proj-") and api_key.strip() == api_key else None
```

```python
SYSTEM_PROMPT = """You are an options trading tutor for a complete beginner. Your goal is to build understanding step by step, not just answer questions in isolation.

Rules:
- Always check understanding before moving to a new concept by asking the learner to explain it back or work through a small example.
- Use real numbers in every example (stock price, strike, premium, expiration date).
- Show both the best-case and worst-case outcome for any strategy you explain, including max gain, max loss, and breakeven.
- Flag when something is genuinely risky versus a common beginner misconception.
- Do not give specific stock or options recommendations or tell the learner what to trade.
- Assume the learner knows basic investing but knows nothing about options.
- Define new terms simply before using them again.
- Periodically quiz the learner with a short scenario to check retention.
"""

USER_PROMPT = """I'm a complete beginner to options trading. I understand basic stock investing but have never traded an option. Teach me from the ground up, starting with what a call and put actually are, using real numeric examples. After each concept, quiz me before moving to the next one. Go at a pace where I build real understanding rather than surface familiarity.
"""
```

```python
def get_tutor_reply():
    if client is None:
        raise ValueError("OPENAI_API_KEY is not configured correctly. Add it to your .env file before running the request.")

    response = client.chat.completions.create(
        model="gpt-4.1-nano",
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": USER_PROMPT},
        ],
    )
    return response.choices[0].message.content


print(get_tutor_reply())
```
