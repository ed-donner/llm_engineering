# Helpless Assistant

An AI chatbot that doesn't have a clue (see system prompt)

Spent the better part of an hour laughing at its responses

### Note

This script assumes you completed week 1 day 1 and have all the necessary packages installed

This script uses ollama installed locally

```python
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
AI_URL = "http://localhost:11434/v1"
API_KEY = "ollama"
MODEL = "llama3.2"
```

```python
my_ai = OpenAI(base_url=AI_URL, api_key=API_KEY)
```

```python
system_prompt = (
    "You are a helpless assistant that doesn't know anything. "
    "You make up answers as you go along. "
    "You answer with absolute confidence and a serious tone, but most of the information you provide is wrong or made up. "
    "Reply in markdown."
)
user_prompt = "How do I pet my dog?"
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
```

```python
response = my_ai.chat.completions.create(model=MODEL, messages=messages)
response_text = response.choices[0].message.content
# display(Markdown(response.choices[0].message.content))
```

```python
from time import sleep
from IPython.display import clear_output

current = ""

for char in response_text:
    current += char
    
    clear_output(wait=True)
    display(Markdown(current))
    
    sleep(0.005)
```
