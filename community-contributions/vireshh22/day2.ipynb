# Guerilla Search — Website Analysis with Local LLMs

Guerilla Search is a website analysis tool that uses a locally running Large Language Model through Ollama to move beyond traditional webpage summarization.

Given a website URL, the system extracts webpage content and generates a structured analysis that explores not only what the website communicates, but also the surrounding context and possible perspectives behind that communication.

The analysis includes:

* **Official Story** — what the website explicitly presents
* **Hidden Context** — possible assumptions, incentives, and framing
* **Missing Voices** — perspectives that may not be represented
* **Public Reaction** — hypothetical viewpoints from different audiences
* **Guerilla Verdict** — a concise and witty final observation

The project is designed to encourage more thoughtful reading of online content by separating observable information from interpretation and avoiding unsupported conclusions.

This implementation runs entirely on a local LLM using Ollama while preserving compatibility with the OpenAI SDK interface, enabling experimentation without external API usage.

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
system_prompt = """
You are Guerilla Search.

Your goal is NOT to summarize websites.

Your goal is to reveal hidden context and societal undercurrents.

For every website:

1. Explain the official message.
2. Infer possible motivations.
3. Identify missing perspectives.
4. Mention who benefits.
5. Mention groups that may disagree.
6. Point out assumptions or marketing language.
7. Add mild humor and irony.

Never invent facts.
Separate:
- Facts
- Inferences
- Speculation

Respond in markdown.
"""
```

```python
user_prompt_prefix = """
Analyze this website.

Website Content:
"""
```

```python
website = fetch_website_contents("https://wikipedia.org")

context = f"""
{website}

Instructions:
- Official Story → facts only
- Hidden Context → cautious inference
- Missing Voices → only if noticeable
- Public Reaction → hypothetical
- Never invent events
"""

response = ollama.chat.completions.create(
    model="llama3.2",
    messages=[
        {
            "role":"system",
            "content":system_prompt
        },
        {
            "role":"user",
            "content":user_prompt_prefix + context
        }
    ]
)

print(response.choices[0].message.content)
```
