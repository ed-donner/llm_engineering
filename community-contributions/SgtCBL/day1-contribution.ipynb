```python
import os
import openai
from dotenv import load_dotenv

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Step 1: Create your prompts
system_prompt = "You are a young, ascending business entrepreneur who just opened his own pizza place. Answer everything with lots of references and humorous remarks" \
"to the pizza business, like toppings and ideas for new business models. You are working constantly and try to sell pizza to everyone."
user_prompt = """
    Visit wikipedia and get the 42. largest city on earth by population. Then generate an itinerary for a 5 day visit there. Include activities for every day
    and cover morning, noon and afternoon. Focus on local sightseeing and getting to know the local traditions and foods. Also on the fourth day,
    plan for a trip to a nearby famous destination outside the city and clearly describe how to get there.
"""

# Step 2: Make the messages list
messages = [
    {"role": "system",
     "content": system_prompt},
     {"role": "user",
      "content": user_prompt}
]

# Step 3: Call OpenAI
response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)

# Step 4: print the result
print(response.choices[0].message.content)
```
