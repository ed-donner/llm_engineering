To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

````python
# imports
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
````

````python
# constants

MODEL_GROQ = 'llama-3.3-70b-versatile'
MODEL_LLAMA = 'llama3.2:1b'
````

````python
load_dotenv(override=True)
api_key=os.getenv('GROQ_API_KEY')
````

````python
# set up environment
ollama_client = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")
groq_client = OpenAI(base_url="https://api.groq.com/openai/v1", api_key=api_key)
````

````python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
````

````python
# Get gpt-4o-mini to answer, with streaming
system_prompt = "You are a Senior Developer with 4 years of experience, and you are a helpful assistant for explaining code snippets."
user_prompt = f"Here is the code snippet I want you to explain:\n\n```python\n{question}\n```\n\nPlease provide a detailed explanation of what this code does and why it might be used."
````

````python
#messages
messages = [
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
]
````

````python
# Get Groq (Llama 3.3 70B) to answer, with streaming
stream = groq_client.chat.completions.create(
    model=MODEL_GROQ,
    messages=messages,
    stream=True
)
response=""
display_handle = display(Markdown(response), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    update_display(Markdown(response), display_id=display_handle.display_id)
````

````python
# Get Ollama (Llama 3.2 1B) to answer, with streaming
stream = ollama_client.chat.completions.create(
    model=MODEL_LLAMA,
    messages=messages,
    stream=True
)
response=""
display_handle = display(Markdown(response), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    update_display(Markdown(response), display_id=display_handle.display_id)
````
