# Week 1 - Day 2 Homework
**For week 1, day 2 homework, I've created a tool that generates an email message given a subject line string, but using local ollama instead of OpenAI API.**

1. It sets a custom system prompt and user prompt prefix.

2. It sets up the messages array in the correct format expected by the OpenAI API compatible ollama API endpoint.

3. It generates an email message according to the `email_subject` string.

4. Finally it displays the ollama markdown formatted response.


```python
# Step 1: Importing libs and creating my prompts.
import os
from IPython.display import Markdown, display
from openai import OpenAI

system_prompt = "You are an expert at generating email contents from subject lines"
user_prompt_prefix = """
    Generate a formatted email message with bullet lists
    and other details given the subject line provided:

"""

# Step 2: Make the OpenAI API messages list with the `email_subject` string.
def messages_for(email_subject):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + email_subject}
    ]
email_subject = "Carrot Cake - New product launch in our Pastry Ltd. website"
messages = messages_for(email_subject)

# Step 3: Call OpenAI Chat endpoint using the "llama3.2" model.
OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
response = ollama.chat.completions.create(model="llama3.2", messages=messages)
email_message = response.choices[0].message.content

# Step 4: Display the result in markdown format.
display(Markdown(email_message))
```
