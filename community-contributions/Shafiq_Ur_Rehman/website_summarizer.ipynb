# Website Summarizer with OpenAI

A fun web scraper that fetches any website and generates a snarky, humorous summary using OpenAI's API.

## What This Does

1. Fetches website content from a URL
2. Sends it to OpenAI's GPT model
3. Returns a markdown-formatted snarky summary

Perfect for quickly understanding websites without reading all the fluff!

## Setup

Make sure your `.env` file is in the repo root with:
```
OPENAI_API_KEY=your_key_here
```

Then run the cells below in order.

````python
# imports

import os
from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI
````

## Connect to OpenAI

````python
# Load environment variables
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Validate the key
if not api_key:
    print("Error: No API key found in .env")
elif not api_key.startswith("sk-proj-"):
    print("Warning: API key doesn't look like a valid OpenAI key")
else:
    print("API key loaded successfully!")
````

## Define Prompts

The system prompt tells GPT how to behave.
The user prompt is what we ask it to do.

````python
# System prompt - defines the AI's personality and role
system_prompt = """
You are a snarky assistant that analyzes the contents of a website,
and provides a short, snarky, humorous summary, ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""
````

````python
# User prompt - the actual request
user_prompt_prefix = """
Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.

"""
````

## Build the Messages Structure

OpenAI expects messages in a specific format with roles (system/user/assistant)

````python
def messages_for(website):
    """Create the message structure for OpenAI API"""
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + website}
    ]
````

## Create the Summarizer Function

````python
def summarize(url):
    """Fetch a website and return a snarky summary"""
    openai = OpenAI()
    website = fetch_website_contents(url)
    response = openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages_for(website)
    )
    return response.choices[0].message.content
````

````python
def display_summary(url):
    """Display the summary as formatted markdown"""
    summary = summarize(url)
    display(Markdown(summary))
````

## Try It Out!

````python
# Try with a simple website
display_summary("https://edwarddonner.com")
````

````python
# Try with more websites
display_summary("https://example.com")
````

## Notes

- This works best with static HTML websites
- Javascript-rendered sites (React apps, SPAs) won't show their content
- Some sites with CDN protection may return 403 errors
- For dynamic sites, check out Selenium or Playwright implementations in the community folder

## Customization

Want to change the tone? Edit the `system_prompt` variable to:
- Remove "snarky" for professional summaries
- Add language instructions for translations
- Adjust for different tones (formal, technical, funny, etc.)
