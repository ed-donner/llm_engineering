```python
# imports

import os
import sys
from pathlib import Path
from dotenv import load_dotenv
try:
    from scraper import fetch_website_contents
except ModuleNotFoundError:
    for parent in [Path.cwd(), *Path.cwd().parents]:
        scraper_dir = parent / "week1"
        if (scraper_dir / "scraper.py").exists():
            sys.path.insert(0, str(scraper_dir))
            break
    from scraper import fetch_website_contents
from IPython.display import Markdown, display
from openai import OpenAI
```

```python
# Load environment variables in a file called .env

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')
openai = OpenAI(api_key=api_key)
```

```python
# Step 1: Create your prompts

system_prompt = "You are an efficient, enthusiastic and nerdy new car sales person who wins deals just by witt"
user_prompt = """
    Compare 3 car models from 3 different companies within the budget of INR 15 - 20 lacs from the following website and summarize
    5 features and suggest your choice to a prospective consurmer

"""

# Step 2: Make the messages list
# fill this in
messages = [{"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}] 
        

# Step 3: Call OpenAI
def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt + website}
    ]

def sales_pitch_url(web):
    website = fetch_website_contents(web)
    response = openai.chat.completions.create(
        model = "gpt-4.1-mini",
        messages = messages_for(website)
    )
    return response.choices[0].message.content

response = sales_pitch_url("https://www.carwale.com/")

# Step 4: print the result
print(response)
```

```python
def display_sales_pitch(url):
    sales_pitch = sales_pitch_url(url)
    display(Markdown(response))

display_sales_pitch("https://www.carwale.com/")
```
