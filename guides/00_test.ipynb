```python
# =================================================================
# UNDERSTANDING GENERATORS
# =================================================================
#
# ── SIMPLE ENGLISH FIRST (imagine you are 10 years old) ──────────
#
# Imagine a vending machine that has 5 chocolates inside.
#
# Option A — Regular function (like a bag):
#   You ask for chocolates → it dumps ALL 5 chocolates into your hand at once.
#   You must hold everything from the start.
#
# Option B — Generator (like the vending machine):
#   You press a button → it gives you 1 chocolate.
#   You press again   → it gives you the next one.
#   It only prepares the NEXT chocolate when YOU ask for it.
#   It does NOT make all 5 at the beginning.
#
# That's it. A generator gives you things ONE AT A TIME, only when asked.
#
# =================================================================


# ── STEP 1: Regular function (the "bag" — gives everything at once) ──
print("=== Regular function ===")

def chocolates_bag():
    return [1, 2, 3, 4, 5]   # makes ALL 5 right now and hands them over

bag = chocolates_bag()
print(bag)          # → [1, 2, 3, 4, 5]   (all 5 are already in memory)


# ── STEP 2: Generator (the "vending machine" — gives one at a time) ──
print("\n=== Generator (vending machine) ===")

def chocolates_vending_machine():
    for i in range(1, 6):
        yield i      # "yield" means: give this one out, then WAIT until asked again

machine = chocolates_vending_machine()  # machine is created, but NO chocolate made yet

print(next(machine))  # you press the button → gives 1  (only 1 is made)
print(next(machine))  # press again          → gives 2
print(next(machine))  # press again          → gives 3
print(next(machine))  # press again          → gives 4
print(next(machine))  # press again          → gives 5
# next(machine) here would raise StopIteration — machine is empty!


# ── STEP 3: for-loop does the "pressing" for you automatically ───
print("\n=== for-loop with generator ===")

for choc in chocolates_vending_machine():
    print("Got chocolate:", choc)   # asks next() behind the scenes each time


# ── STEP 4: Why does this matter? (the real power) ───────────────
print("\n=== Memory difference ===")
#
# Imagine instead of 5 chocolates, you want 10 MILLION numbers.
#
# Regular list — stores ALL 10 million in RAM right now:
#   big_list = list(range(10_000_000))   ← uses ~400 MB of memory!
#
# Generator — stores only the CURRENT number, nothing else:
#   big_gen  = (x for x in range(10_000_000))  ← uses almost ZERO memory!
#
# Both let you loop over them the same way. Generator just doesn't
# prepare everything upfront.

import sys

big_list = list(range(10_000))          # regular list
big_gen  = (x for x in range(10_000))  # generator (the () makes it a generator)

print("List size  :", sys.getsizeof(big_list), "bytes")  # large
print("Generator size:", sys.getsizeof(big_gen), "bytes")  # tiny!


# ── STEP 5: Infinite generator (impossible with a list!) ─────────
print("\n=== Infinite generator ===")
#
# A list of infinite numbers is impossible — it would fill all RAM forever.
# A generator can do it easily — it just makes the NEXT one when asked.

def count_forever(start=1):
    n = start
    while True:      # no end — but that's fine, we only take what we need
        yield n
        n += 1

counter = count_forever()
print(next(counter))   # → 1
print(next(counter))   # → 2
print(next(counter))   # → 3   (could keep going forever, we just stop asking)
```

```python
def count_up_to(limit):
    n = 1
    while n <= limit:
        print(n)
        n += 1
count_up_to(5)

```
