This notebook executes a concise Day 1-5 multi-agent capstone flow.

```python
from pathlib import Path
import subprocess

PROJECT_DIR = Path.cwd()
SCRIPT = PROJECT_DIR / "week8_capstone_multi_agent.py"
SCRIPT
```

```python
import sys

def run_cmd(args):
    cmd = [sys.executable, str(SCRIPT), *args]
    print("$", " ".join(cmd))
    completed = subprocess.run(cmd, capture_output=True, text=True)
    if completed.stdout:
        print(completed.stdout)
    if completed.returncode != 0:
        print(completed.stderr)
        raise RuntimeError(f"Command failed with exit code {completed.returncode}")
```

## Day 1-2: Plan and build

```python
run_cmd(["plan"])
run_cmd(["build"])
run_cmd(["scan", "--limit", "5"])
```

## Day 3-4: Run framework once (scan + rank + notify)

```python
run_cmd(["run-once", "--limit", "5"])
```
