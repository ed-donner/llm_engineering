# Week 8 Exercise: Multi-Agent Debate System for Price Band Classification
---
This notebook implements a deliberative multi-agent debate system for price band classification.

**Architecture:**
- BudgetAdvocateAgent: Identifies value products and budget signals
- MidrangeAdvocateAgent: Expert in mainstream market positioning
- PremiumAdvocateAgent: Specialist in luxury/high-end indicators
- JudgeAgent: Impartial evaluator with systematic scoring rubric

**Key Features:**
1. Specialized Advocate Agents with unique strategies
2. Rigorous Judge Evaluation (5-dimension scoring)
3. Deliberative Process (evidence-based, honest, transparent)
4. Interactive UI (color-coded, real-time, scoring breakdown)

## 1. Import Required Libraries
Import necessary libraries for data handling, visualization, and UI.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, HTML
```

## 2. Define Utility Functions
Reusable functions for evidence grading, confidence calibration, scoring, and feature extraction.

```python
def grade_evidence(features, weights):
    score = 0
    for k, v in features.items():
        score += weights.get(k, 0) * v
    return score

def calibrate_confidence(score, max_score):
    return min(1.0, score / max_score)
```

## 3. Define Advocate Agent Classes
Implement BudgetAdvocateAgent, MidrangeAdvocateAgent, PremiumAdvocateAgent with specialized strategies.

```python
class BudgetAdvocateAgent:
    def __init__(self):
        self.weights = {'brand': 3, 'condition': 2, 'category': 2, 'model': 1}  # Brand critical
    def analyze(self, features):
        score = grade_evidence(features, self.weights)
        confidence = calibrate_confidence(score, sum(self.weights.values()))
        argument = f'Budget: Score={score}, Confidence={confidence:.2f}. Generic brand, used/damaged, older model.'
        return argument, confidence
```

```python
class MidrangeAdvocateAgent:
    def __init__(self):
        self.weights = {'brand': 2, 'condition': 2, 'category': 2, 'model': 2}  # Balanced
    def analyze(self, features):
        score = grade_evidence(features, self.weights)
        confidence = calibrate_confidence(score, sum(self.weights.values()))
        argument = f'Midrange: Score={score}, Confidence={confidence:.2f}. Established brand, current model, standard features.'
        return argument, confidence
```

```python
class PremiumAdvocateAgent:
    def __init__(self):
        self.weights = {'brand': 3, 'features': 2, 'model': 3, 'condition': 1}  # Brand/model critical
    def analyze(self, features):
        score = grade_evidence(features, self.weights)
        confidence = calibrate_confidence(score, sum(self.weights.values()))
        argument = f'Premium: Score={score}, Confidence={confidence:.2f}. Luxury brand, flagship model, professional features.'
        return argument, confidence
```

## 4. Define Judge Agent Class
JudgeAgent with systematic scoring rubric and ambiguity detection.

```python
class JudgeAgent:
    def __init__(self):
        self.dimensions = ['Evidence Grounding', 'Internal Consistency', 'Domain Plausibility', 'Counterpoint Handling', 'Confidence Calibration']
    def evaluate(self, arguments, confidences):
        scores = []
        for arg, conf in zip(arguments, confidences):
            # Dummy scoring: each dimension gets confidence * 2 (for demo)
            score = [conf * 2 for _ in self.dimensions]
            scores.append(score)
        winner = np.argmax([sum(s) for s in scores])
        return scores, winner
```

## 5. Product Feature Extraction
Extract relevant product features for agent analysis.

```python
def extract_features(product):
    # Example: product is dict with keys
    return {
        'brand': product.get('brand_score', 0),
        'condition': product.get('condition_score', 0),
        'category': product.get('category_score', 0),
        'model': product.get('model_score', 0),
        'features': product.get('features_score', 0)
    }
```

## 6. Agent Argument Generation
Each advocate agent generates evidence-based arguments, acknowledges weaknesses, and calibrates confidence.

```python
product = {
    'brand_score': 2,
    'condition_score': 1,
    'category_score': 2,
    'model_score': 1,
    'features_score': 1
}  # Example product
features = extract_features(product)
budget_agent = BudgetAdvocateAgent()
midrange_agent = MidrangeAdvocateAgent()
premium_agent = PremiumAdvocateAgent()
agents = [budget_agent, midrange_agent, premium_agent]
arguments = []
confidences = []
for agent in agents:
    arg, conf = agent.analyze(features)
    arguments.append(arg)
    confidences.append(conf)
```

## 7. Judge Evaluation and Scoring
JudgeAgent evaluates each argument, scores across five dimensions, and determines the winner.

```python
judge = JudgeAgent()
scores, winner = judge.evaluate(arguments, confidences)
print('Scores:', scores)
print('Winner:', ['Budget', 'Midrange', 'Premium'][winner])
```

## 8. Interactive Debate Visualization (UI)
Color-coded, real-time UI to display agent arguments, scores, and debate flow.

```python
colors = ['#4CAF50', '#FFC107', '#2196F3']
agent_names = ['Budget', 'Midrange', 'Premium']
boxes = []
for i, arg in enumerate(arguments):
    box = widgets.HTML(f'<div style="background:{colors[i]};padding:10px;margin:5px;border-radius:5px">{agent_names[i]} Advocate:<br>{arg}</div>')
    boxes.append(box)
display(boxes[i])
score_box = widgets.HTML(f'<b>Judge Scores:</b> {scores}')
display(score_box)
winner_box = widgets.HTML(f'<b>Winner:</b> {agent_names[winner]}')
display(winner_box)
```

## 9. Clarification Mechanism
JudgeAgent requests clarification if ambiguity detected.

```python
def judge_clarification(arguments):
    # Dummy: If all confidences < 0.5, request clarification
    if all('Confidence=0.' in arg for arg in arguments):
        return 'Judge requests clarification from advocates.'
    return 'No clarification needed.'
clarification = judge_clarification(arguments)
print(clarification)
```
