```python
import os
from openai import OpenAI
from agents import ResearcherAgent, AnalyzerAgent, ReporterAgent
```

```python
#Load your OpenAI / OpenRouter API key
# Make sure your OPENAI_API_KEY or OpenRouter API key is set in environment
api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    raise ValueError("Set your OPENAI_API_KEY in environment variables!")

# OpenRouter client
openrouter = OpenAI(
    api_key=api_key,
    base_url="https://openrouter.ai/api/v1"
)

print("OpenRouter client ready.")
```

```python
# nitialize Agents
# Initialize agents with OpenRouter client
researcher = ResearcherAgent(use_llm=True, client=openrouter)
analyzer = AnalyzerAgent(use_llm=True, client=openrouter)
reporter = ReporterAgent()

print("Agents initialized.")
```

```python
# Define topic to research
topic = "Emerging AI tools for small businesses"
print(f"Topic: {topic}")
```

```python
# Research using ResearcherAgent
research = researcher.research(topic)
print("Research complete:\n")
print(research)
```

```python
# Analyze research using AnalyzerAgent
analysis = analyzer.analyze(research)
print("Analysis complete:\n")
print(analysis)
```

```python
# Generate final report using ReporterAgent
report = reporter.report(topic, research, analysis)
print("Final Report:\n")
print(report)
```

```python
# Wrap in a simple function for reuse
def run_pipeline(topic: str):
    research = researcher.research(topic)
    analysis = analyzer.analyze(research)
    report = reporter.report(topic, research, analysis)
    return report

# Example usage
topic2 = "Latest trends in electric vehicles"
final_report = run_pipeline(topic2)
print(final_report)
```

```python
# Import Gradio and define interface function
import gradio as gr

def generate_report(topic: str) -> str:
    if not topic.strip():
        return "Please enter a topic."
    try:
        return run_pipeline(topic.strip())
    except Exception as e:
        return f"Error generating report: {e}"
```

```python
# Launch Gradio Chat UI
gr.Interface(
    fn=generate_report,
    inputs=gr.Textbox(
        label="Enter a topic to research",
        lines=2,
        placeholder="e.g., Emerging AI tools for small businesses"
    ),
    outputs=gr.Textbox(
        label="Generated Report"
    ),
    title="Week 8 — Multi-Agent Research Assistant",
    description="Type a topic, and the Researcher, Analyzer, and Reporter agents will produce a summarized report."
).launch()
```
