# Deal Scanner Agent with Structured Outputs

Retailers need automated deal detection from scraped listings.

We build:

- GPT-4 structured JSON outputs
- Pydantic schema validation
- Simple tool orchestration
- Agent-style loop

## Concepts Applied

- Structured outputs
- Constrained JSON decoding
- Pydantic validation
- Tool-based workflow

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from pydantic import BaseModel
```

```python
load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
MODEL = "gpt-4"
```

schema

```python
class Deal(BaseModel):
    product: str
    original_price: float
    discounted_price: float
    is_good_deal: bool
```

structured agent

```python
def analyze_listing(text):
    response = client.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": "Extract deal info in JSON format."},
            {"role": "user", "content": text}
        ]
    )

    deal = Deal.model_validate_json(response.choices[0].message.content)
    return deal

analyze_listing("Samsung TV originally 700000 KES now 550000 KES limited offer")
```
