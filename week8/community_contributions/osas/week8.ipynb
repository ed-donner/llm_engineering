````python
import gradio as gr
from modal import Cls
````

````python
# -----------------------------
ShoppingAdvisor = Cls.from_name("shopping-advisor", "ShoppingAdvisor")
advisor = ShoppingAdvisor()
````

````python
def evaluate_product(description, price):
    # Call the deployed Modal method synchronously
    result = advisor.evaluate.remote(description, float(price))
    return (
        f"Recommendation: {result.get('recommendation', 'N/A')}\n"
        f"Estimated Price: ${result.get('estimated_price', 'N/A')}\n"
        f"Difference: ${result.get('difference', 'N/A')}\n"
        f"Reason: {result.get('reason', 'N/A')}"
    )
````

````python
with gr.Blocks(title="Shopping Advisor", fill_width=True) as ui:

    gr.Markdown("## 🛍️ Shopping Advisor — Is It a Good Deal?")

    with gr.Row():
        with gr.Column(scale=3):
            description_input = gr.Textbox(
                label="Product Description",
                placeholder="e.g. Sony WH-1000XM4 Wireless Noise Canceling Headphones",
                lines=3
            )
        with gr.Column(scale=1):
            price_input = gr.Number(label="Actual Price ($)", precision=2)

    submit_btn = gr.Button("🔍 Evaluate Deal", variant="primary")

    output = gr.Textbox(
        label="Evaluation Result",
        lines=6,
        interactive=False
    )

    submit_btn.click(
        fn=evaluate_product,
        inputs=[description_input, price_input],
        outputs=output
    )

ui.launch()
````

````python
evaluate_product("Sony WH-1000XM4 Wireless Noise Canceling Headphones", 289.0)
````
