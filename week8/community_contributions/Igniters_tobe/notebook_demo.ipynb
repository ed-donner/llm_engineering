# Multi-Agent Research Assistant Demo

This notebook demonstrates the collaboration between three specialized agents:
1. **Research Agent**: Gathers raw data from the web.
2. **Fact Checker Agent**: Verifies links and extracts key facts.
3. **Writer Agent**: Synthesizes the final explanation.

## Configuration
Ensure you have your OpenRouter API key and SerpAPI key in a `.env` file or set as environment variables.

```python
import os
import sys
from dotenv import load_dotenv

# Add parent directory to path to allow importing the package
current_dir = os.getcwd()
parent_dir = os.path.dirname(current_dir)
if parent_dir not in sys.path:
    sys.path.append(parent_dir)

load_dotenv() # Load OPENROUTER_API_KEY and SERP_API_KEY
```

## Step 1: Research Phase
The Research Agent uses the `web_search` tool.

```python
from Igniters_tobe.agents.research_agent import run_research

query = "What are the top 3 benefits of multi-agent systems in AI?"
raw_research = run_research(query)
print(f"Raw Research Output:\n{raw_research}")
```

## Step 2: Fact Checking Phase
The Fact Checker verifies the research and might use the `url_scraper` tool.

```python
from Igniters_tobe.agents.fact_checker_agent import run_fact_check

verified_facts = run_fact_check(raw_research)
print(f"Verified Facts:\n{verified_facts}")
```

## Step 3: Writing Phase
The Writer Agent produces the final polished response.

```python
from Igniters_tobe.agents.writer_agent import run_writing

final_answer = run_writing(verified_facts)
print(f"Final Answer:\n{final_answer}")
```
