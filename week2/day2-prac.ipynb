```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr # oh yeah!
```

```python
load_dotenv(override=True)
api_key = os.getenv('OPENROUTER_API_KEY')
OPENROUTER_BASE_URL = "https://openrouter.ai/api/v1"

if api_key:
    print("Yes. available")
```

```python
openai = OpenAI(api_key=api_key, base_url= OPENROUTER_BASE_URL)
```

```python
system_prompt = "You are a good helper, always reply in breif and in md format"

def message_gpt(prompt):
    message = [{"role": "system", "content": system_prompt}, {"role": "user", "content": prompt}]
    response = openai.chat.completions.create(model="gpt-4.1-mini", messages= message, max_tokens=150, temperature=0.7)
    result = response.choices[0].message.content
    # display (Markdown(result))
    return result
```

```python
message_gpt("hi")
```

```python
system_prompt = "You are a AI teacher and tells about best things about what user asks in points and markdown format"

message_input = gr.Textbox(label="Your message", info="Type your message", lines=7)
message_output = gr.Markdown(label="Your output",)

view = gr.Interface(
    fn = message_gpt,
    title = "GPT",
    inputs = [message_input],
    outputs = [message_output],
    flagging_mode = "never"
)
view.launch()
```

```python
def stream_gpt(prompt):
    message = [{"role": "system", "content": system_prompt}, {"role": "user", "content": prompt}]
    response = openai.chat.completions.create(model="gpt-4.1-mini", messages= message, max_tokens=150, temperature=0.7, stream=True)
    result = ""
    for chunk in response:
        if chunk.choices[0].delta.content:
            result += chunk.choices[0].delta.content
            yield result
```

```python
system_prompt = "You are a AI teacher and tells about best things about what user asks in points and markdown format"
message_input = gr.Textbox(label="Your message", info="Type your message", lines=7)
message_output = gr.Markdown(label="Your output",)
gr.Interface(
    fn = stream_gpt,
    title = "GPT",
    inputs = [message_input],
    outputs = [message_output]
).launch()
```
