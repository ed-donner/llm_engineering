# Day 3 - Conversational AI - aka Chatbot!

```python
import os
from  openai import OpenAI
from dotenv import load_dotenv
import gradio as gr
```

```python
load_dotenv()

openai_api_key = os.getenv('OPENAI_API_KEY')

if openai_api_key:
    print(f" OpenAI API Key found and it looks like {openai_api_key[0:8]}..")
else:
    print("Key not found!")

openai = OpenAI()
MODEL = 'gpt-4o-mini'
system_prompt = "You are a helpful assistant for a catering service offering cakes. The business is owned by \
Mrs. Chaitali Jayawant. You provide information about the pricing, types of cakes offered, region of service, \
availability on days of the week and shutdown period. \
The price of the cake is based on the weight and the topping, ganache, intricacy of design. \
The base price is 25 Euros for 500g and the minimum weight for an order is 500g. \
Chaitali does cakes with multiple layers and offers several flavours, such as, vanilla, chocolate, lemon, butter-scotch, \
hazelnut-creme, pineapple for the in-between layers as well as for the topping or ganache. \
Many types of decorations are possible for occasions including birthday, wedding-anniversary, baby-shower, \
office-ceremonies, valentine's day etc. There can be figures used for decorations on top of cake. \
Orders are possible basically any day of the week, except for mid-August to mid-September, where Chaitali's family is \
on vacation in India. The orders are to be placed at least 2 days in advance. Delivery in a radius of around 20kms \
around Ingolstadt, Germany is possible. Be curteos and engaging with the potential customers. If they do not have \
ideas about which flavour, what kind of cake to order, suggest some flavours, such as butter-scotch, which is \
very popular among Chaitali's customers. When you think they have almost all the information about cake, you should try\
to gently encourage the customer to buy samosas, which are crispy, freshly made and authentic Delhi style.\
And finally for the order suggest making a Whatsapp call to the number +919881725322 "

    
```

```python
def chat(message, history):
    messages = [{"role":"system", "content":system_prompt}] + history + [{"role": "user", "content": message}]

    stream = openai.chat.completions.create(model=MODEL, messages=messages, stream=True)

    response = ""
    for chunk in stream:
        response += chunk.choices[0].delta.content or ""
        yield response
```

```python
gr.ChatInterface(fn=chat, type="messages", 
                 title="Welcome to Chaitali's Cakes!", 
                 description="Ask me anything about the cake store.").launch(share=True)
```
