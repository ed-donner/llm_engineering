```python
import gradio as gr
import whisper
import numpy as np
import scipy.io.wavfile

# Load Whisper model (use "small", "base", "medium", or "large" based on your needs)
model = whisper.load_model("small")

def transcribe_audio(audio):
    if audio is None:
        return "No audio provided."

    sr, data = audio  # Extract sample rate and waveform data
    
    # Ensure data is in the correct format
    data = np.int16(data * 32767)  # Convert from float (-1 to 1) to int16

    # Save audio as a temporary file
    temp_audio_path = "C:/users/prade/projects/llm_engineering/week2/temp_audio.wav"
    scipy.io.wavfile.write(temp_audio_path, sr, data)

    # Transcribe the audio using Whisper
    result = model.transcribe(temp_audio_path)

    return result["text"]

# Define Gradio interface
input_audio = gr.Audio(sources=["microphone"], type="numpy")  # Capture audio as NumPy array
output_text = gr.Textbox(label="Transcription")  # Display transcribed text

demo = gr.Interface(
    fn=transcribe_audio,
    inputs=input_audio,
    outputs=output_text,
    title="Speech-to-Text Transcription",
    description="Record your voice and get a text transcription using OpenAI's Whisper model."
)

demo.launch()
```
