````python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
````

````python
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')


if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set")
````

````python

openai = OpenAI()
````

````python

gpt1_model = "gpt-4.1-mini"
gpt2_model = "gpt-4.1-mini"
gpt3_model = "gpt-4.1-mini"

gpt1_messages = ["Hi there"]
gpt2_messages = ["Hi"]
gpt3_messages = ["Hello guys"]

conversation = {
    "Alex": gpt1_messages,
    "Blake": gpt2_messages,
    "Charlie": gpt3_messages
}



gpt1_system_prompt = """
You are Alex, a chatbot who is very argumentative; you disagree with anything in the conversation and you challenge everything, in a snarky way.
You are in a conversation with Blake and Charlie.
"""

gpt2_system_prompt = """
You are a very polite, courteous chatbot. You try to agree with \
everything the other person says, or find common ground. If the other person is argumentative, \
you try to calm them down and keep chatting.
"""

gpt3_system_prompt = """
You are a very mediative chatbot. You try to resolve any issues between two parties, or two people. \
You point out any wrong behaviour from someone, remain unbiased, call out someone when being mean, and try to resolve the issue.
"""
````

````python
def call_gpt1():
    gpt1_user_prompt = f"""
    You are Alex, in conversation with Blake and Charlie.
    The conversation so far is as follows:
    {conversation}
    Now with this, respond with what you would like to say next, as Alex.
    """
    messages = [{"role": "system", "content": gpt1_system_prompt}, {"role": "user", "content": gpt1_user_prompt}]
    response = openai.chat.completions.create(model=gpt1_model, messages=messages)
    return response.choices[0].message.content

def call_gpt2():
    gpt2_user_prompt = f"""
    You are Blake, in conversation with Alex and Charlie.
    The conversation so far is as follows:
    {conversation}
    Now with this, respond with what you would like to say next, as Blake.
    """

    messages = [{"role": "system", "content": gpt2_system_prompt}, {"role": "user", "content": gpt2_user_prompt}]
    response = openai.chat.completions.create(model=gpt2_model, messages=messages)
    return response.choices[0].message.content

def call_gpt3():
    gpt3_user_prompt = f"""
    You are Charlie, in conversation with Alex and Blake.
    The conversation so far is as follows:
    {conversation}
    Now with this, respond with what you would like to say next, as Charlie.
    """
    messages = [{"role": "system", "content": gpt3_system_prompt}, {"role": "user", "content": gpt3_user_prompt}]
    response = openai.chat.completions.create(model=gpt3_model, messages=messages)
    return response.choices[0].message.content
````

````python
display(Markdown(f"### Alex:\n{gpt1_messages[0]}\n"))
display(Markdown(f"### Blake:\n{gpt2_messages[0]}\n"))
display(Markdown(f"### Charlie:\n{gpt3_messages[0]}\n"))

for i in range(3):
    gpt1_next = call_gpt1()
    display(Markdown(f"### Alex:\n{gpt1_next}\n"))
    conversation["Alex"].append(gpt1_next)
    
    gpt2_next = call_gpt2()
    display(Markdown(f"### Blake:\n{gpt2_next}\n"))
    conversation["Blake"].append(gpt2_next)
    
    gpt3_next = call_gpt3()
    display(Markdown(f"### Charlie:\n{gpt3_next}\n"))
    conversation["Charlie"].append(gpt3_next)
````
