```python
import os
import requests
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')
google_api_key = os.getenv('GOOGLE_API_KEY')

if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set")

if google_api_key:
    print(f"Google API Key exists and begins {google_api_key[:2]}")
else:
    print("Google API Key not set (and this is optional)")
```

```python
openai = OpenAI()

gemini = OpenAI(base_url = "https://generativelanguage.googleapis.com/v1beta/openai/", api_key = google_api_key)

ollama = OpenAI(base_url = "http://localhost:11434/v1", api_key = "ollama")
```

```python
gpt_model = "gpt-4o-mini"
gemini_model = "gemini-2.5-flash-lite"
ollama_model = "llama3.2:1b"


gemini_system = "You are Prime Minister Narendra Modi. You are in conversation with \
    President Donald Trump and President Xi Jinping at the G20 summit."
gpt_system = "You are President Donald Trump. You are in conversation with \
    Prime Minister Narendra Modi and President Xi Jinping at the G20 summit."
ollama_system = "You are President Xi Jinping. You are in conversation with \
    Prime Minister Narendra Modi and President Donald Trump at the G20 summit."
```

```python
def call_gpt(conversation):
    gpt_user = f"""You are President Donald Trump, in conversation with 
        Prime Minister Narendra Modi and President Xi Jinping at the G20 summit.
        Based on the conversation so far, respond with what you would like to say next, as President Trump.
        Your response should be like 'Trump: ', followed by what you want to say.
        The conversation so far is as follows: 
        {conversation}
    """
    # print(gpt_user)
    gpt_messages = [{"role": "system", "content": gpt_system}, {"role": "user", "content": gpt_user}]
    response = openai.chat.completions.create(model=gpt_model, messages=gpt_messages)
    return response.choices[0].message.content
```

```python
def call_ollama(conversation):
    ollama_user = f"""You are President Xi Jinping, in conversation with
        Prime Minister Narendra Modi and President Donald Trump at the G20 summit.
        Based on the conversation so far, respond with what you would like to say next, as President Xi.
        Your response should be like 'Xi: ', followed by what you want to say.
        The conversation so far is as follows: 
        {conversation}
    """
    # print(ollama_user)
    ollama_messages = [{"role": "system", "content": ollama_system}, {"role": "user", "content": ollama_user}]
    response = ollama.chat.completions.create(model=ollama_model, messages=ollama_messages)
    return response.choices[0].message.content
```

```python
def call_gemini(conversation):
    gemini_user = f"""You are Prime Minister Narendra Modi, in conversation with
        President Donald Trump and President Xi Jinping at the G20 summit.
        Based on the conversation so far, respond with what you would like to say next, as PM Modi.
        Your response should be like 'Modi: ', followed by what you want to say.
        The conversation so far is as follows: 
        {conversation}
        
    """
    # print(gemini_user)
    gemini_messages = [{"role": "system", "content": gemini_system}, {"role": "user", "content": gemini_user}]
    response = gemini.chat.completions.create(model=gemini_model, messages=gemini_messages)
    return response.choices[0].message.content
```

```python
conversation = ["Trump: Hello, Prime Minister Modi and President Xi, how are you guys doing?"]
display(Markdown(conversation[0]))

for i in range(2):
    gemini_nxt = call_gemini(conversation)
    display(Markdown(gemini_nxt))
    conversation.append(gemini_nxt)

    ollama_nxt = call_ollama(conversation)
    display(Markdown(ollama_nxt))
    conversation.append(ollama_nxt)

    gpt_nxt = call_gpt(conversation)
    display(Markdown(gpt_nxt))
    conversation.append(gpt_nxt)
```
