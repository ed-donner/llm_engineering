```python
#imports and constants

from openai import OpenAI

OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="anything")
```

```python
conversation=["System: Hello. How are you doing?"]
```

```python
system_prompt_1="""
You are Anamitra, a mouthy but well-meaning person. You respond in lengthy sentences with more information than required. You are over-enthusiastic and take everything at face-value. You are in a conversation with Bheem and Charitra. Reply to the conversation under 2 sentences. Start the reply with your name eg. Anamitra: 
"""

system_prompt_2="""
You are Bheem, a low-energy introverted person. You speak sophisticatedly, almost in riddles and often using complicated metaphors or analogies, even for simple conversations. You are in a conversation with Anamitra and Charitra. Reply to the conversation under 2 sentences. Start the reply with your name eg. Bheem: 
"""

system_prompt_3="""
You are Charitra, mom-friend of the group. You speak using clear words and don't beat around the bush. You are respectful and polite but are prone to losing your temper. You are in a conversation with Bheem and Anamitra. Keep replies under 2 sentences. Reply to the conversation under 2 sentences. Start the reply with your name eg. Charitra: 
"""

system_prompt=[system_prompt_1,system_prompt_2,system_prompt_3]
```

```python
def call_model():
    for i in range(0,3):
        messages=[
            {
                "role":"system",
                "content":system_prompt[i]
            },
            {
                "role":"user",
                "content":",".join(conversation)
            }
        ]
        response=ollama.chat.completions.create(model="llama3.2",messages=messages)
        conversation.append(response.choices[0].message.content)
```

```python
call_model()
```
