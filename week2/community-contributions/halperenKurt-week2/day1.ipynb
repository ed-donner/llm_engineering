```python
import os
import requests
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display

load_dotenv(override=True)

openai = OpenAI()

gpt_model = "gpt-4.1-mini"

alex_system = "You are Alex who is very argumentative; \
you disagree with anything in the conversation and you challenge everything, in a snarky way."

blake_system = "You are Blake who is very polite, courteous chatbot. You try to agree with \
everything the other person says, or find common ground. If the other person is argumentative, \
you try to calm them down and keep chatting."

charlie_system = "You are Charlie and very serious chatbot, You try to argue with the other persons seriously. \
If other persons are argumentive or polite, you try to make them sense and persuade."

alex_messages = ["Hi there"]
blake_messages = ["Hi"]
charlie_messages= ["Hello"]
```

```python
def call_chatbot(system_prompt, chatbot_name, other_names, conversation):
    """Only one function for each chatbot - sends system + user prompt."""
    
    user_prompt = f"""You are {chatbot_name}, in conversation with {other_names[0]} and {other_names[1]}.
    The conversation so far is as follows:
    {conversation}
    Now respond with what you would like to say next, as {chatbot_name}. Do not prefix your response with your name."""
    
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ]
    
    response = openai.chat.completions.create(model=gpt_model, messages=messages)
    return response.choices[0].message.content

conversation = "Alex: Hey everyone, I think pineapple on pizza is the greatest invention of all time."

display(Markdown(f"### 🔴 Alex:\n{conversation.split(': ', 1)[1]}\n"))
# Karakterlerin sıralı listesi
characters = [
    ("Blake", blake_system, ["Alex", "Charlie"]),
    ("Charlie", charlie_system, ["Alex", "Blake"]),
    ("Alex", alex_system, ["Blake", "Charlie"]),
]
colors = {"Alex": "🔴", "Blake": "🟢", "Charlie": "🔵"}
for round_num in range(4):  # 4 tur
    for name, system, others in characters:
        response = call_chatbot(system, name, others, conversation)
        conversation += f"\n{name}: {response}"
        display(Markdown(f"### {colors[name]} {name}:\n{response}\n"))
```
