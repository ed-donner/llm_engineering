```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr
```

```python
load_dotenv(override=True)
openai_api_key = os.getenv("OPENAI_API_KEY")

if openai_api_key:
    print(f"OpenAI API Key exists adn begins with {openai_api_key[:8]}")
else:
    print("OpenAI API Key does not exist")
```

```python
#Initialze

openai = OpenAI()
MODEL = "gpt-4o-mini"
```

```python
system_message = """ You are a helpful word chain validator.\
     A word chain is considered valid if it consists of two words in the english language. \
     The words must be related to each other and must be used together usually in the english language.\
     Example valid words : Tree House, Light Bulb, Source Code, Weight gain, etc.\
    
    Do not enagage in any conversation with the user. Just inform if the word is valid or not with proper and brief explaination. Use example sentences to explain usage in english language.
    Keep your responses short and brief.\

    If user enters anything else, redirect user and inform the user your role. Give example inputs to help the user.
         """
```

```python
# Let's create a call that streams back results
# If you'd like a refresher on Generators (the "yield" keyword),
# Please take a look at the Intermediate Python guide in the guides folder

def stream_gpt(prompt):
    messages = [
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
      ]
    stream = openai.chat.completions.create(
        model='gpt-4.1-mini',
        messages=messages,
        stream=True
    )
    result = ""
    for chunk in stream:
        result += chunk.choices[0].delta.content or ""
        yield result
```

```python
message_input = gr.Textbox(label="Word Chain Validator", info="Enter a two worded word chain to validate if they are correct", lines=7)
message_output = gr.Markdown(label="Response:")

view = gr.Interface(
    fn=stream_gpt,
    title="GPT", 
    inputs=[message_input], 
    outputs=[message_output], 
    examples=[
        "Light Bulb",
        "Tree House",
        "Animal color",
        "Red white"
        ], 
    flagging_mode="never"
    )
view.launch()
```
