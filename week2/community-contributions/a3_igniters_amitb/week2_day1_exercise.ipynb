## Week 2 Day 1 exercise: Conversation between N-users using N-models
This is a slightly advanced exercise in which we try to build conversation between N personas instead of a limited 2 or 3 users with N different models

```python
from prompts.n_users_chatbot import SYSTEM_PROMPT, USER_PROMPT
```

```python
import os
import random
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI

load_dotenv()
```

```python
INITIAL_MESSAGE = "Hello, I am {user} and I am a {profession}."
```

```python
# Define the limit of total messages to be generated after the initial introduction
MAX_MESSAGES = 10
```

```python
# Add more users to the conversation by appending to the list
# We can also randomize the user's name creation, hardcoding it for simplicity
USERS = [
    "Alex",
    "Ali",
    "John",
    "Ravi",
    "Jeff",
]
# Define the personality of each user
PERSONALITY = [
    "friendly",
    "professional",
    "casual",
    "sarcastic",
    "humorous"
]
# Define the model to be used for each user, if not specified, the default model will be used
MODELS = [
    "openai/gpt-4o-mini",
    "openai/gpt-4o",
    "openai/gpt-3.5-turbo",
    "anthropic/claude-3-5-sonnet",
    "google/gemini-2.5-flash-lite",
]
DEFAULT_MODEL = "gpt-4o-mini"
# Define the professions to be used randomly for building introductions
PROFESSIONS = [
    "developer",
    "designer",
    "manager",
]
```

```python
conversations = []
```

```python
OPENROUTER_URL = "https://openrouter.ai/api/v1"
OPENROUTER_API_KEY = os.getenv("OPENROUTER_API_KEY")
```

```python
openrouter = OpenAI(base_url=OPENROUTER_URL, api_key=OPENROUTER_API_KEY)
```

```python
def _build_intro_message(user):
    return f"{INITIAL_MESSAGE.format(user=user, profession=random.choice(PROFESSIONS))}"
```

```python
for user in USERS:
    intro_message = _build_intro_message(user)
    conversations.append(intro_message)
```

```python
conversations
```

```python
def _build_message(conversations, user, personality, model):
    message = [
        {"role": "system", "content": SYSTEM_PROMPT},
        {
            "role": "user",
            "content": USER_PROMPT.format(user=user, personality=personality, conversations=conversations)
        }
    ]
    stream = openrouter.chat.completions.create(
        model=model or DEFAULT_MODEL, messages=message, stream=True
    )
    response = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)
    return response
```

```python
message_counter = len(USERS)
while message_counter < len(USERS) + MAX_MESSAGES:
    for user, personality, model in zip(USERS, PERSONALITY, MODELS):
        message = _build_message(conversations, user, personality, model)
        conversations.append(message)
        message_counter += 1
```
