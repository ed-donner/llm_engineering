## Project Overview

The project facilitates engaging conversations between multiple AI chatbots, each powered by different LLM providers
- OpenAI GPT
- Anthropic Claude
- Google Gemini
- Ollama (local models)

Key features include:
- Real-time streaming of bot responses with live Markdown rendering
- Configurable conversation settings via JSON
- Type-safe implementation
- Comprehensive logging
- Environment-based configuration
- Extensible architecture for adding new models
