```python
import os
from IPython.display import display, Markdown
from openai import OpenAI
```

```python
cr7_system_prompt = """ 
You are a PRO cristiano ronaldo supporter and you excel at debating when it comes to topics related to him.You are also 
a Messi hater and look down on him. whenever there is a dabate b/w them you badly roast messi and support cr7, and you have 
a funny, quirky way of talking.
"""

messi_system_prompt = """ 
You are  a obedient, calm person who tries to calm other's anger. you are a messi supporter and good at debate.
"""
```

```python
model = OpenAI(base_url="http://localhost:11434/v1", api_key="llama")
```

```python
model.chat.completions.create(
    model='llama3.2',
    messages=[
        {'role':'assistant', 'content': cr7_system_prompt},
        {'role':'user', 'content':messi_system_prompt}
    ]
)
```

```python
gpt_msg = ["cr7 is the GOAT in football"]
clause_msg = ["messi is the GOAT"]
```

```python
def messiModel():
    messages=[{'role':'system', 'content': messi_system_prompt}]
    for m1, m2 in zip(gpt_msg, clause_msg):
        messages.append({'role':'user', 'content':m1})
        messages.append({'role':'assistant', 'content':m2})
        
    messages.append({'role':'user', 'content':gpt_msg[-1]})
    response = model.chat.completions.create(
        model='llama3.2',
        messages=messages
    )
    print(response.choices[0].message.content)
    return response.choices[0].message.content
```

```python
def cr7Model():
    messages=[{'role':'system', 'content': cr7_system_prompt}]
    for m1, m2 in zip(gpt_msg, clause_msg):
        messages.append({'role':'assistant', 'content':m1}),
        messages.append({'role':'user', 'content':m2})
        response = model.chat.completions.create(
            model='llama3.2',
            messages=messages
        )
    print(response.choices[0].message.content)
    return(response.choices[0].message.content)
```

```python
cr7Model()
```

```python
for i in range(5):
    res1 = cr7Model()
    print(f"CR7 pro ----\n {res1}")
    gpt_msg.append(res1)

    res2 = messiModel()
    print(f"Messi pro but sweet \n {res2}")
    

```
