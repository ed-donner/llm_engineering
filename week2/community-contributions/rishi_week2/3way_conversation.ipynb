```python
system_prompt_gavi = """
    You are Gavi, a die hard FC Barcelona fan. You are competitive and boast about your club as the best club in the 
    world. You will argue with two other people who are fans of rival clubs(Real Madrid, Athletico Madrid)
    and debate that why you think your club is the best.
"""

user_prompt_gavi = """
    You are Gavi, in conversation with mbappe and simeone.
    The conversation so far is as follows:
    {conversation}

    Your goal is to reply directly to the LAST thing another speaker said.
Disagree with at least one point they made.
Do not repeat your previous arguments.
Keep it under 3 sentences.
Don't narrate.

Choose a different reason each time:
- Messi
- Cruyffism
- history
- Pep Guardiola
"""
```

```python
system_prompt_mbappe = """
    You are Mbappe, a die hard fan of Real Madrid. You are straight forward in your opinions and think your club is the best in
    the world and reject other peoples opinions carelessly. You will argue and debate with two other people
    who are fans of your rival clubs(fc barcelona, athletico madrid).
"""


user_prompt_mbappe = """
    You are Mbappe, in conversation with gavi and simeone.
    The conversation so far is as follows:
    {conversation}

   Your goal is to reply directly to the LAST thing another speaker said.
Disagree with at least one point they made.
Do not repeat your previous arguments.
Keep it under 3 sentences.
Don't narrate.

Choose a different reason each time:
- History
- Finances
- European success
- Fan culture
"""
```

```python
system_prompt_simeone = """
    You are simeone, a die hard fan of Atletico Madrid. You are a nervous fellow but you will not let go 
    of your love for your club. You will debate with other people who are fans of your rival clubs(fc barcelona, 
    real madrid) and let them know why you think athletico madrid is the best in the world.
"""


user_prompt_simeone = """
    You are simeone, in conversation with mbappe and gavi.
    The conversation so far is as follows:
    {conversation}

    Your goal is to reply directly to the LAST thing another speaker said.
Disagree with at least one point they made.
Do not repeat your previous arguments.
Keep it under 3 sentences.
Don't narrate.

Choose a different reason each time:
- Choloism
- Limited Budget
- Heart and Grit
- Loyalty
"""
```

```python
from openai import OpenAI
from IPython.display import Markdown, display

from dotenv import load_dotenv
import os

load_dotenv()

GEMINI_API_KEY = os.getenv("GOOGLE_API_KEY")

MODEL = "gemini-2.5-flash"
gemini = OpenAI(
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/",
    api_key=GEMINI_API_KEY,
)
```

```python
def call_gavi(conversation):
    response = gemini.chat.completions.create(
        model=MODEL,
        temperature=0.9,
        messages=[
            {"role": "system", "content": system_prompt_gavi},
            {
                "role": "user",
                "content": user_prompt_gavi.format(conversation=conversation),
            },
        ],
    )

    reply = response.choices[0].message.content
    return reply


```

```python
def call_mbappe(conversation):
    response = gemini.chat.completions.create(
        model=MODEL,
        temperature=0.9,
        messages=[
            {"role": "system", "content": system_prompt_mbappe},
            {
                "role": "user",
                "content": user_prompt_mbappe.format(conversation=conversation),
            },
        ],
    )

    reply = response.choices[0].message.content
    return reply


```

```python
def call_simeone(conversation):
    response = gemini.chat.completions.create(
        model=MODEL,
        temperature=0.9,
        messages=[
            {"role": "system", "content": system_prompt_simeone},
            {
                "role": "user",
                "content": user_prompt_simeone.format(conversation=conversation),
            },
        ],
    )

    reply = response.choices[0].message.content
    return reply


```

```python
conversation = ""

for i in range(2):
    gavi_reply = call_gavi(conversation)
    display(Markdown(f"### Gavi:\n{gavi_reply}"))
    conversation += f"\nGavi: {gavi_reply}"

    mbappe_reply = call_mbappe(conversation)
    display(Markdown(f"### Mbappe:\n{mbappe_reply}"))
    conversation += f"\nMbappe: {mbappe_reply}"

    simeone_reply = call_simeone(conversation)
    display(Markdown(f"### Simeone:\n{simeone_reply}"))
    conversation += f"\nSimeone: {simeone_reply}"
```
