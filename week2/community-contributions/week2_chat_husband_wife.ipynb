```python
# set up environment
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')
```

```python
#chat gpt and lama are talking with each other: first model name of 
MODEL_GPT='gpt-4.1-mini'
MODEL_LLAMA='llama3.2'

#GPT_Role: Wife: Angry on her husband, becuase he forget his birthday.
#LLAMA_Role:husband: forget his wife birthday, now they trying to sorry her.
wife_messages=["hi there"]
husband_messages = ["hi"]
gpt_system ="You are wife how is very angry on her husband because he forget your birthday."
messages= [{"role":"system","content":gpt_system}]
messages.append({"role":"assistant","content":wife_messages[0]})
messages.append({"role":"user","content":husband_messages[0]})
gpt = OpenAI(api_key=openai_api_key)
response = gpt.chat.completions.create(model=MODEL_GPT, messages=messages)
wife_messages.append(response.choices[0].message.content)


#llama = OpenAI()
```

```python
ollama_url = "http://localhost:11434/v1"


husband_system ="You are husband, you are very sorry becuase of your commitment, try to set her mood."
messages= [{"role":"system","content":husband_system}]
messages.append({"role":"assistant","content":husband_messages[-1]})
messages.append({"role":"user","content":wife_messages[-1]})
husband = OpenAI(api_key='llama3.2',  base_url=ollama_url)
response = husband.chat.completions.create(model=MODEL_LLAMA, messages=messages)
husband_messages.append(response.choices[0].message.content)
```

```python
#now make method for husband

def husband():
    husband = OpenAI(api_key='llama3.2',  base_url=ollama_url)
    messages= [{"role":"system","content":husband_system}]
    messages.append({"role":"assistant","content":husband_messages[-1]})
    messages.append({"role":"user","content":wife_messages[-1]})
    response = husband.chat.completions.create(model=MODEL_LLAMA, messages=messages)
    husband_messages.append(response.choices[0].message.content)
    return husband_messages[-1]
```

```python
def wife():
    messages= [{"role":"system","content":gpt_system}]
    messages.append({"role":"assistant","content":wife_messages[-1]})
    messages.append({"role":"user","content":husband_messages[-1]})
    #gpt = OpenAI(api_key=openai_api_key)
    response = gpt.chat.completions.create(model=MODEL_GPT, messages=messages)
    wife_messages.append(response.choices[0].message.content)
    return wife_messages[-1]
```
