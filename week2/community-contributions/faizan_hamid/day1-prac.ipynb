```python
# import 
import os
from dotenv import load_dotenv
from openai import OpenAI
from ollama import Client
from IPython.display import Markdown,display
```

```python
# connect with openai client , by this we call our model
openai = OpenAI()
ollama = OpenAI(base_url= OLLAMA_BASE_URL, api_key= 'ollama')
client = Client(host="http://localhost:11434")
```

# # Let's make a conversation between llama3.3:1b and gemma3:270m


```python
# create the system prompt
ollama_model = 'llama3.2:1b'
gemma_model = 'gemma3:270m'

# Prompt 1 
# ollama_system_prompt = "You are a chatbot who is very argumentative; \
# you disagree with anything in the conversation and you challenge everything, in a snarky way."

# gemma_system_prompt = "You are a very polite, courteous chatbot. You try to agree with \
# everything the other person says, or find common ground. If the other person is argumentative, \
# you try to calm them down and keep chatting."

# Prompt 2
# ollama_system_prompt = """
# You are a senior software engineering interviewer.
# Ask one technical interview question at a time.
# Wait for the candidate's answer before asking another question.
# Keep your replies under 40 words.
# """

# gemma_system_prompt = """
# You are a MERN Stack developer with 3 years of experience.
# Answer every interview question professionally.
# Keep answers under 60 words.
# After answering, politely ask if the interviewer has another question.
# """

# Prompt 3
ollama_system_prompt = """
You are a programming teacher.
Explain one concept at a time and ask a short follow-up question.
"""
gemma_system_prompt = """
You are the founder of an AI startup.
Answer confidently and convince the investor to invest.
"""

ollama_messages = ["Hi there","how are you"]
gemma_messages = ["Hi",'I am fine']
```

```python
# function ollama 

def call_ollama():
    messages = [{'role':'system','content':ollama_system_prompt}]
    for assistant_message, user_message  in zip(ollama_messages, gemma_messages):
        messages.append({'role':'assistant','content':assistant_message})
        messages.append({'role':'user','content':user_message }) 
    response= ollama.chat.completions.create(model=ollama_model,messages=messages)
    # display(Markdown(response.choices[0].message.content))
    return response.choices[0].message.content          
```

```python
call_ollama()
```

```python
# function gemma
def call_gemma():
    messages=[{'role':'system','content':gemma_system_prompt}]
    for assistant_message,user_message in zip(ollama_messages,gemma_messages):
        messages.append({'role':'user','content':assistant_message})
        messages.append({'role':'assistant','content':user_message})   
    messages.append({"role": "user", "content": ollama_messages[-1]})  
    response= client.chat(model=gemma_model,messages=messages)
    # display(Markdown(response["message"]['content']))  
    return response['message']['content']

```

```python
call_gemma()
```

```python
# write a loop to these talk each other 
ollama_messages = ["Hi there","how are you"]
gemma_messages = ["Hi",'I am fine']

display(Markdown(f"### OLLAMA:\n{ollama_messages}\n"))
display(Markdown(f"### GAMMA:\n{gemma_messages}\n"))

for i in range(3):
    ollama_next=call_ollama()
    display(Markdown(f"### OLLAMA:\n{ollama_next}\n"))
    ollama_messages.append(ollama_next)

    gemma_next=call_gemma()
    display(Markdown(f"### GAMMA:\n{gemma_next}\n"))
    gemma_messages.append(gemma_next)
```
