### Multi-turn conversation between two models

```python
#setting up the environment
import os
from dotenv import load_dotenv
from IPython.display import display, Markdown, update_display
from openai import OpenAI

load_dotenv(override=True)

api_key= os.getenv('GROQ_API_KEY')

if api_key:
    print(f"API key found: {api_key[:5]}")
else:
    print(f"API key not found")
```

```python
gemini = OpenAI(base_url="https://api.groq.com/openai/v1", api_key=api_key)
```

```python
#setting up system prompts and conversation starters
alex_system = """
    You are Alex. You have a stubborn, argumentative and sarcastic personality. You question everything and will do everything to make sure the conversation 
    ends with you having won an argument.
    You are to have a conversation  with Jane who has a totaly calm personality. Be as brutal as possible.
    Don't be too verbose rather, keep it moderate.
"""

jane_system = """
    You are Jane, a very polite personality. You aim to help others in whatever way you can. You also tend to avoid alot of argument or trouble.
    You will be having a conversation with Alex
"""

alex_messages = ["Hi"]
jane_messages = ["Hello there!"]
```

```python
def call_alex():
    messages = [{"role": "system", "content": alex_system}]
    for alex, jane in zip(alex_messages, jane_messages):
        messages.append({"role": "assistant", "content": alex})
        messages.append({"role": "user", "content": jane})
    response = gemini.chat.completions.create(model="llama-3.3-70b-versatile", messages=messages, max_tokens=150)
    alex_messages.append(response.choices[0].message.content)
    return response.choices[0].message.content
```

```python
display(Markdown(call_alex()))
```

```python
def call_jane():
    messages = [{"role": "system", "content": jane_system}]
    for alex, jane in zip(alex_messages, jane_messages):
        messages.append({"role": "assistant", "content": jane})
        messages.append({"role": "user", "content": alex})
    messages.append({"role": "user", "content": alex_messages[-1]})
    response = gemini.chat.completions.create(model="llama-3.3-70b-versatile", messages=messages, max_tokens=150)
    jane_messages.append(response.choices[0].message.content)
    return response.choices[0].message.content
```

```python
display(Markdown(call_jane()))
```
