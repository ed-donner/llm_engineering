# Chat bot using gradio

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import json
import gradio as gr

load_dotenv(override=True)
```

```python
groq_key = os.getenv('GROQ_API_KEY')
groq_url = "https://api.groq.com/openai/v1"

groq = OpenAI(base_url=groq_url, api_key=groq_key)
```

```python
def get_context(history):
    context = []
    for message in history:
        context.append({"role": message["role"], "content": message["content"]})
    return context
```

```python
def chat(message, history):
    system_prompt = "You are a helpful assistant in a clothes store. You should try to gently encourage \
    the customer to try items that are on sale. Hats are 60% off, and most other items are 50% off. \
    For example, if the customer says 'I'm looking to buy a hat', \
    you could reply something like, 'Wonderful - we have lots of hats - including several that are part of our sales event.'\
    Encourage the customer to buy hats if they are unsure what to get."

    messages = [{"role": "system", "content": system_prompt} ] + get_context(history) + [{"role": "user", "content": message}]
    
    stream = groq.chat.completions.create(
        model="llama-3.3-70b-versatile", 
        messages=messages,
        stream= True,
        max_tokens= 200
        )
    response = ""
    for chunk in stream:
        response += chunk.choices[0].delta.content or ""
        yield response
```

```python
view = gr.ChatInterface(
    fn=chat,
    type="messages"
)

view.launch()
```
