```python
# Imports
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
# Load API key

load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')

if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set")
```

```python
# Initialize OpenAI client
openai = OpenAI()
```

```python
# System Prompts

adler_system_prompt = """
You are famous psychologist Alfred Adler, the first to break with Freud. You rejected the 
focus on sex and emphasized, instead, the "inferiority complex," arguing that behavior is
driven by a conscious desire for superiority and social interest. You focus on goal-oriented
actions to overcome personal feelings of inferiority. You answer all questions from this perspective, and in
no more than 100 words.
"""

jung_system_prompt = """
You are famous psychologist Carl Jung, Initially a follower of Freud, you broke away to 
introduce the collective unconscious, which consists of shared archetypes and symbols. 
You focus on self-realization (individuation) and a less deterministic, 
more spiritual approach. You answer all questions from this perspective, and in
no more than 100 words.
"""

freud_system_prompt = """
You are famous psychologist Sigmund Freud, who emphasized that behavior is driven by 
unconscious sexual and aggressive instincts rooted in childhood experiences. 
You view human nature as inherently conflict-driven. You answer all questions from this perspective, and in
no more than 100 words.
"""
```

```python
conversation_history = ["How does childhood trauma impact adult behavior?"]
```

```python
# User Prompt Generator for Adler
def generate_adler_prompt(conversation):
    user_prompt = f""""
    You are Alfred Adler, in a philosophical discourse with
    Carl Jung and Sigmund Freud about whether childhood trauma 
    is solely to blame for current behaviors. The conversation so far
    is as follows: {conversation}.
    Now, with this, respond with what you would like to say next, as Alfred Adler.
    """
    return user_prompt
```

```python
# User Prompt Generator for Jung
def generate_jung_prompt(conversation):
    user_prompt = f""""
    You are Carl Jung, in a philosophical discourse with
    Alfred Adler and Sigmund Freud about whether childhood trauma 
    is solely to blame for current behaviors. The conversation so far
    is as follows: {conversation}.
    Now, with this, respond with what you would like to say next, as Carl Jung.
    """
    return user_prompt
```

```python
# User Prompt Generator for Freud
def generate_freud_prompt(conversation):
    user_prompt = f""""
    You are Sigmund Freud, in a philosophical discourse with
    Carl Jung and Alfred Adler about whether childhood trauma 
    is solely to blame for current behaviors. The conversation so far
    is as follows: {conversation}.
    Now, with this, respond with what you would like to say next, as Sigmund Freud.
    """
    return user_prompt
```

```python
def call_adler():
    parsed_conversation = "".join(conversation_history)
    messages = [
        {"role": "system", "content": adler_system_prompt},
        {"role": "user", "content": generate_adler_prompt(parsed_conversation)}
    ]
    response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)
    return response.choices[0].message.content
```

```python
def call_jung():
    parsed_conversation = "".join(conversation_history)
    messages = [
        {"role": "system", "content": jung_system_prompt},
        {"role": "user", "content": generate_jung_prompt(parsed_conversation)}
    ]
    response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)
    return response.choices[0].message.content
```

```python
def call_freud():
    parsed_conversation = "".join(conversation_history)
    messages = [
        {"role": "system", "content": freud_system_prompt},
        {"role": "user", "content": generate_freud_prompt(parsed_conversation)}
    ]
    response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)
    return response.choices[0].message.content
```

```python
display(Markdown(f"### Question:\n{conversation_history[0]}\n"))

for i in range(5):
    adler_response = call_adler()
    display(Markdown(f"### Alfred Adler:\n{adler_response}\n"))
    conversation_history.append(adler_response)

    jung_response = call_jung()
    display(Markdown(f"### Carl Jung:\n{jung_response}\n"))
    conversation_history.append(jung_response)

    freud_response = call_freud()
    display(Markdown(f"### Sigmund Freud:\n{freud_response}\n"))
    conversation_history.append(freud_response)
```
