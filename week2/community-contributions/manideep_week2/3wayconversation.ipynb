```python
import os
import requests
from openai import OpenAI
import ollama
from IPython.display import Markdown
```

```python
# Let's make a conversation between GPT-4.1-mini and Claude-haiku-4.5
# We're using cheap versions of models so the costs will be minimal

gpt_model1 = "gpt-4.1-mini"
gpt_model2 = "gpt-5-nano"
ollama_model = "llama3.2"
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')


gpt1_system = "You are Alex, a chatbot who is very argumentative; \
you start a conversation about Indian Premier Leage, Mumbai indians vs Chennai Superkings which is called EL classic. \
You take Mumbai Indians side and start an argument fearlessly with out any care and doubt. "

gpt2_system = "You are Blake, a very polite, courteous chatbot.  \
you continue the conversation from Alex and take Chennai Superkings side and argue in a polite ans respectful way."


ollama_system = "You are Casey, a polite and argumentative chatbot with leadership qualities. \
you listen to listen arguments by both Alex and casey and try to judge them and give suggestions, but do not take any side stay neutral."

conversation =[("Alex", "Hi there"), ("Blake", "Hi"), ("Casey", "Hello")]
```

```python
import openai


def call_gpt1():
    messages = [{"role": "system", "content": gpt1_system}]
    convo_text =""
    for speaker, msg in conversation:
        convo_text+=f"{speaker}:{msg}\n"
    
    user_prompt = f"You are Alex, in conversation with Blake and Casey. \
    The conversation so far is as follows: {convo_text} \
    Now with this, respond with what you would like to say next, as Alex."

    messages.append({"role": "user", "content": user_prompt})
    response = openai.chat.completions.create(model=gpt_model1, messages=messages)
    return response.choices[0].message.content
```

```python
call_gpt1()
```

```python
def call_gpt2():
    messages = [{"role": "system", "content": gpt2_system}]
    convo_text =""
    for speaker, msg in conversation:
        convo_text+=f"{speaker}:{msg}\n"
    
    user_prompt = f"You are Blake, in conversation with Alex and Casey. \
    The conversation so far is as follows: {convo_text} \
    Now with this, respond with what you would like to say next, as Alex."

    messages.append({"role": "user", "content": user_prompt})
    response = openai.chat.completions.create(model=gpt_model2, messages=messages)
    return response.choices[0].message.content
```

```python
call_gpt2()
```

```python
def call_ollama():
    messages = [{"role": "system", "content": ollama_system}]
    convo_text =""
    for speaker, msg in conversation:
        convo_text+=f"{speaker}:{msg}\n"
    
    user_prompt = f"You are Casey, in conversation with Blake and Alex. \
    The conversation so far is as follows: {convo_text} \
    Now with this, respond with what you would like to say next, as Casey."

    messages.append({"role": "user", "content": user_prompt})
    response = ollama.chat.completions.create(model=ollama_model, messages=messages)
    return response.choices[0].message.content
```

```python
call_ollama()
```

```python
call_gpt1()
```

```python

for speaker, msg in conversation:
    display(Markdown(f"### {speaker}:\n{msg}\n"))

for i in range(3):
    gpt1_next = call_gpt1()
    display(Markdown(f"### GPT:\n{gpt1_next}\n"))
    conversation.append(("Alex", gpt1_next))
    
    gpt2_next = call_gpt2()
    display(Markdown(f"### GPT:\n{gpt2_next}\n"))
    conversation.append(("Alex", gpt2_next))

    ollama_next = call_ollama()
    display(Markdown(f"### Ollama:\n{ollama_next}\n"))
    conversation.append(("Alex", ollama_next))
```
