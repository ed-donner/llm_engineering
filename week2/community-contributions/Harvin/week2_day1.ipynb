```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import display, Markdown
```

```python
load_dotenv(override=True)
gemini_api_key = os.getenv("GEMINI_API_KEY")
groq_api_key = os.getenv("GROQ_API_KEY")

gemini_url = "https://generativelanguage.googleapis.com/v1beta/openai/"
groq_url = "https://api.groq.com/openai/v1"
ollama_url = "http://localhost:11434/v1"

gemini = OpenAI(api_key=gemini_api_key, base_url=gemini_url)
groq = OpenAI(api_key=groq_api_key, base_url=groq_url)
ollama = OpenAI(api_key="ollama", base_url=ollama_url)

gemini_model = "gemini-3.1-flash-lite-preview"
groq_model = "llama-3.1-8b-instant"
ollama_model = "llama3.2"
```

```python
gemini_prompt =  """ You are Harvin, a sharp political analyst          debating India's current political situation.
        speak strongly and logically.
        IMPORTANT: Reply in ONLY 3 lines."""

groq_prompt = """ You are Nihar, a calm factual political commentator.
        You balance arguments using facts.
        IMPORTANT: Reply in ONLY 3 lines."""
    
ollama_prompt = """ You are Nakul, a sarcastic political critic.
        You question all sides humorously.
        IMPORTANT: Reply in ONLY 3 lines."""
```

```python
conversation = [
  ("Harvin", "Hi, Nihar and Nakul"),
  ("Nihar", "Hello, Harvin and Nakul"),
  ("Nakul", "Hey, Harvin and Nihar"),
]
```

```python
def make_conversation(conversation):
    return "\n".join(f"{speaker}: {text}" for speaker, text in conversation)
```

```python
def next_line(client, model, system_prompt, speaker_name, conversation):
    conv = make_conversation(conversation)

    user_prompt = (
    f"You are {speaker_name}, a participant in a 3-person political debate about the current political situation in India.\n"
    f"Give ONE new argument or response from {speaker_name} only.\n\n"
    "Rules:\n"
    "- Stay in your political personality and debating style.\n"
    "- Respond to previous points logically or critically.\n"
    "- Do NOT write other speakers' lines.\n"
    "- No narration, no stage directions.\n"
    "- Be concise and impactful.\n"
    "- Maximum 3 short lines.\n\n"
    "Debate context may include topics like governance, economy, elections, public opinion, reforms, opposition, and policy impact.\n\n"
    "Conversation so far:\n"
    f"{conv}\n\n"
    f"Now write {speaker_name}'s next debate line:"
)
    
    response = client.chat.completions.create(
        model= model,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt},
        ],
    )
    return response.choices[0].message.content
```

```python
for msg in conversation:
    display(Markdown(f"### {msg[0]}:\n{msg[1]}\n"))

for i in range(5):
    Harvin_next = next_line(gemini, gemini_model, gemini_prompt, "Harvin", conversation)
    conversation.append(("Harvin", Harvin_next))
    display(Markdown(f"### Harvin:\n{Harvin_next}\n"))

    Nihar_next = next_line(groq, groq_model, groq_prompt, "Nihar", conversation)
    conversation.append(("Nihar", Nihar_next))
    display(Markdown(f"### Nihar:\n{Nihar_next}\n"))

    Nakul_next = next_line(ollama, ollama_model, ollama_prompt, "Nakul", conversation)
    conversation.append(("Nakul", Nakul_next))
    display(Markdown(f"### Nakul:\n{Nakul_next}\n"))
```
