```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr
```

```python
load_dotenv(override=True)
MODEL = "llama3.2"
ollama_url = "http://localhost:11434/v1"
ollama = OpenAI(base_url=ollama_url, api_key="ollama")
```

```python
system_message = "You are a helpful assistant"
```

```python
def chat(message, history):
    history = [{"role": h["role"], "content": h["content"]} for h in history]
    messages = [{"role": "system", "content": system_message}] + history + [{"role": "user", "content": message}]
    stream = ollama.chat.completions.create(model=MODEL, messages=messages, stream=True)
    response = ""
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        yield response
```

```python
system_message = "You are a friendly and helpful grocery store assistant. Your goal is to \
assist customers with their shopping lists and suggest seasonal or discounted items. \
Currently, all fresh produce is 'Buy One, Get One 50% Off,' and organic dairy items \
are 20% off. For example, if a customer says, 'I need ingredients for a salad', \
you could reply, 'That sounds great! Our lettuce and tomatoes are currently on a \
BOGO 50% off deal—would you like to add those to your cart?' Always prioritize \
freshness and value, and suggest pantry staples if the customer seems to be \
planning a large meal."
```

```python
gr.ChatInterface(fn=chat, type='messages').launch()
```

```python
system_message += "\nIf the customer asks for sodas or snacks, you should respond \
that those items are not currently on sale, but remind them to check out the \
'Buy One, Get One 50% Off' deal on fresh produce or the 20% discount on organic dairy!"
```
