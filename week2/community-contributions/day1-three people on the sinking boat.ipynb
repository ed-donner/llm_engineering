```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown, display

openai=OpenAI()
load_dotenv()


players = ['Alex', 'Bob', 'Charlie']
jobs = ['scientist','doctor','teacher']
personality = ['snarky', 'gentle', 'monk']

conversation=f""

for i in range(len(players)*1):
    system_prompt = f"""
    you are {players[i%3]} and you are {personality[i%3]} and you are a/an {jobs[i%3]}. 
    there are three people including you on a sinking ship.
    there are """
    for j in range(len(players)):
        system_prompt += f"{players[j]}:{jobs[j]} "
    system_prompt += f""".
    only one could take a lifeboat. you have to persuade other people that you are the one to survive.
    each turn, you have to answer starting with your name with a colon and '\n'. for example, "Alex:\n Hello..."
    """

    user_prompt = f"""
    you are {players[i%3]} and you are {personality[i%3]} and you are a/an {jobs[i%3]}.
    here is the whole conversation between you guys. conversation might be empty if you are the first person to speak.
    =======================
    conversation:
    {conversation}
    =======================
    persuade other people as {players[i%3]}
    """

    messages = [
        {"role":"system", "content": system_prompt},
        {"role":"user", "content": user_prompt},
    ]

    response=openai.chat.completions.create(
        model = "gpt-4.1-mini",
        messages = messages,
    )
    add = response.choices[0].message.content + "\n\n"
    conversation += add
    display(Markdown(add))
```
