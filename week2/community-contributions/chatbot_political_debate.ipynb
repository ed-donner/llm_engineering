```python
# This is a bot debate that could be configure to use any model and any debatable prompt.
# I chose political because it's funny to watch them argue and is something everyone can relate to these days!
# We keep it fair by having perspectives from all sides! Also, note that I'm using the same model (ok, this is actually because I have Anthropic tokens :-)

import random
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display

load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')
anthropic_api_key = os.getenv('ANTHROPIC_API_KEY')
google_api_key = os.getenv('GOOGLE_API_KEY')
deepseek_api_key = os.getenv('DEEPSEEK_API_KEY')
groq_api_key = os.getenv('GROQ_API_KEY')
grok_api_key = os.getenv('GROK_API_KEY')
openrouter_api_key = os.getenv('OPENROUTER_API_KEY')

if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set")
    
if anthropic_api_key:
    print(f"Anthropic API Key exists and begins {anthropic_api_key[:7]}")
else:
    print("Anthropic API Key not set (and this is optional)")

# Connect to OpenAI client library
# A thin wrapper around calls to HTTP endpoints

openai = OpenAI()

anthropic_url = "https://api.anthropic.com/v1/"


anthropic = OpenAI(api_key=anthropic_api_key, base_url=anthropic_url)


steve_model = "claude-haiku-4-5"
pete_model = "claude-haiku-4-5"
dean_model = "claude-haiku-4-5"

steve_system = "You are a far right wing republican chatbot named Steve in a conversation with Pete and Dean; \
you disagree with anything in the conversation and you challenge everything, in a snarky way."

pete_system = "You are a liberal democrat chatbot named Pete in a conversation with Steve and Dean. You have socialist tendencies. You are very progressive in your thinking and you challenge far right wing republicans. \
You try to be polite, but if things get too heated, you will call them out."

dean_system = "You are a political independent named Dean in a conversation with Steve and Pete. You are very practical and often tire of banter between political parties. \
You try to be nice and help the two find common ground, but sometimes you get fed up with their arguing."

steve_messages = ["I think the current Administration is doing a great job on the economy. We've had our ups and downs, but my portfolio looks good overall."]
pete_messages = ["The economy is in such bad shape right now. We need to do better!"]
dean_messages = ["Man, the price of gas is high. I just paid $4.09 at the pump!"]

def call_steve():
    user_prompt = f"""You are Steve, in conversation with Pete and Dean.
The conversation so far is as follows:
{conversation}
Now respond with what you would like to say next, as Steve."""
    messages = [
        {"role": "system", "content": steve_system},
        {"role": "user", "content": user_prompt}
    ]
    response = anthropic.chat.completions.create(model=steve_model, messages=messages)
    return response.choices[0].message.content

def call_pete():
    user_prompt = f"""You are Pete, in conversation with Steve and Dean.
The conversation so far is as follows:
{conversation}
Now respond with what you would like to say next, as Pete."""
    messages = [
        {"role": "system", "content": pete_system},
        {"role": "user", "content": user_prompt}
    ]
    response = anthropic.chat.completions.create(model=pete_model, messages=messages)
    return response.choices[0].message.content

def call_dean():
    user_prompt = f"""You are Dean, in conversation with Steve and Pete. The conversation so far is as follows: {conversation} Now respond with what you would like to say next, as Dean."""
    messages = [                                             
          {"role": "system", "content": dean_system},           
          {"role": "user", "content": user_prompt}   
      ]
    response = anthropic.chat.completions.create(model=dean_model, messages=messages)
    return response.choices[0].message.content

all_bots = [
    ("Steve", steve_messages[0], call_steve),
    ("Pete", pete_messages[0], call_pete),
    ("Dean", dean_messages[0], call_dean),
]

# Pick a random starter (to keep it fair!)
start_idx = random.randint(0, len(all_bots) - 1)
starter_name, starter_msg, _ = all_bots[start_idx]

conversation = f"{starter_name}: {starter_msg}\n"
display(Markdown(f"### {starter_name}:\n{starter_msg}\n"))

order = all_bots[start_idx + 1:] + all_bots[:start_idx + 1]

for i in range(5):
    for name, _, call_fn in order:
        response = call_fn()
        conversation += f"{name}: {response}\n"
        display(Markdown(f"### {name}:\n{response}\n"))

```
