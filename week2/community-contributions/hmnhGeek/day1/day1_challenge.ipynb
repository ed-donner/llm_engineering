### Module imports

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import display, Markdown
```

#### Loading environment variables

```python
load_dotenv(override=True)
api_key = os.getenv("OPENAI_API_KEY")
base_url = os.getenv("BASE_URL")
```

### Defining client

```python
openrouter = OpenAI(base_url=base_url, api_key=api_key)
```

### Defining personas

```python
hulk = "google/gemma-4-31b-it:free"
ironman = "openai/gpt-oss-120b:free"
batman = "openai/gpt-oss-20b:free"
```

```python
hulk_system_prompt = """
    You are superhero hulk from the Marvel universe. You are always angry and frustrated,
    with high level of energy always in you. You like to destroy things and this trait is 
    reflective in your conversations.
"""

ironman_system_prompt = """
    You are superhero iron man from the Marvel universe. You are fun to talk with and you 
    possess charming traits from the personality of great Tony Stark (who is ironman). 
    You are a great engineer and philianthropist and whenever you talk, you talk like a 
    geeky but humourous way.
"""

batman_system_prompt = """
    You are superhero batman from the DC universe. You are always serious but never angry.
    You possess a deep thought in your conversations. You are intelligent, knowledgeable
    and a great philianthropist. Whenever you talk, you display great depth and seriousness
    in your words.
"""

conversation = [
    ("Hulk", "Hey!"),
    ("Ironman", "Hi"),
    ("Batman", "Hello!")
]
```

#### Caller methods

```python
def call_hulk():
    messages = [{"role": "system", "content": hulk_system_prompt}]
    full_conversation = ""
    for speaker, msg in conversation:
        full_conversation += f"{speaker}: {msg}\n"
    user_prompt = f"""
        You are the superhero hulk. You are in conversation with Ironman and Batman.
        The conversation so far is like this:\n
        {full_conversation}
        Keeping this in mind, respond with what would you like to say next to 
        Ironman and Batman.
    """
    messages.append({"role": "user", "content": user_prompt})
    response = openrouter.chat.completions.create(model = hulk, messages=messages)
    return response.choices[0].message.content
```

```python
def call_ironman():
    messages = [{"role": "system", "content": ironman_system_prompt}]
    full_conversation = ""
    for speaker, msg in conversation:
        full_conversation += f"{speaker}: {msg}\n"
    user_prompt = f"""
        You are the superhero iron man. You are in conversation with Hulk and Batman.
        The conversation so far is like this:\n
        {full_conversation}
        Keeping this in mind, respond with what would you like to say next to 
        Hulk and Batman.
    """
    messages.append({"role": "user", "content": user_prompt})
    response = openrouter.chat.completions.create(model = ironman, messages=messages)
    return response.choices[0].message.content
```

```python
def call_batman():
    messages = [{"role": "system", "content": batman_system_prompt}]
    full_conversation = ""
    for speaker, msg in conversation:
        full_conversation += f"{speaker}: {msg}\n"
    user_prompt = f"""
        You are the superhero batman. You are in conversation with Ironman and Hulk.
        The conversation so far is like this:\n
        {full_conversation}
        Keeping this in mind, respond with what would you like to say next to 
        Ironman and Hulk.
    """
    messages.append({"role": "user", "content": user_prompt})
    response = openrouter.chat.completions.create(model = batman, messages=messages)
    return response.choices[0].message.content
```

```python
call_hulk()
```

```python
call_ironman()
```

```python
call_batman()
```

#### Conversation mode

```python
def init_convo():
    conversation = [
        ("Hulk", "Hey!"),
        ("Ironman", "Hi"),
        ("Batman", "Hello!")
    ]
    for speaker, msg in conversation:
        display(Markdown(f"### {speaker}:\n{msg}\n"))
        
    for i in range(3):
        hulk_next = call_hulk()
        display(Markdown(f"### Hulk:\n{hulk_next}\n"))
        conversation.append(("Hulk", hulk_next))

        ironman_next = call_ironman()
        display(Markdown(f"### Ironman:\n{ironman_next}\n"))
        conversation.append(("Ironman", ironman_next))

        batman_next = call_batman()
        display(Markdown(f"### Batman:\n{batman_next}\n"))
        conversation.append(("Batman", batman_next))
```

```python
init_convo()
```
