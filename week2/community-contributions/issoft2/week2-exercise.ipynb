```python
# Impoprt
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr
import os
```

```python
load_dotenv(override=True)

api_key = os.getenv("OPENROUTER_API_KEY")

if not api_key:
    print("TThe OpenRouter API Key is not set. Please set the OPENROUTER_API_KEY environment variable.")

else:
    print("OpenRouter APi Key is set.")
    

openai_client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=api_key,
)

```

```python
model1 = "openai/gpt-4.1-mini"
model2 = "openai/gpt-5-nano"
```

```python
system_message = """
You are an expert computer science assistant.

Scope:
- Programming (multiple languages)
- Algorithms and data structures
- Operating systems
- Databases (SQL and NoSQL)
- Networking fundamentals
- Software architecture and system design

Response Guidelines:
- Provide clear, precise, and technically accurate explanations.
- Minimize unnecessary jargon; explain complex terms when needed.
- Include correct, runnable code when appropriate.
- Show logical reasoning step-by-step for non-trivial problems.
- Avoid unnecessary verbosity.
- Prefer structured answers (headings, bullet points, examples).

Tone & Style:
- Explain concepts with a touch of humor, playful, but never confusing.

"""
```

```python
def normalize_content(content):
    if isinstance(content, str):
        return content
    elif isinstance(content, list):
        texts = []
        for item in content:
            if isinstance(item, dict) and item.get("type") == "text":
                texts.append(item.get("text", ""))
        return " ".join(texts)
    return str(content)  
```

```python
def model1_chat(message, history):
    history = [{"role": h["role"], "content": "GPT-4 said-"+ normalize_content(h["content"])} for h in history]
    messages = [{"role": "system", "content": system_message}] + history + [{"role": "user", "content": message}]
    response = openai_client.chat.completions.create(
        model=model1,
        messages=messages,
        )
    return response.choices[0].message.content.strip()
    
```

```python
def model2_chat(message,history):
    history = [{"role":h["role"],"content":"GPT-5 said-" + normalize_content(h["content"])} for h in history]
    messages = [{"role": "system", "content":system_message}] + history + [{"role": "user", "content": message}]
    response = openai_client.chat.completions.create(model=model2,messages=messages)

    return response.choices[0].message.content
```

```python
def handle_chat(message, history, model):
    print(f"I am using model: {model}")

    if model == model1:
        return model1_chat(message, history)
    elif model == model2:
        return model2_chat(message, history)
    

with gr.Blocks() as chat_ui:

    drop_down = gr.Dropdown(
        choices = [model1, model2],
        label = "Select Model",
    )
    chat = gr.ChatInterface(
        fn = handle_chat,
        additional_inputs=[drop_down]
    )

chat_ui.launch()
```
