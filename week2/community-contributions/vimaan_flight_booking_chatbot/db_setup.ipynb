```python
import sqlite3
```

```python
conn = sqlite3.connect("vimaan.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS passengers (
    passenger_id INTEGER PRIMARY KEY,
    first_name TEXT NOT NULL,
    middle_name TEXT,
    last_name TEXT NOT NULL,
    age INTEGER NOT NULL CHECK(age >= 0)
);
""")
```

```python
# cursor.execute("""
# INSERT INTO passengers (first_name, middle_name, last_name, age)
# VALUES (?, ?, ?, ?)
# """, ("Shyam ", "Kumar", "Singh", 30))
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS bookings (
    pnr_number TEXT NOT NULL CHECK(pnr_number GLOB '[A-Za-z0-9]*'),
    passenger_id INTEGER NOT NULL,
    datetime_of_journey DATETIME NOT NULL,
    flight_number INTEGER NOT NULL,
    flight_company TEXT,
    PRIMARY KEY (pnr_number, passenger_id)
);
""")
```
