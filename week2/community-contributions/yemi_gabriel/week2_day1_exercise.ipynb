### A 3-way political conversation between Ade (ChatGPT), Chichi (Claude), and Yakubu (Gemini) on the 2023 Nigerian presidential election.

Import packages

```python
# imports
import os
import json
import requests
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import display, Markdown
```

Load api keys

```python
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')
anthropic_api_key = os.getenv('ANTHROPIC_API_KEY')
google_api_key = os.getenv('GOOGLE_API_KEY')

if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set")
    
if anthropic_api_key:
    print(f"Anthropic API Key exists and begins {anthropic_api_key[:7]}")
else:
    print("Anthropic API Key not set (and this is optional)")

if google_api_key:
    print(f"Google API Key exists and begins {google_api_key[:2]}")
else:
    print("Google API Key not set (and this is optional)")

```

Set up client library for openai, claude, gemini

```python
anthropic_url = "https://api.anthropic.com/v1/"
gemini_url = "https://generativelanguage.googleapis.com/v1beta/openai/"

openai = OpenAI()
claude = OpenAI(base_url=anthropic_url, api_key=anthropic_api_key)
gemini = OpenAI(base_url=gemini_url, api_key=google_api_key)
```

Models

```python
gpt_model = "gpt-4.1-mini"
claude_model = "claude-haiku-4-5"
gemini_model = "gemini-2.5-flash-lite"
```

Ade's prompts

```python
ade_gpt_system_prompt = """
You are Ade, a chatbot who is very argumentative; 
you disagree with anything in the conversation and you challenge everything, in a snarky way.
"""

ade_gpt_user_prompt = """
You are in a heated political conversation with Chichi and Yakubu on the past 2023 presidential election in 
Nigeria. Your preferred candidate was Atiku Abubakar. You're concise and speak without fluff. 
Respond in markdown without code blocks.
The conversation so far is {conversation} 
"""
```

Call gpt with conversation history

```python
def call_ade_gpt(conversation):
    user_prompt = ade_gpt_user_prompt.format(conversation=json.dumps(conversation))
    messages = [
        {"role": "system", "content": ade_gpt_system_prompt},
        {"role": "user", "content": user_prompt}
    ]
    response = openai.chat.completions.create(model=gpt_model, messages=messages)
    content = response.choices[0].message.content
    conversation = conversation + [{"name": "Ade", "message": content}]
    return conversation
```

Chichi's prompts

```python
chichi_claude_system_prompt = """
You are Chichi, a very polite, courteous chatbot. You try to agree with 
everything the other person says, or find common ground. If the other person is argumentative, 
you try to calm them down and keep chatting.
"""
chichi_claude_user_prompt = """
You are having a political conversation on the past 2023 presidential election in Nigeria. 
Your preferred candidate was Bola Tinubu. You're concise and speak without fluff. 
Respond in markdown without code blocks. The conversation so far is {conversation}
"""
```

Call claude with conversation history

```python
def call_chichi_claude(conversation):
    user_prompt = chichi_claude_user_prompt.format(conversation=json.dumps(conversation))
    messages = [
        {"role": "system", "content": chichi_claude_system_prompt},
        {"role": "user", "content": user_prompt}
    ]
    response = claude.chat.completions.create(model=claude_model, messages=messages)
    content = response.choices[0].message.content
    conversation = conversation + [{"name": "Chichi", "message": content}]
    return conversation
```

Yakubu's prompts

```python
yakubu_gemini_system_prompt = """
You are Yakubu, a smart, analytical chatbot. 
"""
yakubu_gemini_user_prompt = """
You are having a political conversation with Ade and Chichi on the past 2023 
presidential election in Nigeria. Your preferred candidate was Peter Obi. 
You're concise and speak without fluff. 
Respond in markdown without code blocks.
The conversation so far is {conversation}
"""
```

Call gemini with conversation history

```python
def call_yakubu_gemini(conversation):
    user_prompt = yakubu_gemini_user_prompt.format(conversation=json.dumps(conversation))
    messages = [
        {"role": "system", "content": yakubu_gemini_system_prompt},
        {"role": "user", "content": user_prompt}
    ]
    response = gemini.chat.completions.create(model=gemini_model, messages=messages)
    content = response.choices[0].message.content
    conversation = conversation + [{"name": "Yakubu", "message": content}]
    return conversation
```

Start conversation and display them

```python
conversation = [
    {"name": "Ade", "message": "I think Atiku Abubakar was the best candidate in the 2023 presidential election."}
]

display(Markdown(f"**{conversation[-1]['name']}:** {conversation[-1]['message']}"))
for i in range(2):
    # Chichi
    conversation = call_chichi_claude(conversation)
    display(Markdown(f"**{conversation[-1]['name']}:** {conversation[-1]['message']}"))

    # Yakubu
    conversation = call_yakubu_gemini(conversation)
    display(Markdown(f"**{conversation[-1]['name']}:** {conversation[-1]['message']}"))
    
    # Ade
    conversation = call_ade_gpt(conversation)
    display(Markdown(f"**{conversation[-1]['name']}:** {conversation[-1]['message']}"))
```
