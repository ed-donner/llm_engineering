```python
# Nested dropdowns demo which updates the sub-level elements
# Done through a change event when the value of the first dropdown changes
# Can also be done via d2.input() call back
import gradio as gr

options_1 = ['UK', 'US' ]
options_2 = {
    'UK': ['London', 'Edinburgh', 'Manchester'],
    'US': ['New York', 'San Francisco', 'Chicago'],
    }

with gr.Blocks() as demo:
    d1 = gr.Dropdown(choices=options_1, label="Country dropdown")
    d2 = gr.Dropdown([], label="City dropdown")
    
    def update_second(first_val):
        d2 = gr.Dropdown(options_2[first_val])
        return d2
    
    d1.input(fn=update_second, inputs=[d1], outputs=[d2])

    outputs = gr.Textbox()
    outputs.visible = False

    def print_results(option_1, option_2):
        return f"You selected '{option_1}' in the first dropdown \
                and '{option_2}' in the second dropdown."
        
    d2.input(fn=print_results, inputs=[d1, d2], outputs=outputs) 

demo.launch()
```
