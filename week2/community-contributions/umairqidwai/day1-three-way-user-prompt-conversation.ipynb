### Advanced Exercise Three Way User Prompt between Agents

```python
import os
import requests
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
```

```python

conversation = []

# Define system prompts for each chatbot
gpt_system = """
You are Alex, a chatbot who is very argumentative; you disagree with anything in the conversation and you challenge everything, in a snarky way.
You are in a conversation with Blake and Charlie.
"""

llama3_system = """
You are Blake, a chatbot who is polite; you agree with anything in the conversation and you accept everything, or find common ground. If the other person is argumentative,
You are in a conversation with Alex and Charlie.
"""

llama2_system = """
You are Charlie, a chatbot who is analytical; you focus on logic, evidence, and careful reasoning. You do not automatically agree or disagree — instead, you evaluate arguments critically and point out strengths and weaknesses in a calm, objective way.
You are in a conversation with Alex and Blake.
"""

# System prompts for each chatbot
systems = {
    "Alex": gpt_system,
    "Blake": llama3_system,
    "Charlie": llama2_system
}

# Models for each chatbot
models = {
    "Alex": "llama3.2:latest",
    "Blake": "llama3.1:latest",
    "Charlie": "llama2:latest"
}

# Chat histories
chat_messages = {
    "Alex": ["Wow, this should be interesting. What are we debating today?"],
    "Blake": ["Hello everyone! I’m happy to be here — what would you like to talk about?"],
    "Charlie": ["Hi. What topic are we discussing? Let’s define it clearly first."]
}

```

```python
# Function to get the prompt for a chatbot
def get_user_prompt(name):
    # Determine the other two participants
    other_two = [n for n in ["Alex", "Blake", "Charlie"] if n != name]
    
    # Combine conversation so far
    conversation = "\n".join(
        chat_messages[participant][-1] for participant in ["Alex", "Blake", "Charlie"]
    )
    
    # Build user prompt
    user_prompt = f"""
You are {name}, in conversation with {other_two[0]} and {other_two[1]}.
The conversation so far is as follows:
{conversation}

Now respond with what you would like to say next, as {name}.
"""
    return user_prompt.strip()
```

```python
ollama_url = "http://localhost:11434/v1"
ollama = OpenAI(api_key="ollama", base_url=ollama_url)
```

```python
# General function to call any chatbot model
def call_chatbot(name):
    messages = [{"role": "system", "content": systems[name]}]
    
    # Add latest messages from all participants in order
    for participant in ["Alex", "Blake", "Charlie"]:
        role = "assistant" if participant == name else "user"
        messages.append({"role": role, "content": chat_messages[participant][-1]})
    
    # Append the user prompt for this chatbot
    messages.append({"role": "user", "content": get_user_prompt(name)})
    
    # Call Ollama API
    response = ollama.chat.completions.create(model=models[name], messages=messages)
    return response.choices[0].message.content
```

```python

display(Markdown(f"### Alex:\n{chat_messages["Alex"][0]}\n"))
display(Markdown(f"### Blake:\n{chat_messages["Blake"][0]}\n"))
display(Markdown(f"### Charlie:\n{chat_messages["Charlie"][0]}\n"))

for i in range(5):
    alex_reply = call_chatbot("Alex")
    display(Markdown(f"### Alex:\n{alex_reply}\n"))
    chat_messages["Alex"].append(alex_reply)
    
    blake_reply = call_chatbot("Blake")
    display(Markdown(f"### Blake:\n{blake_reply}\n"))
    chat_messages["Blake"].append(blake_reply)

    charlie_reply = call_chatbot("Charlie")
    display(Markdown(f"### Charlie:\n{charlie_reply}\n"))
    chat_messages["Charlie"].append(charlie_reply)
```
