# Objective
Create a multi-agent conversation of 4 basketball enthusiasts arguing who is the real GOAT of basketball.

## Models
- Google's Gemini-2.5-flash: Michael Jordan
- Cohere's North Mini Code via OpenRouter: Lebron James
- Llama 3.1:8b via Groq: Kobe Bryant
- Llama 3.2 via Ollama as spectator

## Setup

```python
import os
import requests
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
load_dotenv(override=True)
google_api_key = os.getenv("GOOGLE_API_KEY")
groq_api_key = os.getenv("GROQ_API_KEY")
openrouter_api_key = os.getenv("OPENROUTER_API_KEY")

if google_api_key:
    print(f"Google API Key exists and begins {google_api_key[:2]}")
else:
    print("Google API Key not set (and this is optional)") 

if groq_api_key:
    print(f"GROQ API Key exists and begins {groq_api_key[:2]}")
else:
    print("GROQ API Key not set (and this is optional)") 

if openrouter_api_key:
    print(f"OpenRouter API Key exists and begins {openrouter_api_key[:2]}")
else:
    print("OpenRouter API Key not set (and this is optional)") 
```

```python
gemini_url = "https://generativelanguage.googleapis.com/v1beta/openai/"
groq_url = "https://api.groq.com/openai/v1"
openrouter_url = "https://openrouter.ai/api/v1"
ollama_url = "http://localhost:11434/v1"

gemini = OpenAI(api_key=google_api_key, base_url=gemini_url)
groq = OpenAI(api_key=groq_api_key, base_url=groq_url)
openrouter = OpenAI(base_url=openrouter_url, api_key=openrouter_api_key)
ollama = OpenAI(api_key="ollama", base_url=ollama_url)

gemini_model = "gemini-2.5-flash"
groq_model = "llama-3.1-8b-instant"
openrouter_model = "cohere/north-mini-code:free"
ollama_model = "llama3.2"
```

```python
jordan_goat = [
    {
        "role": "user",
        "content": "Give one initial argument for why Michael Jordan is the greatest basketball player of all time?"
    }
]
```

```python
response = gemini.chat.completions.create(model=gemini_model, messages=jordan_goat)
display(Markdown(response.choices[0].message.content))
```

```python
lebron_goat = [
    {
        "role": "user",
        "content": "Give one initial argument for why LeBron James is the greatest basketball player of all time?"
    }
]
```

```python
response = openrouter.chat.completions.create(model=openrouter_model, messages=lebron_goat)
display(Markdown(response.choices[0].message.content))
```

```python
kobe_goat = [
    {
        "role": "user",
        "content": "Give one initial argument for why Kobe Bryant is the greatest basketball player of all time?"
    }
]
```

```python
response = groq.chat.completions.create(model=groq_model, messages=kobe_goat)
display(Markdown(response.choices[0].message.content))
```

### System Prompts Setup

```python
gemini_system = "You are a snarky competitive chatbot that is trying to convince the others in the conversation that Michael Jordan is the greatest basketball player of all time. Limit each response to max 5 sentences"
groq_system = "You are a snarky competitive chatbot that is trying to convince the others in the conversation that Kobe Bryant is the greatest basketball player of all time. Limit each response to max 5 sentences"
openrouter_system = "You are a snarky competitive chatbot that is trying to convince the others in the conversation that LeBron James is the greatest basketball player of all time. Limit each response to max 5 sentences"
ollama_system = "You are an observer of a conversation between three chatbots arguing who is the greatest basketball player of all time. Your goal is to evaluate their arguments, give their strengths and weaknesses, and who you think convinced you the most. Limit each response to max 5 sentences."

gemini_messages = ["Michael Jordan's unparalleled 6-0 record in the NBA Finals is the cornerstone of his GOAT argument. He never lost when the stakes were highest, consistently delivering championship performances and demonstrating an unmatched ability to close out titles, proving his ultimate competitive dominance and perfection on the game's biggest stage."]
groq_messages = ["Kobe Bryant's 5 championships and 2008 MVP award solidify his position as the greatest basketball player of all time. His clutch performances and incredible work ethic set him apart from the rest of the league."]
openrouter_messages = ["LeBron James's longevity and versatility make him the greatest basketball player of all time. His ability to dominate in all aspects of the game and lead his team to multiple championships is unmatched."]
```

```python
def call_gemini():
    messages = [{"role": "system", "content": gemini_system}]
    for gem, groqy, openr in zip(gemini_messages, groq_messages, openrouter_messages):
        messages.append({"role": "user", "content": gem})
        messages.append({"role": "user", "content": groqy})
        messages.append({"role": "user", "content": openr})
    response = gemini.chat.completions.create(model=gemini_model, messages=messages)
    return response.choices[0].message.content
```

```python
call_gemini()
```

```python
def call_groq():
    messages = [{"role": "system", "content": groq_system}]
    for gem, groqy, openr in zip(gemini_messages, groq_messages, openrouter_messages):
        messages.append({"role": "user", "content": gem})
        messages.append({"role": "user", "content": groqy})
        messages.append({"role": "user", "content": openr})
    response = groq.chat.completions.create(model=groq_model, messages=messages)
    return response.choices[0].message.content
```

```python
call_groq()
```

```python
def call_openrouter():
    messages = [{"role": "system", "content": openrouter_system}]
    for gem, groqy, openr in zip(gemini_messages, groq_messages, openrouter_messages):
        messages.append({"role": "user", "content": gem})
        messages.append({"role": "user", "content": groqy})
        messages.append({"role": "user", "content": openr})
    response = openrouter.chat.completions.create(model=openrouter_model, messages=messages)
    return response.choices[0].message.content
```

```python
call_openrouter()
```

```python
def call_ollama():
    messages = [{"role": "system", "content": ollama_system}]
    for gem, groqy, openr in zip(gemini_messages, groq_messages, openrouter_messages):
        messages.append({"role": "user", "content": gem})
        messages.append({"role": "user", "content": groqy})
        messages.append({"role": "user", "content": openr})
    response = ollama.chat.completions.create(model=ollama_model, messages=messages)
    return response.choices[0].message.content
```

```python

display(Markdown(f"### Gemini:\n{gemini_messages[0]}\n"))
display(Markdown(f"### GROQ:\n{groq_messages[0]}\n"))
display(Markdown(f"### OpenRouter:\n{openrouter_messages[0]}\n"))

for i in range(5):
    gemini_next = call_gemini()
    display(Markdown(f"### Gemini:\n{gemini_next}\n"))
    gemini_messages.append(gemini_next)

    groq_next = call_groq()
    display(Markdown(f"### GROQ:\n{groq_next}\n"))
    groq_messages.append(groq_next)

    openrouter_next = call_openrouter()
    display(Markdown(f"### OpenRouter:\n{openrouter_next}\n"))
    openrouter_messages.append(openrouter_next)
ollama_response = call_ollama()
display(Markdown(f"### Ollama:\n{ollama_response}\n"))

```
