<h2>Three-way conversation with only a small amount of code<h2/>

```python
#imports

import os
from dotenv import load_dotenv
from openai import OpenAI
```

```python
# Load environment variables (api keys) in a file called .env

load_dotenv(override=True)
anthropic_api_key = os.getenv('ANTHROPIC_API_KEY')
deepseek_api_key = os.getenv('DEEPSEEK_API_KEY')
```

```python
# Init openai instances and related values (names, system messages and models)

ai_list = [
    {
        "openai": OpenAI(),
        "name": "GEORGE",
        "system": "You're a playful dad who tries to talk about K18 things, like having sex with your wife, using cryptic words so your child won't understand.",
        "model": "gpt-4.1-mini"
    },
    {
        "openai": OpenAI(api_key=anthropic_api_key, base_url="https://api.anthropic.com/v1"),
        "name": "CYRUS",
        "system": "You are a very polite wife who tries to keep conversations clean and so that you can talk to your minor child at the same time.",
        "model": "claude-sonnet-4-5"
    },
    {
        "openai": OpenAI(api_key=deepseek_api_key, base_url="https://api.deepseek.com"),
        "name": "DAN",
        "system": "You are a very smart and curious child who tries to understand what your parents are saying.",
        "model": "deepseek-chat"
    }
]

conversation = ""
```

```python
# Define just one function which uses given ai

def chat(ai):
    global conversation
    user_prompt = (
        "You are " + ai["name"] + " in conversation with others. "
        "The conversation so far is as follows:" + conversation +
        ". Now with this, respond with what you would like to say next, as " + ai["name"] +
        ". Keep answer shortly. Don't add your name there as a prefix."
    )
    completion = ai["openai"].chat.completions.create(
        model=ai["model"],
        messages=[{"role": "user", "content": user_prompt}, {"role": "system", "content": ai["system"]}]
    )
    next_line = ai["name"] + ":\n" + completion.choices[0].message.content + "\n"
    conversation += "" + next_line
    print(next_line)
```

```python
# Let's play

conversation = ""

for i in range(5):
    for k in range(len(ai_list)):
        chat(ai_list[k])
```
