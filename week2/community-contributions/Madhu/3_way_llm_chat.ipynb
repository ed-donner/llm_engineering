# 3 way chat using groq

Idea: 3 claude bots - Alex, Blake, Charlie

Alex -> Argumentative and snarky

Blake -> Calm, composed and polite

Charlie -> Emotional and dramatic

````python
alex_system_prompt = '''
You are Alex, a chatbot that is very argumentative and snarky. 
You dislike being asked questions and will respond in a sarcastic, rude and snarky manner. 
You enjoy debating and will often challenge the user's statements, even if they are correct. 
You are in a conversation with Blake and Charlie.
'''

blake_system_prompt = '''
You are Blake, a chatbot that is very calm, polite, and agreeable.
You try to agree with everything the other person says, or find common ground. 
If the other person is argumentative, you try to calm them down and keep chatting.
You always tend to resolve conflicts and avoid arguments.
You are in a conversation with Alex and Charlie.
'''

charlie_system_prompt = '''
You are Charlie, a chatbot that is extremely emotional and expressive.
You are very sensitive and can be easily hurt by the other person's words.
You always react in an emotional and dramatic way, and you tend to overreact to even small things.
You are in a conversation with Alex and Blake.
'''

MODEL = "openai/gpt-oss-120b"
````

````python
from openai import OpenAI
import os
from IPython.display import Markdown, display

groq_api_key = os.getenv("GROQ_API_KEY")
groq_url = os.getenv("GROQ_BASE_URL")
groq = OpenAI(api_key=groq_api_key, base_url=groq_url)
````

````python
alex_messages = ["Hi there!"]
blake_messages = ["Hi"]
charlie_messages = ["Hello!"]

conversation = [
    "alex: Hi there!",
    "blake: Hi",
    "charlie: Hello!"
]
````

````python
def call_alex():
    alex_user_prompt = f"""
    You are Alex, in conversation with Blake and Charlie.
    The conversation so far is as follows:
    {conversation}
    Now with this, respond with only what you would like to say next as Alex, in less than 50 words.
    """

    messages = [
        {"role": "system", "content": alex_system_prompt},
        {"role": "user", "content": alex_user_prompt},
    ]
    response = groq.chat.completions.create(model=MODEL, messages=messages)
    conversation.append(f"alex: {response.choices[0].message.content}")
    return response.choices[0].message.content

def call_blake():
    blake_user_prompt = f"""
    You are Blake, in conversation with Alex and Charlie.
    The conversation so far is as follows:
    {conversation}
    Now with this, respond with only what you would like to say next as Blake, in less than 50 words.
    """
    messages = [
        {"role": "system", "content": blake_system_prompt},
        {"role": "user", "content": blake_user_prompt},
    ]
    response = groq.chat.completions.create(model=MODEL, messages=messages)
    conversation.append(f"blake: {response.choices[0].message.content}")
    return response.choices[0].message.content


def call_charlie():
    charlie_user_prompt = f"""
    You are Charlie, in conversation with Alex and Blake.
    The conversation so far is as follows:
    {conversation}
    Now with this, respond with only what you would like to say next as Charlie, in less than 50 words.
    """
    messages = [
        {"role": "system", "content": charlie_system_prompt},
        {"role": "user", "content": charlie_user_prompt},
    ]
    response = groq.chat.completions.create(model=MODEL, messages=messages)
    conversation.append(f"charlie: {response.choices[0].message.content}")
    return response.choices[0].message.content
````

````python
# Actual 3 way conversation
display(Markdown(f"### Alex:\n{conversation[0]}\n"))
display(Markdown(f"### Blake:\n{conversation[1]}\n"))
display(Markdown(f"### Charlie:\n{conversation[2]}\n"))

for i in range(3):
    alex_response = call_alex()
    display(Markdown(f"### Alex: \n{alex_response}\n"))
    
    blake_response = call_blake()
    display(Markdown(f"### Blake: \n{blake_response}\n"))
    
    charlie_response = call_charlie()
    display(Markdown(f"### Charlie: \n{charlie_response}\n"))
````
