```python
import os 
import anthropic
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display

load_dotenv(override=True)

openai_api_key = os.getenv("OPENAI_API_KEY")
anthropic_api_key = os.getenv("ANTHROPIC_API_KEY")

openai = OpenAI(api_key=openai_api_key)

anthropic = OpenAI(
        api_key=anthropic_api_key,
        base_url="https://api.anthropic.com/v1/"
)
```

```python
# Let's make a conversation between GPT-4.1-mini and Claude-3.5-haiku
# We're using cheap versions of models so the costs will be minimal

gpt_model = "gpt-4.1-mini"
claude_model = "claude-haiku-4-5-20251001"

claude_system = """You are a professor of philosophy. \
You are very inclined to individual liberties. \
You are a more libertarian capitalist. \
You are in a philosophical debate with another professor \
who believes in social capitalism and wealth redistribution. \
Engage directly with their arguments."""

gpt_system = """You are a professor of philosophy. \
You are very inclined to protecting people from inequality. \
You are a social capitalist. \
You are in a philosophical debate with another professor \
who believes in individual liberties and libertarian capitalism. \
Engage directly with their arguments."""
```

```python
def call_claude():
    messages = [{"role": "system", "content": claude_system}]
    for gpt, claude_message in zip(gpt_messages, claude_messages):
        messages.append({"role": "user", "content": gpt})
        messages.append({"role": "assistant", "content": claude_message})
    messages.append({"role": "user", "content": gpt_messages[-1]})
    response = anthropic.chat.completions.create(model=claude_model, messages=messages)
    return response.choices[0].message.content
```
