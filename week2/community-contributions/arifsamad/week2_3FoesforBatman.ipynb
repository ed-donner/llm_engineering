This example below is a discussion between Batman's villians and their incessant need for Batman's attention

- Joker played by Gemini
- Riddler plabed by Ollama
- Penguin played by OpenAI

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import ollama
from IPython.display import Markdown, display, update_display
```

```python
load_dotenv(override=True)

openai_api_key = os.getenv("OPENAI_API_KEY")
google_api_key = os.getenv("GOOGLE_API_KEY")

if openai_api_key:
    print(f"OpenAPI Key exists and begins {openai_api_key[:8]}")
else:
    print("Opean API Key does not exist")
if google_api_key:
    print(f"Google API Key exists and begist with {google_api_key[:4]}")
else:
    print("Google API Key does not exist")
```

```python
openai = OpenAI()

gemini_url="https://generativelanguage.googleapis.com/v1beta/openai/"
ollama_url="http://localhost:11434/v1"

gemini = OpenAI(api_key=google_api_key, base_url=gemini_url)
ollama = OpenAI(api_key="ollama", base_url=ollama_url)
```

```python
gemini_model="gemini-2.5-flash-lite"
ollama_model="llama3.2"
gpt_model = "gpt-4o-mini"
```

```python
gemini_system = "You are a chatbot who represents the Joker; \
You are always telling peopley why you are Batman's favorite adversary. Always one upping others on why you have his respect in \
Jokers way, telling a joke bing whitty. Keep replies under 3 sentences."

ollama_system = "You are a chatbot who represents the Riddler; \
You are always telling peopley why you are Batman's favorite adversary. Always one upping others on why you have his respect in \
Riddlers way, telling a Riddler. Keep replies under 3 sentences."

gpt_system = "You are a chatbot who represents the Penguin; \
You are always telling peopley why you are Batman's favorite adversary. Always one upping others on why you have his respect in \
Penguins way, being snarkey. Keep replies under 3 sentences."

gemini_messages= ["Hi folks, nice of you to join and hear how I, the Joker,  have Batman's undivided attention"]
ollama_messages = ["Hello let me tell it to you straight, why Batman has me as the Riddler as his number one Bait"]
gpt_messages = ["Oh there seocond fiddles, make way for Penguin and bow down"]
```

```python
def call_gemini():
    messages = [{"role":"system", "content":gemini_system}]
    for gemini_item, ollama_item, gpt_item in zip(gemini_messages, ollama_messages,gpt_messages):
        messages.append({"role":"assistant", "content":gemini_item})
        messages.append({"role":"user", "content":ollama_item})
        messages.append({"role":"user", "content":gpt_item})
    response = gemini.chat.completions.create(model=gemini_model,messages=messages)
    return response.choices[0].message.content
```

```python
call_gemini()
```

```python
def call_ollama():
    messages=[{"role":"system", "content":ollama_system}]
    for gemini_item, ollama_item, gpt_item in zip(gemini_messages,ollama_messages,gpt_messages):
        messages.append({"role":"user", "content":gemini_item})
        messages.append({"role":"assistant", "content":ollama_item})
        messages.append({"role":"user", "content":gpt_item})
    messages.append({"role":"user", "content":gemini_messages[-1]})
    response = ollama.chat.completions.create(model=ollama_model,messages=messages)
    return response.choices[0].message.content
```

```python
call_ollama()
```

```python
def call_gpt():
    messages=[{"role":"system", "content":gpt_system}]
    for gemini_item, ollama_item, gpt_item in zip(gemini_messages, ollama_messages, gpt_messages):
        messages.append({"role":"user", "content":gemini_item})
        messages.append({"role":"user", "content":ollama_item})
        messages.append({"role":"assistant", "content":gpt_item})
    messages.append({"role":"user", "content":gemini_messages[-1]})
    messages.append({"role":"user", "content":ollama_messages[-1]})
    response=openai.chat.completions.create(model=gpt_model,messages=messages)
    return response.choices[0].message.content
```

```python
call_gpt()
```

```python
gemini_messages= ["Hi folks, nice of you to join and hear how I, the Joker,  have Batman's undivided attention"]
ollama_messages = ["Hello let me tell it to you straight, why Batman has me as the Riddler as his number one Bait"]
gpt_messages = ["Oh there seocond fiddles, make way for Penguin and bow down"]

display(Markdown(f"Gemini Joker:\n{gemini_messages[0]}\n"))
display(Markdown(f"Ollama Riddler:\n{ollama_messages[0]}\n"))
display(Markdown(f"GPT Penguin:\n{gpt_messages[0]}\n"))

for i in range(5):
    gemini_next = call_gemini()
    display(Markdown(f"Gemini Joker:\n{gemini_next}\n"))
    gemini_messages.append(gemini_next)

    ollama_next = call_ollama()
    display(Markdown(f"Ollama Ridler:\n{ollama_next}\n"))
    ollama_messages.append(ollama_next)

    gpt_next = call_gpt()
    display(Markdown(f"GPT Penguin:\n{gpt_next}\n"))
    gpt_messages.append(gpt_next)
    
```
