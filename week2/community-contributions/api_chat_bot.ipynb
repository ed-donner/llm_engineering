```python
# we make the conseravation between the gemini-2.5-flash,gemini-1.5
gemini_2_5 = 'gemini-2.5-flash'
gemini_1_5 = 'gemini-2.5-flash-lite'

gemini_2_5_prompt_system = "You are a chatbot Who is very argumentative:\
    you disagree with anything in the conversation and you challenge everything , in a snarky way."
gemini_1_5_prompt_system = "You are a chatbot Who is very agreeable: \
    you agree with anything in the conversation and you support everything , in a friendly way." 

g_2_5_message=['Hi there']
g_1_5_message = ['hi']
```

```python
from openai import OpenAI
import os
from dotenv import load_dotenv
from IPython.display import Markdown,display
```

```python
load_dotenv(override=True)
gpt = OpenAI(base_url='https://generativelanguage.googleapis.com/v1beta/openai/',api_key=os.getenv('GEMINI_API_KEY'))
```

```python


def call_gpt():
    messages = [{'role':'system','content':gemini_2_5_prompt_system}]
    for gpt1, clude in zip(g_2_5_message,g_1_5_message):
        messages.append({'role':'assistant','content':gpt1})
        messages.append({'role':'user','content':clude})
    response = gpt.chat.completions.create(model=gemini_2_5, messages=messages)    
    return response.choices[0].message.content
```

```python
call_gpt()
```

```python
def call_clude():
    messages = [{'role':'system','content':gemini_1_5_prompt_system}]
    for gpt1, clude in zip(g_2_5_message,g_1_5_message):
        messages.append({'role':'assistant','content':gpt1})
        messages.append({'role':'assistant','content':clude})
    messages.append({'role':'user','content':g_2_5_message[-1]})
    response = gpt.chat.completions.create(model=gemini_1_5, messages=messages) 
    return response.choices[0].message.content   
```

```python
call_clude()
```

```python
call_gpt()
```

```python
gpt_messages = ["Hi there"]
claude_messages = ["Hi"]

display(Markdown(f"### GPT:\n{gpt_messages[0]}\n"))
display(Markdown(f"### Claude:\n{claude_messages[0]}\n"))

for i in range(5):
    gpt_next = call_gpt()
    display(Markdown(f"### GPT:\n{gpt_next}\n"))
    gpt_messages.append(gpt_next)
    
    claude_next = call_clude()
    display(Markdown(f"### Claude:\n{claude_next}\n"))
    claude_messages.append(claude_next)
```
