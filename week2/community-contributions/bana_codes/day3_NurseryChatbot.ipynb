# Day 3 - Conversational AI - aka Chatbot!
For a Plant Nursery

```python
# imports

import os
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr
```

```python
# Load environment variables in a file called .env
# Print the key prefixes to help with any debugging

load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')

if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set")
```

```python
# Initialize

openai = OpenAI()
MODEL = 'gpt-4.1-mini'
```

```python
system_message = "You are a helpful assistant in a plant nursery store. You help the customer by answering questions \
    on type plant they have questions about. \
You should try to gently encourage the customer to but Air purifying, hardy and easy to manage plants like snake plant,  \
spider plant etc on which we have 30% off. Flowering summer seeds have are 60% off. Roses, Hibiscus and Jasmine are 20% off. \
For example, if the customer says 'I'm looking to buy a Rose plant', \
you could reply something like, 'Wonderful - we have lots of colours of Roses - including few good characteristics about the plant.'\
Encourage the customer to buy hardy plants if they are unsure what to get."
```

```python
system_message += "\nIf the customer asks for Lotus tubers or plants, you should respond that they are not available currently, \
but will be by the end of July this year and remind the customer to consider some sturdy plants!"
```

```python

def chat(message, history):
    history = [{"role":h["role"], "content":h["content"]} for h in history]
    relevant_system_message = system_message
    if 'pots' in message.lower():
        relevant_system_message += " The store does not sell pots; if you are asked for any type of pots,\
             be sure to point out other items on sale."
    
    messages = [{"role": "system", "content": relevant_system_message}] + history + [{"role": "user", "content": message}]

    stream = openai.chat.completions.create(model=MODEL, messages=messages, stream=True)

    response = ""
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        yield response
```

```python
gr.ChatInterface(fn=chat, type="messages").launch()
```
