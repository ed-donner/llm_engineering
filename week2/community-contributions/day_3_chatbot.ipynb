```python
import os 
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import display, Markdown
```

```python
load_dotenv(override=True)
client = OpenAI(base_url='https://generativelanguage.googleapis.com/v1beta/openai/',api_key= os.getenv('GEMINI_API_KEY'))

response = client.chat.completions.create(model='gemini-2.5-flash',messages=[{"role": "user", "content": "hellow"}])
display(Markdown(response.choices[0].message.content))
```

```python
def opposite_message(prompt):
    system_message = 'You are naughty assistant,You will always give opposition answer of prompt'
    message = [
        {'role': 'system', 'content': system_message },
        {'role': 'user', 'content': prompt}
    ]
    response = client.chat.completions.create(model='gemini-2.5-flash',messages=message)
    return response.choices[0].message.content
```

```python
import gradio as gr
```

```python
gr.Interface(fn=opposite_message, inputs='textbox',outputs='textbox', flagging_mode='never').launch()
```
