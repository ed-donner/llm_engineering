```python
sys_prompt = """ 
you are a debate partner.  Whatever the user says, find a logical reason to politely disagree with them
"""

def chat(message, history):
    stream = caller.chat.completions.create(
        model="llama3.2",
        messages=[
            {'role':'system', 'content':sys_prompt}] + history + [{'role':'user', 'content':message}],
        stream=True
    )
    result=""
    for chunk in stream:
        result += chunk.choices[0].delta.content or ""
        yield result
```
