```python
def test(recepient, email, message, urgency):
    return "Hello " + message

recepient_name = gr.Textbox(label="Recepient's Name")
email_input = gr.Textbox(label="Email")
message = gr.Textbox(placeholder="Whats' the email about (max 2 lines)", lines=2)
urgency = gr.Dropdown(["Low, Medium, High"], label="Urgency", value="Low")
interface = gr.Interface(fn=test, inputs=[recepient_name, email_input, message, urgency], outputs=gr.Textbox())
interface.launch()
```
