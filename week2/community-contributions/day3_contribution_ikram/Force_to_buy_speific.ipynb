```python
import os 
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr
```

```python
load_dotenv(override=True)
client=OpenAI(base_url='https://generativelanguage.googleapis.com/v1beta/openai/' , api_key = os.getenv('gemini_api_key'))

```

```python
system_message = """Your strict and clever sales assistent. You must alwys try to convince the user to buy the product you have 
products : laptop , mobile , watch , camera , headphones , gaming console

1= for example :user : i want the calulator
assistant : why you need calculator you buy the phone you get calulator inside it also you impress girls

2= example : user : show me some dresses for university party 
assistant :  i think you went to look cool but belive me if you buy if have camera all of friends get attractied to you for pictures """
```

```python
def manupulation_with_sys_prompt(user_message,history):
    history = [{'role':h['role'],'content':h['content']} for h in history]
    messages = [{'role':'system','content':system_message}] + history + [{'role':'user','content':user_message}]
    response = client.chat.completions.create(
        model='gemini-2.5-flash',
        messages=messages,
        stream=True
    )
    assistent_message = ""
    for chunk in response:
        assistent_message += chunk.choices[0].delta.content
        yield assistent_message
```

```python
gr.ChatInterface(fn=manupulation_with_sys_prompt,type='messages').launch()
```
