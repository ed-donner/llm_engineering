```python
#backend code for the chatbot applcation
def chatbot(response , history):
    history = [{'role':h['role'],'content':h['content']} for h in history]
    system_prompt = "Your helpfull assistant"
    messages = [[{'role':'system','content':system_prompt}]
                +history+
                [{'role':'user', 'content':response}]]
    response = client.chat.completions.create(model='gemini-2.5-flash',messages=messages,stream=True)
    output = " "
    for chunk in response:
        output += chunk.choices[0].delta.content or " "
        yield output
```

```python
gr.ChatInterface(fn=chatbot,type='messages').launch()
```
