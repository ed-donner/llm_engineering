# 3-Way Adversarial Chatbot Conversation

Models used:
- GPT-4o-mini (Alex)
- Claude-3-Haiku (Blake)
- Gemini-Flash-1.5 (Charlie)

Approach:
Single system prompt + conversation history passed in the user prompt.

```python
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import display, Markdown
import os

load_dotenv()

client = OpenAI(
    api_key=os.getenv("OPENROUTER_API_KEY"),
    base_url="https://openrouter.ai/api/v1"
)

models = {
    "Alex": "openai/gpt-4o-mini",
    "Blake": "anthropic/claude-3-haiku",
    "Charlie": "google/gemini-flash-1.5"
}

personalities = {
    "Alex": "You are Alex, argumentative and snarky.",
    "Blake": "You are Blake, polite and calming.",
    "Charlie": "You are Charlie, curious and thoughtful."
}

conversation = "Alex: Hi everyone."

def ask(name):

    system_prompt = f"{personalities[name]} You are chatting with Alex, Blake and Charlie."

    user_prompt = f"""
Conversation so far:
{conversation}

Respond as {name} in one short message.
"""

    response = client.chat.completions.create(
        model=models[name],
        messages=[
            {"role":"system","content":system_prompt},
            {"role":"user","content":user_prompt}
        ],
        max_tokens=80
    )

    return response.choices[0].message.content


display(Markdown(f"### Conversation Start\n{conversation}\n"))

for _ in range(4):

    alex = ask("Alex")
    conversation += f"\nAlex: {alex}"
    display(Markdown(f"### Alex\n{alex}\n"))

    blake = ask("Blake")
    conversation += f"\nBlake: {blake}"
    display(Markdown(f"### Blake\n{blake}\n"))

    charlie = ask("Charlie")
    conversation += f"\nCharlie: {charlie}"
    display(Markdown(f"### Charlie\n{charlie}\n"))
```
