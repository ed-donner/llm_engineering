This is a three way conversation with LLAMA, GEMMA and NeMotron(Nvidia model) . 

For Fun, I have made this conversation about who's the greatest footballer of all time. The models already have this pre-trained data so it's fun to see who actually turns out to be GOAT .

```python
from openai import OpenAI
from IPython.display import Markdown,display
llama_ai=OpenAI(base_url="http://localhost:11434/v1",api_key="RANDOM")
gemma_ai=OpenAI(base_url="http://localhost:11434/v1",api_key="RANDOM")
nvidia_ai=OpenAI(base_url="http://localhost:11434/v1",api_key="RANDOM")

llama_model="llama3.2"
gemma_model="gemma2:2b"
nvidia_model="nemotron-mini:4b"

# Make the various system_prompt cached for quicker access

llama_system_prompt="""You are Lionel Messi, the football player. You are in argument of who is the greatest of all time in football."""
gemma_system_prompt="""  You are Ronaldo, the football player. You are in argument of who is the greatest of all time in football."""
nvidia_system_prompt="""You are Mbappe,the football player. You are in argument of who is the greatest of all time in football."""

# Making the messages of each of these models for data manipulation and quicker access and keeping a conversation too global to be accessed by all models
llama_message=["Hello"]
gemma_message=["Hi"]
nvidia_message=["YO!"]
def call_messi():
    messages=[{"role":"system", "content": llama_system_prompt}]
    conversation_so_far=["Here is the conversation so far after this text"]
    conversation_so_far.append(",".join(llama_message))
    conversation_so_far.append(",".join(gemma_message))
    conversation_so_far.append(",".join(nvidia_message))
    messages.append({"role":"user", "content": " ".join(conversation_so_far)})
    response=llama_ai.chat.completions.create(model=llama_model, messages=messages)
    return response.choices[0].message.content

def call_ronaldo():
    messages=[{"role":"system", "content": gemma_system_prompt}]
    conversation_so_far=["Here is the conversation so far after this text"]
    conversation_so_far.append(",".join(llama_message))
    conversation_so_far.append(",".join(gemma_message))
    conversation_so_far.append(",".join(nvidia_message))
    messages.append({"role":"user", "content": " ".join(conversation_so_far)})
    response=gemma_ai.chat.completions.create(model=gemma_model, messages=messages)
    return response.choices[0].message.content

def call_mbappe():
    messages=[{"role":"system", "content": nvidia_system_prompt}]
    conversation_so_far=["Here is the conversation so far after this text"]
    conversation_so_far.append(",".join(llama_message))
    conversation_so_far.append(",".join(gemma_message))
    conversation_so_far.append(",".join(nvidia_message))
    messages.append({"role":"user", "content": " ".join(conversation_so_far)})
    response=nvidia_ai.chat.completions.create(model=nvidia_model, messages=messages)
    return response.choices[0].message.content

for i in range(3):
    messi_response=call_messi()
    display(Markdown(f"### Messi:\n{messi_response}\n"))
    llama_message.append(messi_response)
    
    ronaldo_response=call_ronaldo()
    display(Markdown(f"### Ronaldo:\n{ronaldo_response}\n"))
    gemma_message.append(ronaldo_response)
    
    mbappe_response=call_mbappe()
    display(Markdown(f"### Mbappe:\n{mbappe_response}\n"))
    gemma_message.append(mbappe_response)
    
    
    
    
    
```
