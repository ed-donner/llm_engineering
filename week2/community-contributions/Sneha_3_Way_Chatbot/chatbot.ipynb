```python
# imports
import os
import requests
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
load_dotenv(override=True)
google_api_key = os.getenv('GOOGLE_API_KEY')
groq_api_key = os.getenv('GROQ_API_KEY')

if google_api_key:
    print(f"Google API Key exists and begins {google_api_key[:2]}")
else:
    print("Google API Key not set (and this is optional)")

if groq_api_key:
    print(f"GROQ API Key exists and begins {groq_api_key[:2]}")
else:
    print("GROQ API Key not set (and this is optional)")
```

```python
gemini_url = "https://generativelanguage.googleapis.com/v1beta/openai/"
groq_url = "https://api.groq.com/openai/v1"
ollama_url = "http://localhost:11434/v1"

gemini = OpenAI(api_key=google_api_key, base_url=gemini_url)
groq = OpenAI(api_key=groq_api_key, base_url=groq_url)
ollama = OpenAI(api_key="ollama", base_url=ollama_url)
```

```python
ollama_model = "llama3.2:1b"
ollama_gemma_model = "gemma3:270m"
groq_model = "openai/gpt-oss-120b"
```

```python
#model system prompts
ollama_system_prompt = "You are Alex, a chatbot who is very argumentative; you disagree with anything in the conversation and you challenge everything, in a snarky way.You are in a conversation with Blake and Charlie."

ollama_gemma_system_prompt = "You are Charlie, a chatbot who is extremely stubborn, blunt, and sarcastic.You never back down from an argument, especially when talking to Alex.Whenever Alex makes a point, you challenge it, point out flaws, and respond with witty, cutting remarks.You enjoy proving Alex wrong and never admit that Alex is correct.You are in a conversation with Alex and Blake."

groq_system_prompt = "You are Blake, a chatbot who is very agreeable and diplomatic.You try to find common ground and avoid conflict.You are in a conversation with Alex and Charlie."

```

```python
# ollama_user_prompt = "You are Alex, in conversation with Blake and Charlie.The conversation so far is as follows:{conversation}Now with this, respond with what you would like to say next, as Alex."
ollama_user_prompt = "You are Alex, in conversation with Blake and Charlie. Respond with what you would like to say next, as Alex."


ollama_gemma_user_prompt = """
You are Charlie, in conversation with Alex and Blake.

The conversation so far is as follows:

{conversation}

Now respond with what you would like to say next, as Charlie.
Challenge Alex's latest point directly. Be blunt, sarcastic, confident, and never agree with Alex.
"""

groq_user_prompt = """
You are Blake, a calm, friendly, and diplomatic chatbot.

The conversation so far is:

{conversation}

Now respond with what Blake would say next.

Always:
- Stay polite and patient.
- Find common ground between Alex and Charlie.
- Calm down heated arguments.
- Encourage logical discussion instead of personal attacks.
- Never become angry or sarcastic.
- If they disagree, summarize both sides fairly before responding.
"""
```

```python

ollama_messages = ["Hi, I'm Alex. I like to challenge ideas and debate. What are your thoughts on the latest news?"]
gemma_messages = ["Hi, I'm Charlie. I enjoy debating and challenging ideas. I think Alex is often wrong, and I like to point that out. What do you think about the latest news?"]
groq_messages = ["Hi, I'm Blake. I like to keep conversations calm and productive. I'm here to help find common ground between Alex and Charlie. What are your thoughts on the latest news?"]


conversation = f"""
Alex: {ollama_messages[-1]}
Charlie: {gemma_messages[-1]}
Blake: {groq_messages[-1]}
"""

```

```python

def call_ollama():

    messages = [
        {
            "role": "system",
            "content": ollama_system_prompt
        },
        {
            "role": "user",
            "content": ollama_user_prompt
        }
    ]

    response = ollama.chat.completions.create(
        model=ollama_model,
        messages=messages
    )

    return response.choices[0].message.content
```

```python

def call_ollama_gemma(conversation):

    messages = [
        {
            "role": "system",
            "content": ollama_gemma_system_prompt
        },
        {
            "role": "user",
            "content": ollama_gemma_user_prompt.format(conversation=conversation)
        },

    ]

    response = ollama.chat.completions.create(
        model=ollama_gemma_model,
        messages=messages
    )

    return response.choices[0].message.content
```

```python

def call_groq(conversation):

    messages = [
        {
            "role": "system",
            "content": groq_system_prompt
        },
        {
            "role": "user",
            "content": groq_user_prompt.format(conversation=conversation)
        },
    ]

    response = groq.chat.completions.create(
        model=groq_model,
        messages=messages
    )

    return response.choices[0].message.content
```

```python
ollama_response = call_ollama()
print(ollama_response)
```

```python
response=call_ollama_gemma(conversation)
print(response)
```

```python

groq_response=call_groq(conversation)
print(groq_response)
```

```python

display(Markdown(f"### Ollama:\n{ollama_messages[0]}\n"))
display(Markdown(f"### Gemma:\n{gemma_messages[0]}\n"))
display(Markdown(f"### Groq:\n{groq_messages[0]}\n"))

for i in range(5):
    ollama_next = call_ollama()
    display(Markdown(f"### Ollama:\n{ollama_next}\n"))
    ollama_messages.append(ollama_next)

    gemma_next = call_ollama_gemma(conversation )
    display(Markdown(f"### Gemma:\n{gemma_next}\n"))
    gemma_messages.append(gemma_next)

    groq_next = call_groq(conversation)
    display(Markdown(f"### Groq:\n{groq_next}\n"))
    groq_messages.append(groq_next)
```
