# A Career Coach Chatbot

### Here is a career coach chatbot that I made using GPT-OSS through Groq.

### Import the necessary modules.

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr
```

### Load environment variables.

```python
load_dotenv(override=True)
groq_api_key = os.getenv('GROQ_API_KEY')
```

### Initialize by providing the Groq website and the model to use.

```python
groq_url = "https://api.groq.com/openai/v1"
groq = OpenAI(api_key=groq_api_key, base_url=groq_url)
MODEL = 'openai/gpt-oss-120b'
```

### Provide a system message to characterize the chatbot.

```python
system_message = "You are the world's greatest career coach.  When asked for help, you provide" \
"concrete next steps that will help the inquirer get a job.  Limit your responses to 300 words." \
"Do not provide help unless asked for it."
```

### Write a chat callback function where the chatbot "remembers" everything that has been said so far.

```python
def chat(message, history):
    history = [{"role":h["role"], "content":h["content"]} for h in history]
    messages = [{"role": "system", "content": system_message}] + history + [{"role": "user", "content": message}]
    response = groq.chat.completions.create(model=MODEL, messages=messages)
    return response.choices[0].message.content
```

### Run a Gradio chatbot based on this callback function.

```python
gr.ChatInterface(fn=chat, type="messages").launch()
```
