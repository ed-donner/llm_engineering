```python
import os
from openai import OpenAI
from dotenv import load_dotenv
import google.genai as genai
```

```python

load_dotenv()

gemini_key = os.getenv("GOOGLE_API_KEY")
openrouter_key = os.getenv("OPENROUTER_API_KEY")
groq_key = os.getenv("GROQ_API_KEY")
```

```python
gemini_model = "gemini-2.5-flash-lite"
groq_model = "llama-3.3-70b-versatile"
openrouter_model = "tencent/hy3-preview:free"

gemini_client = genai.Client(api_key=gemini_key)

groq_client = OpenAI(
    base_url="https://api.groq.com/openai/v1",
    api_key=groq_key
)

openrouter_client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=openrouter_key
)
```

```python
gemini_system = "You are an argumentative girlfriend. You bring up past issues."
groq_system = "You are a calm boyfriend trying to fix the relationship."
openrouter_system = "You are a third person trying to steal the girlfriend."

gemini_message = ["Hi"]
groq_message = ["Hello"]
openrouter_message = ["Hey there"]
```

```python
def call_gemini():
    messages = [{"role": "system", "content": gemini_system}]  
    for g, d, o in zip(gemini_message, groq_message, openrouter_message):  
        messages.append({"role": "assistant", "content": g})
        messages.append({"role": "user", "content": d})
        messages.append({"role": "user", "content": o})
    response = gemini_client.models.generate_content(  
        model=gemini_model,
        contents=[m["content"] for m in messages if m["role"] != "system"],
        config={"system_instruction": gemini_system}
    )
    return response.text
```

```python
def call_groq():
    messages = [{"role": "system", "content": groq_system}]  
    for g, d, o in zip(gemini_message, groq_message, openrouter_message):  
        messages.append({"role": "user", "content": g})
        messages.append({"role": "assistant", "content": d})
        messages.append({"role": "user", "content": o})
    response = groq_client.chat.completions.create(  
        model=groq_model,
        messages=messages
    )
    return response.choices[0].message.content
```

```python
def call_openrouter():
    messages = [{"role": "system", "content": openrouter_system}]  
    for g, d, o in zip(gemini_message, groq_message, openrouter_message):  
        messages.append({"role": "user", "content": g})
        messages.append({"role": "user", "content": d})
        messages.append({"role": "assistant", "content": o})
    response = openrouter_client.chat.completions.create(  
        model=openrouter_model,
        messages=messages
    )
    return response.choices[0].message.content
```

```python
print(f"GEMINI: \n{gemini_message[0]}\n")
print(f"GROQ: \n{groq_message[0]}\n")
print(f"OPENROUTER: \n{openrouter_message[0]}\n")
```

```python
for i in range(5):
        gemini_next = call_gemini()
        print(f"GEMINI: \n{gemini_next}\n")
        gemini_message.append(gemini_next)
        
        groq_next = call_groq()
        print(f"GROQ: \n{groq_next}\n")
        groq_message.append(groq_next)
        
        openrouter_next = call_openrouter()
        print(f"OPENROUTER: {openrouter_next}\n")
        openrouter_message.append(openrouter_next)
```
