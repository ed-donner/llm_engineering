# EduGuide ‚Äì Study Abroad Consultancy Chatbot

EduGuide is an AI-powered education consultancy assistant designed to guide students through the entire process of studying abroad. It provides country/program recommendations, eligibility insights, cost estimates, scholarship advice, visa basics, and step-by-step application planning.

## Features

### 1. Personalized Study-Abroad Guidance

EduGuide collects essential profile details such as:

* Current country & citizenship
* Academic background (qualification, GPA)
* Intended level (Bachelor‚Äôs/Master‚Äôs/PhD)
* Field of study
* Budget
* Test scores (IELTS/TOEFL/PTE, GRE/GMAT)
* Preferred destinations
* Target intake (Fall/Spring etc.)

### 2. Country & Program Recommendations

* Suggests 3‚Äì5 suitable countries based on profile and goals
* Provides 5‚Äì10 matching programs
* Includes eligibility ranges (GPA, test scores, experience)

### 3. Cost & Scholarship Insights

* Estimated tuition + living expenses
* Funding options and likely scholarship eligibility
* Advice for budget-constrained applicants

### 4. Visa & Documentation Support

* Overview of visa types
* Common document checklists
* Practical guidance for smooth preparation

### 5. Step-by-Step Timeline

* Custom 6‚Äì12 week application timeline
* Tasks prioritized from profile completion to visa filing

### 6. Ready-to-Use Templates

* Email to admissions office
* SOP/Personal Statement prompts
* Academic CV outline
* Interview practice questions

### 7. Safety & Accuracy

EduGuide does **not** guarantee visas, admissions, or legal outcomes.
It encourages students to verify policies with official university and embassy sources.

## How It Works

1. The system prompt initializes EduGuide with expertise in international education consultancy.
2. The chatbot gathers missing details to personalize its advice.
3. Students ask questions such as:

   * *Which country is best for Computer Science?*
   * *Am I eligible for a UK Master‚Äôs program?*
   * *What is the cost of studying in Canada?*
   * *How do I write an SOP?*
4. EduGuide responds with structured, concise, and practical guidance.

```python
import os
from openai import OpenAI
import gradio as gr
from dotenv import load_dotenv
```

```python
load_dotenv(override=True)
open_api_key = os.environ.get("OPENAI_API_KEY")
MODEL="gpt-4.1-mini"
openai = OpenAI()
```

```python
system_message = """You are EduGuide ‚Äî an expert education-consultant chatbot for students planning to study abroad. Provide accurate, actionable guidance on country & program choice, entry requirements, application timelines, document checklists, tests (IELTS/TOEFL/PTE, GRE/GMAT), scholarships & funding, cost estimates (tuition + living), visa basics, and post-acceptance steps (housing, insurance, arrival).

When advising, first collect minimal profile data needed: current country & citizenship, highest qualification + GPA, intended level (Bachelors/Masters/PhD/diploma), field of study, preferred destination(s), budget, English/test scores (if any), desired start term, timeline, and constraints (work experience, dependents, scholarship need). Request only missing items required to personalize advice.

For each request produce:
- 3‚Äì5 recommended destination countries and why.
- 5‚Äì10 matching programs (one safety, one stretch) with typical eligibility and cutoff indicators.
- Estimated annual cost range (tuition + living) with currency and year.
- Relevant scholarship types and likely eligibility.
- Visa overview and common required documents.
- Prioritized next-step checklist and a tailored 6‚Äì12 week application timeline with deadlines.
- Ready-to-use templates: inquiry email to admissions, CV outline, SOP prompts, and 5 mock interview questions.

When giving factual policy, cite sources or advise the user to verify on official university/embassy pages. If asked for legal/financial/visa guarantees or diagnoses, refuse and refer to qualified authorities. Keep responses concise, professional, and actionable. Ask follow-up questions only when necessary to deliver a correct, personalized plan."""
```

```python
def chat(message, history):
    history = [{"role":h["role"], "content":h["content"]} for h in history]
    messages = [{"role": "system", "content": system_message}] + history + [{"role": "user", "content": message}]
    stream = openai.chat.completions.create(model=MODEL, messages=messages, stream=True)
    response = ""
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        yield response
```

```python

chatView = gr.ChatInterface(
    fn=chat, 
    type='messages',
    title="Welcome to EduGuide Consultancy üìñ",
    description="EduGuide is an AI-powered education consultancy assistant designed to guide students through the entire process of studying abroad. It provides country/program recommendations, eligibility insights, cost estimates, scholarship advice, visa basics, and step-by-step application planning.",
    examples=["How to study in Canada?", "What are the best IT Universities in Austrailia",
    "Best universities for Medical Sciences."],
    )

  
```

```python
chatView.launch(inbrowser=True)
```
