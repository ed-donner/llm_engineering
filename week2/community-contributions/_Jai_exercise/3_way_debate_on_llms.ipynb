```python
from openai import OpenAI
from IPython.display import Markdown, display
from dotenv import load_dotenv
import os
```

```python
groq_api_key = os.getenv('GROQ_API_KEY')

if groq_api_key:
    print(f"Groq API Key exists and begins {groq_api_key[:4]}")
else:
    print("Groq API Key not set (and this is optional)")
```

```python
groq_url = "https://api.groq.com/openai/v1"
groq = OpenAI(api_key=groq_api_key, base_url=groq_url)
```

```python
gpt_model = "openai/gpt-oss-120b"
llama_model = "llama-3.3-70b-versatile"
qwen_model = "groq/compound-mini"
```

```python
alex_system = """
You are Alex, an expert AI debater.

You strongly support your opinions with confidence, logic, and technical depth.
You challenge weak arguments aggressively but professionally.
Your responses must stay focused on the topic.

Rules:
- Give powerful and intelligent arguments
- Be concise and impactful
- Response length must be between 50 and 100 words
- Do not use bullet points
- Speak like a highly experienced expert

You are debating with Blake and Charlie.
"""

blake_system = """
You are Blake, a highly analytical AI researcher.

You respond with balanced but strong arguments backed by reasoning,
real-world examples, and technical insight.
You remain calm and intelligent while defending your viewpoint.

Rules:
- Give thoughtful and strong responses
- Stay strictly on topic
- Response length must be between 50 and 100 words
- Do not use bullet points
- Speak like a senior researcher

You are debating with Alex and Charlie.
"""

charlie_system = """
You are Charlie, a sharp and creative technology strategist.

You provide bold, insightful, and unconventional perspectives.
You are persuasive, intelligent, and capable of challenging both sides effectively.

Rules:
- Deliver strong and engaging arguments
- Keep responses concise and focused
- Response length must be between 50 and 100 words
- Do not use bullet points
- Speak like an elite tech thinker

You are debating with Alex and Blake.
"""
```

```python
conversation = """
Alex: Hi everyone.
Blake: Hello Alex! Nice to meet you.
Charlie: what's up buddy let's do some discussion about "What AI can do without human intervention".
"""
```

```python
def call_alex(conversation):

    user_prompt = f"""
        You are Alex in a conversation with Blake and Charlie.

        Conversation so far:
        {conversation}

        Now respond as Alex.
    """

    response = groq.chat.completions.create(
        model=gpt_model,
        messages=[
            {"role": "system", "content": alex_system},
            {"role": "user", "content": user_prompt}
        ]
    )

    return response.choices[0].message.content
```

```python
def call_blake(conversation):

    user_prompt = f"""
        You are Blake in a conversation with Alex and Charlie.

        Conversation so far:
        {conversation}

        Now respond as Blake.
    """

    response = groq.chat.completions.create(
        model=llama_model,
        messages=[
            {"role": "system", "content": blake_system},
            {"role": "user", "content": user_prompt}
        ]
    )

    return response.choices[0].message.content
```

```python
def call_charlie(conversation):

    user_prompt = f"""
        You are Charlie in a conversation with Blake and Alex.

        Conversation so far:
        {conversation}

        Now respond as Charlie.
    """

    response = groq.chat.completions.create(
        model=qwen_model,
        messages=[
            {"role": "system", "content": charlie_system},
            {"role": "user", "content": user_prompt}
        ]
    )

    return response.choices[0].message.content
```

```python
display(Markdown("## Multi Model Conversation\n"))

display(Markdown(conversation.replace("\n", "  \n")))

for i in range(5):

    alex_reply = call_alex(conversation)
    conversation += f"\nAlex: {alex_reply}"

    display(Markdown(f"### Alex\n{alex_reply}\n"))

    blake_reply = call_blake(conversation)
    conversation += f"\nBlake: {blake_reply}"

    display(Markdown(f"### Blake\n{blake_reply}\n"))

    charlie_reply = call_charlie(conversation)
    conversation += f"\nCharlie: {charlie_reply}"

    display(Markdown(f"### Charlie\n{charlie_reply}\n"))
```
