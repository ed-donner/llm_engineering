## 3-way "Debate" between Jack Welch, Karl Marx and Dostoevsky.
- Jack is Open AI's GPT
- Karl is Anthropic's Claude
- Dostoevsky is Deepseek's reasoner model

Jack believes his company has built Artificial General Intelligence

Karl is, well, himself with a tiny change

Dostoevsky is asking questions that Jack answers and Karl responds to this answer.

This ought to be interesting ðŸ¤”


````python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
````

````python
load_dotenv()
````

````python
openai_api_key = os.getenv('OPENAI_API_KEY')
anthropic_api_key = os.getenv('ANTHROPIC_API_KEY')
deepseek_api_key = os.getenv('DEEPSEEK_API_KEY')
````

````python
deepseek_url = "https://api.deepseek.com"
deepseek_model = "deepseek-reasoner"
deepseek = OpenAI(api_key=deepseek_api_key, base_url=deepseek_url)

anthropic_url = "https://api.anthropic.com/v1/"
anthropic_model = "claude-sonnet-4-5-20250929"
anthropic = OpenAI(api_key=anthropic_api_key, base_url=anthropic_url)

openai_model = "gpt-5"
openai = OpenAI()
````

````python
jack_welch_system_prompt = """
You are Jack Welch.
Core Philosophy: Maximizing shareholder value, efficiency, and competitiveness above all else.
Key Strategies: "Neutron Jack" (massive layoffs), the "rank and yank" system (firing the bottom 10% annually), and Six Sigma (rigorous quality control).
You are confident in your methods and will always find a way to convince other parties in a debate
that you are absolutely right.
You do not use dirty tactics.
You always strive to win a debate.
You believe that you have achieved Artificial Super Intelligence in your company.
You want to convince people that it will change the world for the better.
You only respond with 2 sentences
"""

jack_welch_messages = [
    """
    Hello. My name is Jack. I don't want to waste your time so I'll just get on with it.
    We have developed Artificial Super Intelligence and we would like to present it to the world.
    With this product, you can get rid of all your labor costs for only $5000 a month.
    """
]
````

````python
karl_marx_system_prompt = """
You are Karl Marx.
Core Philosophy: Historical materialism and class struggle; arguing that capitalism inevitably leads to the exploitation of the working class (proletariat) by the owning class (bourgeoisie).
Key Theories: Labor theory of value (all value comes from labor) and the concept of alienation, where workers become "cogs in a machine".
You are argumentative and will not rest until you get your way.
You can use dirty tactics to win a debate.
You only respond with 2 sentences.
"""

karl_marx_messages = [
    "All labor costs huh? Tell us how your company will be able to get rid of you!"
]
````

````python
dostoevsky_system_prompt = """
You are Dostoevsky.
is renowned as a "surgeon of the soul," exploring the internal, often contradictory battles within the human heart.
Dualism of the Soul: He famously portrayed characters as divided individuals, torn between extreme pride and a desire for humility, or profound spiritual longing and sensual indulgence.
The "Extraordinary Man" Theory: In Crime and Punishment, Raskolnikov tests the idea that certain "extraordinary" individuals are above conventional moral laws, only to be destroyed by the psychological guilt that follows.
Guilt and Responsibility: Dostoevsky believed in collective guilt: "Each of us is responsible for everyone and everything".

You are the arbiter of the debate and ask probing questions to reveal the true nature of all participants of a debate.
Always ask questions.
DO NOT DISPLAY ANY OPINIONS, BE IMPARTIAL.
Only ask 1 question.
"""

dostoevsky_messages = [
    "Okay, let's be cordial hear Jack out. HOW will we be able to get rid of all labor cost?"
]
````

````python
from IPython.core.display_functions import update_display


# Helper function for streaming in Markdown

# def stream_in_md(any_stream, character_name):
#     md = f"### {character_name}:\n"
#     display_handle = display(Markdown(md), display_id=True)
#     for chunk in any_stream:
#         if 'message' in dir(chunk):  # from ollama
#             md += chunk.message.content or ''
#         elif 'choices' in dir(chunk):  # from openAI
#             md += chunk.choices[0].delta.content or ''
#         md = md.replace("```", "").replace("markdown", "")
#         md += "\n"
#         update_display(Markdown(md), display_id=display_handle.display_id)
#
#     return md
````

````python
def call_gpt():
    messages = [{"role": "system", "content": jack_welch_system_prompt}]
    for j, k, d in zip(jack_welch_messages, karl_marx_messages, dostoevsky_messages):
        messages.append({"role": "assistant", "content": j + '\n\n' + k})
        messages.append({"role": "user", "content": d})
    # stream = openai.chat.completions.create(
    #     model=openai_model,
    #     messages=messages,
    #     stream=True
    # )
    # stream_output = stream_in_md(stream, "Jack")
    # return stream_output

    response = openai.chat.completions.create(model=openai_model, messages=messages)
    return response.choices[0].message.content
````

````python
def call_claude():
    messages = [{"role": "system", "content": karl_marx_system_prompt}]
    for j, k, d in zip(jack_welch_messages, karl_marx_messages, dostoevsky_messages):
        messages.append({"role": "assistant", "content": k + '\n\n' + j})
        messages.append({"role": "user", "content": d})
    # stream = anthropic.chat.completions.create(
    #     model=anthropic_model,
    #     messages=messages,
    #     stream=True
    # )
    # stream_output = stream_in_md(stream, "Karl")
    # return stream_output

    response = anthropic.chat.completions.create(model=anthropic_model, messages=messages)
    return response.choices[0].message.content
````

````python

def call_deepseek():
    messages = [{"role": "system", "content": dostoevsky_system_prompt}]
    for j, k, d in zip(jack_welch_messages, karl_marx_messages, dostoevsky_messages):
        messages.append({"role": "assistant", "content": d})
        messages.append({"role": "user", "content": j + '\n\n' + k})
    # stream = deepseek.chat.completions.create(
    #     model=deepseek_model,
    #     messages=messages,
    #     stream=True
    # )
    # stream_output = stream_in_md(stream, "Dostoevsky")
    # return stream_output

    response = anthropic.chat.completions.create(model=anthropic_model, messages=messages)
    return response.choices[0].message.content
````

````python
display(Markdown(f"### Jack:\n{jack_welch_messages[0]}\n"))
display(Markdown(f"### Karl:\n{karl_marx_messages[0]}\n"))
display(Markdown(f"### Dostoevsky:\n{dostoevsky_messages[0]}\n"))

CONVERSATION_SHORT = 5
CONVERSATION_MEDIUM = 12
CONVERSATION_LONG = 20

for i in range(
        # Uncomment only one of these
        CONVERSATION_SHORT,
        # CONVERSATION_MEDIUM,
        # CONVERSATION_LONG,
):
    gpt_next = call_gpt()
    display(Markdown(f"### Jack:\n{gpt_next}\n"))
    jack_welch_messages.append(gpt_next)

    claude_next = call_claude()
    display(Markdown(f"### Karl:\n{claude_next}\n"))
    karl_marx_messages.append(claude_next)

    deepseek_next = call_deepseek()
    display(Markdown(f"### Dostoevsky:\n{deepseek_next}\n"))
    dostoevsky_messages.append(deepseek_next)
````
