```python
import os
import json
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr
import sqlite3

load_dotenv(override=True)
```

```python
# llm = OpenAI(base_url="http://127.0.0.1:11434/v1", api_key="")
llm = OpenAI(base_url="https://api.groq.com/openai/v1", api_key=os.getenv('GROQ_API_KEY'))

system_prompt = "You are a help AI assistant. Your task is to answer user questions. Only respond in markdown format."
```

```python
def answer_question(query, history, model):
    model = model
    messages = [{"role": "system", "content": system_prompt}]
    messages += [{"role": h["role"], "content": h["content"]} for h in history] + [{"role": "user", "content": query}]
    
    answer = ""
    chunks = llm.chat.completions.create(model=model, messages=messages, stream=True)

    for chunk in chunks:
        answer += chunk.choices[0].delta.content or ''
        yield answer
```

```python
with gr.Blocks() as ui:
    gr.Markdown("# Personal Q&A Assistant")
    
    model = gr.Dropdown(
        choices=["openai/gpt-oss-120b", "llama-3.3-70b-versatile", "qwen/qwen3-32b"],
        value="openai/gpt-oss-120b",
        label="Select Model"
    )
    
    gr.ChatInterface(
        fn=answer_question,
        additional_inputs=[model]
    )

```

```python
ui.launch()
```
