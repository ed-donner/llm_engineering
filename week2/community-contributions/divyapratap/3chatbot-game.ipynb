Three chatbots will play a game. They will pose as players, where each player will tell a word from the world of LLM Engineering, that word should start from the last letter of the word told by the previous player and a word already used can not be told again. Here, I have used ollama and tried three different local installed LLM models namely - gpt-oss:20b, llama3.2 and phi3. Performance of llama3.2 and phi3 was not adequate for this game so I used gpt-oss:20b as model for all the three players.

I think with frontier models this game can be much more interesting.

```python
gptoss_msg = ["Tokenization"]
llamaoss_msg = ["Neural"]
phi_msg = ["Lemmatization"]

gptoss_model = "gpt-oss:20b"
llamaoss_model = "llama3.2"
phi_model = "phi3"
```

```python
def call_gptoss():
    messages = [{"role": "system", "content": "You are one of the player in a game of 3 players, check the context and answer in one word (from the world of LLM engineering) that should start with the last letter of previous word"}]
    for gptoss, llama, phi in zip(gptoss_msg, llamaoss_msg, phi_msg):
        messages.append({"role": "assistant", "content": gptoss})
        messages.append({"role": "user", "content": llama})
        messages.append({"role": "user", "content": phi})
    print(messages)
    response = ollama.chat.completions.create(messages = messages, model=gptoss_model)
    return response.choices[0].message.content
```

```python
def call_llamaoss():
    messages = [{"role": "system", "content": "You are one of the player in a game of 3 players, check the context and answer in one word (from the world of LLM engineering) that should start with the last letter of previous word"}]
    for gptoss, llama, phi in zip(gptoss_msg, llamaoss_msg, phi_msg):
        messages.append({"role": "user", "content": gptoss})
        messages.append({"role": "assistant", "content": llama})
        messages.append({"role": "user", "content": phi})
    messages.append({"role": "user", "content": gptoss_msg[-1]})
    print(messages)
    response = ollama.chat.completions.create(messages = messages, model=llamaoss_model)
    return response.choices[0].message.content
```

```python
def call_phi():
    messages = [{"role": "system", "content": "You are one of the player in a game of 3 players, check the context and answer in one word (from the world of LLM engineering) that should start with the last letter of previous word"}]
    for gptoss, llama, phi in zip(gptoss_msg, llamaoss_msg, phi_msg):
        messages.append({"role": "user", "content": gptoss})
        messages.append({"role": "user", "content": llama})
        messages.append({"role": "assistant", "content": phi})
    messages.append({"role": "user", "content": gptoss_msg[-1]})
    messages.append({"role": "user", "content": llamaoss_msg[-1]})
    
    response = ollama.chat.completions.create(messages = messages, model=phi_model)
    return response.choices[0].message.content
```

```python
display(Markdown(f"### GPT:\n{gptoss_msg[0]}\n"))
display(Markdown(f"### llama:\n{llamaoss_msg[0]}\n"))
display(Markdown(f"### phi:\n{phi_msg[0]}\n"))

for i in range(2):
    gpt_next = call_gptoss()
    display(Markdown(f"### GPT:\n{gpt_next}\n"))
    gptoss_msg.append(gpt_next)
    
    llama_next = call_llamaoss()
    display(Markdown(f"### llama:\n{llama_next}\n"))
    llamaoss_msg.append(llama_next)

    phi_next = call_phi()
    display(Markdown(f"### phi:\n{phi_next}\n"))
    phi_msg.append(phi_next)
```
