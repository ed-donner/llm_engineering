```python
"""
Module: 3-Way Multi-Agent Financial Debate Orchestrator
Description: Simulates a multi-turn market efficiency debate between three distinct 
             LLM personas (Alex, Blake, Charlie) using a single-variable script buffer.
Author: Mark Byron
Date: June 2026
"""

import datetime

# 1. Initialize the single, shared conversation script
conversation = "Alex: The stock market is a caffeine-fueled pendulum and anyone calling it efficient is delusional."

# 2. Define the script-based helper function
def get_next_reply(model_id, persona_name, current_script):
    prompt = f"""
    You are {persona_name}. 
    The conversation so far is as follows:
    {current_script}
    
    Now with this, respond with what you would like to say next, as {persona_name}.
    Keep your response short (under 2 sentences) and stay completely in character. Do not prefix your response with your name.
    """
    
    response = litellm.completion(
        model=model_id, 
        messages=[{"role": "user", "content": prompt}]
    )
    
    return response.choices[0].message.content.strip()

# 3. Run the turn-taking loop
print(f"Starting Script:\n{conversation}\n" + "-"*40)

# Turn A: Blake (The Polite One) responds
blake_says = get_next_reply(BLAKE_MODEL, "Blake", conversation)
conversation += f"\n\nBlake: {blake_says}"

# Turn B: Charlie (Gemini Mediator) responds to both
charlie_says = get_next_reply(CHARLIE_MODEL, "Charlie", conversation)
conversation += f"\n\nCharlie: {charlie_says}"

# 4. Generate Pull Request Summary and Timestamp
current_time = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")

pr_header = f"""
======================================================================
PULL REQUEST POST-EXECUTION REPORT
Timestamp: {current_time}
======================================================================
Experiment: 3-Way Multi-Agent Dialogue Loop
Orchestration Framework: LiteLLM Router Implementation
Personas Engaged: 
  - Alex (GPT-4o-Mini): Market Cynic
  - Blake (Claude-3-Haiku): Diplomatic Realist
  - Charlie (Gemini-2.5-Flash): Analytical Mediator
======================================================================
"""

# 5. Print the complete output
print(pr_header)
print("FINAL TRANSCRIPT:")
print(conversation)
```
