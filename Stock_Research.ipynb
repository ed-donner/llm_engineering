```python
import pandas as pd # 'pd' is the standard nickname for pandas

# 1. We create a list of dictionaries
research_data = [
    {"Ticker": "AAPL", "Price": 312.51, "PE": 32.53},
    {"Ticker": "MSFT", "Price": 426.99, "PE": 22.08},
    {"Ticker": "GOOGL", "Price": 390.13, "PE": 26.92},
    {"Ticker": "TSLA", "Price": 442.10, "PE": 176.16},
    {"Ticker": "NVDA", "Price": 214.25, "PE": 16.93}
]

# 2. Turn that list into a DataFrame (the spreadsheet)
df = pd.DataFrame(research_data)

# 3. Sort by P/E Ratio (Lowest to Highest)
df_sorted = df.sort_values(by="PE")

# 4. Show the result
df_sorted
```

```python
import yfinance as yf

# 1. Create your watch list
watch_list = ["AAPL", "MSFT", "GOOGL", "TSLA", "NVDA"]

print(f"{'Ticker':<10} | {'Price':<10} | {'P/E Ratio':<10}")
print("-" * 35)

# 2. The Loop: "For every ticker in my list..."
for ticker in watch_list:
    stock = yf.Ticker(ticker)
    info = stock.info
    
    price = info.get('currentPrice', 0)
    pe = info.get('forwardPE', 0)
    
    # 3. Print a formatted row
    print(f"{ticker:<10} | ${price:<9.2f} | {pe:<10.2f}")
```

```python
# We already imported yfinance as yf, so we can just use it
ticker = "AAPL"
stock = yf.Ticker(ticker)
info = stock.info

# Pulling specific research metrics
current = info.get('currentPrice')
target = info.get('targetMeanPrice')
pe_ratio = info.get('forwardPE')
dividend = info.get('dividendYield', 0) * 100 # Convert to percentage

# Logic: Is it 'undervalued' based on analyst targets?
diff = ((target - current) / current) * 100

print(f"--- {ticker} RESEARCH REPORT ---")
print(f"Current Price:  ${current:.2f}")
print(f"Analyst Target: ${target:.2f}")
print(f"P/E Ratio:      {pe_ratio:.2f}")
print(f"Dividend Yield: {dividend:.2f}%")
print(f"Upside Potential: {diff:.1f}%")
```
